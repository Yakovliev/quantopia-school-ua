Темна матерія — фундаментальна концепція в астрофізиці та космології, яка прагне пояснити кілька спостережуваних явищ у Всесвіті, які неможливо пояснити лише видимою (баріонною) матерією. Незважаючи на те, що темна матерія становить приблизно 27% від загальної маси-енергії у Всесвіті, вона залишається невловимою, оскільки вона не взаємодіє з електромагнітними силами, тобто не випромінює, не поглинає та не відбиває світло, що робить її невидимою для сучасних телескопічних інструментів.

1. Докази темної матерії
a. Криві обертання Галактики:

Спостереження: зірки в галактиках, особливо у зовнішніх регіонах, обертаються навколо галактичного центру приблизно з тією ж швидкістю, що й зірки поблизу центру.
Очікування: згідно з фізикою Ньютона та розподілом видимої матерії, зірки, розташовані далі від центру, повинні обертатися повільніше.
Підсумок: плоскі криві обертання свідчать про наявність додаткової, невидимої маси, яка справляє гравітаційний вплив, який ми приписуємо темній матерії.

b. Гравітаційне лінзування:

Спостереження: масивні об’єкти, такі як скупчення галактик, відхиляють світло від об’єктів позаду них, ефект, передбачений загальною теорією відносності Ейнштейна.
Невідповідність: кількість видимої маси не може пояснити ступінь спостережуваного лінзування.
Підсумок: додаткова маса, яка вважається темною матерією, сприяє ефекту гравітаційної лінзи.
в. Космічний мікрохвильовий фон (CMB):

Спостереження: CMB надає знімок раннього Всесвіту, що показує хвилинні коливання температури.
Аналіз: закономірності цих коливань узгоджуються із Всесвітом, що містить темну матерію, яка вплинула на формування великомасштабних структур.
d. Формування великомасштабної структури:

Спостереження: розподіл і формування галактик і скупчень галактик протягом космічного часу.
Підсумок: темна матерія забезпечує необхідні гравітаційні риштування для того, щоб звичайна матерія збиралася разом і утворювала спостережувані структури.

2. Характеристики темної матерії
Несвітиться: не випромінює електромагнітне випромінювання та не взаємодіє з ним.
Небаріонний: Ймовірно, складається з частинок, відмінних від протонів і нейтронів.
Холодний або теплий: відноситься до швидкостей частинок; «холодна» темна матерія рухається повільніше порівняно зі швидкістю світла, сприяючи формуванню структури.
Без зіткнень: взаємодіє в основному через силу тяжіння з незначною взаємодією через інші сили.
3. Кандидати в темну матерію
a. Слабко взаємодіючі масивні частинки (WIMP):

Опис: гіпотетичні частинки, які взаємодіють через слабку ядерну силу та гравітацію.
Статус: провідний кандидат, хоча пряме виявлення залишається невловимим, незважаючи на значні експериментальні зусилля.
b. Аксіони:

Опис: Надзвичайно легкі частинки, запропоновані для вирішення сильної проблеми CP у квантовій хромодинаміці (КХД).
Статус: ще один перспективний кандидат, його шукають такі експерименти, як Експеримент темної матерії Axion (ADMX).
в. Стерильні нейтрино:

Опис: тип нейтрино, який не взаємодіє через стандартну слабку силу, лише через гравітацію.
Статус: може пояснити деяку темну матерію, хоча їх точна роль ще досліджується.
d. Масивні компактні гало-об’єкти (MACHO):

Опис: звичайна матерія у формі чорних дір, нейтронних зірок, коричневих карликів або інших компактних об’єктів.
Статус: ймовірно, вносить лише невелику частку темної матерії на основі обмежень спостережень.
4. Виявлення та дослідження
Пряме виявлення: такі експерименти, як LUX, XENON і PandaX, спрямовані на спостереження частинок темної матерії, які взаємодіють зі звичайною матерією у високочутливих детекторах.
Непряме виявлення: пошук продуктів анігіляції або розпаду темної матерії, таких як гамма-промені, нейтрино або космічні промені, за допомогою телескопів і космічних обсерваторій.
Пошуки на колайдерах: експерименти на прискорювачах елементарних частинок, як-от Великий адронний колайдер (LHC), намагаються створити частинки темної матерії шляхом зіткнень високої енергії.
