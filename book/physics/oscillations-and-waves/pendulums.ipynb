# Математичний і пружинний маятники

Перш ніж ми почнемо, важливо зрозуміти, що таке **фізична модель**. У реальному світі процеси надзвичайно складні. Наприклад, коливання кульки на пружині залежить від маси, форми та розмірів кульки, властивостей пружини, опору повітря, температури і навіть від того, як обертається Земля. Врахувати все це дуже складно.

Тому у фізиці створюють **ідеалізовані моделі** - спрощені версії реальних об'єктів або явищ, в яких ми нехтуємо несуттєвими на даному етапі факторами. Це дозволяє виділити головні закономірності. Пружинний та математичний маятники - це саме такі моделі.

## Коливання пружинного маятника

> **Пружинний маятник** - це фізична модель коливальної системи, що складається з тіла масою $m$, підвішеного на абсолютно пружній пружині жорсткістю $k$, коливання в якій відбуваються без тертя.

"Абсолютно пружна" означає, що пружина ідеально підкоряється закону Гука. "Без тертя" означає, що ми нехтуємо опором повітря та внутрішнім тертям в пружині. Тобто, ми розглядаємо ідеальні, незатухаючі коливання.

Давайте з'ясуємо, чому взагалі виникають коливання. Розглянемо вантаж масою $m$ на горизонтальній пружині, що прикріплений пружиною з жорсткістю $k$ до вертикальної стіни. Це модель **горизонтального пружинного маятника**.

1.  **Положення рівноваги.** У положенні $x=0$ пружина не деформована, жодних сил в горизонтальній площині на тіло не діє, і воно може перебувати в стані спокою. Це і є положення рівноваги.
2.  **Виведення з рівноваги.** Відхилимо тіло праворуч на відстань $x$. Пружина розтягнеться, і виникне **сила пружності** $F_{пруж}$, яка, згідно із законом Гука, намагатиметься повернути тіло назад до положення рівноваги.

> **Сила пружності** $F_{пруж}$ - це сила, що виникає в тілі при його деформації і напрямлена в бік, протилежний зміщенню частинок тіла. Вона визначається законом Гука:
>
> $$F_{пруж} = -kx$$
>
> де $k$ - це **жорсткість пружини**, а $x$ - видовження (деформація) пружини. Знак "мінус" вказує на те, що сила пружності завжди напрямлена протилежно до зміщення. **Жорсткість пружини** $k$ показує, яку силу треба прикласти, щоб розтягнути (або стиснути) пружину на одиницю довжини. **Одиниця в СІ:** ньютон на метр (Н/м).

3.  **Виникнення коливань.** Ця повертаюча сила надає тілу прискорення, спрямованого до положення рівноваги. За другим законом Ньютона

    $$\vec{F}_{пруж} + m\vec{g} + \vec{N} = m\vec{a}$$

    Оскільки сила тяжіння $m\vec{g}$ і сила нормальної реакції опори $\vec{N}$ взаємно компенсуються, то 
    
    $$\vec{F}_{пруж} = m\vec{a}$$

    При проектуванні цього рівняння на горизонтальну вісь ОХ отримуємо

    $$ma_x = -kx$$

    Звідси ми можемо виразити прискорення:

    $$a_x = -\frac{k}{m}x$$

## Виведення формули періоду коливань пружинного маятника

А тепер згадаємо матеріал попередньої сторінки. Ми встановили, що **динамічною ознакою гармонічних коливань** є рівняння:

$$a_x = -\omega^2x$$

Це рівняння говорить, що рух є гармонічним коливанням, якщо прискорення тіла прямо пропорційне його зміщенню і напрямлене в протилежний бік.

Порівняємо два отримані рівняння:

$$a_x = -\frac{k}{m}x \quad \text{та} \quad a_x = -\omega^2x$$

Ми бачимо, що вони мають абсолютно ідентичну структуру! Це доводить, що **вільні коливання пружинного маятника є гармонічними**. Зіставивши ці два вирази, ми легко знаходимо квадрат циклічної частоти:

$$\omega^2 = \frac{k}{m}$$

Отже, **циклічна частота** коливань пружинного маятника дорівнює:

$$
\omega = \sqrt{\frac{k}{m}}
$$

А знаючи циклічну частоту, ми можемо вивести формулу для **періоду коливань** $T$. Зв'язок між ними нам відомий: $T = \frac{2\pi}{\omega}$. Підставимо наш вираз для $\omega$:

$$
T = \frac{2\pi}{\sqrt{\frac{k}{m}}}
$$

Після спрощення отримуємо остаточну формулу для періоду коливань пружинного маятника:

$$
T = 2\pi\sqrt{\frac{m}{k}}
$$

**Аналіз формули:**
* Період коливань тим більший, чим більша маса вантажу $m$ (більш інертне тіло важче "розгойдати").
* Період коливань тим менший, чим більша жорсткість пружини $k$ (більш жорстка пружина швидше повертає тіло до положення рівноваги).
* Важливо: Період коливань пружинного маятника залежить тільки від маси тіла $m$ та жорсткості пружини $k$, і не залежить від амплітуди коливань (при пружних коливаннях) або гравітаційних умов (на Землі, в космосі, на Місяці тощо). Цю властивість можна використати для визначення маси об'єктів у невагомості, коли звичайні ваги не функціонують: знаючи жорсткість пружини $k$ і вимірявши період коливань $T$, можна обчислити масу $m$.
* Властивість незалежності від амплітуди називають *ізохронністю* коливань.

Примітка: для вертикального маятника (Рис. 2) положення рівноваги зміщується вниз під дією сили тяжіння, однак сила тяжіння є сталою, тому вона не впливає на характер коливань навколо нового положення рівноваги, і формула періоду коливань залишається тією ж самою.

## Перетворення енергії в пружинному маятнику

Під час коливань відбувається безперервне перетворення енергії. В ідеальній системі без тертя повна механічна енергія $E$ залишається сталою. Вона складається з кінетичної та потенціальної енергій.

> **Кінетична енергія** $E_k$ - це енергія руху.
>
> $$E_k = \frac{mv^2}{2}$$
>
> **Одиниця в СІ:** джоуль (Дж).

> **Потенціальна енергія** $E_p$ пружно деформованого тіла - це енергія, накопичена в пружині.
>
> $$E_p = \frac{kx^2}{2}$$
>
> **Одиниця в СІ:** джоуль (Дж).

Таким чином, повна механічна енергія системи:

$$E = E_k + E_p = \frac{mv^2}{2} + \frac{kx^2}{2} = \text{const}$$

Давайте прослідкуємо за перетвореннями енергії протягом одного коливання:
1.  **В крайньому положенні ($x = A$)**: Тіло на мить зупиняється ($v=0$), отже, кінетична енергія $E_k=0$. Пружина максимально розтягнута (або стиснута), тому потенціальна енергія максимальна: $E_p = \frac{kA^2}{2}$. Вся енергія системи - потенціальна: $E = \frac{kA^2}{2}$.
2.  **Рух до положення рівноваги**: Пружина стискається, повертаючи тіло. Її потенціальна енергія зменшується, а швидкість тіла зростає, тобто $E_p$ перетворюється на $E_k$.
3.  **В положенні рівноваги ($x = 0$)**: Пружина не деформована, отже $E_p=0$. Швидкість тіла максимальна ($v = v_{max}$), тому кінетична енергія теж максимальна: $E_k = \frac{mv_{max}^2}{2}$. Вся енергія системи - кінетична: $E = \frac{mv_{max}^2}{2}$. Якщо початок відліку $t=0$ вести від моменту, коли зміщення тіла $x = A$, то положення рівноваги досягається в момент часу $t = T/4$.
4.  **Рух до іншого крайнього положення**: Тіло за інерцією продовжує рух, стискаючи (або розтягуючи) пружину. Швидкість зменшується, а деформація зростає. $E_k$ перетворюється назад на $E_p$.
5.  **В іншому крайньому положенні ($x = -A$)**: Процес повторюється, тіло знову зупиняється ($E_k=0$), а потенціальна енергія знову максимальна ($E_p = \frac{k(-A)^2}{2} = \frac{kA^2}{2}$). Якщо початок відліку $t=0$ вести від моменту, коли зміщення тіла $x = A$, то зміщення $x = -A$ досягається в момент часу $t = T/2$, що якраз відповідає половині одного коливання.

У другій половині періоду процес повторюється в зворотному напрямку: візок рухається вправо, проходить положення рівноваги в момент часу $t = \frac{3T}{4}$ і досягає початкового положення максимального відхилення вправо при $t = T$, завершуючи повний цикл коливання. Після цього процес повторюється.

Ці перетворення відбуваються безперервно, підтримуючи коливальний рух.

## Математичний маятник

Тверде тіло, яке може коливатися відносно осі, що проходить через точку підвісу, називають **фізичним маятником**. Наприклад, гойдалка, вантаж на мотузці, маятник годинника. Ще одним прикладом є люстра, що висить на стелі. При виведенні з рівноваги вона починає коливатися. Однак аналіз таких коливань є складним, оскільки на них впливають розміри та форма тіла, властивості провода, на якому підвішена люстра та інші фактори.

Для спрощення дослідження коливань використовують ідеалізовану модель - математичний маятник. 

Щоб створити таку модель, необхідно:
1. використати нитку, довжина якої значно перевищує розміри підвішеного тіла;
2. забезпечити, щоб маса нитки була набагато меншою за масу тіла, що дозволяє розглядати тіло як матеріальну точку;
3. нитка має бути нерозтяжною, щоб відстань від тіла до точки підвісу залишалася сталою.

> **Математичний маятник** - це ідеалізована коливальна система, що складається з матеріальної точки масою $m$, підвішеної на невагомій та нерозтяжній нитці довжиною $l$, яка коливається в полі тяжіння без тертя.

*  **Матеріальна точка** - тіло, розмірами якого можна знехтувати в умовах даної задачі.
*  **Невагома і нерозтяжна нитка** - ідеалізація, що означає, що маса та деформація нитки не впливають на рух.

## Динаміка коливань математичного маятника

Розглянемо кульку на нитці, відхилену від положення рівноваги на невеликий кут $\alpha$. Які сили на неї діють?
1.  **Сила тяжіння** $F_т = mg$, напрямлена вертикально вниз.
2.  **Сила натягу нитки** $T_н$, напрямлена вздовж нитки.

Рівнодійна цих сил і змушує маятник коливатися. Щоб проаналізувати рух, розкладемо силу тяжіння на дві складові:
*   **Нормальна складова** $F_n = mg \cos\alpha$, напрямлена вздовж нитки. Вона зрівноважується силою натягу нитки: $T_н = F_n$.
*   **Тангенціальна (дотична) складова** $F_{\tau} = mg \sin\alpha$. Вона напрямлена по дотичній до траєкторії руху (дуги кола) і завжди спрямована до положення рівноваги.

Саме тангенціальна складова є **повертаючою силою**. Отже, рівняння руху вздовж траєкторії має вигляд:

$$ma = -F_{\tau} = -mg \sin\alpha$$

$$a = -g \sin\alpha$$

Знак "мінус" знову вказує, що сила є повертаючою.

## Виведення формули періоду: наближення малих кутів

Рівняння $a = -g \sin\alpha$ є складним, і такий рух, строго кажучи, не є гармонічним. Однак, якщо кут відхилення $\alpha$ є малим (зазвичай до 5-10 градусів), ми можемо зробити дуже влучне спрощення. Для малих кутів, виражених в радіанах, виконується наближена рівність (хоча вам це наближення може здаватись чимось новим і незрозумілим, однак це наближчення дуже часто застосовується у фізиці для спрощення багатьох виведень та запису формул):

$$\sin\alpha \approx \alpha$$

Для того, щоб переконатися у справедливості цього наближення, можна виконати наступні кроки:
1. Візьмемо довільний малий кут, наприклад, 5 градусів.
2. Переведемо цей кут в радіани:

    $$\text{радіани} = \text{градуси} \times \frac{\pi}{180}$$

    Ця формула походить від того, що повне коло становить $360^\circ$, або $2 \pi$ радіан. Відповідно, $180^\circ$ дорівнює $\pi$ радіан.

    Підставимо $5^\circ$ у формулу:

    $$\text{радіани} = 5 \times \frac{\pi}{180} = \frac{5\pi}{180} = \frac{\pi}{36} \approx 0.087266$$

3. Розрахуємо синус 5 градусів (можете використати для цього будь-який калькулятор). Результат буде приблизно 0.087156, що доволі близько до значення 5 градусів, що виражені в радіанах. Для менших кутів точність буде ще більшою.

Для малих кутів горизонтальне зміщення $x$ від положення рівноваги майже дорівнює довжині дуги $s$. А довжина дуги пов'язана з кутом як $s \approx l \sin\alpha \approx l \alpha$, де $l$ - довжина нитки. Отже, ми можемо записати $x \approx l\alpha$, звідки 

$$\alpha \approx \frac{x}{l}$$

Підставимо ці наближення в наше рівняння для повертаючої сили:

$$F_{\tau} = m g \sin\alpha \approx mg\alpha \approx mg\frac{x}{l}$$

Тепер знову застосуємо другий закон Ньютона $F = ma_x$:

$$ma_x = -\frac{mg}{l}x$$

Скоротивши масу $m$ з обох боків, отримуємо:

$$a_x = - \frac{g}{l}x$$

Ми знову отримали рівняння, яке є **динамічною ознакою гармонічних коливань** $a_x = -\omega^2x$. Це означає, що **при малих кутах відхилення коливання математичного маятника є гармонічними**.

Порівнюючи два рівняння, знаходимо квадрат циклічної частоти:

$$\omega^2 = \frac{g}{l}$$

Звідки **циклічна частота** коливань математичного маятника:

$$
\omega = \sqrt{\frac{g}{l}}
$$

І, нарешті, використовуючи зв'язок $T = \frac{2\pi}{\omega}$, виводимо формулу для періоду коливань математичного маятника, відому як **формула Гюйгенса**:

$$
T = \frac{2\pi}{\sqrt{\frac{g}{l}}}
$$

$$
T = 2\pi\sqrt{\frac{l}{g}}
$$

**Аналіз формули:**
*   Період коливань залежить від **довжини нитки** $l$ та **прискорення вільного падіння** $g$. Чим довша нитка, тим більший період.
*   Період коливань **не залежить від маси** підвішеного тіла та (для малих кутів) від **амплітуди** коливань. Цю властивість незалежності від амплітуди називають *ізохронністю* коливань. Саме завдяки їй маятники стали основою для створення точних годинників.

Формула Гюйгенса дозволяє використовувати математичний маятник для визначення прискорення вільного падіння: вимірявши довжину нитки і період коливань, можна обчислити значення $g$ у даній місцевості:

$$
T = 2\pi\sqrt{\frac{l}{g}}
$$

$$
T^2 = 4\pi^2 \frac{l}{g}
$$

$$
g = \frac{4\pi^2 l}{T^2}
$$

## Перетворення енергії в математичному маятнику

Як і у пружинного маятника, тут теж відбувається перетворення енергії. Повна механічна енергія $E = E_k + E_p$ зберігається. Але тепер потенціальна енергія - це **гравітаційна потенціальна енергія**.

$$E_p = mgh$$

де $h$ - висота тіла над найнижчим положенням (положенням рівноваги).

**В крайньому положенні (максимальне відхилення)**: Тіло на мить зупиняється ($v=0$), тому $E_k=0$. Воно знаходиться на максимальній висоті $h_{max}$, отже потенціальна енергія максимальна: $E_p = mgh_{max}$.

**В положенні рівноваги (найнижча точка)**: Висота $h=0$, тому $E_p=0$. Швидкість тіла максимальна ($v = v_{max}$), отже кінетична енергія максимальна: $E_k = \frac{mv_{max}^2}{2}$.

Протягом коливань потенціальна енергія висоти перетворюється на кінетичну енергію руху і навпаки, забезпечуючи неперервний рух.

## Підсумок: Пружинний та Математичний маятники

| Характеристика | Пружинний маятник | Математичний маятник |
| :--- | :--- | :--- |
| **Визначення** | Фізична модель, що складається з тіла масою $m$ на абсолютно пружній пружині жорсткістю $k$. | Ідеалізована система, що складається з матеріальної точки масою $m$ на невагомій, нерозтяжній нитці довжиною $l$. |
| **Повертаюча сила** | Сила пружності, що підкоряється закону Гука $F_{пруж} = -kx$. | Тангенціальна складова сили тяжіння $F_{\tau} = -mg \sin\alpha$. Для малих кутів $F_{\tau} \approx -mg\frac{x}{l}$. |
| **Рівняння руху (для гармонічних коливань)** | Прискорення прямо пропорційне зміщенню $a_x = -\frac{k}{m}x$. | При малих кутах відхилення прискорення пропорційне зміщенню $a_x = -\frac{g}{l}x$. |
| **Циклічна частота** | $\omega = \sqrt{\frac{k}{m}}$ | $\omega = \sqrt{\frac{g}{l}}$ |
| **Період коливань** | $T = 2\pi\sqrt{\frac{m}{k}}$ | $T = 2\pi\sqrt{\frac{l}{g}}$ (формула Гюйгенса) |
| **Залежність періоду** | Залежить від **маси тіла** $m$ та **жорсткості пружини** $k$. | Залежить від **довжини нитки** $l$ та **прискорення вільного падіння** $g$. |
| **Незалежність періоду (Ізохронність)** | Не залежить від амплітуди коливань (при пружних деформаціях) та від $g$. | Не залежить від **маси тіла** $m$ та (при малих кутах) від **амплітуди**. |
| **Повна механічна енергія** | $E = \frac{mv^2}{2} + \frac{kx^2}{2} = \text{const}$ | $E = \frac{mv^2}{2} + mgh = \text{const}$ |
| **Енергія в крайніх точках ($x=\pm A$)** | Швидкість $v=0$. Кінетична енергія $E_k = 0$. Потенціальна енергія максимальна $E_p = \frac{kA^2}{2}$. | Швидкість $v=0$. Кінетична енергія $E_k = 0$. Потенціальна енергія максимальна $E_p = mgh_{max}$. |
| **Енергія в положенні рівноваги ($x=0$)** | Швидкість максимальна $v=v_{max}$. Потенціальна енергія $E_p=0$. Кінетична енергія максимальна $E_k = \frac{mv_{max}^2}{2}$. | Швидкість максимальна $v=v_{max}$. Потенціальна енергія $E_p=0$ (при $h=0$). Кінетична енергія максимальна $E_k = \frac{mv_{max}^2}{2}$. |
