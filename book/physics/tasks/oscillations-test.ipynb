# Механічні коливання. Самостійна робота

## Механічні коливання

**Задача 1**

Варіант 1. Механічний годинник для своєї роботи використовує коливання маятника, який здійснює 1800 коливань за 15 хвилин. Визначте період та частоту цих коливань.

Варіант 2. Крила комара під час польоту здійснюють 12000 коливань за 20 секунд. Визначте період та частоту цих коливань.

Варіант 3. Під час шторму невеличкий човен на хвилях здійснив 90 коливань вгору-вниз за 3 хвилини. Визначте період та частоту коливань човна.

---

**Задача 2**

Варіант 1. Комаха, що летить, робить 1200 змахів крилами за 4 секунди. Визначте період та частоту цих коливань.

Варіант 2. Гойдалка в парку атракціонів здійснює 45 повних коливань за 90 секунд. Визначте період та частоту цих коливань.

Варіант 3. Маятник настінного годинника здійснив 900 коливань за 15 хвилин. Визначте період та частоту цих коливань.

---

**Задача 3**

Варіант 1. Гілка дерева під дією вітру за 40 секунд здійснює 20 повних коливань. Амплітуда цих коливань становить 30 см. Визначте частоту та період коливань гілки.

Варіант 2. Поршень у двигуні автомобіля на холостих обертах здійснює 200 повних коливань (вгору-вниз) за 10 секунд. Амплітуда його руху становить 4 см. Визначте частоту та період коливань поршня.

Варіант 3. Струна гітари після щипка здійснює 880 повних коливань за 2 секунди. Амплітуда початкового коливання становить 3 мм. Визначте частоту та період коливань струни.

---

**Задача 4**

Варіант 1. Коливання точки на поверхні води описуються рівнянням: $x = 0.05 \cos(10\pi t)$, де всі величини вказані в СІ.
1. Чому дорівнює амплітуда коливань?
2. Визначте циклічну частоту.
3. Обчисліть період та частоту коливань.

Варіант 2. Коливання вантажу, підвішеного на пружині, описуються рівнянням: $x = 0.1 \cos(5\pi t)$, де всі величини вказані в СІ.
1. Чому дорівнює амплітуда коливань?
2. Визначте циклічну частоту.
3. Обчисліть період та частоту коливань.

Варіант 3. Рівняння коливань струни музичного інструменту має вигляд: $x = 0.001 \cos(880\pi t)$, де всі величини вказані в СІ.
1. Чому дорівнює амплітуда коливань?
2. Визначте циклічну частоту.
3. Обчисліть період та частоту звуку, який видає струна.
