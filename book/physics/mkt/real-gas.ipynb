# Реальні гази

## Вступ: Від ідеального до реального

У шкільному курсі фізики ми багато працюємо з моделлю **ідеального газу**. Це чудова і проста модель, яка добре працює в багатьох ситуаціях, наприклад, для опису повітря у кімнаті за звичайних умов (нормального атмосферного тиску та кімнатної температури).

Давайте пригадаємо основні припущення цієї моделі:
1.  **Молекули - це матеріальні точки.** Вони не мають власного об'єму.
2.  **Молекули не взаємодіють між собою на відстані.** Вони взаємодіють лише під час пружних зіткнень, як більярдні кулі.

Стан ідеального газу описується знаменитим **рівнянням стану ідеального газу**:

$$pV = \nu RT$$

де:
*   $p$ - тиск газу (одиниця в СІ: Паскаль, Па);
*   $V$ - об'єм газу (одиниця в СІ: метр кубічний, м³);
*   $\nu$ - кількість речовини (одиниця в СІ: моль, моль);
*   $T$ - абсолютна температура (одиниця в СІ: Кельвін, К);
*   $R$ - універсальна газова стала, $R \approx 8.314$ Дж/(моль·К).

Але що станеться, якщо ми почнемо сильно стискати газ (збільшувати тиск) або сильно його охолоджувати (зменшувати температуру)? У цих умовах модель ідеального газу перестає працювати. Газ починає поводитись інакше, і врешті-решт він може навіть перетворитися на рідину. Такий газ, поведінка якого залежить від взаємодії та розміру молекул, називається **реальним газом**.

Наша мета - зрозуміти, чому модель ідеального газу дає збій, і як її можна виправити, щоб вона описувала реальні гази.

---

## Що не так з ідеальним газом?

У реальному світі молекули - це не просто точки. Вони мають цілком конкретний, хоч і дуже малий, **власний об'єм**. Отже, перше, що ми повинні врахувати: **реальний об'єм, доступний для руху молекул, є меншим за об'єм посудини.**

Крім того, між ними існують **сили міжмолекулярної взаємодії**.

Ці сили мають подвійну природу:
*   **Сили притягання:** Діють на відносно більших відстанях (порівняно з розміром молекул). Саме завдяки цим силам газ при достатньому охолодженні та стисненні може перетворитися на рідину. Молекули ніби "чіпляються" одна за одну.
*   **Сили відштовхування:** Проявляються на дуже малих відстанях, коли електронні оболонки молекул починають "накладатися" одна на одну. Саме ці сили не дають молекулам "злипнутися" в одну точку і відповідають за те, що речовина має власний об'єм.

Отже, друге, що ми маємо врахувати: **молекули реального газу притягуються одна до одної, що впливає на тиск, який газ чинить на стінки посудини.**

## Рівняння стану Ван-дер-Ваальса

Голландський фізик **Ян Ван-дер-Ваальс** у 1873 році запропонував геніально просте виправлення рівняння стану ідеального газу, яке враховувало б ці два фактори. Його ідея полягала не в тому, щоб створити нове рівняння з нуля, а в тому, щоб "покращити" вже існуюче, ввівши до нього дві поправки: одну для об'єму, іншу - для тиску.

Давайте пройдемо його шляхом. Ми візьмемо рівняння ідеального газу $pV = \nu RT$ і по черзі змінимо в ньому $V$ і $p$.

### Перша поправка: врахування власного об'єму молекул

Як ми вже з'ясували, вільний об'єм, у якому можуть рухатись молекули, менший за загальний об'єм посудини $V$. На скільки менший? Він менший на величину, яку можна назвати "забороненим об'ємом" - тим простором, який зайнятий самими молекулами.

Ван-дер-Ваальс ввів величину $b$, яка позначає об'єм, що фактично "займає" один моль молекул. Ця величина враховує не тільки власний об'єм молекул, а й те, що центри двох молекул не можуть наблизитись один до одного на відстань, меншу за діаметр молекули. Тому виявляється, що цей "виключений об'єм" приблизно в **4 рази більший за сумарний об'єм самих молекул**.

*   $b$ - **стала Ван-дер-Ваальса**, яка є поправкою на власний об'єм молекул (одиниця в СІ: м$^3$/моль).

Якщо в посудині знаходиться $\nu$ молів газу, то сумарний "заборонений" об'єм буде дорівнювати $\nu b$. Отже, вільний об'єм, доступний для руху, буде не $V$, а $V_{\text{вільний}} = V - \nu b$.

Тепер ми можемо замінити $V$ у рівнянні ідеального газу на цей вираз:

$$p(V - \nu b) = \nu RT$$

Це вже перший крок до рівняння реального газу.

**Пояснення "виведення" $b$ (концептуальне):**

Уявіть собі дві сферичні молекули радіусом $r$. Коли вони зближуються, центр однієї молекули не може наблизитися до центру іншої на відстань меншу, ніж сума їхніх радіусів, тобто $2r$. Це означає, що навколо кожної молекули є так званий "виключений об'єм", куди *центр* іншої молекули не може проникнути.

Для *пари* молекул цей "виключений об'єм" становить сферу радіусом $2r$. Об'єм цієї сфери:

$$ V_{\text{викл, пара}} = \frac{4}{3}\pi (2r)^3 = 8 \cdot \frac{4}{3}\pi r^3 $$

Ми знаємо, що об'єм однієї молекули $V_{\text{мол}} = \frac{4}{3}\pi r^3$. Отже, $V_{\text{викл, пара}} = 8 \cdot V_{\text{мол}}$.

Цей об'єм стосується *двох* молекул. Якщо ми хочемо знайти, який виключений об'єм "припадає" на *одну* молекулу в середньому, ми повинні розділити цей об'єм на два, оскільки кожна молекула робить свій внесок у створення цього виключеного об'єму для іншої:

$$ V_{\text{викл, на 1 мол.}} = \frac{V_{\text{викл, пара}}}{2} = \frac{8 \cdot V_{\text{мол}}}{2} = 4 \cdot V_{\text{мол}} $$

Таким чином, виключений об'єм на одну молекулу приблизно в 4 рази перевищує її власний об'єм, як ми згадували вище.

Якщо ми маємо $\nu$ молів газу, а в одному молі $N_A$ молекул (число Авогадро), то поправка $b$ враховує цей виключений об'єм для одного моля:

$$ b = 4 \cdot N_A \cdot V_{\text{мол}} $$

Ця поправка зменшує об'єм у рівнянні ідеального газу. Отже, рівняння стану тепер виглядає так:

$$ p(V - \nu b) = \nu RT $$

### Друга поправка: врахування міжмолекулярного притягання

Тепер розберемося з тиском. Тиск газу на стінки посудини створюється ударами молекул об ці стінки. Розглянемо молекулу, яка летить до стінки. В ідеальному газі на неї нічого не діє. У реальному газі її з усіх боків притягують інші молекули.

*   Якщо молекула знаходиться **всередині об'єму**, ці сили притягання в середньому скомпенсовані, бо її тягнуть у всі боки однаково.
*   Але якщо молекула знаходиться **біля стінки**, то з одного боку (з боку стінки) молекул немає, а з іншого (з боку об'єму газу) - є. Виникає рівнодійна сила, спрямована **вглиб газу**.

Ця сила "пригальмовує" молекулу перед ударом об стінку. Внаслідок цього удар стає слабшим, а отже, і тиск, який ми вимірюємо манометром ($p$), буде **меншим** за той "ідеальний" тиск, який був би за відсутності сил притягання.

Нам потрібно додати до виміряного тиску $p$ якусь поправку, яку назвемо **внутрішнім** або **молекулярним тиском** $p_м$. Тобто, в рівнянні стану ми маємо використовувати не реальний виміряний тиск $p$, а "ідеальний" тиск $p_{\text{ідеальний}} = p + p_м$.

Тепер найцікавіше: як знайти цей внутрішній тиск $p_м$? Давайте поміркуємо.

1.  Сила, що діє на одну окрему молекулу біля стінки, пропорційна концентрації молекул у газі. Чим більше молекул навколо, тим сильніше вони її "тягнуть" назад. Концентрація $n = \frac{N}{V}$, де $N$ - кількість молекул, або ж $n = \frac{\nu N_A}{V} \sim \frac{\nu}{V}$. Чим більша концентрація, тим сильніше молекули притягують ті, що біля стінки.
2.  Кількість молекул, що вдаряються об стінку за одиницю часу, також пропорційна їхній концентрації і, в свою чергу, пропорційна $\frac{\nu}{V}$. Чим більше молекул в одиниці об'єму, тим частіше вони б'ють по стінці.

Отже, загальне зменшення тиску (внутрішній тиск $p_м$) пропорційне добутку цих двох факторів:

$$ p_м \sim \left(\text{концентрація}\right) \cdot \left(\text{концентрація}\right) $$

$$ p_м \sim \frac{\nu}{V} \cdot \frac{\nu}{V} $$

або

$$p_м \sim \left(\frac{\nu}{V}\right)^2$$

Щоб перейти від пропорційності до рівності, вводять ще одну константу, яка залежить від конкретного газу і характеризує інтенсивність міжмолекулярного притягання. Назвемо її $a$.

*   $a$ - **стала Ван-дер-Ваальса**, що характеризує сили міжмолекулярного притягання (одиниця в СІ: Па·м$^6$/моль$^2$).

Таким чином, формула для внутрішнього тиску набуває вигляду:

$$ p_м = \frac{a\nu^2}{V^2}$$

Оскільки молекулярний тиск $p_м$ зменшує "реальний" тиск газу, щоб отримати тиск, який був би, якби не було притягання (тобто "ідеальний" тиск), ми повинні до виміряного тиску $p$ додати цю поправку:

$$ p_{\text{ідеальний}} = p + p_м = p + \frac{a\nu^2}{V^2} $$

Тепер ми можемо внести і цю поправку в наше рівняння (оскільки в рівнянні стану ідеального газу фігурує саме тиск ідеального газу, а не реального, то ми й шукали, як виразити тиск ідеального газу через тиск реального). Замість "ідеального" тиску $p_{\text{ідеальний}}$ ми підставляємо вираз $p + p_м$:

$$\left(p + \frac{a\nu^2}{V^2}\right)(V - \nu b) = \nu RT$$

Це і є знамените **рівняння стану реального газу Ван-дер-Ваальса**!

Воно враховує і власний об'єм молекул (через сталу $b$), і сили притягання між ними (через сталу $a$). Сталі $a$ і $b$ для кожного газу визначаються експериментально і є в довідниках.

## Аналіз рівняння Ван-дер-Ваальса. Ізотерми реального газу

Отже, ми отримали рівняння стану реального газу, яке враховує ключові відмінності від ідеальної моделі:

$$\left(p + \frac{a\nu^2}{V^2}\right)(V - \nu b) = \nu RT$$

На перший погляд, воно виглядає не набагато складнішим за рівняння стану ідеального газу. Але його математична структура та фізичні наслідки значно глибші. Щоб це побачити, давайте проведемо деякі алгебраїчні перетворення. Наша мета - виразити це рівняння у вигляді полінома відносно об'єму $V$.

### Математичне перетворення рівняння

Спочатку розкриємо дужки:

$$p(V - \nu b) + \frac{a\nu^2}{V^2}(V - \nu b) = \nu RT$$

$$pV - p\nu b + \frac{a\nu^2}{V} - \frac{ab\nu^3}{V^2} = \nu RT$$

Тепер, щоб позбутися від $V$ та $V^2$ у знаменниках, помножимо обидві частини рівняння на $V^2$:

$$pV \cdot V^2 - p\nu b \cdot V^2 + \frac{a\nu^2}{V} \cdot V^2 - \frac{ab\nu^3}{V^2} \cdot V^2 = \nu RT \cdot V^2$$

Після спрощення отримуємо:

$$pV^3 - p\nu bV^2 + a\nu^2V - ab\nu^3 = \nu RTV^2$$

І наостанок, перенесемо всі члени в один бік і згрупуємо їх за степенями змінної $V$:

$$pV^3 - p\nu bV^2 - \nu RTV^2 + a\nu^2V - ab\nu^3 = 0$$

Винесемо $V^2$ за дужки в другому та третьому членах:

$$pV^3 - (p\nu b + \nu RT)V^2 + a\nu^2V - ab\nu^3 = 0$$

Що ми отримали? Це **кубічне рівняння** відносно об'єму $V$. І це є фундаментальною відмінністю від рівняння ідеального газу ($pV - \nu RT = 0$), яке є лінійним відносно $V$.

**Що це означає?** Лінійне рівняння при заданих $p$ і $T$ завжди має лише один розв'язок для $V$. А ось кубічне рівняння може мати **один** або **три** дійсних корені. Саме ця математична особливість дозволяє рівнянню Ван-дер-Ваальса описувати такий складний фізичний процес, як перехід газу в рідину.

### 1. Ізотерма при високій температурі $T > T_c$

Верхня крива на графіку відповідає високій температурі. Ми бачимо, що вона дуже схожа на звичайну гіперболу, яка описує ізотерму ідеального газу. Чому так?

*   При високій температурі кінетична енергія молекул ($\sim T$) є дуже великою. На її фоні енергія міжмолекулярного притягання (пов'язана зі сталою $a$) стає незначною.
*   Також при високій температурі (і не надто високому тиску) газ займає великий об'єм $V$, тому поправка на власний об'єм молекул $\nu b$ стає дуже малою порівняно з $V$.

Таким чином, при високих $T$ і великих $V$ доданками $\frac{a\nu^2}{V^2}$ та $\nu b$ можна знехтувати, і рівняння Ван-дер-Ваальса перетворюється на рівняння ідеального газу. У цьому випадку наше кубічне рівняння має лише один дійсний корінь для $V$ при будь-якому $p$.

### 2. Ізотерма при низькій температурі $T < T_c$

Нижня крива є набагато цікавішою (синя крива на Рис. 1). Вона має характерну хвилеподібну ділянку. Давайте пройдемось по ній справа наліво, ніби ми стискаємо газ у циліндрі під поршнем.

*   **Ділянка праворуч (газоподібний стан):** Ми починаємо стискати газ. Його об'єм $V$ зменшується, а тиск $p$ плавно зростає. Тут речовина поводиться як газ.
*   **Хвиляста ділянка:** Модель Ван-дер-Ваальса прогнозує, що при досягненні певного об'єму тиск почне падати, а потім знову стрімко зростати. Частина цієї ділянки, де об'єм і тиск зростають одночасно, є абсолютно нефізичною. Така система була б нестабільною.

**Що ж відбувається в реальному експерименті?**
У реальності, коли ми стискаємо газ і досягаємо певної точки, коли починається процес **конденсації**. Газ починає перетворюватися на рідину. Під час цього процесу, поки в посудині співіснують обидві фази (рідина і насичена пара), тиск і температура залишаються **сталими**. На графіку цей процес відповідає не хвилі, а прямій горизонтальній лінії (на Рис. 1 ця лінія не зображена).
*   Коли ми рухаємось по цій лінії справа наліво, кількість газу зменшується, а кількість рідини зростає. У лівій точці перетину цієї горизонтальної лінії та теоретичної ізотерми весь газ перетворився на рідину.
*   **Ділянка ліворуч (рідкий стан):** Рідини дуже погано стискаються. Тому навіть незначне зменшення об'єму вимагає величезного зростання тиску. Ось чому ізотерма в цій області йде майже вертикально вгору.

З точки зору нашого кубічного рівняння, в області тисків, де відбувається конденсація, воно має **три дійсних корені** для об'єму $V$. Найменший корінь відповідає об'єму рідини, найбільший - об'єму газу (пари), а середній корінь не має фізичного сенсу (він лежить на нестабільній ділянці).

Підсумуємо:
*   **Газова фаза:** При великих об'ємах і низьких тисках (права частина графіка) газ поводиться як ідеальний - тиск повільно зростає зі зменшенням об'єму.
*   **Фазовий перехід (конденсація):** При певних значеннях тиску і температури газ починає конденсуватися в рідину. На реальних ізотермах (які ми отримуємо експериментально), ця ділянка є **горизонтальною**. Це означає, що при постійній температурі тиск залишається незмінним, поки весь газ не перетвориться на рідину. У цей час співіснують дві фази: газ і рідина.
*   **Рідка фаза:** Після повної конденсації газу в рідину (ліва частина графіка) об'єм продовжує зменшуватися, але тиск зростає дуже різко. Це пояснюється тим, що рідина майже нестислива.
*   **Ізотерма Ван-дер-Ваальса та S-подібна крива:** Рівняння Ван-дер-Ваальса передбачає хвилясту S-подібну криву з мінімумом і максимумом замість горизонтальної ділянки фазового переходу. Ділянка між мінімумом і максимумом (де об'єм зменшується при зростанні тиску) є **фізично нестабільною** і не реалізується в чистому вигляді. Вона відповідає метастабільним станам, таким як перегріта рідина (ділянка зростання тиску після мінімуму) або переохолоджена пара (ділянка спаду тиску перед максимумом). Насправді, ці стани нестійкі і швидко переходять у стабільні двофазні системи.

**Приклад**. Уявіть, що ми стискаємо газ в циліндрі при постійній температурі.
1.  **Починаємо з великого об'єму (газ):** Поршень легко рухається, тиск зростає повільно.
2.  **Досягаємо точки конденсації:** Раптом, при подальшому русі поршня (зменшенні об'єму), тиск перестає зростати. Це тому, що газ починає перетворюватися на рідину. Енергія йде на формування міжмолекулярних зв'язків, а не на стиснення.
3.  **Весь газ сконденсувався (рідина):** Поршень майже неможливо рухати далі. Навіть невелике зменшення об'єму викликає дуже різке зростання тиску, оскільки рідина майже нестислива.

Рівняння Ван-дер-Ваальса було першою моделлю, яка могла якісно описати ці явища фазового переходу.

### 3. Критична ізотерма $T = T_c$

Між двома розглянутими режимами існує проміжний, особливий випадок. Це ізотерма, що відповідає **критичній температурі** $T_c$.

*   На цій ізотермі хвиляста ділянка вироджується, "стягуючись" в одну точку - **точку перегину K**. Ця точка називається **критичною точкою**.
*   У критичній точці всі три дійсних корені нашого кубічного рівняння зливаються в один.
*   **Фізичний сенс критичної точки:** При температурах, вищих за критичну $T > T_c$, газ неможливо перетворити на рідину, яким би великим не був тиск. Речовина перебуває в особливому **надкритичному стані**, де зникає різниця між рідиною і газом.

Отже, рівняння Ван-дер-Ваальса, попри свою відносну простоту, не тільки краще описує поведінку газів при високих тисках, а й якісно пояснює одне з найважливіших явищ у фізиці - фазовий перехід газ-рідина та існування критичної точки. Це був величезний крок уперед у розумінні природи речовини.

## Обмеження рівняння Ван-дер-Ваальса

Хоча рівняння Ван-дер-Ваальса стало величезним кроком вперед, воно є наближеним і має свої обмеження:

*   **Наближений характер сталих $a$ і $b$:** Вважається, що $a$ і $b$ є константами, незалежними від температури і тиску. Насправді, вони можуть дещо змінюватися, особливо для різних діапазонів умов. Це робить рівняння менш точним у широкому діапазоні температур і тисків.
*   **Неточний опис фазового переходу:** Як ми бачили, S-подібна крива не точно відображає горизонтальну ділянку конденсації. Вона лише якісно показує існування фазового переходу та метастабільних станів (перегрітої рідини та переохолодженої пари), але не може точно описати рівновагу між рідкою та газоподібною фазами.
*   **Недостатня точність для всіх речовин:** Рівняння краще працює для газів з простими, неполярними молекулами (наприклад, благородних газів) та при не надто екстремальних умовах. Для складних молекул, сильно полярних речовин або при дуже високих тисках чи низьких температурах його точність значно знижується.
*   **Ігнорування багаточастинкових взаємодій:** Модель Ван-дер-Ваальса враховує лише парні взаємодії між молекулами, ігноруючи складніші взаємодії, які можуть виникати, коли багато молекул знаходяться близько одна до одної.

Незважаючи на ці обмеження, рівняння Ван-дер-Ваальса залишається фундаментальним у термодинаміці та статистичній фізиці, оскільки воно вперше успішно врахувало реальні властивості молекул та якісно пояснило явище конденсації газів. Воно заклало основу для розробки більш складних і точних рівнянь стану.
