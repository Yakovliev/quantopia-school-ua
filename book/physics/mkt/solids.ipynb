# Будова та властивості твердих тіл

На перший погляд, усі тверді тіла здаються однаковими - вони просто тверді. Але якщо зазирнути всередину, на рівень атомів та молекул, ми побачимо два абсолютно різні світи. За внутрішньою структурою тверді тіла поділяються на кристалічні та аморфні.

## Кристали та аморфні тіла

Уявіть собі бібліотеку. В одному випадку всі книги розставлені на полицях в ідеальному порядку: за алфавітом, за жанрами, за розміром. Ви можете легко передбачити, де знайти наступну книгу. Це аналогія до **кристалічного тіла**.

А тепер уявіть, що всі книги з бібліотеки просто звалили на одну велику купу. Порядок відсутній. Це аналогія до **аморфного тіла**.

Тепер перейдемо до наукових термінів.

> **Тверде тіло** - це агрегатний стан речовини, що характеризується стабільністю форми та об'єму, а також сильною взаємодією між частинками (атомами, молекулами або іонами), які здійснюють лише коливання навколо певних положень рівноваги.

Існує два основні типи твердих тіл: кристалічні та аморфні. Розглянемо їх детальніше.

### Кристалічні тіла

> **Кристалічні тіла** - це тверді тіла, атоми, молекули або іони яких розташовані у просторі суворо впорядковано, утворюючи регулярну просторову структуру, яка повторюється. Цей порядок зберігається на великих відстанях.

Цей ідеальний порядок називається **далеким порядком**. Він означає, що знаючи положення кількох частинок, ми можемо точно передбачити, де буде знаходитись частинка на будь-якій, навіть дуже великій, відстані.

Найменший структурний елемент, що періодично повторюється в ґратці, називається **елементарною коміркою**. Усю кристалічну ґратку можна уявити як нескінченну кількість таких "цеглинок", складених разом.

*Приклади кристалічних тіл:* кухонна сіль (NaCl), цукор, алмаз, графіт, усі метали (залізо, мідь, алюміній), кварц.

### Аморфні тіла

> **Аморфні тіла** - це тверді тіла, в яких частинки розташовані хаотично, без чіткої періодичної структури.

В аморфних тілах існує лише **ближній порядок**. Це означає, що кожен атом має певних сусідів, розташованих на певній відстані, але ця структура не повторюється регулярно по всьому об'єму тіла. Уявіть собі натовп людей - кожен стоїть поруч з кількома іншими, але немає чітких рядів чи шаблонів, як у параді. Аморфні тіла за своєю структурою більше нагадують дуже-дуже в'язкі, "застиглі" рідини, де молекули також мають лише ближній порядок.

Під дією тривалих навантажень аморфні тіла можуть повільно "текти", змінюючи свою форму. Ця "текучість" залежить від речовини. Хоча скло і є аморфним тілом, але зафіксувати за нормальних умов (температур), що скло "тече" практично неможливо навіть на масштабі століть або тисячоліть, бо зміни дуже і дуже повільні. Натомість "текучість" смоли зафіксувати простіше.

*Ключова відмінність:* На відміну від кристалів, аморфні тіла не мають фіксованої температури плавлення. При нагріванні вони не плавляться миттєво, а поступово розм'якшуються і переходять у рідкий стан.

*Приклади аморфних тіл:* скло, смола, бурштин, віск, пластилін, більшість пластмас.

### Далекий порядок, ближній порядок та кристалічна ґратка

> **Далекий порядок** - це строга, періодична впорядкованість у розташуванні частинок, що повторюється на значних відстанях, які в тисячі та мільйони разів перевищують розміри самих частинок. Це як цегляна стіна, де кожна цеглина розташована на своєму місці, і ми можемо передбачити розташування цеглин далеко від початкової точки.

> **Ближній порядок** - це впорядкованість у розташуванні частинок лише у межах кількох міжатомних відстаней. Наприклад, кожен атом має певну кількість сусідів, розташованих на певній відстані.

У кристалічних тілах ми спостерігаємо як ближній, так і далекий порядок. Однак у аморфних тілах та рідинах існує тільки ближній порядок, але немає далекого порядку.

Основою порядку в кристалічних тілах є **кристалічна ґратка**.

> **Кристалічна ґратка** - це ідеалізована, абстрактна сукупність точок у просторі (вузлів ґратки), що періодично повторюються. Атоми, молекули або іони, що утворюють кристалічне тіло, розташовуються *у* або *навколо* цих вузлів ґратки.

Ці вузли ґратки не є порожніми точками, а місцями, де з найбільшою ймовірністю перебувають частинки. Частинки в кристалічній ґратці не нерухомі, вони постійно коливаються навколо цих вузлів.

Для опису кристалічної ґратки використовують поняття **елементарної комірки**.

> **Елементарна комірка** - це найменший об'єм кристала, періодичне повторення якого в просторі дозволяє побудувати всю кристалічну ґратку. Це як "базовий будівельний блок" кристала.

Ми не будемо заглиблюватися в усі 14 типів кристалічних ґраток Браве (переглянути їх можна ось тут https://uk.wikipedia.org/wiki/Кристалічна_ґратка і тут https://en.wikipedia.org/wiki/Unit_cell), але розглянемо найпростіші та найпоширеніші кубічні та гексагональні ґратки для кращого розуміння.

*   **Проста кубічна ґратка:** Частинки розташовані лише у вершинах куба. Це найпростіший тип, але зустрічається рідко (наприклад, полоній).
*   **Об'ємноцентрована кубічна ґратка (ОЦК):** Частинки розташовані у вершинах куба та одна частинка у його центрі. Приклади: залізо ($\alpha$-Fe), вольфрам, хром.
*   **Гранецентрована кубічна ґратка (ГЦК):** Частинки розташовані у вершинах куба та по одній частинці в центрі кожної грані. Приклади: мідь, срібло, золото, алюміній.
*   **Гексагональна кристалічна ґратка:** Ця ґратка має шестикутну симетрію. Частинки розташовані у вершинах шестикутних призм, у центрі верхньої та нижньої граней, а також три частинки розташовані всередині призми на середній висоті. Приклади: магній, цинк, титан.

### Монокристали та полікристали

Кристалічні тіла можуть бути двох видів: монокристали та полікристали.

> **Монокристал** - це кристалічне тіло, що складається з однієї суцільної кристалічної ґратки, тобто має єдину, безперервну структуру без розривів чи меж. У монокристалі орієнтація кристалічної ґратки однакова в усьому об'ємі.

Приклади: дорогоцінне каміння (алмаз, рубін, сапфір), кристали кварцу, кремнію, які використовуються в електроніці.

> **Полікристал** - це тверде тіло, що складається з великої кількості дрібних, хаотично орієнтованих кристалів, які називаються **кристалітами** або **кристалічними зернами**. Ці кристаліти зростаються між собою, утворюючи межі зерен. 

Більшість металів та сплавів, які ми використовуємо щодня (наприклад, сталь, мідний дріт), є полікристалами.

### Дефекти Кристалічної Ґратки

Важливо розуміти, що ідеальні кристали існують переважно в теорії. Реальні кристали завжди містять **дефекти кристалічної ґратки**. Це порушення ідеальної періодичності розташування атомів. Дефекти можуть бути точковими (наприклад, відсутність атома у вузлі - **вакансія**, або зайвий атом у міжвузлі), лінійними (**дислокації**) або об'ємними (наприклад, пори). Наявність і тип дефектів значно впливають на механічні, електричні, оптичні та інші властивості матеріалів, визначаючи їхню міцність, пластичність, електропровідність тощо.

## Анізотропія та ізотропія

Ми розглянули будову твердих тіл, тепер давайте подивимося, як ця будова впливає на властивості матеріалів.

> **Анізотропія** - це залежність фізичних властивостей тіла (механічних, теплових, електричних, оптичних) від обраного в ньому напрямку. На противагу цьому, **ізотропія** - це незалежність, однаковість фізичних властивостей в усіх напрямках.

**Чому монокристали анізотропні?** У монокристалі атоми розташовані впорядковано. Відстані між атомами та сили взаємодії між ними можуть бути різними в різних напрямках. Тому, наприклад, розтягувати кристал вздовж однієї осі може бути легше, ніж вздовж іншої, або тепло буде поширюватися швидше в одному напрямку, ніж в іншому.

Наприклад, деревина є анізотропним матеріалом: вона набагато міцніша вздовж волокон, ніж упоперек. Інший приклад - композитні матеріали, де властивості спеціально розробляються бути анізотропними для забезпечення міцності у певних напрямках, що критично важливо в авіабудуванні або космічній техніці.

**Чому аморфні тіла ізотропні?** Через хаотичне розташування частинок. Якщо немає впорядкованості в певному напрямку, то і властивості не можуть від цього напрямку залежати.

**Чому полікристали часто поводяться як ізотропні?** Хоча кожен окремий кристаліт анізотропний, у полікристалі вони орієнтовані випадковим чином. В результаті, усереднені властивості в будь-якому напрямку виявляються однаковими.

## Поліморфізм

> **Поліморфізм** - це здатність речовини існувати в різних кристалічних формах (поліморфними модифікаціями), що відрізняються своєю кристалічною ґраткою і, як наслідок, фізичними властивостями, але мають однаковий хімічний склад. 

Для простих речовин це явище називають **алотропією**. Термін "поліморфізм" іноді в ширшому сенсі використовують і для простих речовин, але "алотропія" є більш специфічним і точним терміном у цьому випадку.

> **Поліморфний перехід** - це перетворення однієї кристалічної модифікації речовини в іншу під впливом зміни температури, тиску або інших зовнішніх умов.

Яскравим прикладом є вуглець:
*   **Графіт** (м'який, чорний, електропровідний) має шарувату гексагональну ґратку. Атоми утворюють шари, зв'язок всередині шару міцний, а між шарами - слабкий. Тому графіт м'який, шари легко зсуваються (олівець залишає слід).
*   **Алмаз** (найтвердіший мінерал, прозорий, діелектрик) має гранецентровану кубічну ґратку. Кожен атом міцно пов'язаний з чотирма іншими, утворюючи надзвичайно міцну тривимірну ґратку. Це робить алмаз найтвердішою природною речовиною.

Це дві поліморфні модифікації однієї й тієї ж речовини - вуглецю. Перехід графіту в алмаз можливий за дуже високих температур і тисків.

Інший приклад - олово. За низьких температур (нижче +13,2 °С) "біле олово" (металеве, пластичне) перетворюється на "сіре олово" (крихкий порошок), що є причиною "олов'яної чуми" - руйнування олов'яних виробів на морозі.

## Ключові Відмінності Кристалічних та Аморфних Тіл

| Ознака               | Кристалічні тіла                                   | Аморфні тіла                                           |
| :------------------- | :------------------------------------------------- | :----------------------------------------------------- |
| **Внутрішня будова** | Далекий та ближній порядок, періодична ґратка      | Лише ближній порядок, хаотичне розташування частинок   |
| **Температура плавлення** | Мають чітку температуру плавлення. При досягненні температури плавлення відбувається різкий перехід речовини з твердого стану в рідкий, оскільки руйнується далекий порядок по всьому об'єму одночасно, що вимагає певної енергії. | Не мають чіткої температури плавлення, розм'якшуються поступово в діапазоні температур. Це відбувається тому, що їм бракує далекого порядку, і для переходу в рідкий стан не потрібно руйнувати впорядковану структуру по всьому об'єму; вони поступово втрачають свою в'язкість. |
| **Анізотропія/Ізотропія** | Монокристали анізотропні, полікристали ізотропні | Ізотропні                                              |
| **Приклади**         | Сіль, кварц, метали                                | Скло, смола, пластмаси, бурштин                        |


## Рідкі кристали

Існує дивовижний проміжний стан речовини, який поєднує властивості як рідин, так і кристалів. Це **рідкі кристали**.

**Рідкі кристали** - це органічні речовини, молекули яких зазвичай мають видовжену або дископодібну форму, і які в певному температурному діапазоні (між температурою плавлення твердого кристала та температурою переходу в ізотропну рідину) виявляють одночасно текучість рідин та оптичну анізотропію кристалів.

У звичайній рідині молекули розташовані хаотично. У кристалічному твердому тілі вони жорстко зафіксовані в ґратці. У рідких кристалах молекули можуть вільно переміщатися, як у рідині, але при цьому зберігають певну орієнтацію одна щодо одної.

Завдяки здатності змінювати орієнтацію молекул під дією електричного поля рідкі кристали стали основою для дисплеїв (LCD) у моніторах, телевізорах та смартфонах.

Залежно від ступеня впорядкованості розрізняють три основні типи рідких кристалів:
*   **Нематичний рідкий кристал:** Молекули (зазвичай видовжені) орієнтовані паралельно одна до одної, але їхні центри мас розташовані хаотично. Уявіть собі коробку сірників, де всі сірники лежать паралельно, але не утворюють чітких шарів. Це найпоширеніший тип, що використовується в LCD-дисплеях (Liquid Crystal Display).
*   **Смектичний рідкий кристал:** Молекули також орієнтовані паралельно одна до одної, але вони додатково утворюють шари. Молекули можуть вільно рухатися в межах свого шару, але не можуть легко переміщатися між шарами. Це як стопка паперу, де кожен аркуш - це шар, а на аркуші волокна орієнтовані в одному напрямку. Смектичні рідкі кристали більш впорядковані, ніж нематики.
*   **Холестеричний рідкий кристал:** Ці рідкі кристали є різновидом нематичних, але їхні молекули розташовані шарами, причому в кожному наступному шарі напрямок орієнтації молекул трохи повернутий відносно попереднього. Це створює спіральну структуру, яка реагує на температуру, змінюючи крок спіралі. Зміна кроку спіралі призводить до зміни довжини хвилі світла, що відбивається від кристала, тому ми спостерігаємо зміну кольору (наприклад, у рідкокристалічних термометрах або індикаторах температури).

## Підсумок

| Термін / Поняття | Визначення та Характеристики |
| :--- | :--- |
| **Основні стани** | |
| **Тверде тіло** | Агрегатний стан речовини, що характеризується стабільністю форми та об'єму. Частинки здійснюють лише коливання навколо положень рівноваги. |
| **Кристалічні тіла** | Тверді тіла, в яких атоми, молекули або іони розташовані у суворо впорядкованому порядку, утворюючи регулярну просторову структуру. |
| **Аморфні тіла** | Тверді тіла, в яких частинки розташовані хаотично, без чіткої періодичної структури. За будовою нагадують дуже в'язкі рідини. |
| **Рідкі кристали** | Проміжний стан речовини, що поєднує текучість рідин та оптичну анізотропію кристалів. |
| **Структурний порядок** | |
| **Далекий порядок** | Строга періодична впорядкованість у розташуванні частинок, що повторюється на великих відстанях. Характерний для кристалічних тіл. |
| **Ближній порядок** | Впорядкованість у розташуванні частинок лише на відстані кількох міжатомних проміжків. Властивий як кристалічним, так і аморфним тілам, а також рідинам. |
| **Кристалічна ґратка** | Ідеалізована сукупність точок (вузлів) у просторі, що періодично повторюються і в яких розташовані частинки кристала. |
| **Елементарна комірка** | Найменший структурний елемент, періодичне повторення якого в просторі утворює всю кристалічну ґратку. |
| **Типи кристалічних тіл** | |
| **Монокристал** | Тіло, що складається з однієї суцільної кристалічної ґратки. Приклади: алмаз, кристали кварцу. |
| **Полікристал** | Тіло, що складається з безлічі дрібних, хаотично орієнтованих кристалів (кристалітів). Приклади: більшість металів, сталь. |
| **Фізичні властивості** | |
| **Ізотропія** | Незалежність, однаковість фізичних властивостей (механічних, оптичних, теплових) в усіх напрямках. Властивість аморфних тіл та полікристалів. |
| **Анізотропія** | Залежність фізичних властивостей тіла від обраного в ньому напрямку. Властивість монокристалів. |
| **Поліморфізм** | Здатність речовини існувати в різних кристалічних формах (поліморфних модифікаціях), що відрізняються кристалічною ґраткою та властивостями. |
| **Алотропія** | Поліморфізм для простих речовин. Приклад: вуглець (графіт та алмаз). |
| **Температура плавлення** | **Кристали:** мають чітку температуру плавлення. **Аморфні тіла:** не мають чіткої температури плавлення, а поступово розм'якшуються в певному діапазоні температур. |
| **Різновиди рідких кристалів** | |
| **Нематичні** | Молекули орієнтовані паралельно, але їхні центри мас розташовані хаотично. |
| **Смектичні** | Молекули орієнтовані паралельно і утворюють шари. Більш впорядковані, ніж нематичні. |
| **Холестеричні** | Різновид нематичних, де напрямок орієнтації молекул у кожному наступному шарі повернутий, створюючи спіральну структуру. Реагують на зміну температури зміною кольору. |
