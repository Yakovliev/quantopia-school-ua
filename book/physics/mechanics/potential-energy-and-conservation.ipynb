# Потенціальна енергія та закон збереження енергії

## Потенціальна енергія

Уявімо собі яблуко, що висить на гілці. Воно нерухоме, тому його кінетична енергія дорівнює нулю. Але якщо гілка зламається, яблуко почне падати, набираючи швидкість, а отже, і кінетичну енергію. Звідки взялася ця енергія? Вона була "накопичена" в яблуці завдяки його положенню відносно Землі. Або візьмемо стрілу, натягнуту на тятиві лука. Стріла нерухома, але варто відпустити тятиву, і стріла полетить з величезною швидкістю. Енергія була накопичена у вигнутій дузі лука.

Ця "накопичена" енергія, яка залежить від взаємного розташування тіл або частин одного тіла, і є потенціальною енергією.

> **Потенціальна енергія $E_p$** - це частина енергії фізичної системи, що виникає завдяки взаємодії між тілами, які складають систему, або внаслідок взаємодії із зовнішніми відносно цієї системи тілами.

Як і будь-який інший вид енергії, в системі СІ потенціальна енергія вимірюється в **Джоулях (Дж)**.

### Консервативні (потенціальні) сили

Поняття потенціальної енергії має сенс лише для певного класу сил, які називаються **консервативними** або **потенціальними**.

> **Консервативні сили** - це сили, робота яких не залежить від траєкторії руху тіла, а визначається лише його початковим і кінцевим положенням.

Ключова властивість таких сил: **робота консервативної сили на будь-якій замкненій траєкторії дорівнює нулю.**

**Приклади консервативних сил:**
*   **Сила тяжіння:** Робота, яку виконує сила тяжіння при переміщенні тіла з точки А в точку Б, не залежить від того, чи кинули ми тіло по параболі, чи спустили по похилій площині. Вона залежить лише від різниці висот початковою точкою та кінцевою точкою.
*   **Сила пружності:** Робота, яку виконує пружина при розтягуванні чи стисканні, залежить лише від початкової та кінцевої деформації, а не від того, як швидко чи повільно ми її розтягували.

**А які сили не є консервативними?**

Найяскравіший приклад - **сила тертя ковзання**. Її робота *дуже* залежить від пройденого шляху. Якщо ми пересунемо ящик з точки А в точку В по прямій, робота сили тертя буде одна. А якщо ми пересунемо його по довгій звивистій траєкторії, то робота сили тертя буде значно більшою. Робота сили тертя по замкненому шляху ніколи не дорівнює нулю - вона завжди від'ємна, оскільки спрямована проти руху тіла. Для таких неконсервативних сил поняття потенціальної енергії не вводять.

## Потенціальна енергія тіла в полі тяжіння Землі

Давайте розглянемо найпростіший випадок: тіло масою $m$ переміщується з початкової точки А, що знаходиться на висоті $h_0, до кінцевої точки В, що знаходиться на висоті $h$ над поверхнею Землі. Ми розглядаємо загальний випадок, тому точка В не обов'язково буде рівно під точкою А, ця точка В може бути трохи в стороні. Таким чином, у нас є безліч різних 

На тіло діє стала сила тяжіння $\vec{F_т}$, напрямлена вертикально вниз. Знайдемо роботу цієї сили.

1.  За визначенням, робота $A = Fs\cos\alpha$.
2.  Модуль сили тяжіння $F_т = mg$.
3.  Модуль переміщення по вертикалі $s = h - h_0$ (якщо тіло підіймається) або $s = h_0 - h$ (якщо опускається).
4.  Якщо тіло підіймається, то вектор сили тяжіння напрямлений протилежно вектору переміщення, тому кут $\alpha = 180^\circ$, а $\cos(180^\circ) = -1$.
    $$
    A_т = F_т \cdot s \cdot \cos(180^\circ) = mg(h-h_0)(-1) = -mgh + mgh_0
    $$
5.  Отже, робота сили тяжіння дорівнює:
    $$
    A_т = mgh_0 - mgh
    $$

Ця формула надзвичайно важлива. Вона показує, що робота сили тяжіння дорівнює різниці деякої величини, що залежить від висоти. Цю величину $mgh$ ми й називаємо **потенціальною енергією тіла, піднятого над Землею**.

$$
E_p = mgh
$$

Тепер ми можемо переписати нашу формулу для роботи через потенціальну енергію: $A_т = E_{p0} - E_p$. Бачимо, що робота сили тяжіння дорівнює різниці потенціальних енергій у початковому та кінцевому станах. Цей результат є чудовою ілюстрацією загального правила, яке ми сформулюємо трохи згодом.

**Важливий момент:** Формула $E_p = mgh$ визначає потенціальну енергію відносно певного **нульового рівня**. Це рівень, на якому ми домовились вважати потенціальну енергію рівною нулю ($h=0$). Ми можемо обрати цей рівень довільно: поверхня столу, підлога кімнати, поверхня Землі. Сама по собі величина потенціальної енергії не має глибокого фізичного змісту, зміст має лише її **зміна** $\Delta E_p$.

## Потенціальна енергія пружно деформованого тіла

Тепер розглянемо іншу консервативну силу — силу пружності. Нехай пружина жорсткістю $k$ розтягується (або стискається) з початкового положення $x_0$ до кінцевого $x$. Сила пружності, що діє з боку пружини, за законом Гука дорівнює $F_{\text{пр}} = -kx$. Знак "мінус" показує, що сила завжди напрямлена в бік, протилежний деформації.

Сила пружності не є сталою, вона залежить від видовження $x$. Роботу такої сили зручно знаходити як площу фігури під графіком залежності $F_{\text{пр}}(x)$.

1.  Графік залежності $F_{\text{пр}}(x)$ від деформації $x$ є прямою лінією.
2.  Робота сили пружності при переміщенні тіла від $x_0$ до $x$ дорівнює площі трапеції під графіком. Проте простіше порахувати цю роботу як різницю робіт, необхідних для розтягування пружини з нуля до $x_0$ і з нуля до $x$.
3.  Робота сили пружності при розтягуванні пружини від 0 до $x$ дорівнює площі трикутника під графіком, але взятої зі знаком "мінус" (бо сила напрямлена протилежно розтягуванню): $A_{0 \to x} = - \frac{1}{2} x \cdot (kx) = -\frac{kx^2}{2}$.
4.  Таким чином, робота при переході від $x_0$ до $x$ дорівнює:
    $$
    A_{\text{пр}} = A_{0 \to x} - A_{0 \to x_0} = \left(-\frac{kx^2}{2}\right) - \left(-\frac{kx_0^2}{2}\right) = \frac{kx_0^2}{2} - \frac{kx^2}{2}
    $$
Ми знову бачимо, що робота консервативної сили пружності дорівнює різниці деякої величини. Цю величину ми називаємо **потенціальною енергією пружно деформованого тіла**.

$$
E_p = \frac{kx^2}{2}
$$

І знову ми можемо переписати формулу для роботи через потенціальну енергію: $A_{\text{пр}} = E_{p0} - E_p$. Ми вдруге переконалися в існуванні загального зв'язку між роботою консервативної сили та зміною потенціальної енергії.

Тут за нульовий рівень потенціальної енергії природно вважати стан, коли пружина не деформована ($x=0$).

## Теорема про потенціальну енергію

Тепер, розглянувши два конкретні приклади, ми можемо сформулювати загальну теорему, яка їх об'єднує.

> **Теорема про потенціальну енергію:** Робота консервативної сили дорівнює зміні потенціальної енергії системи, взятій з протилежним знаком.

$$
A_{\text{конс}} = -(E_p - E_{p0}) = E_{p0} - E_p = -\Delta E_p
$$

де:
*   $A_{\text{конс}}$ — робота консервативної сили.
*   $E_p$ — кінцева потенціальна енергія.
*   $E_{p0}$ — початкова потенціальна енергія.

**Фізичний зміст:** Коли консервативна сила виконує додатну роботу (наприклад, сила тяжіння при падінні тіла), потенціальна енергія системи зменшується. І навпаки, коли консервативна сила виконує від'ємну роботу (ми підіймаємо тіло вгору, а сила тяжіння напрямлена вниз), потенціальна енергія системи зростає. Робота відбувається ніби "за рахунок" запасеної потенціальної енергії.

## Принцип мінімуму потенціальної енергії

Існує цікава закономірність: будь-яка механічна система прагне зайняти положення, в якому її потенціальна енергія буде мінімальною. Це положення є **стійкою рівновагою**.

*   **Аналогія:** Кулька, що лежить на дні сферичної чаші, перебуває в положенні стійкої рівноваги. Її потенціальна енергія мінімальна. Якщо її трохи відхилити, вона повернеться назад.

# Повна механічна енергія та її збереження

Тепер, коли ми знаємо про два види механічної енергії — кінетичну (енергію руху) та потенціальну (енергію взаємодії), — ми можемо ввести поняття повної механічної енергії.

> **Повна механічна енергія $E$** — це сума кінетичної та потенціальної енергій тіла (або системи тіл).
> $$
> E = E_k + E_p
> $$

Ця величина є однією з найважливіших у фізиці, оскільки за певних умов вона залишається незмінною.

## Закон збереження повної механічної енергії

Для того, щоб сформулювати цей закон, нам потрібне поняття **замкненої (або ізольованої) системи тіл**.

> **Замкнена система тіл** — це сукупність тіл, які взаємодіють лише між собою і не взаємодіють з іншими тілами, що не входять до цієї системи.

На практиці це означає, що зовнішні сили або відсутні, або їхня дія скомпенсована.

Тепер ми готові сформулювати один з фундаментальних законів природи.

> **Закон збереження повної механічної енергії:** У замкненій системі тіл, де діють лише консервативні сили (сила тяжіння та сила пружності), повна механічна енергія системи залишається сталою.
> $$
> E = E_k + E_p = \text{const}
> $$

Це означає, що для двох будь-яких моментів часу повна енергія системи однакова:

$$
E_{k0} + E_{p0} = E_k + E_p
$$

### Виведення закону збереження механічної енергії

Цей закон не є постулатом, його можна вивести з уже відомих нам теорем. Давайте зробимо це.

1.  Згідно з **теоремою про кінетичну енергію**, робота всіх сил, що діють на тіло, дорівнює зміні його кінетичної енергії: $A_{\text{всіх сил}} = \Delta E_k = E_k - E_{k0}$.
2.  В нашій системі діють лише консервативні сили, отже $A_{\text{всіх сил}} = A_{\text{конс}}$.
3.  Згідно з **теоремою про потенціальну енергію**, робота консервативних сил дорівнює: $A_{\text{конс}} = -\Delta E_p = E_{p0} - E_p$.
4.  Прирівняємо вирази для роботи з пунктів 2 і 3:
    $$
    \Delta E_k = -\Delta E_p
    $$
    $$
    E_k - E_{k0} = E_{p0} - E_p
    $$
5.  Перегрупуємо доданки: перенесемо всі величини з індексом "0" (початковий стан) в один бік, а без індексу (кінцевий стан) — в інший.
    $$
    E_{k0} + E_{p0} = E_k + E_p
    $$
    Що й треба було довести.

### Перетворення енергії

Закон збереження енергії не означає, що кінетична та потенціальна енергії не змінюються. Навпаки, вони постійно **перетворюються** одна в одну, але їхня сума залишається незмінною.

**Приклад:** Математичний маятник.
*   У крайньому верхньому положенні (точці А) маятник на мить зупиняється. Його швидкість дорівнює нулю, отже $E_k=0$. Висота максимальна, отже потенціальна енергія $E_p$ максимальна. Вся енергія — потенціальна.
*   Коли маятник рухається до нижнього положення (точка В), його висота зменшується, а швидкість зростає. $E_p$ перетворюється в $E_k$.
*   У найнижчій точці (В) висота мінімальна (можна вважати її $h=0$), отже $E_p=0$. Швидкість максимальна, отже $E_k$ максимальна. Вся енергія стала кінетичною.
*   Далі, рухаючись до іншого крайнього положення (точка С), маятник сповільнюється, але підіймається. $E_k$ знову перетворюється в $E_p$.
*   У будь-якій проміжній точці повна енергія $E = E_k + E_p$ буде такою ж, як і в крайніх точках.

## Роль неконсервативних сил: сила тертя

А що станеться, якщо в системі діють неконсервативні сили, наприклад, сила тертя? У такому разі повна механічна енергія **не зберігається**. Вона зменшується.

Робота сили тертя $A_{\text{тертя}}$ завжди від'ємна. Вона "забирає" механічну енергію з системи. У цьому випадку зв'язок між роботою і зміною енергії виглядає так:

> Робота всіх неконсервативних сил (наприклад, сили тертя), що діють у системі, дорівнює зміні повної механічної енергії системи.
> $$
> A_{\text{тертя}} = \Delta E = E - E_0 = (E_k + E_p) - (E_{k0} + E_{p0})
> $$

Оскільки $A_{\text{тертя}} < 0$, то і $\Delta E < 0$, що означає $E < E_0$ — кінцева механічна енергія менша за початкову.

## Загальний закон збереження енергії

Куди ж зникає механічна енергія, коли діє сила тертя? Чи вона просто зникає? Ні.

**Енергія нікуди не зникає й нізвідки не з'являється, вона лише перетворюється з одного виду на інший або переходить від одного тіла до іншого.**

Це — **загальний закон збереження енергії**, один з найголовніших у всій науці.

Коли сила тертя виконує роботу, механічна енергія перетворюється в інший вид енергії — **внутрішню енергію**. Це енергія хаотичного руху та взаємодії молекул і атомів, з яких складаються тіла. Простіше кажучи, тіла нагріваються.

**Приклад:** Коли автомобіль гальмує, його величезна кінетична енергія не зникає. Вона перетворюється на внутрішню енергію гальмівних колодок та дисків, які сильно нагріваються. Так само відбувається при **непружному ударі** (наприклад, коли шматок пластиліну падає на підлогу і прилипає до неї) — кінетична енергія частково переходить у нагрівання та деформацію тіл.

Таким чином, якщо враховувати всі види енергії (механічну, внутрішню, електромагнітну, ядерну тощо), то повна енергія ізольованої системи завжди залишається незмінною. Закон збереження механічної енергії є лише окремим, хоч і дуже важливим, випадком цього всеосяжного закону природи.



### Теорема про потенціальну енергію

Існує прямий зв'язок між роботою консервативної сили та зміною потенціальної енергії.

> **Теорема про потенціальну енергію:** Робота консервативної сили дорівнює зміні потенціальної енергії системи, взятій з протилежним знаком.

$$
A_{\text{конс}} = -(E_p - E_{p0}) = E_{p0} - E_p = -\Delta E_p
$$

де:
*   $A_{\text{конс}}$ - робота консервативної сили.
*   $E_p$ - кінцева потенціальна енергія.
*   $E_{p0}$ - початкова потенціальна енергія.

**Фізичний зміст:** Коли консервативна сила виконує додатну роботу (наприклад, сила тяжіння при падінні тіла), потенціальна енергія системи зменшується. І навпаки, коли консервативна сила виконує від'ємну роботу (ми підіймаємо тіло вгору, а сила тяжіння напрямлена вниз), потенціальна енергія системи зростає. Робота відбувається ніби "за рахунок" запасеної потенціальної енергії.

## Види потенціальної енергії

Давайте розглянемо дві найважливіші для шкільного курсу фізики потенціальні енергії.

### Потенціальна енергія тіла, піднятого над Землею

Розглянемо тіло масою $m$, яке підіймають з висоти $h_0$ на висоту $h$ над поверхнею Землі. Сила тяжіння $F_т = mg$ напрямлена вертикально вниз. Переміщення $s = h - h_0$ напрямлене вертикально вгору. Кут між вектором сили та вектором переміщення становить $180^\circ$.

1.  Знайдемо роботу, яку виконує сила тяжіння:
    $$
    A_т = F_т \cdot s \cdot \cos(180^\circ) = mg(h-h_0)(-1) = -mgh + mgh_0 = mgh_0 - mgh
    $$

2.  Тепер застосуємо теорему про потенціальну енергію:
    $$
    A_т = E_{p0} - E_p
    $$

3.  Порівнюючи два вирази для роботи сили тяжіння, ми можемо зробити висновок:
    $$
    E_p = mgh
    $$

**Важливий момент:** Формула $E_p = mgh$ визначає потенціальну енергію відносно певного **нульового рівня**. Це рівень, на якому ми домовились вважати потенціальну енергію рівною нулю ($h=0$). Ми можемо обрати цей рівень довільно: поверхня столу, підлога кімнати, поверхня Землі. Сама по собі величина потенціальної енергії не має глибокого фізичного змісту, зміст має лише її **зміна** $\Delta E_p$.

### Потенціальна енергія пружно деформованого тіла

Тепер розглянемо пружину жорсткістю $k$. Нехай у недеформованому стані її довжина дорівнює $x_0 = 0$. Щоб розтягнути пружину на величину $x$, ми маємо прикласти зовнішню силу $F_{\text{зовн}}$, яка за модулем дорівнює силі пружності, що виникає в пружині (за законом Гука, $F_{\text{пр}} = kx$), тобто $F_{\text{зовн}}=kx$.

Ця сила не є сталою, вона зростає від 0 до $kx$ по мірі розтягування пружини. Як ми пам'ятаємо з попереднього матеріалу, у такому випадку роботу можна знайти як площу під графіком залежності сили від переміщення.

1.  Графік залежності $F_{\text{зовн}}(x)$ є прямою лінією, що проходить через початок координат.
2.  Робота, яку ми виконуємо, дорівнює площі трикутника під цим графіком:
    $$
    A_{\text{зовн}} = \text{Площа} = \frac{1}{2} \cdot \text{основа} \cdot \text{висота} = \frac{1}{2} \cdot x \cdot F_{\text{зовн}}(x) = \frac{1}{2} \cdot x \cdot (kx) = \frac{kx^2}{2}
    $$
3.  Ця виконана нами робота "запасається" у пружині у вигляді потенціальної енергії. Отже, потенціальна енергія пружно деформованої (розтягнутої або стиснутої) пружини дорівнює:
    $$
    E_p = \frac{kx^2}{2}
    $$

Тут за нульовий рівень потенціальної енергії природно вважати стан, коли пружина не деформована ($x=0$).

### Принцип мінімуму потенціальної енергії

Існує цікава закономірність: будь-яка механічна система прагне зайняти положення, в якому її потенціальна енергія буде мінімальною. Це положення є **стійкою рівновагою**.

*   **Аналогія:** Кулька, що лежить на дні сферичної чаші, перебуває в положенні стійкої рівноваги. Її потенціальна енергія мінімальна. Якщо її трохи відхилити, вона повернеться назад.

# Повна механічна енергія та її збереження

Тепер, коли ми знаємо про два види механічної енергії - кінетичну (енергію руху) та потенціальну (енергію взаємодії), - ми можемо ввести поняття повної механічної енергії.

> **Повна механічна енергія $E$** - це сума кінетичної та потенціальної енергій тіла (або системи тіл).
> $$
> E = E_k + E_p
> $$

Ця величина є однією з найважливіших у фізиці, оскільки за певних умов вона залишається незмінною.

## Закон збереження повної механічної енергії

Для того, щоб сформулювати цей закон, нам потрібне поняття **замкненої (або ізольованої) системи тіл**.

> **Замкнена система тіл** - це сукупність тіл, які взаємодіють лише між собою і не взаємодіють з іншими тілами, що не входять до цієї системи.

На практиці це означає, що зовнішні сили або відсутні, або їхня дія скомпенсована.

Тепер ми готові сформулювати один з фундаментальних законів природи.

> **Закон збереження повної механічної енергії:** У замкненій системі тіл, де діють лише консервативні сили (сила тяжіння та сила пружності), повна механічна енергія системи залишається сталою.
> $$
> E = E_k + E_p = \text{const}
> $$

Це означає, що для двох будь-яких моментів часу повна енергія системи однакова:

$$
E_{k0} + E_{p0} = E_k + E_p
$$

### Виведення закону збереження механічної енергії

Цей закон не є постулатом, його можна вивести з уже відомих нам теорем. Давайте зробимо це.

1.  Згідно з **теоремою про кінетичну енергію**, робота всіх сил, що діють на тіло, дорівнює зміні його кінетичної енергії: $A_{\text{всіх сил}} = \Delta E_k = E_k - E_{k0}$.
2.  В нашій системі діють лише консервативні сили, отже $A_{\text{всіх сил}} = A_{\text{конс}}$.
3.  Згідно з **теоремою про потенціальну енергію**, робота консервативних сил дорівнює: $A_{\text{конс}} = -\Delta E_p = E_{p0} - E_p$.
4.  Прирівняємо вирази для роботи з пунктів 2 і 3:
    $$
    \Delta E_k = -\Delta E_p
    $$
    $$
    E_k - E_{k0} = E_{p0} - E_p
    $$
5.  Перегрупуємо доданки: перенесемо всі величини з індексом "0" (початковий стан) в один бік, а без індексу (кінцевий стан) - в інший.
    $$
    E_{k0} + E_{p0} = E_k + E_p
    $$
    Що й треба було довести.

### Перетворення енергії

Закон збереження енергії не означає, що кінетична та потенціальна енергії не змінюються. Навпаки, вони постійно **перетворюються** одна в одну, але їхня сума залишається незмінною.

**Приклад:** Математичний маятник.
*   У крайньому верхньому положенні (точці А) маятник на мить зупиняється. Його швидкість дорівнює нулю, отже $E_k=0$. Висота максимальна, отже потенціальна енергія $E_p$ максимальна. Вся енергія - потенціальна.
*   Коли маятник рухається до нижнього положення (точка В), його висота зменшується, а швидкість зростає. $E_p$ перетворюється в $E_k$.
*   У найнижчій точці (В) висота мінімальна (можна вважати її $h=0$), отже $E_p=0$. Швидкість максимальна, отже $E_k$ максимальна. Вся енергія стала кінетичною.
*   Далі, рухаючись до іншого крайнього положення (точка С), маятник сповільнюється, але підіймається. $E_k$ знову перетворюється в $E_p$.
*   У будь-якій проміжній точці повна енергія $E = E_k + E_p$ буде такою ж, як і в крайніх точках.

## Роль неконсервативних сил: сила тертя

А що станеться, якщо в системі діють неконсервативні сили, наприклад, сила тертя? У такому разі повна механічна енергія **не зберігається**. Вона зменшується.

Робота сили тертя $A_{\text{тертя}}$ завжди від'ємна. Вона "забирає" механічну енергію з системи. У цьому випадку зв'язок між роботою і зміною енергії виглядає так:

> Робота всіх неконсервативних сил (наприклад, сили тертя), що діють у системі, дорівнює зміні повної механічної енергії системи.
> $$
> A_{\text{тертя}} = \Delta E = E - E_0 = (E_k + E_p) - (E_{k0} + E_{p0})
> $$

Оскільки $A_{\text{тертя}} < 0$, то і $\Delta E < 0$, що означає $E < E_0$ - кінцева механічна енергія менша за початкову.

## Загальний закон збереження енергії

Куди ж зникає механічна енергія, коли діє сила тертя? Чи вона просто зникає? Ні.

**Енергія нікуди не зникає й нізвідки не з'являється, вона лише перетворюється з одного виду на інший або переходить від одного тіла до іншого.**

Це - **загальний закон збереження енергії**, один з найголовніших у всій науці.

Коли сила тертя виконує роботу, механічна енергія перетворюється в інший вид енергії - **внутрішню енергію**. Це енергія хаотичного руху та взаємодії молекул і атомів, з яких складаються тіла. Простіше кажучи, тіла нагріваються.

**Приклад:** Коли автомобіль гальмує, його величезна кінетична енергія не зникає. Вона перетворюється на внутрішню енергію гальмівних колодок та дисків, які сильно нагріваються. Так само відбувається при **непружному ударі** (наприклад, коли шматок пластиліну падає на підлогу і прилипає до неї) - кінетична енергія частково переходить у нагрівання та деформацію тіл.

Таким чином, якщо враховувати всі види енергії (механічну, внутрішню, електромагнітну, ядерну тощо), то повна енергія ізольованої системи завжди залишається незмінною. Закон збереження механічної енергії є лише окремим, хоч і дуже важливим, випадком цього всеосяжного закону природи.
