# Активність Сонця

Сонце — не просто яскрава незмінна куля на небі. Насправді, наша зірка — це динамічний і мінливий об'єкт, який проходить через періоди різної активності. Сонячна активність — це комплекс різноманітних явищ і процесів, що відбуваються на Сонці та пов'язані з його магнітним полем. Ці явища включають сонячні плями, спалахи, викиди корональної маси та інші процеси, які значно впливають на космічну погоду та стан нашої планети.

## Цикли сонячної активності

**11-річний цикл**

Найвідоміший цикл сонячної активності триває приблизно 11 років (фактично від 9 до 14 років). Протягом цього періоду кількість сонячних плям та інших проявів активності зростає від мінімуму до максимуму, а потім знову зменшується.
Під час максимуму сонячної активності:

На Сонці з'являється багато сонячних плям
Відбувається більше сонячних спалахів
Частіше виникають корональні викиди маси
Посилюється сонячний вітер

Під час мінімуму:

Сонячних плям мало або взагалі немає
Спалахи відбуваються рідко
Інші прояви активності також значно зменшуються

**22-річний цикл (цикл Хейла)**

Крім 11-річного циклу, існує також 22-річний магнітний цикл, відомий як цикл Хейла. Протягом 11-річного циклу магнітна полярність сонячних плям залишається незмінною: в одній півкулі Сонця плями мають переважно північну полярність, а в іншій — південну. Але після завершення 11-річного циклу, полярність змінюється на протилежну. Таким чином, повний магнітний цикл, коли полярність повертається до початкового стану, триває приблизно 22 роки.

## Прояви сонячної активності

**Сонячні плями**

Сонячні плями — це темні ділянки на видимій поверхні Сонця (фотосфері), які виглядають темнішими через нижчу температуру порівняно з оточуючими областями. Температура сонячної плями становить близько 4,000°C, тоді як звичайна температура фотосфери — близько 5,500°C.
Плями виникають там, де сильні магнітні поля прориваються через поверхню Сонця і блокують нормальний потік тепла з нижніх шарів. Великі плями можуть досягати розмірів, більших за діаметр Землі!
Цікавий факт: сонячні плями вперше систематично спостерігав Галілео Галілей у 1610 році, що допомогло довести, що Сонце обертається.

**Хромосферні спалахи**

Сонячні спалахи — це потужні вибухи енергії, які відбуваються, коли енергія, накопичена в магнітних полях поблизу сонячних плям, раптово вивільняється. Спалах проявляється як раптове яскраве свічення в певній області Сонця, переважно в хромосфері (шарі атмосфери Сонця над фотосферою).
Спалахи класифікують за їхньою яскравістю в рентгенівському діапазоні:

Клас A: найслабші
Клас B: слабкі
Клас C: помірні
Клас M: сильні
Клас X: найпотужніші (кожен наступний спалах класу X вдвічі потужніший за попередній)

Спалах може тривати від декількох хвилин до кількох годин і вивільняє величезну кількість енергії — еквівалент мільйонів водневих бомб!

**Протуберанці**

Протуберанці — це величезні хмари сонячного газу (плазми), які підтримуються магнітними полями і піднімаються над поверхнею Сонця. Ці структури можуть існувати тижнями або навіть місяцями і досягати висоти в сотні тисяч кілометрів. Коли ми спостерігаємо їх збоку (на краю сонячного диска), вони виглядають як яскраві петлі або арки. Коли ми бачимо їх зверху (на фоні сонячного диска), вони виглядають як темні нитки і називаються волокнами.
Інколи протуберанці можуть раптово вибухати, викидаючи величезні об'єми сонячної речовини в космос, що призводить до корональних викидів маси.

**Корональні викиди маси (КВМ)**

Корональні викиди маси — це величезні хмари намагніченої плазми, які викидаються з сонячної корони в міжпланетний простір. Це одні з найдраматичніших проявів сонячної активності. КВМ можуть містити мільярди тонн матеріалу і рухатися зі швидкістю від 250 до 3000 км/с.
КВМ часто (але не завжди) пов'язані з сонячними спалахами і вибухами протуберанців. Якщо такий викид спрямований у бік Землі, він може досягти нашої планети приблизно за 1-3 дні і викликати сильні геомагнітні бурі.

## Сонячний вітер і геліосфера

**Що таке сонячний вітер?**

Сонячний вітер — це постійний потік заряджених частинок (переважно електронів і протонів), які випромінюються Сонцем у всіх напрямках. Цей потік виникає через високу температуру сонячної корони (понад 1 мільйон градусів), яка дозволяє частинкам "втікати" від гравітаційного притягання Сонця.
Сонячний вітер поділяється на два типи:

Повільний сонячний вітер — зі швидкістю 300-400 км/с
Швидкий сонячний вітер — зі швидкістю 750-800 км/с, який виходить із корональних дір (областей з відкритими магнітними полями)

Геліосфера — це область космічного простору навколо Сонця, на яку впливає сонячний вітер. Вона простягається далеко за орбіту Плутона і утворює своєрідний "бульбашку" в міжзоряному середовищі. Геліосфера захищає Сонячну систему від галактичних космічних променів — високоенергетичних частинок, що надходять із глибокого космосу.

## Вплив сонячної активності на Землю

**Магнітосфера Землі**

Магнітосфера Землі — це область навколо нашої планети, де домінує земне магнітне поле. Вона діє як щит, відхиляючи більшість заряджених частинок сонячного вітру. Без магнітосфери сонячний вітер міг би поступово "здути" атмосферу Землі, як це сталося з Марсом.
Коли сонячний вітер стикається з магнітосферою, він деформує її, стискаючи з боку Сонця (до відстані близько 10 земних радіусів) і розтягуючи з протилежного боку у довгий "хвіст" (який може простягатись на сотні земних радіусів).

**Магнітні бурі**

Магнітна буря — це тимчасове збурення магнітосфери Землі, викликане сонячним вітром або корональними викидами маси. Під час магнітної бурі частинки сонячного вітру "притискають" магнітосферу і частково проникають усередину, викликаючи значні коливання магнітного поля Землі.
Магнітні бурі класифікують за шкалою G від G1 (слабкі) до G5 (екстремальні). Сильні магнітні бурі можуть:

Пошкоджувати супутники
Викликати збої в електромережах
Порушувати радіозв'язок
Впливати на навігаційні системи
Створювати ризики для астронавтів

**Полярні сяйва**

Одним із найкрасивіших проявів сонячної активності на Землі є полярні сяйва (аврори). Вони виникають, коли заряджені частинки сонячного вітру, захоплені магнітним полем Землі, стикаються з атомами і молекулами у верхніх шарах атмосфери (зазвичай на висоті 80-300 км).
Під час цих зіткнень атоми атмосфери збуджуються і потім випромінюють світло різних кольорів:

Зелений (найпоширеніший) — від атомів кисню на висоті близько 100-150 км
Червоний — від кисню на більших висотах (200-300 км)
Фіолетовий і блакитний — від азоту

Зазвичай полярні сяйва видно лише поблизу магнітних полюсів Землі (Аврора Бореаліс на півночі та Аврора Австраліс на півдні), але під час сильних геомагнітних бур вони можуть бути видимі на набагато нижчих широтах.



## Космічна погода

Термін "космічна погода" описує зміни в навколоземному космічному середовищі, викликані сонячною активністю. Подібно до земної погоди, космічна погода може бути "спокійною" або "бурхливою".
Для моніторингу космічної погоди використовуються:

Сонячні обсерваторії на Землі
Космічні апарати, такі як Solar Dynamics Observatory (SDO) та Solar and Heliospheric Observatory (SOHO)
Супутники, що відстежують сонячний вітер (наприклад, ACE і DSCOVR)

Прогнозування космічної погоди важливе для:

Захисту супутників
Безпеки астронавтів
Планування радіозв'язку
Управління електромережами
Навігації літаків, особливо на полярних маршрутах

## Механізми сонячної активності: сонячне динамо

Основним механізмом, що викликає сонячну активність, вважається "сонячне динамо" — процес генерації і підсилення магнітних полів всередині Сонця. Він працює подібно до електричного генератора (динамо-машини), перетворюючи механічну енергію руху плазми в магнітну енергію.
Ключові елементи сонячного динамо:

Диференційне обертання: різні частини Сонця обертаються з різними швидкостями. Екваторіальні області здійснюють один оберт приблизно за 25 днів, а полярні — за 35 днів.
Конвекція: рух гарячої плазми від центру до поверхні і охолодженої плазми назад до центру.
Меридіональна циркуляція: повільний рух плазми від екватора до полюсів на поверхні і назад до екватора в глибинах Сонця.

Ці рухи, взаємодіючи з іонізованою плазмою, генерують електричні струми, які, в свою чергу, створюють магнітні поля. З часом ці поля підсилюються, скручуються і переплітаються, накопичуючи енергію, яка потім вивільняється у формі різних проявів сонячної активності.

## Історичні спостереження сонячної активності

Систематичні спостереження за сонячними плямами почалися в XVII столітті після винаходу телескопа. Особливо важливими були спостереження Семюеля Генріха Швабе, який у 1843 році виявив регулярний цикл появи сонячних плям, і Рудольфа Вольфа, який почав вести індекс сонячних плям, відомий зараз як число Вольфа.
Цікаві історичні періоди:

Мінімум Маундера (1645-1715): період надзвичайно низької сонячної активності, коли сонячні плями майже не спостерігалися. Цей період збігся з найхолоднішою фазою "малого льодовикового періоду" в Європі.
Мінімум Дальтона (1790-1830): ще один період зниженої сонячної активності, хоч і не такий глибокий, як мінімум Маундера.
Сонячна супер-буря 1859 року (подія Керрінгтона): найпотужніша зареєстрована геомагнітна буря, викликана величезним сонячним спалахом. Полярні сяйва спостерігалися навіть поблизу екватора, а телеграфні системи вийшли з ладу, деякі навіть викликали пожежі.

## Практичне значення: вплив на технології

Вплив на супутники
Сонячна активність може пошкоджувати супутники кількома способами:

Зарядження частинками може викликати електростатичні розряди
Нагрівання атмосфери під час сонячної активності збільшує опір для супутників на низьких орбітах
Радіація може пошкодити електроніку

Вплив на електромережі
Геомагнітні бурі індукують електричні струми в довгих провідниках (лініях електропередач), що може викликати перевантаження трансформаторів і навіть повне відключення електроенергії. Найвідоміший такий випадок стався в 1989 році в Квебеку (Канада), коли 6 мільйонів людей залишилися без електрики на 9 годин.
Вплив на радіозв'язок
Сонячні спалахи можуть порушувати радіозв'язок, особливо на високих частотах. Рентгенівське та ультрафіолетове випромінювання від спалахів іонізує верхні шари атмосфери, що впливає на поширення радіохвиль.
Радіаційна небезпека для астронавтів і екіпажів літаків
Сонячні частинки можуть бути небезпечними для астронавтів за межами захисної магнітосфери Землі, наприклад, під час місій на Місяць або Марс. Навіть екіпажі та пасажири літаків, особливо на полярних маршрутах, отримують підвищену дозу радіації під час сильних сонячних подій.

Сонце – не статичний об'єкт. Його активність змінюється з часом, найвідомішим проявом чого є 11-річний сонячний цикл (іноді 22-річний, якщо враховувати зміну магнітної полярності).


Сонячний Цикл: Протягом приблизно 11 років кількість сонячних плям, спалахів, протуберанців та інших проявів активності зростає до максимуму, а потім спадає до мінімуму. Цей цикл пов'язаний зі складними процесами генерації та переорганізації магнітного поля Сонця всередині конвективної зони, що посилюється диференціальним обертанням.

Космічна Погода: Сонячна активність (спалахи, КВМ, сонячний вітер) безпосередньо впливає на Землю та космічний простір навколо неї. Це явище називають космічною погодою. Потужні викиди можуть викликати:
Полярні сяйва (Аврори).
Порушення радіозв'язку та роботи супутників GPS.
Збої в роботі енергосистем на Землі.
Підвищення рівня радіації для космонавтів та пасажирів літаків на великих висотах.

Феномен сонячної активності характеризується варіаціями кількості плям на поверхні Сонця та сумарною площею, яку вони займають. Наукові спостереження демонструють, що внутрішня частина плями відзначається доволі високою температурою, яка досягає показника 4500 К. Водночас, на тлі значно гарячішої фотосфери з температурним режимом 5780 К, ці утворення візуально сприймаються як темні області. Виникає логічне запитання: який механізм спричиняє зниження температурних показників усередині плями? Враховуючи, що сонячні плями здатні зберігати свою структуру впродовж декількох місяців, науковцями була висунута теорія про існування певного процесу, що перешкоджає конвективному руху плазми в межах плями та забезпечує підтримку температурного градієнту. Сучасні дослідження підтверджують, що роль такого "ізолятора" виконує потужне магнітне поле, яке, взаємодіючи з електрично зарядженими компонентами плазми, пригнічує конвекційні потоки в області плями.

Додатковою загадкою сонячної активності залишається її циклічність - послідовність змін кількісних показників плям, що повторюється з періодичністю приблизно 11 років.

Структурно сонячні плями об'єднані між собою магнітними силовими лініями, подібно до полюсів у магнітних системах, причому кожне таке утворення характеризується власною полярністю. Аналогічно до неможливості фізичного розділення північного та південного полюсів магніту, сонячні плями функціонують виключно як парні структури з протилежними магнітними полярностями. При врахуванні полярності плям, повний цикл сонячної активності становить близько 22 років.

## Вплив сонячної активності на Землю

Астрономічні дослідження Сонця, проведені за допомогою супутникових систем та автоматичних міжпланетних станцій, виявили інтенсивне корпускулярне випромінювання - потік фундаментальних частинок (протонів, нейтронів, електронів). Зокрема, хромосферні спалахи, що виникають у безпосередній близькості до плям, супроводжуються вивільненням колосального обсягу енергії, порівнянного з випромінюванням цілісної фотосфери Сонця. Важливо диференціювати спалахи від протуберанців. **Протуберанці** є постійними елементами сонячної активності - це компактні холодні водневі хмаровидні утворення, що піднімаються до корони та переміщуються вздовж магнітних силових ліній. Протуберанці забезпечують циркуляцію речовини між хромосферним шаром та короною.

**Хромосферний спалах** представляє собою короткочасне суттєве підвищення світлової інтенсивності локалізованої ділянки хромосфери Сонця, що супроводжується вибуховим вивільненням речовини та енергії, акумульованої в магнітному полі сонячних плям.

Спалахові явища формуються в просторі між двома плямами з протилежною полярністю, коли протягом кількох годин відбувається стрімке підвищення температури в цій зоні до рівня $5 \cdot 10^6$ К, що супроводжується генерацією енергії порядку $10^21-10^25$ Дж - показника, майже еквівалентного світимості Сонця у видимому діапазоні спектра. В процесі спалаху енергія випромінюється переважно в невидимих спектральних діапазонах (радіо-, ультрафіолетовому та рентгенівському). Одночасно з цим у міжпланетний простір відбувається викид потоків заряджених частинок, що рухаються зі швидкістю до 20 000 км/с. Після кількагодинного польоту такі корпускулярні потоки можуть досягати земної атмосфери, викликаючи збурення магнітного поля планети та іонізацію верхніх шарів атмосфери, що візуалізується у формі яскравих полярних сяйв.

**Магнітна буря** характеризується збуренням магнітного поля Землі внаслідок сонячного спалаху. Такі періоди супроводжуються неполадками радіокомунікаційних систем та електронного обладнання, а також погіршенням фізіологічного стану людей.

**Увага!** При проведенні спостережень Сонця категорично забороняється безпосередній візуальний контакт із сонячним диском як неозброєним оком, так і через телескопічні прилади без застосування спеціалізованих світлофільтрів!

Сонячні плями
Сонячні плями — це темні області на фотосфері Сонця з сильним магнітним полем. Напруженість магнітного поля в плямах у тисячі разів перевищує напруженість поля в спокійних областях фотосфери. Через сильне магнітне поле конвекція в цих областях пригнічується, що призводить до зниження температури (приблизно 4 000 К порівняно з 5 700 К у навколишній фотосфері) і, відповідно, до зменшення яскравості.


Сонячні плями зазвичай з'являються групами і можуть існувати від кількох днів до кількох місяців. Кількість і площа плям змінюються з циклом приблизно в 11 років.

Сонячні спалахи і корональні викиди маси
Сонячні спалахи — це потужні вибухи на поверхні Сонця, під час яких вивільняється енергія, еквівалентна мільярдам мегатонн тротилу. Спалахи тривають від кількох хвилин до кількох годин і супроводжуються викидом високоенергетичного випромінювання (від радіохвиль до гамма-променів) та прискоренням заряджених частинок.
Корональні викиди маси (КВМ) — це масштабні викиди плазми і магнітного поля з сонячної корони в міжпланетний простір. КВМ можуть виносити мільярди тонн сонячної речовини зі швидкістю до 3 000 км/с. Якщо КВМ спрямований у бік Землі, він може спричинити геомагнітні бурі, полярні сяйва та різноманітні порушення в роботі технічних систем.

Сонячний вітер
Сонячний вітер — це потік заряджених частинок (переважно протонів та електронів), що постійно витікає з сонячної корони. Швидкість сонячного вітру варіюється від 300 до 800 км/с залежно від активності Сонця та області, з якої він витікає:

Повільний сонячний вітер (300-400 км/с) походить з областей із замкнутими магнітними петлями
Швидкий сонячний вітер (600-800 км/с) витікає з корональних дір — областей з відкритими магнітними лініями

Сонячний вітер формує геліосферу і взаємодіє з магнітними полями планет, створюючи магнітосфери. Взаємодія сонячного вітру з магнітосферою Землі призводить до різноманітних геофізичних явищ, включаючи полярні сяйва та геомагнітні бурі.

Механізми сонячного динамо — як саме генерується магнітне поле Сонця і що визначає 11-річний цикл сонячної активності?
