# Задачі. Кінематика

## Рівномірний рух

**Задача 1**: Дрон-кур'єр

Квадрокоптер вилетів з бази, доставив пакунок на відстань 1.2 км по прямій і повернувся назад на базу. Визначте шлях, пройдений дроном, та модуль його переміщення за весь час польоту.

**Задача 2**

Автомобіль проїхав по прямій дорозі 3 км на схід, а потім повернув і проїхав 4 км на північ. Визначити пройдену відстань (шлях) та модуль переміщення.

**Задача 3**

Перевести км/год в м/с або навпаки:
* 10 м/с
* 30 м/с
* 36 км/год
* 72 км/год
* 18 км/год
* 54 км/год
* перша космічна швидкість: 28 440 км/год

**Задача 4**: Квадрокоптер і доставка

Квадрокоптер служби доставки вилетів зі складу, пролетів по прямій 800 метрів до будинку клієнта, передав посилку і тим самим маршрутом повернувся назад на склад. Весь політ тривав 4 хвилини. Визначте:
1. Шлях, який подолав квадрокоптер.
2. Модуль його переміщення за весь час польоту.
3. Середню швидкість польоту (на основі шляху).

**Задача 5**: Автопілот у Tesla

Автомобіль Tesla, рухаючись на автопілоті по прямому відрізку шосе зі сталою швидкістю 90 км/год, перебував у точці з координатою $x_0 = 10$ км (відносно початку відліку - міста). Якою буде його координата через 2 години руху в тому ж напрямку? Запишіть рівняння руху автомобіля.

**Задача 6**: Рух на травелаторі в аеропорту

Пасажир іде по рухомій доріжці (травелаторі) в аеропорту. Його швидкість відносно доріжки становить 1 м/с, а швидкість самої доріжки відносно підлоги термінала - 1.5 м/с. Який шлях подолає пасажир відносно підлоги термінала за 2 хвилини, якщо він рухається:
1. У напрямку руху доріжки?
2. Проти напрямку руху доріжки?

**Задача 7**: Хто швидший: дрон чи велосипедист?

Спортивний дрон летить зі швидкістю 20 м/с. Професійний велогонщик їде зі швидкістю 75 км/год. Порівняйте їхні швидкості, попередньо перевівши одну з них в одиниці іншої. Хто з них рухається швидше?

**Задача 8**: GPS-трекер кур'єра

Кур'єр на велосипеді виїхав з офісу і, згідно з даними GPS-трекера, проїхав 400 метрів на схід, а потім повернув і проїхав 300 метрів на північ до першого пункту доставки.
1. Зобразіть на малюнку траєкторію руху та вектор переміщення.
2. Обчисліть шлях, який подолав кур'єр.
3. Обчисліть модуль переміщення кур'єра (тобто, на якій відстані по прямій від офісу він опинився).

**Задача 9**: Зустріч роботів на складі Amazon

Два автоматизовані роботи-візки рухаються назустріч один одному по прямому довгому коридору складу довжиною 200 м. Перший робот стартує з нульової позначки зі швидкістю 2 м/с. Другий робот стартує одночасно з ним з позначки 200 м зі швидкістю 3 м/с.
1. Напишіть рівняння руху для кожного робота (взявши за напрямок осі ОХ рух першого робота).
2. Знайдіть час, через який роботи зустрінуться.
3. Знайдіть координату їхньої зустрічі.

**Задача 10**: Маневри міського дрона

Дрон-розвідник для моніторингу трафіку злетів з даху хмарочоса, пролетів 600 м на схід, потім 400 м на північ, після чого розвернувся і пролетів 300 м на захід, де й завис над перехрестям.
1. Який загальний шлях подолав дрон?
2. Знайдіть модуль кінцевого переміщення дрона відносно точки старту.

**Задача 11**

Людина їде на електросамокаті до університету. Відстань у 1.8 км вона долає за 5 хвилин. Вважаючи рух рівномірним, визначте швидкість самоката в м/с.

**Задача 12**

Автомобіль, рухаючись рівномірно і прямолінійно по шосе, знаходився в точці з координатою $x_0 = 15$ км. Через 10 хвилин його координата стала x = 35 км. Напишіть рівняння руху автомобіля $x(t)$ і визначте, де він буде через пів години після початку спостереження. (Вісь ОХ спрямована вздовж шосе).

**Задача 13**: Рух на ескалаторі

Людина йде по ескалатору в торговому центрі. Швидкість руху ескалатора відносно підлоги становить 0.5 м/с, а швидкість людини відносно ескалатора - 1 м/с. Знайдіть швидкість людини відносно підлоги у двох випадках:
1. Людина йде у напрямку руху ескалатора.
2. Людина йде проти руху ескалатора.

**Задача 14**: Фітнес-трекер і пробіжка

Спортсмен пробіг 5 кіл на стандартному 400-метровому стадіоні за 10 хвилин, повернувшись у точку старту. Його фітнес-трекер зафіксував усі дані. Визначте:
1. Пройдений шлях спортсмена.
2. Його переміщення за весь час бігу.
3. Середню шляхову швидкість спортсмена в м/с.
4. Середню векторну швидкість (швидкість переміщення).

**Задача 15**: Доставка за допомогою дрона

Дрон-кур'єр доставляє посилку. Перші 2 км шляху він пролетів зі швидкістю 15 м/с, після чого через сильний вітер знизив швидкість до 10 м/с і дрон пролетів ще 1 км. Знайдіть середню шляхову швидкість дрона на всьому шляху.

**Задача 16**: Зустріч роботів-кур'єрів

З двох складів, відстань між якими 600 м, одночасно назустріч один одному виїхали два роботи-кур'єри. Рівняння руху першого робота $x_1(t) = 1.5t$, а другого - $x_2(t) = 600 - 2.5t$. (Вісь ОХ спрямована від першого складу до другого).
1. Опишіть рух кожного робота (початкова координата, швидкість, напрямок).
2. Знайдіть час і місце їхньої зустрічі.
3. Побудуйте графіки руху $x(t)$ для обох роботів в одній системі координат.

**Задача 17**: Обгін на автостраді

Два автомобілі їдуть по прямій автостраді в одному напрямку. Швидкість першого автомобіля 90 км/год, а другого - 108 км/год. Яка відносна швидкість другого автомобіля відносно першого? Якщо в початковий момент часу відстань між ними була 200 м, через який час другий автомобіль наздожене перший?

**Задача 18**: Марсохід "Perseverance"

Марсохід рухається по поверхні Марса. Рівняння його руху вздовж осі ОХ має вигляд $x(t) = 10 - 0.2t$, а вздовж осі OY - $y(t) = -5 + 0.15t$ (усі величини в СІ).
1. Визначте початкові координати марсохода.
2. Знайдіть проекції його швидкості $v_x$ та $v_y$.
3. Знайдіть модуль його миттєвої швидкості (підказка: використовуйте теорему Піфагора для проекцій швидкості).
4. Напишіть рівняння траєкторії руху $y(x)$.

## Рівноприскорений прямолінійний рух

**Задача 1**

Електросамокат рушає з місця і, рухаючись прямолінійно, за 4 секунди розвиває швидкість 7 м/с. З яким прискоренням рухався електросамокат?

**Задача 2**

Відео-блогер запускає свій квадрокоптер вертикально вгору зі стану спокою. Прискорення дрона становить 2 $\text{м}/\text{с}^2$. Яку висоту набере дрон за перші 5 секунд польоту?

**Задача 3**

Під час тест-драйву електромобіль розганяється зі стану спокою з постійним прискоренням 4 $\text{м}/\text{с}^2$. Яку швидкість матиме автомобіль, проїхавши перші 50 метрів?

**Задача 4**

Квадрокоптер, що доставляє посилку, летів горизонтально зі швидкістю 10 м/с. Оператор дав команду на прискорення, і дрон почав рухатись із прискоренням 0.5 $\text{м}/\text{с}^2$. Якою буде швидкість дрона через 10 секунд після початку прискорення?

**Задача 5**
Система екстреного гальмування безпілотного автомобіля, що рухався зі швидкістю 72 км/год, спрацювала, і автомобіль зупинився, пройшовши шлях 40 метрів. Визначте прискорення автомобіля під час гальмування та час, який для цього знадобився.

**Задача 6**
Космічний апарат, що рухається у відкритому космосі зі швидкістю 200 м/с, вмикає на 1 хвилину маневрові двигуни, які створюють прискорення 0.2 $\text{м}/\text{с}^2$ у напрямку руху. Який шлях подолає апарат за час роботи двигунів?

**Задача 7**

Швидкісний ліфт у хмарочосі починає рух з першого поверху і перші 3 секунди рухається з прискоренням 2 $\text{м}/\text{с}^2$, після чого їде з постійною швидкістю. На якій висоті буде ліфт через 3 секунди після старту, і якої швидкості він досягне? На якій висоті буде ліфт через 5 секунд після старту.

**Задача 8**
Зі складу, розташованого на прямому шосе, виїжджає перший робот-кур'єр зі швидкістю 3 м/с. Через 10 секунд слідом за ним зі складу виїжджає другий робот, який рухається з прискоренням 0.4 $\text{м}/\text{с}^2$. Через який час і на якій відстані від складу другий робот наздожене першого?

**Задача 9**

Посадковий модуль входить в атмосферу Марса і починає гальмувати. На останній ділянці гальмування довжиною 120 м його швидкість зменшилася з 50 м/с до 10 м/с. Вважаючи рух на цій ділянці рівносповільненим, знайдіть прискорення модуля та час, за який він пройшов цю ділянку.

**Задача 10**

Автомобіль, оснащений антиблокувальною системою (ABS), рухається по прямій дорозі зі швидкістю 72 км/год. Водій помічає перешкоду і починає екстрене гальмування. Гальмівний шлях автомобіля склав 20 метрів.
1. З яким прискоренням (за модулем) гальмував автомобіль?
2. Скільки часу тривало гальмування?

**Задача 11**

Зі складу, що знаходиться в точці з координатою x=0, виїжджає робот-кур'єр і рухається рівноприскорено з прискоренням 0.4 $\text{м}/\text{с}^2$. Одночасно з ним із пункту видачі з координатою $x=150$ м назустріч виїжджає інший робот-кур'єр, який рухається рівномірно зі швидкістю 5 м/с.
1. Запишіть рівняння руху $x(t)$ для обох роботів.
2. Знайдіть час та координату їхньої зустрічі.

**Задача 12**

Безпілотний автомобіль рухається зі швидкістю 20 м/с. Бортовий комп'ютер виявляє перешкоду на відстані 55 м і через 0.5 секунди (час на обробку даних) активує гальмівну систему, яка забезпечує прискорення -5 $\text{м}/\text{с}^2$.
Запитання: Чи встигне автомобіль зупинитися перед перешкодою? Підтвердьте розрахунками. (Підказка: загальний гальмівний шлях складається зі шляху, пройденого за час реакції, та шляху, пройденого під час гальмування).

**Задача 13**

Смартфон випадково випадає з балкона 10-го поверху, що знаходиться на висоті 30 метрів. Скільки часу триватиме падіння і якою буде швидкість смартфона в момент удару об землю? Опором повітря знехтувати, прискорення вільного падіння вважати рівним 9.8 $\text{м}/\text{с}^2$.
