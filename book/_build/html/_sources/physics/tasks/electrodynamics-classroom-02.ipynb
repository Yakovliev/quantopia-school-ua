# Електродинаміка. Задачі 2

## Електрорушійна сила та закон Ома для повного кола

**Задача 1**

Ліхтарик живиться від двох послідовно з'єднаних батарейок типу АА. Електрорушійна сила (ЕРС) кожної батарейки становить 1.5 В, а внутрішній опір - 0.25 Ом. Опір лампочки розжарення ліхтарика дорівнює 5.5 Ом. Знайдіть силу струму, що протікає через лампочку.

**Задача 2**

Портативний зарядний пристрій (power bank) має ЕРС 5 В і внутрішній опір 0.2 Ом. Під час зарядки смартфона він забезпечує струм силою 2 А.
1.  Який загальний опір (зовнішній + внутрішній) у цьому колі?
2.  Який ефективний опір самого смартфона в процесі зарядки?

**Задача 3**

Автомобільний акумулятор з ЕРС 12.6 В має дуже малий внутрішній опір, що дорівнює 0.02 Ом. Визначте струм короткого замикання, який виникне, якщо випадково з'єднати його клеми товстим мідним дротом (опором дроту знехтувати). Поясніть, чому така ситуація є вкрай небезпечною.

**Задача 4**

Акумулятор дрона виконав роботу 3600 Дж для переміщення заряду величиною 200 Кл через повне електричне коло. Чому дорівнює електрорушійна сила цього акумулятора?

**Задача 5**

Для живлення базової станції системи супутникового інтернету Starlink використовують сонячну панель з ЕРС 24 В та внутрішнім опором 1.5 Ом. Станція споживає струм 2 А.
1.  Яка напруга на клемах сонячної панелі під час роботи станції?
2.  Яка потужність втрачається на внутрішньому опорі панелі, перетворюючись на тепло?

**Задача 6**

Власник будинку на колесах хоче забезпечити автономне освітлення за допомогою двох однакових сонячних панелей. Кожна панель має ЕРС 12 В і внутрішній опір 2 Ом. Система освітлення має загальний опір 5 Ом. Розгляньте два варіанти підключення панелей:
1.  Який струм буде протікати через систему освітлення, якщо з'єднати панелі послідовно?
2.  Який струм буде протікати, якщо з'єднати їх паралельно?
3.  Яке з'єднання є ефективнішим для даного навантаження?

**Задача 7**

Нова батарейка для фотоапарата має ЕРС 3 В і внутрішній опір 0.3 Ом. Після тривалого використання її внутрішній опір зріс до 2.5 Ом, а ЕРС практично не змінилася. Фотоапарат для роботи потребує струм не менше 0.8 А.
1.  Який максимальний струм може віддати нова батарейка при короткому замиканні?
2.  Чи зможе фотоапарат працювати зі старою батарейкою, якщо його власний опір становить 3 Ом? Відповідь обґрунтуйте розрахунками.

**Задача 8**

Електрична схема квадрокоптера захищена запобіжником. Акумуляторна збірка має ЕРС 14.8 В та внутрішній опір 0.05 Ом. У номінальному режимі польоту двигуни створюють навантаження з еквівалентним опором 1.2 Ом.
1.  Розрахуйте силу струму в номінальному режимі.
2.  Визначте струм короткого замикання, що може виникнути при пошкодженні ізоляції.
3.  Із доступних запобіжників (5 А, 10 А, 15 А, 30 А) оберіть мінімальний номінал, який не перегорить під час нормальної роботи, але захистить схему при короткому замиканні.

**Задача 9**

Для живлення наукового обладнання в експедиції потрібне джерело з напругою не менше 8 В при силі струму 2 А. У розпорядженні є багато однакових елементів живлення з ЕРС 1.5 В і внутрішнім опором 0.2 Ом. Опір обладнання становить 4 Ом.
1.  Скільки елементів потрібно з'єднати послідовно, щоб отримати необхідний струм?
2.  Чи буде при цьому напруга на обладнанні відповідати вимогам (не менше 8 В)?
3.  Запропонуйте комбіновану схему з'єднання (кілька паралельних гілок, кожна з яких складається з послідовно з'єднаних елементів), яка дозволить виконати обидві умови.

**Задача 10**

Розглядається система живлення електромобіля. Батарея складається з великої кількості елементів, її сумарна ЕРС становить 400 В, а сумарний внутрішній опір - 0.25 Ом.
1.  Знайдіть опір зовнішнього кола (двигун та трансмісія), при якому корисна потужність, що виділяється на ньому, буде максимальною.
2.  Обчисліть цю максимальну корисну потужність.
3.  Яким буде коефіцієнт корисної дії (ККД) батареї в цьому режимі? ККД визначається як відношення корисної потужності до повної потужності, що розвивається джерелом.
