# Електродинаміка. Задачі 1

## Електричний заряд. Закон Кулона

**Задача 1**

Уявіть, що ви працюєте в лабораторії, яка досліджує властивості нових матеріалів для акумуляторів. Ви маєте дві однакові маленькі металеві кульки. Після контакту з катодом та анодом одна кулька набула заряду +32 нКл, а інша -14 нКл. Згодом ці кульки випадково торкнулися одна одної. Яким стане заряд кожної кульки після дотику?

**Задача 2**

Під час роботи лазерного принтера крихітні частинки тонера (порошкової фарби) отримують електричний заряд, щоб потім притягнутися до паперу в потрібних місцях. Припустимо, що дві однакові частинки тонера, що знаходяться у вакуумі на відстані 1 мм одна від одної, мають однакові негативні заряди. Сила їх взаємного відштовхування дорівнює $8.1 \cdot 10^{-10}$ Н.
1.  Обчисліть заряд кожної частинки тонера.
2.  Скільком елементарним зарядам відповідає заряд однієї частинки?

**Задача 3**

Два квадрокоптери під час нічного шоу з дронами зблизились на мінімальну відстань у 5 метрів. Корпус одного з них внаслідок тертя об повітря набув заряду +2 мкКл, а іншого - заряду -3 мкКл. З якою силою вони взаємодіяли в цей момент? Ця сила була силою притягання чи відштовхування?

**Задача 4**

У сучасних смартфонах використовуються сенсорні екрани, робота яких базується на зміні електричного поля. Уявіть, що два точкові заряди, які моделюють взаємодію пальця з екраном, знаходяться на відстані 3 см один від одного у вакуумі. Заряд першого становить +4·10⁻⁸ Кл, а другого +1·10⁻⁸ Кл. Знайдіть силу, з якою перший заряд діє на другий.

**Задача 5**

В одній з моделей іонного двигуна для космічних апаратів два іони аргону, що втратили по одному електрону, рухаються паралельно один одному. Знайдіть силу електричного відштовхування між цими іонами, коли відстань між ними становить 10 нм. Вважайте, що іони є точковими зарядами.

**Задача 6**

Електростатичні фільтри використовуються для очищення повітря від пилу та алергенів. У такому фільтрі частинка пилу масою 10⁻⁹ г і зарядом +1000e потрапляє в область між двома точковими зарядами. Перший заряд, +1.6 нКл, знаходиться ліворуч від частинки на відстані 2 см. Другий заряд, -1.6 нКл, знаходиться праворуч від частинки на відстані 4 см. Знайдіть модуль і напрямок результуючої сили, що діє на частинку пилу.

**Задача 7**

У дослідженнях з нанотехнологій маніпулюють окремими атомами за допомогою електричного поля. Два точкові заряди, $q_1 = +4e$ та $q_2 = +9e$, закріплені на відстані 20 см один від одного. Де між ними потрібно розмістити третій точковий заряд $q_3$, щоб він перебував у рівновазі? Чи залежить положення цього заряду від його знаку та величини?

**Задача 8**

Під час електростатичного фарбування автомобіля краплі фарби заряджають, щоб вони рівномірно осідали на металевій поверхні кузова. Розглянемо дві краплі фарби, що мають однакові заряди. Коли вони знаходяться на відстані 2 см одна від одної, сила відштовхування між ними становить $F$. Якою має стати відстань між краплями, щоб сила взаємодії зменшилася в 16 разів?

**Задача 9**

В основі деяких медичних діагностичних методів лежить взаємодія заряджених біологічних молекул. Уявіть, що три точкові заряди розташовані у вершинах рівностороннього трикутника зі стороною 10 см. Заряди $q_A = +1$ мкКл і $q_B = +1$ мкКл знаходяться у вершинах A і B відповідно, а заряд $q_C = -1$ мкКл - у вершині C. Знайдіть модуль і напрямок сили, що діє на заряд у вершині C.

**Задача 10**

Сонячне вітрило - це технологія для руху космічних апаратів, що використовує тиск сонячного світла. Уявіть спрощену модель, де тиск створюється відштовхуванням заряджених частинок сонячного вітру від зарядженої поверхні вітрила. Нехай точкова частинка сонячного вітру із зарядом $q_1 = +e$ наближається до вітрила. Вітрило можна змоделювати двома нерухомими точковими зарядами $q_2 = +1000e$ та $q_3 = +1000e$, що знаходяться на відстані 2 мкм один від одного. Частинка сонячного вітру рухається вздовж перпендикуляра, проведеного через середину відрізка, що з'єднує заряди $q_2$ та $q_3$. Знайдіть силу, що діє на частинку, коли вона знаходиться на відстані 1 мкм від цього відрізка.

## Напруженість електричного поля

**Задача 1**

У сучасному очищувачі повітря використовується електростатичний фільтр. Пилинка, що пролітає через нього, отримує заряд $q = 3.2 \cdot 10^{-19}$ Кл. З якою силою електричне поле фільтра з напруженістю $E = 5000$ Н/Кл діє на цю пилинку?

**Задача 2**

Під час роботи мікросхеми смартфона на одному з її мікроскопічних елементів може накопичуватися статичний заряд, що дорівнює приблизно $q = -5$ нКл ($5 \cdot 10^{-9}$ Кл).
1.  Знайдіть модуль напруженості електричного поля, яке створює цей заряд на відстані $r = 2$ см від нього.
2.  Який напрямок матиме вектор напруженості в цій точці?

**Задача 3**

Уявіть датчик, що складається з двох точкових зарядів: позитивного $+q$ та негативного $-q$, розташованих на невеликій відстані один від одного. Як буде напрямлений вектор напруженості $\vec{E}$ у точці, що лежить точно посередині між цими зарядами? Зробіть схематичний малюнок силових ліній для цієї системи.

**Задача 4**

Медичний прилад для фізіотерапії створює електричне поле за допомогою зарядженого аплікатора, який можна вважати точковим зарядом $q = 20$ мкКл ($2 \cdot 10^{-5}$ Кл). На якій відстані від аплікатора напруженість електричного поля дорівнюватиме безпечному для певних процедур значенню $E = 2000$ Н/Кл?

**Задача 5**

У струменевому принтері краплина чорнила масою $m = 1.3 \cdot 10^{-10}$ кг і зарядом $q = -1.5 \cdot 10^{-13}$ Кл влітає в область, де існує однорідне електричне поле з напруженістю $E = 1.2 \cdot 10^6$ Н/Кл, спрямованою вертикально вгору.
1.  Знайдіть силу, що діє на краплину з боку електричного поля, та вкажіть її напрямок.
2.  Обчисліть прискорення, якого набуває краплина під дією цієї сили.

**Задача 6**

Дослідницький дрон вимірює електричне поле грозової хмари, яку можна умовно вважати точковим зарядом. На відстані 500 м від центру хмари прилад зафіксував напруженість $E_1 = 3000$ Н/Кл.
1.  Як зміниться напруженість поля, якщо дрон підніметься вище і опиниться на відстані 1.5 км від центру хмари?
2.  Обчисліть нове значення напруженості $E_2$.

**Задача 7**

Два статичних заряди, що виникли на компонентах електронної плати, $q_1 = +1$ нКл та $q_2 = +4$ нКл, знаходяться на відстані 30 см один від одного. Знайдіть точку на лінії, що з'єднує ці заряди, в якій напруженість сумарного електричного поля дорівнює нулю.

**Задача 8**

Генератор Ван де Граафа, що використовується в наукових експериментах для прискорення частинок, має металеву сферу радіусом $R = 20$ см, на якій накопичено заряд $q = -10$ мкКл ($ -1 \cdot 10^{-5}$ Кл).
1.  Яка напруженість електричного поля на поверхні сфери?
2.  Якою буде напруженість на відстані 1 м від центру сфери?

**Задача 9**

У лабораторії для дослідження властивостей аерозолів мікроскопічну кульку масою $m = 10$ мг ($10^{-5}$ кг) із зарядом $q = -2 \cdot 10^{-8}$ Кл утримують у нерухомому стані за допомогою однорідного електричного поля.
1.  Визначте модуль напруженості $\vec{E}$ цього електричного поля.
2.  Який напрямок має вектор напруженості $\vec{E}$?

**Задача 10**

Уявіть модель іонної пастки, яка використовується в мас-спектрометрах. Чотири однакові позитивні заряди $+q$ розташовані у вершинах квадрата.
1.  Чому дорівнює напруженість електричного поля в центрі цього квадрата? Поясніть свою відповідь, спираючись на властивості силових ліній та принцип симетрії.
2.  Як зміниться ситуація, якщо один із зарядів прибрати? Визначте напрямок та модуль напруженості поля в центрі квадрата, якщо сторона квадрата дорівнює $a$.

## Електричний струм та закон Ома

**Задача 1**

Акумулятор сучасного смартфона має ємність 4500 мА·год. Зарядний пристрій забезпечує силу струму 2 А.
1.  Який електричний заряд (в кулонах) накопичує повністю заряджений акумулятор?
2.  Скільки часу (в годинах) потрібно для повної зарядки цього смартфона, якщо вважати процес зарядки рівномірним?

**Задача 2**

Для освітлення рекламної вивіски використовують світлодіодну стрічку, яка підключена до блоку живлення з напругою 12 В. Визначте силу струму, що протікає через стрічку, якщо її загальний опір становить 60 Ом.

**Задача 3**

Ви вирішили підключити потужні аудіоколонки до підсилювача за допомогою мідного дроту.
1.  Обчисліть опір мідного дроту довжиною 15 м з площею поперечного перерізу 2.5 мм². Питомий опір міді становить 0.017 Ом·мм²/м.
2.  Як зміниться опір, якщо для економії ви використаєте дріт такої ж довжини, але з площею перерізу 1.5 мм²?

**Задача 4**

Через USB-порт комп'ютера, який використовується для живлення настільного вентилятора, проходить струм силою 500 мА.
1.  Який заряд пройде через порт за 10 хвилин роботи вентилятора?
2.  Скільки електронів пройде через поперечний переріз провідника за цей час? (Заряд електрона $e \approx 1.6 \cdot 10^{-19}$ Кл).

**Задача 5**

Нагрівальний елемент в сучасному 3D-принтері виготовлений з ніхромового дроту довжиною 80 см і площею поперечного перерізу 0.2 мм². Елемент підключений до напруги 24 В. Питомий опір ніхрому становить 1.1 Ом·мм²/м.
1.  Обчисліть опір нагрівального елемента.
2.  Знайдіть силу струму, що проходить через нього під час роботи принтера.

**Задача 6**

Для зарядки ноутбука використовується кабель USB Type-C довжиною 2 метри. Кабель складається з двох мідних провідників (для струму), кожен довжиною 2 м. Сила струму під час зарядки досягає 3 А. Площа поперечного перерізу кожного провідника становить 0.34 мм². (Питомий опір міді $\rho \approx 0.017$ Ом·мм²/м).
1.  Знайдіть загальний опір струмопровідних ліній у кабелі (враховуйте шлях струму "туди" і "назад").
2.  Обчисліть, яка частина напруги "втрачається" на самому кабелі під час зарядки.

**Задача 7**

Акумуляторна батарея електричного самоката має напругу 36 В. Двигун самоката під час рівномірного руху можна розглядати як споживач з постійним опором 4.5 Ом.
1.  Яка сила струму живить двигун самоката під час руху?
2.  Якщо ємність батареї становить 10 А·год, на який час безперервної роботи вистачить її заряду за таких умов?

**Задача 8**

Невелика сонячна панель на туристичному рюкзаку в сонячний день генерує напругу 6 В. Її використовують для живлення портативного ліхтарика, опір лампи якого становить 30 Ом.

**Задача 9**

Згідно з його технічними характеристиками, для нормальної роботи через нього має протікати струм 20 мА, а падіння напруги на ньому становить 2 В. Джерелом живлення слугує батарейка на 9 В. Щоб світлодіод не згорів, послідовно з ним потрібно підключити резистор, який "забере" на себе надлишок напруги.

**Задача 10**

Сучасні фітнес-браслети можуть вимірювати склад тіла (відсоток жиру та м'язів), пропускаючи через тіло дуже слабкий, безпечний струм (біоімпедансний аналіз). Уявімо спрощену модель: пристрій прикладає до двох точок на руці напругу 0.5 В і вимірює струм.

## Послідовне та паралельне з'єднання провідників

**Задача 1**

Для декорування кімнати використовують світлодіодну стрічку, що складається з 50 маленьких світлодіодів, з'єднаних послідовно. Опір кожного світлодіода становить 40 Ом. Визначте загальний опір усієї стрічки.

**Задача 2**

Два ігрові пристрої, підключені до портативного зарядного пристрою (power bank), можна розглядати як два паралельно з'єднані резистори. Опір першого пристрою в режимі зарядки становить 25 Ом, а другого - 30 Ом.
1. Який із пристроїв споживає більший струм?
2. Знайдіть загальний опір обох пристроїв.

**Задача 3**

Новорічна гірлянда складається з 20 лампочок, з'єднаних послідовно. Гірлянда підключена до мережі з напругою 220 В.
1. Яка напруга припадає на кожну окрему лампочку?
2. Якщо опір кожної лампочки 22 Ом, яка сила струму протікає в гірлянді?

**Задача 4**

У сучасних автомобілях для освітлення панелі приладів використовують декілька світлодіодів, що підключені паралельно до бортової мережі (12 В). Припустімо, що для підсвітки кнопок клімат-контролю паралельно підключено 3 світлодіоди з однаковим опором 150 Ом кожен. Знайдіть загальний опір цієї ділянки кола.

**Задача 5**

Квадрокоптер має чотири безколекторні двигуни, які для стабільної роботи з'єднані послідовно з платою контролера польоту. Під час польоту через усі двигуни проходить однаковий струм 5 А. Загальна напруга, що подається на ланцюг двигунів, становить 16 В.
1. Знайдіть загальний опір усіх чотирьох двигунів.
2. Яка напруга припадає на кожен окремий двигун, якщо їхні опори однакові?

**Задача 6**

Дві галогенні лампи в фарах автомобіля підключені паралельно до акумулятора, напруга якого 12 В. Опір однієї лампи становить 4 Ом.
1. Визначте загальну силу струму, що споживають обидві лампи.
2. Під час руху одна з ламп перегоріла (коло розімкнулося). Як змінилася загальна сила струму, що йде від акумулятора?

**Задача 7**

Для живлення потужного ігрового ноутбука використовується блок живлення, який створює на виході напругу 20 В. Усередині ноутбука система живлення містить два модулі пам'яті, з'єднані паралельно. Перший модуль споживає струм 0.5 А, а другий - 0.7 А.
1. Яка напруга подається на кожен із модулів пам'яті?
2. Який загальний струм споживають обидва модулі?
3. Обчисліть опір кожного модуля.
