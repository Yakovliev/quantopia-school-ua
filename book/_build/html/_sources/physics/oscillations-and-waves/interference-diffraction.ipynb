# Інтерференція і дифракція

Уявімо, що ми кинули у воду не один, а два камінці одночасно. Від кожного з них побіжать кола хвиль. У тому місці, де ці кола зустрінуться, почнеться щось цікаве: поверхня води стане дуже неспокійною, з'являться ділянки, де вода здіймається дуже високо, та ділянки, де вона залишається майже нерухомою. Це і є візуальний прояв інтерференції.

## Принцип суперпозиції хвиль

В основі всіх явищ, пов'язаних із накладанням хвиль, лежить фундаментальний **принцип суперпозиції**.

> **Принцип суперпозиції (накладання) хвиль:** Якщо в середовищі одночасно поширюється декілька хвиль, то коливання кожної частинки середовища є векторною сумою коливань, які ця частинка здійснювала б, беручи участь у кожному з хвильових процесів окремо.

Простими словами, хвилі не "штовхаються" і не заважають одна одній. Вони просто проходять одна крізь одну, і в кожній точці простору їхні зміщення додаються. Після того, як вони розійшлись, кожна хвиля продовжує свій рух так, ніби зустрічі й не було. Саме цей принцип ми вже застосовували, коли розглядали утворення стоячих хвиль.

## Інтерференція механічних хвиль

> **Інтерференція** - це явище накладання двох або більше когерентних хвиль, внаслідок якого в одних точках простору спостерігається стійке в часі посилення, а в інших - стійке в часі послаблення результуючих коливань.

Ключовим словом тут є "когерентних". Не будь-які хвилі створять стійку картину, яку ми можемо спостерігати.

> **Когерентні хвилі** - це хвилі, які мають однакову частоту (і, відповідно, довжину хвилі) та незмінну в часі різницю фаз.

Джерела, що випромінюють такі хвилі, називаються **когерентними джерелами**. Коливання від таких джерел є синхронними.

### Умови максимуму та мінімуму інтерференції

Давайте розберемося, від чого залежить, посилиться хвиля чи послабиться у певній точці. Уявімо два когерентних джерела $S_1$ та $S_2$, що створюють хвилі. Ми будемо спостерігати за деякою точкою $M$ у просторі.

До точки $M$ хвиля від джерела $S_1$ проходить відстань $d_1$, а від джерела $S_2$ - відстань $d_2$.

> **Різниця ходу хвиль** $\Delta d$ - це різниця відстаней, які проходять дві хвилі від своїх джерел до даної точки простору.

$$
\Delta d = |d_2 - d_1|
$$

Саме від цієї величини і залежить результат інтерференції.

#### Умова посилення коливань (інтерференційний максимум)

Щоб у точці $M$ коливання максимально посилились, необхідно, щоб гребінь однієї хвилі приходив одночасно з гребенем іншої, а западина - одночасно із западиною. Це означає, що хвилі мають прийти в точку $M$ в однаковій фазі.

Така ситуація виникає, коли на відрізку, що дорівнює різниці ходу, вкладається ціле число довжин хвиль.

> **Умова інтерференційного максимуму:** різниця ходу хвиль дорівнює цілому числу довжин хвиль.

Математично це записується так:

$$
\Delta d = m \lambda
$$

де:
*   $\lambda$ - довжина хвилі (в метрах, м).
*   $m$ - будь-яке ціле число ($m = 0, 1, 2, 3, \dots$), яке називають **порядком максимуму**.

#### Умова послаблення коливань (інтерференційний мінімум)

Щоб коливання в точці $M$ максимально послабились (в ідеалі, повністю зникли), необхідно, щоб гребінь однієї хвилі приходив одночасно із западиною іншої. Це означає, що хвилі мають прийти в точку $M$ у протифазі (різниця фаз дорівнює $\pi$, $3\pi$, $5\pi$ і т.д.).

Така ситуація виникає, коли на відрізку, що дорівнює різниці ходу, вкладається непарне число півхвиль.

> **Умова інтерференційного мінімуму:** різниця ходу хвиль дорівнює непарному числу півхвиль.

Математично це можна записати двома еквівалентними способами:

$$
\Delta d = (2m + 1) \frac{\lambda}{2} \quad \text{або} \quad \Delta d = \left(m + \frac{1}{2}\right) \lambda
$$

де:
*   $\lambda$ - довжина хвилі (в метрах, м).
*   $m$ - будь-яке ціле число ($m = 0, 1, 2, 3, \dots$), яке називають **порядком мінімуму**.

### Інтерференційна картина та закон збереження енергії

Сукупність усіх максимумів і мінімумів у просторі називається **інтерференційною картиною**. Вона має вигляд чергування областей з великою амплітудою коливань та областей, де коливання майже відсутні.

Тут може виникнути слушне питання: "А куди зникає енергія у точках мінімумів?". Відповідь криється у **законі збереження енергії**. Інтерференція не знищує і не створює енергію. Вона лише **перерозподіляє** її у просторі. Енергія, яка "зникла" з точок мінімумів, просто перейшла в точки максимумів, роблячи коливання там ще інтенсивнішими. Загальна енергія, що переноситься хвилями, залишається незмінною.

## Дифракція механічних хвиль

Ми звикли вважати, що хвилі, як і світло, поширюються прямолінійно. Якщо на шляху хвилі стоїть перешкода, за нею має утворитися "тінь". Однак, ми часто спостерігаємо інше. Наприклад, ми можемо чути звук з-за рогу будинку, хоча джерело звуку не бачимо. Звукова хвиля якимось чином оминула перешкоду.

> **Дифракція** - це явище огинання хвилями перешкод або відхилення від прямолінійного поширення при проходженні через отвори.

Дифракція, як і інтерференція, є фундаментальною властивістю будь-якого хвильового процесу.

### Умови спостереження дифракції

Чому ж ми чітко чуємо звук за рогом, але не бачимо світло? Адже світло - це теж хвиля. Виявляється, дифракція проявляється тим яскравіше, чим розміри перешкоди або отвору ($d$) ближчі до довжини хвилі ($\lambda$).

> Умова чіткого спостереження дифракції: $d \approx \lambda$ або $d < \lambda$.

Довжина звукових хвиль у повітрі становить від сантиметрів до метрів, що є порівнянним з розмірами дверей, будинків, дерев. Тому звук легко їх огинає. Довжина видимого світла надзвичайно мала (близько $400-700$ нанометрів), тому для спостереження дифракції світла потрібні дуже малі перешкоди або отвори.

### Принцип Гюйгенса-Френеля

Як же пояснити це огинання перешкод? Для цього ми скористаємося потужним інструментом, який називається **принципом Гюйгенса-Френеля**. Він складається з двох частин.

1.  **Принцип Гюйгенса (Хрістіан Гюйгенс, 1678 р.):** Кожна точка середовища, до якої дійшов **фронт хвилі**, сама стає джерелом вторинних сферичних хвиль. Нове положення фронту хвилі в наступний момент часу є огинаючою поверхнею усіх цих вторинних хвиль.

Цей принцип добре пояснює прямолінійне поширення, відбивання та заломлення хвиль, але не саму дифракцію.

2.  **Доповнення Френеля (Огюстен Френель, 1815 р.):** Вторинні хвилі, що випромінюються цими джерелами, є **когерентними** і **інтерферують** між собою.

Отже, **принцип Гюйгенса-Френеля** стверджує:

> Хвильовий рух у будь-якій точці простору є результатом інтерференції вторинних когерентних хвиль, джерелами яких є кожна точка хвильового фронту.

Давайте подивимось, як це працює. Уявімо плоску хвилю, що падає на екран з вузькою щілиною.

*   Згідно з принципом Гюйгенса, кожна точка фронту хвилі, яка досягла щілини, стає джерелом вторинних сферичних хвиль.
*   Ці вторинні джерела є когерентними.
*   За щілиною ці вторинні хвилі поширюються у всіх напрямках (включаючи область "геометричної тіні") та інтерферують між собою.
*   Результат цієї інтерференції і є та хвиля, що поширюється за перешкодою.

Таким чином, явище дифракції тісно і нерозривно пов'язане з інтерференцією. Можна сказати, що дифракція - це результат інтерференції когерентних вторинних хвиль. Саме цей процес і призводить до того, що хвиля "затікає" в область тіні, огинаючи перешкоду.
