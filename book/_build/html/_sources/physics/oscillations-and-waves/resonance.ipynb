# Резонанс

З матеріалу попередніх сторінок ми вже знаємо, що коливання бувають вільними та вимушеними. Вільні коливання з часом завжди затухають через втрати енергії. Щоб коливання були незатухаючими, потрібно, щоб коливальна система отримувала енергію ззовні. Це відбувається під час вимушених коливань, коли на систему діє зовнішня періодична сила.

Але що станеться, якщо "підштовхувати" коливальну систему з певною, особливою частотою? Мабуть, кожен з нас інтуїтивно знає відповідь, згадуючи, як розгойдував гойдалку. Щоб розгойдати її якнайсильніше, ми підштовхуємо її в такт її власному руху. Цей простий приклад ілюструє дуже цікаве і важливе фізичне явище - резонанс.

---

## Вимушені коливання та умова виникнення резонансу

Давайте згадаємо, що **вимушені коливання** - це коливання, які відбуваються під дією зовнішньої періодичної сили, яку ми назвемо **змушувальною силою** $F_{зм}$. Ця сила змінюється з часом за гармонічним законом, наприклад:

$$
F_{зм}(t) = F_0 \cos(\omega t + \varphi_0) \quad \text{або} \quad F_{зм}(t) = F_0 \sin(\omega t + \varphi_0)
$$

* Тут $F_0$ - це **амплітуда змушувальної сили**, тобто її максимальне значення. **Одиниця в СІ:** ньютон (Н). 
* Величина $\omega$ - це **циклічна частота змушувальної сили**, яка визначає, як швидко ця сила змінюється. **Одиниця в СІ:** радіан на секунду (рад/с). 
* Величина $\varphi_0$ - це початкова фаза коливань періодичної сили (фаза коливань періодичної сили у початковий момент відліку часу, коли $t = 0$). **Одиниця в СІ:** радіан (рад). 

У багатьох випадках та задачах відлік часу починають в момент, коли початкова фаза коливань дорівнює нулю $\varphi_0 = 0$, тому вищевказані рівняння для періодичної сили спрощуються:

$$
F_{зм}(t) = F_0 \cos(\omega t) \quad \text{або} \quad F_{зм}(t) = F_0 \sin(\omega t)
$$

Вибір косинуса чи синуса для опису сили, що змінюється за періодичним гармонічним законом, залежить від задачі та початкових умов. Наприклад, якщо в початковий момент часу $t=0$ сила дорівнює нулю, то логічніше обрати формулу $F_{зм}(t) = F_0 \sin(\omega t)$, яка задовольняє початкову умову. Натомість, якщо в початковий момент часу $t=0$ сила має максимальне значення, то обирають формулу $F_{зм}(t) = F_0 \cos(\omega t)$, оскільки в початковий момент часу значення косинуса набуває максимального значення (косинус нуля дорівнює одиниці) і як наслідок змушувальна сила також набуває максимального значення $F_{зм}(t = 0) = F_0$.

Кожна коливальна система (гойдалка, вантаж на пружині, струна гітари) має свою **власну частоту коливань** $\omega_0$ (або $\nu_0$), яка залежить від її фізичних параметрів (маси, жорсткості, довжини). Виявляється, що амплітуда вимушених коливань, які встановляться в системі, критично залежить від того, наскільки близькою є частота змушувальної сили $\omega$ до власної частоти системи $\omega_0$.

> **Резонанс** - це явище різкого зростання амплітуди вимушених коливань, яке відбувається за умови, що частота $\nu$ (або циклічна частота $\omega$) зовнішньої періодичної сили наближається до власної частоти $\nu_0$ (або власної циклічної частоти $\omega_0$) коливальної системи.

Отже, **умова виникнення резонансу**:

$$
\omega \approx \omega_0 \quad \text{або еквівалентно} \quad \nu \approx \nu_0
$$

Примітка 1: Раніше, ми вже продемонстрували, що $\omega = 2 \pi \nu$, що і доводить еквівалентність двох вищевказаних формул.

Примітка 2: У контексті теми, що ми тут розглядаємо, ми вважаємо поняття змушувальної сили та зовнішної періодичної сили еквівалентними і взаємозамінними. Тому іноді будемо використовувати або одне, або інше поняття.

## Амплітуда вимушених коливань. Резонансна крива

Коли на систему починає діяти змушувальна сила, спочатку коливання мають складний характер, але через деякий час перехідний процес завершується, і в системі встановлюються **усталені коливання**. Частота цих усталених коливань завжди дорівнює частоті змушувальної сили $\omega$. А от амплітуда цих коливань залежить як від амплітуди сили $F_0$, так і від співвідношення частот $\omega$ та $\omega_0$.

В ідеальній системі, де зовсім немає тертя, при точному збігу частот ($\omega = \omega_0$) амплітуда коливань зростала б до нескінченності. Це тому, що змушувальна сила постійно надає системі енергію, а механізмів для її розсіювання немає.

Однак у реальному світі завжди існує **сила тертя** (опір повітря, внутрішнє тертя в матеріалах). Тертя виконує роль "гальма", розсіюючи механічну енергію і перетворюючи її на тепло. Саме сила тертя обмежує амплітуду коливань під час резонансу, не даючи їй стати нескінченною.

Залежність амплітуди $A$ усталених вимушених коливань від частоти змушувальної сили $\omega$ можна зобразити графічно. Такий графік називається **резонансною кривою** (Рис. 1).

Проаналізувавши ці криві, ми можемо зробити кілька важливих висновків:
1.  Найбільша амплітуда коливань (пік на графіку) досягається тоді, коли частота змушувальної сили $\omega$ дуже близька до власної частоти системи $\omega_0$.
2.  Висота та гострота цього піку залежать від сили тертя (затухання) в системі.
3.  **Чим менша сила тертя**, тим вищий і гостріший пік резонансу. Це означає, що амплітуда при резонансі буде дуже великою, але резонанс буде спостерігатися лише у вузькому діапазоні частот, дуже близьких до власної частоти.
4.  **Чим більша сила тертя**, тим нижчим і більш "розмитим" є пік. Амплітуда в максимумі менша, але система помітно реагує на зовнішню силу в ширшому діапазоні частот навколо власної частоти.

Рис. 1. Резонансні криві. Зміна амплітуди залежно від відносної частоти $\omega / \omega_0$ (відношення (циклічної) частоти зовнішньої сили до власної (циклічної) частоти системи) та сили тертя в системі гармонічного маятника. Вісь абсцис позначає значення відносної частоти $\omega / \omega_0$. Вісь ординат позначає амплітуду коливань. Резонанс наступає при значенні відносної частоти близької до одиниці, тобто коли частота зовнішньої сили близька до власної частоти системи. ## Фізичний механізм резонансу

Чому ж амплітуда так сильно зростає? Справа в узгодженості дії зовнішньої сили з власним рухом системи.

Уявімо знову гойдалку. Щоб її розгойдати, ми штовхаємо її в той момент, коли вона досягає крайнього положення і починає рух у потрібному нам напрямку. Наші поштовхи (змушувальна сила) збігаються за частотою з власною частотою гойдалки. Кожен поштовх додає невелику порцію енергії до системи. Оскільки поштовхи відбуваються в такт руху, енергія накопичується, і амплітуда (висота підйому) зростає. Проте зі збільшенням амплітуди посилюються й втрати енергії, що врешті-решт знову приводить до усталеного режиму - просто з більшою амплітудою порівняно з початковою.

Якщо ж штовхати гойдалку не в такт (з іншою частотою), то іноді поштовх буде допомагати руху (додавати енергію), а іноді - заважати (поглинати енергію самої гойдалки). В результаті ефективного накопичення енергії не відбудеться, і амплітуда коливань буде малою.

Таким чином, **резонанс - це умова найбільш ефективної передачі енергії від зовнішнього джерела до коливальної системи**.

## Прояви резонансу в природі та техніці

Резонанс - це явище, яке може бути як надзвичайно корисним, так і вкрай руйнівним.

### Шкідливі прояви резонансу

Коли резонанс виникає там, де його не чекають, наслідки можуть бути катастрофічними.
*   **Такомський міст (США, 1940 р.).** Цей висячий міст зруйнувався через кілька місяців після відкриття. Хоча справжньою причиною був складний аеродинамічний ефект (аеропружний флаттер), цей випадок часто наводять як приклад руйнівних коливань, що нагадують резонанс, коли періодичні сили, що виникають під дією вітру, призвели до наростання амплітуди крутильних коливань моста до критичних значень.
*   **Вібрації в техніці.** Незбалансовані частини, що обертаються в машинах (наприклад, вал двигуна або колінвал), створюють вібрації. Якщо частота цих вібрацій збіжиться з власною частотою корпусу машини чи літака, це може призвести до сильного деренчання, втоми матеріалу і, зрештою, до руйнування конструкції.
*   **Землетруси.** Під час землетрусу коливання ґрунту є періодичною зовнішньою силою. Якщо частота цих коливань збіжиться з власною частотою будівлі, вона почне коливатися з величезною амплітудою, що може призвести до її повного руйнування.

### Корисні прояви резонансу

Люди навчилися використовувати потужну силу резонансу у своїх цілях.
*   **Музичні інструменти.** Звучання гітари, скрипки або фортепіано - це результат резонансу. Коливання струни (джерело) змушує коливатися корпус інструмента (деку). Дека сконструйована так, щоб її власні частоти збігалися з частотами коливань струн. Завдяки акустичному резонансу звук значно підсилюється.
*   **Радіо та телебачення.** Щоб налаштувати радіоприймач на улюблену станцію, ми обертаємо ручку настройки. При цьому ми змінюємо параметри електричного коливального контуру всередині приймача, змінюючи його власну частоту. Коли власна частота контуру збігається з частотою радіохвилі, що йде від передавальної станції, в контурі виникає електричний резонанс. Сигнал цієї станції різко підсилюється, а сигнали всіх інших - ні.
*   **Магнітно-резонансна томографія (МРТ).** Це сучасний метод медичної діагностики, що використовує явище ядерного магнітного резонансу. Ядра атомів водню в тілі людини поміщають у сильне магнітне поле і змушують резонувати під дією радіохвиль. Це дозволяє отримувати детальні зображення внутрішніх органів та м'яких тканин.

## Як боротися зі шкідливим резонансом?

Оскільки резонанс може бути небезпечним, інженери розробили кілька способів боротьби з ним.
1.  **Демпфування (збільшення тертя).** Як ми бачили на резонансній кривій, збільшення тертя значно зменшує амплітуду коливань при резонансі. В конструкції вводять спеціальні пристрої - демпфери (амортизатори в автомобілях, сейсмічні демпфери в хмарочосах), які ефективно розсіюють енергію коливань.
2.  **Зміна власної частоти системи.** Якщо не можна змінити частоту зовнішньої сили, можна змінити власну частоту системи, щоб вони не збігалися. Власна частота залежить від маси та жорсткості. Наприклад, укріпивши конструкцію (збільшивши жорсткість) або додавши масу, можна змінити її власну частоту і "вивести" її з резонансу.
3.  **Зміна частоти змушувальної сили.** Якщо це можливо, намагаються змінити частоту зовнішньої дії. Наприклад, швидкість обертання двигунів вибирають такою, щоб вона не збігалася з власними частотами конструкції, на якій вони встановлені.

Таким чином, резонанс - це фундаментальне фізичне явище з подвійною природою. Розуміння його законів дозволяє нам створювати дивовижні технології та уникати катастрофічних руйнувань.

| **Змушувальна сила** $F_{зм}$    | Зовнішня періодична сила, що підтримує коливання. Описується гармонічним законом $F_{зм}(t) = F_0 \cos(\omega t + \varphi_0)$, де $F_0$ - амплітуда сили, а $\omega$ - її циклічна частота.                                                              |

| **Фізичний механізм резонансу**     | Найбільш ефективна передача та накопичення енергії від зовнішнього джерела до коливальної системи. Це відбувається, коли дія сили збігається в такт із власним рухом системи.                                                                         |
| **Прояви резонансу**                | **Корисні:** акустика музичних інструментів, налаштування радіоприймачів на певну частоту, магнітно-резонансна томографія (МРТ). <br> **Шкідливі:** вібрації в техніці, руйнування мостів та будівель під дією вітру чи землетрусів. |
