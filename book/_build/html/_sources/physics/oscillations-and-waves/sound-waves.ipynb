# Звукові хвилі

Кожного дня ми живемо у світі, наповненому звуками: спів птахів, музика, шум міста, голоси людей. Але що таке звук насправді? З точки зору фізики, звук - це механічна хвиля. А розділ фізики, що вивчає звукові явища, їх виникнення, поширення та сприйняття, називається **акустикою**.

> **Звукові (або акустичні) хвилі** - це поздовжні механічні хвилі, що поширюються в пружних середовищах (газах, рідинах, твердих тілах) і сприймаються людським вухом у певному діапазоні частот.

Важливо наголосити на трьох ключових моментах цього означення:
1.  **Звук - це механічна хвиля.** Це означає, що для його поширення обов'язково потрібне середовище (повітря, вода, метал тощо). У вакуумі, де немає частинок, звук поширюватися не може. Класичним доказом цього є експеримент із дзвоником під скляним ковпаком, з-під якого викачують повітря: звук дзвоника поступово зникає.
2.  **Звук - це поздовжня хвиля.** Як ми пам'ятаємо з попередньої теми, у поздовжніх хвилях частинки середовища коливаються вздовж напрямку поширення хвилі.
3.  **Сприйняття вухом.** Хоча з фізичної точки зору звуком можна назвати будь-яку поздовжню хвилю, зазвичай під цим терміном розуміють саме ті хвилі, які ми можемо почути.

## Природа та поширення звуку

Уявімо собі джерело звуку, наприклад, **камертон** - простий пристрій у вигляді металевої "виделки", який після удару починає коливатися з певною частотою.

> **Джерело звуку** - це будь-яке тіло, що здійснює вимушені або власні згасаючі коливання з частотою, яка лежить у чутному діапазоні (чутний діапазон відповідає частотам від 20 Гц до 20 кГц).

Коли ніжка камертона рухається, наприклад, праворуч, вона стискає прилеглий шар повітря, створюючи в ньому область підвищеного тиску. Ця область називається **згущенням**. Завдяки пружності повітря, це згущення передається сусіднім шарам, поширюючись у просторі.

Потім ніжка камертона починає рухатися у зворотному напрямку, ліворуч. У просторі, де вона щойно була, тиск повітря стає нижчим за атмосферний. Ця область зниженого тиску називається **розрідженням**. Це розрідження так само поширюється в просторі, слідуючи за згущенням.

Таким чином, коливання джерела звуку створюють у середовищі послідовність згущень і розріджень, що чергуються і поширюються від джерела. Це і є поздовжня звукова хвиля.

Коливання тиску в звуковій хвилі відбуваються навколо середнього, атмосферного тиску. Відхилення тиску від атмосферного в даній точці середовища називається **звуковим тиском** $p$. **Одиниця в СІ:** паскаль (Па). Саме цей змінний тиск і діє на нашу барабанну перетинку, викликаючи відчуття звуку.

## Швидкість поширення звуку

Швидкість, з якою поширюються ці згущення та розрідження, називається **швидкістю звуку** $v$. Вона не залежить від джерела звуку (тобто від гучності чи висоти тону), а визначається виключно властивостями середовища, в якому звук поширюється, а саме його пружністю та густиною.

*   **У газах** швидкість звуку відносно невелика, оскільки молекули газу розташовані далеко одна від одної і слабко взаємодіють. Швидкість звуку в повітрі сильно залежить від температури: з її підвищенням швидкість зростає. При температурі 20 °C **швидкість звуку в повітрі** становить приблизно **343 м/с**.
*   **У рідинах** частинки знаходяться набагато ближче, а сили взаємодії між ними значно сильніші. Тому рідини є більш пружними, ніж гази. Наприклад, **швидкість звуку у воді** становить близько **1500 м/с**.
*   **У твердих тілах** атоми та молекули мають найміцніші зв'язки, що робить їх дуже пружними. Тому в твердих тілах швидкість звуку є найбільшою. Наприклад, у сталі вона може досягати 5000-6000 м/с.

Цікавий факт: почувши грім через кілька секунд після спалаху блискавки, ми можемо оцінити відстань до грози. Оскільки швидкість світла величезна (~300 000 км/с), ми бачимо спалах практично миттєво. Затримка звуку, наприклад, в 3 секунди означає, що гроза знаходиться на відстані приблизно $3 \text{ с} \cdot 343 \text{ м/с} \approx 1 \text{ км}$.

## Характеристики звуку

Звук, як і будь-яку хвилю, можна описати набором об'єктивних фізичних параметрів. Однак наше сприйняття звуку є суб'єктивним. Давайте розберемося, як ці дві сторони однієї медалі пов'язані між собою.

### Об'єктивні (фізичні) характеристики

Це параметри самої звукової хвилі, які можна виміряти приладами.
*   **Частота** $\nu$ - це кількість коливань у звуковій хвилі за одну секунду. **Одиниця в СІ:** герц (Гц). Частота звуку визначається виключно частотою коливань його джерела. Людське вухо здатне сприймати звукові хвилі з частотами в діапазоні приблизно від 20 Гц до 20 000 Гц (20 кГц).
*   **Амплітуда звукової хвилі** $A$ - це максимальна зміна тиску (тобто максимальний звуковий тиск) або максимальне зміщення частинок середовища від положення рівноваги.
*   **Інтенсивність звуку** $I$ - це фізична величина, що дорівнює енергії, яку звукова хвиля переносить за одиницю часу через одиницю площі поверхні, перпендикулярної до напрямку поширення хвилі. Простими словами, це потужність звуку на одиницю площі. **Одиниця в СІ:** ват на квадратний метр (Вт/м$^2$). Інтенсивність звуку пропорційна квадрату його амплітуди ($I \propto A^2$).

### Суб'єктивні (фізіологічні) характеристики

Це характеристики того, як наш слуховий апарат та мозок інтерпретують фізичні параметри звуку.

> **Висота звуку** - це фізіологічна характеристика, що визначається переважно **частотою** звукової хвилі. Дозволяє розрізняти звуки як "високі" та "низькі".

Чим більша частота коливань, тим вищим нам здається звук. Наприклад, писк комара - це звук високої частоти, а гудіння джмеля - низької. У музиці високі ноти відповідають високим частотам, а низькі - низьким.

> **Гучність звуку** - це фізіологічна характеристика, що визначається переважно **амплітудою (або інтенсивністю)** звукової хвилі.

Чим більша амплітуда коливань, тим більша інтенсивність звуку і тим гучнішим він нам здається. Окрім амплітуди, на сприйняття гучності впливає і частота: за однакової інтенсивності звуки середніх частот (1-5 кГц) здаються нам гучнішими, ніж звуки дуже низьких або дуже високих частот. 

Наше сприйняття гучності не є лінійним. Воно підпорядковується логарифмічному закону. Це означає, що для того, аби звук здався нам вдвічі гучнішим, його інтенсивність потрібно збільшити приблизно в 10 разів.

Через таку особливість нашого слуху, для вимірювання гучності використовують не абсолютні значення інтенсивності, а відносну логарифмічну шкалу - **шкалу децибелів (дБ)**.

> **Рівень інтенсивності звуку** $L$ (який ми сприймаємо як гучність) вимірюється в **децибелах** (дБ).

За нульовий рівень (0 дБ) прийнято **поріг чутності** - мінімальну інтенсивність звуку, яку може почути людина ($I_0 = 10^{-12}$ Вт/м²). Рівень інтенсивності звуку обчислюється за формулою:

$$
L = 10 \lg\left(\frac{I}{I_0}\right)
$$

Ось кілька прикладів рівнів гучності:
*   Шелест листя: 10 дБ
*   Тиха розмова: 40 дБ
*   Гучна розмова, шум у класі: 60-70 дБ
*   Шум потяга в метро: 100 дБ
*   Рок-концерт, реактивний двигун: 120 дБ

Звук з рівнем інтенсивності близько 120-130 дБ викликає больові відчуття і називається **больовим порогом**. Тривалий вплив звуку гучністю понад 85 дБ може призвести до незворотного пошкодження слуху.

> **Тембр звуку** - це "забарвлення" або "якість" звуку, яка дозволяє нам відрізняти звуки однакової висоти та гучності, створені різними джерелами (наприклад, скрипкою та фортепіано, що грають одну й ту ж ноту, або голос двох різних людей).

Чим же визначається тембр? Виявляється, будь-який звук, що видається реальним джерелом (голосом, музичним інструментом), не є простою гармонічною хвилею з однією частотою. Він є складною сумішшю кількох коливань.

*   **Основний тон** - це найнижча частота у цьому складному звуці. Саме вона визначає висоту звуку, яку ми сприймаємо.
*   **Обертони** - це всі інші, вищі частоти, що присутні у звуці. Зазвичай їхні частоти є кратними частоті основного тону (в такому випадку їх називають гармоніками).

Сукупність частот основного тону та обертонів, а також їхніх відносних інтенсивностей, називається **спектром звукової хвилі**.

> **Тембр** звуку визначається його **спектром**, тобто набором і відносною інтенсивністю різних обертонів. Чим багатший спектр обертонів, тим "соковитішим", "насиченішим" є тембр.

Саме унікальний набір обертонів надає неповторного звучання кожному музичному інструменту та людському голосу.

*   **Музичний звук (або музичний тон)** - це звук, що має чітко виражений основний тон та набір гармонійних обертонів. Його коливання є періодичними.
*   **Шум** - це звук, що є хаотичною сумішшю величезної кількості різних частот, без чітко вираженого основного тону та гармонік. Його коливання є неперіодичними (наприклад, шурхіт, скрип, шипіння).

## Акустичний резонанс

Ми вже знайомі з явищем резонансу як різкого зростання амплітуди вимушених коливань. В акустиці це явище відіграє надзвичайно важливу роль.

> **Акустичний резонанс** - це явище різкого зростання амплітуди коливань тіла (або об'єму повітря) під дією звукової хвилі, частота якої збігається з однією з власних частот коливань цього тіла.

Будь-яке тіло має свою **частоту власних коливань**. Якщо на нього діє звукова хвиля з такою ж частотою, тіло починає активно поглинати енергію цієї хвилі і коливатися з великою амплітудою.

Пристрої, призначені для підсилення звуку за допомогою резонансу, називаються **резонаторами**.

Класичний приклад - **камертон** на **резонансному ящику**. Сам по собі камертон видає дуже тихий звук, бо його тонкі ніжки погано передають коливання повітрю. Але якщо поставити його на дерев'яний ящик, відкритий з одного боку, звук стає набагато гучнішим. Чому? Тому що розміри ящика підібрані так, щоб власна частота коливань повітря всередині нього збігалася з частотою камертона. Повітряний стовп у ящику починає резонувати, і тепер коливається великий об'єм повітря, що створює гучний звук.

Саме цей принцип лежить в основі всіх струнних та духових музичних інструментів. Корпус гітари чи скрипки, труба саксофона - все це складні **резонатори**, які підсилюють слабкі коливання струни або губ музиканта і формують унікальний тембр інструмента.

## Пов'язані з звуком явища

*   **Інтерференція звуку.** Оскільки звук - це хвиля, він може інтерферувати. Якщо дві звукові хвилі від когерентних джерел (наприклад, двох динаміків, що відтворюють той самий тон) накладаються, то в просторі виникають точки, де звук посилюється (максимуми), і точки, де він послаблюється (мінімуми). Цей ефект використовується, наприклад, у системах активного шумозаглушення в навушниках.
*   **Дифракція звуку.** Звукові хвилі здатні огинати перешкоди. Довжина звукових хвиль, які ми чуємо, становить від кількох сантиметрів до кількох метрів. Це порівнянно з розмірами повсякденних об'єктів (двері, меблі, будівлі). Тому звук легко огинає ці перешкоди. Саме завдяки дифракції ми можемо чути розмову людей в сусідній кімнаті, навіть якщо двері відчинені лише трохи.

## Як ми чуємо? Людське вухо

Наше вухо - це дивовижний біологічний прилад для сприйняття звукових хвиль. Процес слуху можна розділити на кілька етапів:
1.  **Зовнішнє вухо** (вушна раковина та слуховий прохід) вловлює звукові хвилі і спрямовує їх до **барабанної перетинки**.
2.  **Середнє вухо.** Під дією звукового тиску **барабанна перетинка** починає коливатися. Ці коливання передаються на систему з трьох кісточок (молоточок, коваделко та стремінце), які діють як важіль, підсилюючи коливання приблизно у 20 разів.
3.  **Внутрішнє вухо.** Підсилені коливання передаються до **завитки** - спірального, заповненого рідиною органу.
4.  Всередині завитки коливання рідини змушують вібрувати основну мембрану. Уздовж цієї мембрани розташовані тисячі чутливих **волоскових клітин**. Різні ділянки мембрани резонують на різні частоти: високі частоти змушують коливатися ділянки біля основи завитки, а низькі - біля її вершини.
5.  Коливання **волоскових клітин** перетворюються на електричні імпульси, які по слуховому нерву надходять у мозок, де й формується остаточне відчуття звуку з його висотою, гучністю та тембром.

## Інфразвук та Ультразвук

Діапазон частот, який ми чуємо, - лише мала частина всього спектру механічних хвиль.

### Інфразвук

> **Інфразвук** - це механічні хвилі з частотою, нижчою за поріг людського слуху, тобто **частота інфразвуку** < 20 Гц.

Ми не чуємо інфразвук, але він може впливати на наш організм, викликаючи відчуття тривоги, втоми або навіть паніки. Інфразвукові хвилі мають велику довжину, слабо поглинаються середовищем і можуть поширюватися на сотні й тисячі кілометрів.

Джерелами інфразвуку є землетруси, виверження вулканів, сильний вітер, океанські шторми, а також працюючі двигуни, промислові вібрації. Деякі тварини, наприклад, слони та кити, використовують інфразвук для спілкування на великих відстанях.

### Ультразвук

> **Ультразвук** - це механічні хвилі з частотою, вищою за поріг людського слуху, тобто **частота ультразвуку** > 20 000 Гц (20 кГц).

На відміну від інфразвуку, ультразвук має дуже коротку довжину хвилі. Це надає йому ряд корисних властивостей: він добре відбивається від перешкод і його можна сфокусувати у вузький пучок. Ці властивості знайшли широке застосування в науці, техніці та медицині.

В основі більшості застосувань ультразвуку лежить **ультразвукова локація** (або ехолокація) - метод виявлення та визначення положення об'єктів за допомогою відбитих від них ультразвукових хвиль.

*   **У природі:** Кажани та дельфіни використовують ехолокацію для орієнтації в просторі та полювання. Вони випромінюють короткі ультразвукові імпульси і, аналізуючи час затримки та характеристики відлуння, створюють детальну "звукову картину" навколишнього світу.
*   **У техніці:**
    *   **Сонар** (або **ехолот**) - пристрій, що використовується на кораблях для вимірювання глибини, дослідження рельєфу морського дна, пошуку косяків риби чи підводних човнів.
    *   **Дефектоскоп** - прилад для виявлення дефектів (тріщин, порожнин) всередині непрозорих виробів, наприклад, у металевих деталях або рейках. Ультразвуковий імпульс, відбившись від дефекту, повертається назад і фіксується приладом.
*   **У медицині:** Ультразвукове дослідження (УЗД) є безпечним і поширеним методом діагностики, що дозволяє отримувати зображення внутрішніх органів.

| Поняття / Характеристика | Опис та ключові властивості | Формула / Позначення / Приклади |
| :--- | :--- | :--- |
| **Акустика** | Розділ фізики, що вивчає звукові явища. | |
| **Звукова хвиля** | Поздовжня механічна хвиля, що поширюється в пружному середовищі та сприймається вухом. | Потребує середовища, не поширюється у вакуумі. |
| **Джерело звуку** | Будь-яке тіло, що коливається з частотою в чутному діапазоні. | Чутний діапазон: 20 Гц – 20 кГц |
| **Поширення звуку** | Чергування згущень (підвищений тиск) і розріджень (знижений тиск) середовища. | |
| **Звуковий тиск** $p$ | Відхилення тиску від атмосферного у точці середовища. | Одиниця в СІ: паскаль (Па) |
| **Швидкість звуку** $v$ | Швидкість поширення звукової хвилі. Залежить від середовища, а не від джерела. | У повітрі: ≈343 м/с; у воді: ≈1500 м/с; у сталі: ≈5000-6000 м/с |
| --- | --- | --- |
| **Об'єктивні характеристики** | Параметри, що вимірюються приладами. | |
| **Частота** $\nu$ | Кількість коливань за одиницю часу. Визначає висоту звуку. | Одиниця в СІ: герц (Гц) |
| **Інтенсивність звуку** $I$ | Енергія, що її переносить хвиля за одиницю часу через одиницю площі. Визначає гучність. | Одиниця в СІ: Вт/м². $I \propto A^2$ (де $A$ - амплітуда) |
| **Спектр звуку** | Сукупність частот основного тону та обертонів, а також їх відносних інтенсивностей. Визначає тембр. | |
| --- | --- | --- |
| **Суб'єктивні характеристики** | Характеристики сприйняття звуку людиною. | |
| **Висота звуку** | Суб'єктивна оцінка частоти звуку ("високі" та "низькі" звуки). | Визначається частотою $\nu$ |
| **Гучність звуку** | Суб'єктивна оцінка інтенсивності звуку. Сприйняття підпорядковується логарифмічному закону. | Визначається інтенсивністю $I$ |
| **Рівень інтенсивності** $L$ | Логарифмічна міра гучності звуку. | $L = 10 \lg\left(\frac{I}{I_0}\right)$, де $I_0 = 10^{-12}$ Вт/м² (поріг чутності). Одиниця: децибел (дБ) |
| **Тембр звуку** | "Забарвлення" звуку, що дозволяє відрізняти джерела з однаковою гучністю та висотою. | Визначається спектром (набором обертонів) |
| --- | --- | --- |
| **Акустичний резонанс** | Явище різкого зростання амплітуди коливань, коли частота звукової хвилі збігається з власною частотою тіла. | Приклади: резонаторний ящик камертона, корпус гітари. |
| **Інфразвук** | Механічні хвилі з частотою, нижчою за поріг людського слуху. | Частота < 20 Гц |
| **Ультразвук** | Механічні хвилі з частотою, вищою за поріг людського слуху. | Частота > 20 кГц |
| **Ехолокація** | Метод визначення положення об'єктів за допомогою відбитих від них (ультра)звукових хвиль. | Приклади: сонар (ехолот), УЗД, дефектоскопія, орієнтація кажанів та дельфінів. |
