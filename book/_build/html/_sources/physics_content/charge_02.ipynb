# Напруженість електричного поля та силові лінії

## Вступ до поняття напруженості електричного поля

Для повного розуміння природи електричного поля необхідно ввести фізичну величину, яка характеризує його властивості. Розглянемо силу $\vec{F}$, що діє на точковий заряд $q$, розміщений у певній точці електричного поля. 

## Основні характеристики та залежності

Встановлено, що сила $\vec{F}$, яка діє на заряд, прямо пропорційна величині цього заряду $q$. Математично це можна записати як $\vec{F} \sim q$. Це фундаментальне співвідношення можна продемонструвати простим експериментом: якщо помістити в одну й ту саму точку поля заряд $2q$, то сила, що діятиме на нього, становитиме $2\vec{F}$.

## Напруженість як векторна величина

Особливу увагу варто приділити відношенню $\frac{\vec{F}}{q}$. Це векторна величина, яка має унікальні властивості:

1. Не залежить від модуля заряду
2. При зміні знаку заряду змінює свій напрям на протилежний
3. Характеризує властивості поля в конкретній точці простору

## Визначення напруженості

Напруженість електричного поля $\vec{E}$ визначається як відношення сили, що діє на пробний позитивний заряд, до величини цього заряду:

$\vec{E} = \frac{\vec{F}}{q}$

При цьому одиниця вимірювання напруженості в СІ: $[E] = \frac{\text{Н}}{\text{Кл}}$ або $\frac{\text{В}}{\text{м}}$

## Напруженість поля точкового заряду

Для точкового заряду $q$, розміщеного у вакуумі, напруженість поля на відстані $r$ визначається за формулою:

$E = \frac{|q|}{4\pi\varepsilon_0r^2}$

Важливо відзначити, що напруженість:
- Обернено пропорційна квадрату відстані від заряду
- Є векторною величиною, напрямленою вздовж радіуса від позитивного заряду або до негативного
- Залежить від середовища, в якому знаходиться заряд

## Силові лінії електричного поля

Для наочного представлення електричного поля використовують силові лінії - криві, дотичні до яких у кожній точці співпадають з напрямком вектора напруженості електричного поля. Густота силових ліній пропорційна модулю напруженості поля в даній точці.

# Напруженість електричного поля. Силові лінії

## Характеристика електричного поля

Для повноцінного розуміння та "роботи" з поняттям електричного поля необхідно ввести фізичні величини, які його характеризують. Основною такою величиною є сила $\vec{F}$, з якою поле діє в певній точці на точковий заряд $q$. 

Важливо розуміти, що ця сила має дві ключові залежності:
- залежить від значення заряду $q$
- залежить від властивостей самого електричного поля

При цьому залежність $\vec{F}$ від $q$ є прямо пропорційною. Це можна продемонструвати наступним чином: якщо об'єднати два однакові заряди $q$, то на такий подвійний заряд $2q$ поле діятиме з силою $2\vec{F}$. Це відповідає й означенню електричного заряду.

## Відношення $\frac{\vec{F}}{q}$ та його властивості

Розглянемо відношення $\frac{\vec{F}}{q}$. Це векторна величина, яка має особливі характеристики:

1. Не залежить від модуля заряду - якщо модуль заряду зміниться в певну кількість разів, сила $\vec{F}$ зміниться в таку ж кількість разів
2. Не залежить від знаку заряду $q$ - при зміні заряду на протилежний, напрям сили $\vec{F}$ також зміниться на протилежний (тобто замість $\vec{F}$ матимемо $-\vec{F}$)

## Напруженість електричного поля

Відношення $\frac{\vec{F}}{q}$ для певної точки поля не залежить від заряду $q$! Саме тому це відношення характеризує виключно електричне поле в зазначеній точці. Це відношення називають напруженістю електричного поля.

Більш формально: напруженість $\vec{E}$ електричного поля чисельно дорівнює силі, з якою поле в даній точці діє на одиничний позитивний заряд:

$\vec{E} = \frac{\vec{F}}{q}$

Напруженість є силовою характеристикою електричного поля. Одиниця напруженості в СІ: $[E] = \frac{\text{Н}}{\text{Кл}}$, хоча частіше використовують $\frac{\text{В}}{\text{м}}$.

## Напруженість поля точкового заряду

Для визначення напруженості електричного поля, яке створює нерухомий точковий заряд $q$ в точці на відстані $r$ від нього у вакуумі, використовуємо наступний підхід:

1. Розміщуємо в цю точку малий "пробний" заряд $q_1$
2. Згідно з означенням напруженості поля: $\vec{E} = \frac{\vec{F}}{q_1}$
3. За законом Кулона: $F = \frac{|q||q_1|}{4\pi\varepsilon_0r^2}$

Таким чином, отримуємо формулу для модуля напруженості:

$E = \frac{|q|}{4\pi\varepsilon_0r^2}$

Важливо зазначити, що напруженість є векторною величиною, тому потрібно визначати й напрям $\vec{E}$:
- Для позитивного заряду вектор $\vec{E}$ направлений від заряду
- Для негативного заряду вектор $\vec{E}$ направлений до заряду

Як видно з формули, напруженість поля точкового заряду спадає обернено пропорційно квадрату відстані від заряду.

## Силові лінії електричного поля

У складніших випадках отримати прості формули для визначення $\vec{E}$ може бути складно. Тому для наочного зображення електричного поля використовують так звані силові лінії електричного поля.

# Напруженість електричного поля та силові лінії

Для того щоб краще зрозуміти природу електричного поля, нам потрібно ввести важливу фізичну величину, яка його характеризує. Розглянемо, як це працює на практиці.

Уявімо, що ми помістили в електричне поле пробний заряд $q$. На цей заряд починає діяти сила $\vec{F}$. Важливо відмітити, що ця сила прямо пропорційна значенню заряду. Тобто, якщо ми візьмемо два однакові заряди $q$, то сила, що діє на них, подвоїться і становитиме $2\vec{F}$. 

Виходячи з цього, ми можемо ввести дуже важливу величину - відношення $\frac{\vec{F}}{q}$. Це векторна величина має цікаву властивість: вона не залежить від величини заряду! Скільки б разів ми не змінювали заряд, це відношення залишається незмінним. Більше того, якщо змінити знак заряду на протилежний, то напрямок сили теж зміниться на протилежний ($-\vec{F}$).

Це відношення $\frac{\vec{F}}{q}$ називають **напруженістю електричного поля** і позначають як $\vec{E}$:

$\vec{E} = \frac{\vec{F}}{q}$

Напруженість електричного поля $\vec{E}$ чисельно дорівнює силі, з якою поле діє на одиничний позитивний заряд у даній точці. Одиниця вимірювання напруженості в СІ - $[\vec{E}] = \frac{\text{Н}}{\text{Кл}}$ або, як частіше використовують, В/м.

Розглянемо конкретний приклад: напруженість поля точкового заряду $q$ на відстані $r$ у вакуумі. Використовуючи закон Кулона, можемо записати:

$E = \frac{|q|}{4\pi\epsilon_0r^2}$

При цьому вектор напруженості $\vec{E}$ направлений:
- від позитивного заряду (лінії розходяться)
- до негативного заряду (лінії сходяться)

Важливо розуміти, що напруженість поля спадає обернено пропорційно квадрату відстані від заряду. Це означає, що чим далі ми від джерела поля, тим слабше воно діє.

*Примітка: електричне поле є складовою електромагнітного поля, дія якого на заряджені частинки не залежить від швидкості їх руху.*

# Напруженість електричного поля. Силові лінії

## Фізична характеристика електричного поля

Для повного розуміння природи електричного поля необхідно ввести важливі фізичні величини, які його характеризують. Розглянемо силу $\vec{F}$, яка діє в певній точці електричного поля на пробний заряд $q$. Ця сила має цікаву властивість - вона залежить як від значення заряду $q$, так і від характеристик самого електричного поля. При цьому спостерігається пряма пропорційність: якщо ми візьмемо дві однакові кульки із зарядами $q$, то на кожну з них поле діятиме із силою $\vec{F}$, а на їх "об'єднання" - із силою $2\vec{F}$.

## Поняття напруженості електричного поля

Особливу увагу варто приділити відношенню $\frac{\vec{F}}{q}$. Це векторна величина, яка має унікальну властивість - вона не залежить від модуля заряду. Тобто, скільки б разів ми не змінювали заряд, це відношення залишається незмінним. Більше того, воно не залежить навіть від знаку заряду - при зміні заряду на протилежний напрям сили $\vec{F}$ також змінюється на протилежний.

Це відношення $\frac{\vec{F}}{q}$ називають напруженістю електричного поля і позначають як $\vec{E}$:

$\vec{E} = \frac{\vec{F}}{q}$

## Одиниці вимірювання та практичне застосування

Напруженість є силовою характеристикою електричного поля. Одиниця напруженості в СІ: $[E] = \frac{\text{Н}}{\text{Кл}}$ або, як частіше використовують, В/м.

Для практичного визначення напруженості поля точкового заряду $q$ на відстані $r$ у вакуумі використовується формула:

$E = \frac{F}{|q_1|} = \frac{|q|}{4\pi\varepsilon_0r^2}$

Важливо розуміти, що напруженість - це векторна величина. Її напрям визначається так:
- Для позитивного заряду вектор $\vec{E}$ направлений від заряду
- Для негативного заряду вектор $\vec{E}$ направлений до заряду

При цьому модуль напруженості спадає обернено пропорційно квадрату відстані від заряду.

> **Примітка**: Електричне поле є складовою електромагнітного поля, дія якого на заряджені частинки не залежить від швидкості їх руху.
