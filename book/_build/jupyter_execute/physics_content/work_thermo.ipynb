# Робота в термодинаміці

**Зв'язок роботи газу та внутрішньої енергії**

При зміні об'єму газу його внутрішня енергія також змінюється, якщо зовнішні сили виконують роботу. Розглянемо два випадки:

1. При стисненні газу (газ виконує від'ємну роботу):
   - Зовнішні сили виконують додатну роботу
   - Швидкість руху молекул збільшується і, як наслідок, температура газу зростає в результаті чого зростає і внутрішня енергія

2. При розширенні газу (газ виконує додатну роботу):
   - Газ виконує роботу проти зовнішніх сил
   - Швидкість руху молекул зменшується (молекули передають свій імпульс поршню чи іншому тілу, що обмежує об'єм газу, однак рухається під дією цього газу)
   - Як наслідок, температура газу знижується в результаті чого зменшується і внутрішня енергія

---

**Обчислення роботи газу**

Робота газу при зміні його об'єму визначається за формулою (це звичайна формула, відома з механіки):

$$
A = F \cdot s \cdot \cos \alpha,
$$

де $ F $ — сила тиску газу, $ s $ — переміщення поршня, $ \alpha $ — кут між силою і переміщенням.


У випадку ізобарного процесу (тобто тиск постійний), сила, що діє з боку газу на поршень, визначається як $F = pS$, де $p$ - тиск газу, а $S$ - площа поршня (ця формула отримується із формули означення тиску $p = \frac{F}{S}$). Переміщення поршня $s = l_2 - l_1$. Таким чином, робота газу при ізобарному розширенні дорівнює:

$$A = pS(l_2 - l_1) = p(S l_2 - S l_1) = p(V_2 - V_1) = p\Delta V$$

де:
- $ p $ — тиск газу,
- $ \Delta V = V_2 - V_1 $ — зміна об'єму газу.

---

**Геометрична інтерпретація роботи**

Роботу газу можна представити геометрично як площу під графіком залежності $p(V)$:

- При ізобарному процесі робота чисельно дорівнює площі прямокутника
- При довільному процесі робота дорівнює площі криволінійної трапеції
- При ізохорному процесі ($V = const$) об'єм не змінюється ($ \Delta V = 0 $), тому робота дорівнює нулю $A = 0$.

Примітка: тут невдовзі будуть додані рисунки для кращої ілюстрації.

---

**Підсумки**

- Робота газу виконується лише під час зміни об'єму ($ \Delta V \neq 0 $).
- Робота газу залежить від шляху переходу з початкового стану в кінцевий.
- При ізобарному процесі робота обчислюється за формулою:
  $$ A = p \Delta V. $$
- При ізобарному розширенні робота газу додатна, при стисненні - від'ємна.
- У випадку ізохорного процесу ($ V = const $), робота не виконується ($ A = 0 $).
- Геометрично робота газу дорівнює площі під графіком $ p(V) $.

При відсутності теплообміну з довкіллям:
   - Якщо над газом виконується додатна робота (при цьому сам газ виконує від'ємну роботу), то його внутрішня енергія збільшується.
   - Якщо газ сам виконує роботу (додатню роботу), його внутрішня енергія зменшується.

Примітка: тут невдовзі будуть додані приклади задач та приклади коду.
