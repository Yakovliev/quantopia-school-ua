# Спектральний аналіз

Коли сонячне світло проходить крізь призму, ми бачимо дивовижне явище - білий промінь розділяється на різнокольорові смуги, утворюючи веселку кольорів від червоного до фіолетового. Це явище відкрив ще Ісаак Ньютон у XVII столітті, і саме він назвав цю кольорову смугу спектром.

Однак справжнє значення спектрів для науки стало зрозумілим лише у XIX столітті, коли вчені виявили, що спектри містять унікальну інформацію про хімічний склад речовин, їхню температуру та багато інших фізичних властивостей. Спектри стали своєрідними "відбитками пальців" атомів і молекул, за якими ми можемо ідентифікувати речовини навіть на величезних космічних відстанях.

> **Спектр** - це розподіл інтенсивності випромінювання за частотами або довжинами хвиль. Коли світло розкладається на складові компоненти (наприклад, коли біле світло проходить через призму), ми отримуємо спектр.

Ключове рівняння, що пов'язує частоту ($\nu$) і довжину хвилі ($\lambda$):

$$\nu = \frac{c}{\lambda}$$

де $c$ - швидкість світла ($3 \times 10^8$ м/с).


## Види спектрів

У природі існує три основні типи спектрів, і кожен з них несе свою унікальну інформацію про джерело випромінювання.

> **Неперервний спектр** - спектр, у якого монохроматичні складові заповнюють без розривів інтервал довжин хвиль, в межах якого відбувається випромінювання.

Неперервні спектри

- Характеризуються безперервним розподілом енергії за всіма довжинами хвиль
- Випромінюються твердими, рідкими тілами або стисненим газом, нагрітими до високих температур
- Приклад: спектр абсолютно чорного тіла, випромінювання від лампи розжарювання

Класичним прикладом джерела з неперервним спектром є лампа розжарювання. Нагріта металева спіраль випромінює світло всіх видимих частот, утворюючи суцільну кольорову смугу у спектрі.

> Коли розріджений газ збуджується (наприклад, нагріванням або електричним розрядом), його атоми випромінюють світло тільки певних, строго визначених довжин хвиль. Це створює **лінійчастий спектр випромінювання**, де на темному фоні видно яскраві кольорові лінії.

Наприклад, якщо у полум'я потрапляє невелика кількість кухонної солі, воно набуває яскраво-жовтого забарвлення. Цей ефект пояснюється присутністю у складі солі елемента Натрію, атоми якого відповідальні за характерне жовте світіння. Під впливом високої температури полум'я атоми Натрію отримують додаткову енергію та переходять у збуджений енергетичний стан. При поверненні до основного стану надлишкова енергія вивільняється у формі **електромагнітного випромінювання**. Згідно з постулатами квантової теорії Бора, це випромінювання має **строго визначені частоти та, відповідно, довжини хвиль**. У випадку Натрію найінтенсивніше випромінювання припадає на довжини хвиль, що відповідають жовтій ділянці видимого спектра.

Експериментальні дослідження підтверджують, що при достатньому нагріванні атоми будь-якого хімічного елемента здатні випромінювати світло. Якщо пропустити це випромінювання через призму, то вузький пучок світла розкладається на декілька окремих пучків. Для розріджених газів в атомарному стані на екрані спектрографа можна спостерігати **окремі кольорові лінії з широкими темними проміжками між ними**. Такий набір ліній і називають **лінійчастим спектром випромінювання**.

Варто підкреслити, що **спектральні лінії конкретного хімічного елемента утворюють унікальний набір, який відрізняється від спектрів усіх інших елементів**. Це дозволяє використовувати спектр як своєрідний "відбиток пальця" або "паспорт" хімічного елемента.

Лінійчасті спектри випромінювання

- Виникають, коли збуджені атоми переходять у стани з нижчою енергією
- Представлені яскравими кольоровими лініями на темному фоні
- Кожна лінія відповідає конкретному електронному переходу


> Якщо світло з неперервним спектром (наприклад, від розжареного тіла) пропустити через холодний розріджений газ, то атоми газу поглинатимуть фотони саме тих енергій, які вони могли б випромінити у збудженому стані. У результаті, на фоні яскравого неперервного спектра з'являються темні лінії - **лінійчастий спектр поглинання**.

Положення темних ліній у спектрі поглинання точно відповідає положенню яскравих ліній у спектрі випромінювання того ж елемента.

Лінійчасті спектри поглинання

- Виникають, коли світло проходить через речовину, і атоми поглинають фотони певних енергій
- Представлені темними лініями на яскравому кольоровому фоні
- Темні лінії з'являються на тих самих довжинах хвиль, що й яскраві лінії у спектрі випромінювання того ж елемента

Лінійчасті спектри (поглинання та випромінювання)

- Складаються з окремих спектральних ліній на певних довжинах хвиль
- Характерні для атомів у газоподібному стані
- Кожен хімічний елемент має унікальний "відбиток пальців" з ліній

### Смугасті спектри
- Складаються з груп близько розташованих ліній, що формують смуги
- Характерні для молекул
- Виникають через додаткові коливальні та обертальні енергетичні рівні

## 3. Основи спектрального аналізу

Спектральний аналіз — це метод визначення хімічного складу речовини за її спектром. Він базується на тому, що:

1. Кожен хімічний елемент випромінює світло з унікальним набором частот
2. Ці частоти відповідають переходам електронів між енергетичними рівнями

Принцип спектрального аналізу можна описати рівнянням Бора:

$$E_2 - E_1 = h\nu$$

де:
- $E_2$ і $E_1$ — енергії рівнів
- $h$ — стала Планка ($6.626 \times 10^{-34}$ Дж·с)
- $\nu$ — частота випромінювання

Основні етапи спектрального аналізу:
1. Збудження атомів речовини (нагріванням, електричним розрядом тощо)
2. Розкладання випроміненого світла у спектр
3. Реєстрація та аналіз отриманого спектру


## 5. Спектри розріджених газів vs стиснених газів, рідин та твердих тіл

### Розріджені гази
- Дають чіткі лінійчасті спектри
- Атоми взаємодіють слабко, енергетичні рівні не збурені
- Лінії вузькі і чітко визначені

### Стиснені гази
- При збільшенні тиску спектральні лінії розширюються
- Причина: зіткнення між атомами змінюють енергетичні рівні (ефект Штарка)
- Формула розширення лінії: $\Delta \lambda \propto n$, де $n$ — концентрація атомів

### Рідини та тверді тіла
- Дають переважно неперервні спектри
- Атоми тісно взаємодіють, енергетичні рівні перекриваються
- У кристалах можуть спостерігатися смуги поглинання, пов'язані зі структурою кристалічної ґратки

## 6. Молекулярні спектри (смугасті спектри)

Молекулярні спектри складніші за атомні через додаткові ступені свободи молекул:

1. **Електронні переходи** — дають основну структуру спектра
2. **Коливальні рівні** — виникають через коливання атомів у молекулі
3. **Обертальні рівні** — виникають через обертання молекули як цілого

Енергія молекули може бути представлена як:

$$E = E_{електронна} + E_{коливальна} + E_{обертальна}$$

Коливальна енергія двоатомної молекули (гармонічне наближення):

$$E_{коливальна} = (v + \frac{1}{2})h\nu_0$$

де:
- $v$ — коливальне квантове число (0, 1, 2, ...)
- $\nu_0$ — частота коливань

Обертальна енергія двоатомної молекули:

$$E_{обертальна} = \frac{h^2}{8\pi^2I}J(J+1)$$

де:
- $I$ — момент інерції молекули
- $J$ — обертальне квантове число (0, 1, 2, ...)



## Основи спектрального аналізу

Спектральний аналіз — це потужний метод дослідження речовин за їхніми спектрами. Його основи були закладені німецькими вченими Густавом Кірхгофом та Робертом Бунзеном у 1859-1860 роках.

### Принципи якісного та кількісного аналізу

Спектральний аналіз дозволяє:

1. **Якісно** визначити, які хімічні елементи присутні в досліджуваній речовині (за положенням ліній у спектрі)
2. **Кількісно** визначити концентрацію цих елементів (за інтенсивністю ліній)

Для проведення спектрального аналізу використовують спеціальні прилади — спектрометри (або спектрографи). Основними елементами спектрометра є:
- Вхідна щілина, через яку проходить досліджуване випромінювання
- Диспергуючий елемент (призма або дифракційна ґратка), який розкладає світло в спектр
- Детектор, який реєструє розподіл інтенсивності у спектрі

## Лінійчасті спектри: природа та інтерпретація

Щоб зрозуміти природу лінійчастих спектрів, ми повинні звернутися до квантової теорії. На початку XX століття Нільс Бор запропонував модель атома, яка пояснювала дискретний характер спектрів.

Згідно з моделлю Бора, електрони в атомі можуть перебувати лише на певних дозволених орбітах з фіксованими значеннями енергії. Коли електрон переходить з вищого енергетичного рівня на нижчий, він випромінює фотон, енергія якого дорівнює різниці енергій цих рівнів:

$$E_{\text{фотона}} = E_{\text{вищого рівня}} - E_{\text{нижчого рівня}} = h\nu$$

де $h$ — стала Планка, а $\nu$ — частота випроміненого світла.

Оскільки енергетичні рівні в атомах квантовані (приймають лише дискретні значення), то й енергії випромінених фотонів також квантовані, що призводить до появи окремих спектральних ліній.

### Серії спектральних ліній

У спектрі атома водню спостерігаються кілька серій ліній, які відповідають переходам електрона на різні кінцеві енергетичні рівні:

1. **Серія Лаймана** — переходи на перший (найнижчий) рівень. Ці лінії лежать в ультрафіолетовій області спектра.

2. **Серія Бальмера** — переходи на другий рівень. Ці лінії лежать у видимій області спектра і позначаються як $H_\alpha$ (червона), $H_\beta$ (блакитна), $H_\gamma$ (фіолетова) тощо.

3. **Серія Пашена** — переходи на третій рівень. Ці лінії лежать в інфрачервоній області.

Довжини хвиль ліній в цих серіях можна обчислити за формулою Рідберга:

$$\frac{1}{\lambda} = R \left(\frac{1}{n_1^2} - \frac{1}{n_2^2}\right)$$

де $R$ — стала Рідберга, $n_1$ — номер нижнього рівня, $n_2$ — номер верхнього рівня ($n_2 > n_1$).

## Спектри різних станів речовини

Спектри суттєво залежать від агрегатного стану речовини та умов, у яких вона перебуває.

### Спектри розріджених газів

Розріджені гази, де атоми знаходяться на значних відстанях один від одного і слабко взаємодіють, дають чіткі лінійчасті спектри. Кожен атом можна розглядати як майже ізольовану систему, і його енергетичні рівні практично не збурюються сусідніми атомами.

### Спектри стиснених газів

При підвищенні тиску газу відстані між атомами зменшуються, і їхня взаємодія стає суттєвішою. Це призводить до розширення спектральних ліній — явища, відомого як **тиснове розширення**. При дуже високому тиску лінії можуть розширитися настільки, що перекриються, утворюючи майже неперервний спектр.

### Спектри рідин та твердих тіл

У рідинах і твердих тілах атоми розташовані дуже близько один до одного і сильно взаємодіють. Це призводить до значного розширення і навіть розщеплення енергетичних рівнів. 

У кристалах дискретні енергетичні рівні атомів розщеплюються на енергетичні зони, що містять велику кількість близьких рівнів. Це призводить до того, що спектри твердих тіл часто мають характер широких смуг або стають неперервними.

## Молекулярні спектри

На відміну від атомів, молекули мають додаткові ступені свободи — вони можуть обертатися та коливатися. Це ускладнює їхні енергетичні спектри.

### Смугасті спектри

Молекули зазвичай дають **смугасті спектри** — групи близько розташованих ліній, які при спостереженні з низькою роздільною здатністю здаються суцільними смугами.

Смугасті спектри виникають через те, що кожному електронному переходу відповідає багато можливих коливальних і обертальних переходів. Це призводить до розщеплення кожної "чистої" електронної лінії на серію близьких ліній, які формують спектральну смугу.

### Структура молекулярних спектрів

Молекулярний спектр містить інформацію про:

1. **Електронну структуру** молекули — визначає загальне розташування спектральних смуг
2. **Коливальні рівні** — створюють структуру всередині кожної електронної смуги
3. **Обертальні рівні** — утворюють тонку структуру всередині коливальних смуг

Аналіз молекулярних спектрів дозволяє визначити:
- Міжатомні відстані в молекулі
- Силу хімічних зв'язків
- Моменти інерції молекули
- Потенціальну енергію взаємодії атомів



**Лінійчасті спектри можна спостерігати тільки у розріджених газах, де атоми перебувають в ізольованому стані**. Наприклад, спостерігаючи через призму за світінням газорозрядної трубки, можна бачити характерний лінійчастий спектр газу, яким вона заповнена.

При збільшенні густини газу відбуваються зміни в спектрі: спектральні лінії стають ширшими, а коли взаємодія між частинками (атомами, молекулами, іонами) досягає значної інтенсивності, окремі лінії зливаються, утворюючи **неперервний спектр**. Ось чому **тверді тіла, рідини та щільні гази у нагрітому стані демонструють неперервний спектр випромінювання**.

> Важлива закономірність: газ найбільш інтенсивно поглинає електромагнітні хвилі тих самих довжин, які він випромінює у нагрітому стані. Тому **темні лінії в спектрі поглинання розташовані точно на тих самих місцях, де у спектрі випромінювання спостерігаються яскраві лінії**.

## Молекулярні спектри

Спектри газів у молекулярному стані суттєво відрізняються від атомарних спектрів. Вони представлені не окремими лініями, а **системою смуг, що складаються з великої кількості дуже близько розташованих ліній**. Такі спектри отримали назву **смугастих спектрів**.

Формування смугастого спектра обумовлене складнішою структурою молекул порівняно з окремими атомами. При отриманні енергії в молекулах відбуваються одночасно кілька типів енергетичних переходів: електрони в атомах переходять на вищі енергетичні рівні, виникають коливання атомів відносно положення рівноваги всередині молекули, а також змінюється обертальний рух молекули як цілого. Відповідно до законів квантової фізики, енергії коливальних та обертальних рухів також можуть набувати лише дискретних значень. В результаті кожен електронний енергетичний рівень розщеплюється на множину коливальних підрівнів. Це суттєво збільшує кількість можливих енергетичних переходів при поверненні молекули до основного стану, що призводить до появи надзвичайно великої кількості спектральних ліній, які зливаються у характерні широкі смуги.

## Основи спектрального аналізу

Як було вже зазначено, вивчення лінійчастих спектрів випромінювання та поглинання одноатомних газів дає можливість визначити їхній хімічний склад. Це можливо завдяки тому, що кожен газ в атомарному стані характеризується власним унікальним набором спектральних ліній (тобто специфічним набором довжин хвиль). Положення цих ліній залишається незмінним незалежно від методу збудження атомів. На сьогодні науковці визначили та систематизували спектри всіх відомих хімічних елементів, створивши детальні спектральні таблиці.

Аналізуючи інтенсивність спектральних ліній, можна також визначити концентрацію елементів у газовій суміші: більша кількість атомів певного елемента відображається у вигляді яскравіших ліній у спектрі випромінювання або більш виражених темних ліній у спектрі поглинання.

> **Спектральним аналізом** називається метод якісного та кількісного визначення складу речовини на основі дослідження її спектра.

За допомогою спектрального аналізу було здійснено чимало наукових відкриттів, зокрема виявлено раніше невідомі хімічні елементи. Першим таким елементом став Цезій (від латинського слова *caesius*, що означає "блакитний"). Цей надзвичайно рідкісний на Землі елемент отримав свою назву через наявність у його спектрі випромінювання двох яскравих блакитних ліній.

Спектральний аналіз став революційним методом дослідження в астрофізиці. Завдяки йому вчені змогли встановити хімічний склад зірок, міжзоряних газових хмар та інших космічних об'єктів, перебуваючи на відстані мільйонів світлових років від них.

Завдяки своїй універсальності та надзвичайній чутливості (метод дозволяє виявити присутність елемента навіть якщо його маса становить лише $10^{-10}$ г) спектральний аналіз знайшов широке застосування в багатьох галузях науки і техніки - хімії, металургії, ядерній фізиці та інших сферах.

Важливо зауважити, що для молекулярних газів спектральний аналіз проводиться за їхніми характерними смугастими спектрами. Для твердих і рідких речовин прямий спектральний аналіз неможливий - спочатку необхідно перевести досліджуваний зразок у газоподібний стан.

## Звідки ми знаємо властивості далеких зір

Історія вивчення спектрів зірок розпочалась у 1802 році, коли британський науковець Вільям Хайд Волластон (1766–1828) вперше спостерігав темні лінії в спектрі сонячного світла. Розглядаючи сонячне випромінювання через спектроскоп, він помітив декілька темних смуг, що перетинали райдужний спектр. Однак Волластон помилково вважав їх дефектом оптичної призми. Лише через 17 років німецький фізик Йозеф Фраунгофер (1787–1826) довів, що ці лінії є природною властивістю сонячного світла. Сьогодні темні лінії поглинання в спектрі Сонця називають **фраунгоферовими лініями** на його честь.

Дослідження ліній поглинання в зоряних спектрах дозволяє астрофізикам визначати фундаментальні характеристики небесних тіл: їхній елементний склад, температурні показники, тиск атмосфери, швидкість обертання та інші параметри.

Відповідно до закону випромінювання Макса Планка, підвищення температури зорі призводить до зміщення максимуму енергії випромінювання в напрямку короткохвильової (фіолетової) частини спектра. Аналізуючи співвідношення інтенсивності випромінювання різних спектральних діапазонів, астрономи можуть **досить точно визначати температуру поверхневих шарів зірок**.

**Ефект Допплера** став неоціненним інструментом для вимірювання швидкості руху космічних об'єктів, визначення відстаней до них, а також виявлення екзопланет. Суть цього ефекту полягає в зміні спостережуваної довжини хвилі залежно від відносного руху джерела та спостерігача: при наближенні джерела довжина хвилі зменшується (частота зростає), а при віддаленні - збільшується (частота зменшується). Завдяки цьому фізичному явищу було зроблено фундаментальне відкриття: спектральні лінії далеких галактик демонструють систематичне зміщення в бік довгохвильової (червоної) частини спектра, що свідчить про їхнє віддалення від нашої галактики з колосальними швидкостями.



## Фраунгоферові лінії та їхнє значення для астрономії

На початку XIX століття німецький оптик Йозеф Фраунгофер помітив темні лінії в спектрі Сонця. Ці лінії, які пізніше були названі **фраунгоферовими лініями**, виявилися ключовими для розуміння хімічного складу Сонця та інших зір.

### Походження фраунгоферових ліній

Фраунгоферові лінії — це лінії поглинання, які утворюються, коли світло від гарячої фотосфери Сонця проходить через холодніші зовнішні шари сонячної атмосфери (хромосферу). Атоми в цих шарах поглинають певні довжини хвиль, створюючи характерні темні лінії в спектрі.

Найбільш помітні фраунгоферові лінії були позначені Фраунгофером літерами. Наприклад:
- **A** і **B** — лінії, викликані поглинанням кисню в земній атмосфері
- **C** — червона лінія водню $H_\alpha$
- **D** — жовта дублетна лінія натрію
- **E** — лінія заліза
- **F** — блакитна лінія водню $H_\beta$
- **G** — лінія заліза та кальцію
- **H** і **K** — фіолетові лінії кальцію

### Значення для астрофізики

Відкриття Фраунгофера стало основою для розвитку астрофізики. Аналіз фраунгоферових ліній дозволив:

1. Визначити хімічний склад Сонця та інших зір
2. Виявити нові хімічні елементи (наприклад, гелій було вперше виявлено в спектрі Сонця)
3. Вимірювати магнітні поля на Сонці (за ефектом Зеємана — розщепленням спектральних ліній у магнітному полі)
4. Вивчати динаміку сонячної атмосфери (за доплерівським зсувом ліній)

## Вимірювання температури зорі за її спектром

Одним із найважливіших параметрів зорі є її температура. Існує кілька методів визначення температури зорі за її спектром.

### Абсолютно чорне тіло та закон Планка

**Абсолютно чорне тіло** (АЧТ) — це ідеалізований об'єкт, який повністю поглинає все електромагнітне випромінювання, що падає на нього. При нагріванні АЧТ випромінює світло зі спектром, який описується законом Планка:

$$B_\lambda(T) = \frac{2hc^2}{\lambda^5} \cdot \frac{1}{e^{hc/\lambda kT} - 1}$$

де:
- $B_\lambda$ — спектральна світність на одиницю довжини хвилі
- $h$ — стала Планка
- $c$ — швидкість світла
- $\lambda$ — довжина хвилі
- $k$ — стала Больцмана
- $T$ — абсолютна температура

Зорі, особливо їхні фотосфери, досить добре апроксимуються моделлю абсолютно чорного тіла.

### Закон зміщення Віна

З закону Планка випливає закон зміщення Віна, який встановлює, що довжина хвилі, на яку припадає максимум випромінювання, обернено пропорційна температурі:

$$\lambda_{max} = \frac{b}{T}$$

де $b \approx 2.9 \cdot 10^{-3}$ м·К — стала Віна.

Це означає, що чим гарячіша зоря, тим коротші довжини хвиль переважають у її випромінюванні:
- Холодні червоні зорі (T ≈ 3000 K) мають максимум у червоній або навіть інфрачервоній області
- Зорі середньої температури, як наше Сонце (T ≈ 5800 K), мають максимум у жовто-зеленій області
- Найгарячіші блакитні зорі (T ≈ 30000 K) мають максимум у синій або ультрафіолетовій області

### Метод колірної температури

Одним із практичних методів визначення температури зорі є вимірювання її колірного індексу — різниці видимих зоряних величин, виміряних у різних спектральних смугах (наприклад, B-V, де B — блакитна, а V — візуальна смуга). 

Колірний індекс пов'язаний з температурою зорі, оскільки відображає нахил її спектра. Для зір головної послідовності можна використовувати емпіричні залежності типу:

$$T_{eff} \approx \frac{7000 \text{ K}}{(B-V) + 0.56}$$

### Метод іонізаційної рівноваги

Інший метод базується на аналізі ліній іонізованих атомів різних ступенів іонізації. Відношення інтенсивностей ліній, що належать різним іонізаційним станам одного й того ж елемента (наприклад, нейтрального заліза Fe I і одноразово іонізованого заліза Fe II), залежить від температури згідно з рівнянням Саха:

$$\frac{N_{i+1}}{N_i} \cdot \frac{N_e}{N_0} = \frac{2Z_{i+1}}{Z_i} \left(\frac{2\pi m_e kT}{h^2}\right)^{3/2} e^{-\chi_i/kT}$$

де:
- $N_i$, $N_{i+1}$ — концентрації атомів у станах іонізації $i$ та $i+1$
- $N_e$ — концентрація електронів
- $Z_i$, $Z_{i+1}$ — статистичні суми станів
- $\chi_i$ — потенціал іонізації
- $T$ — температура

## Ефект Доплера та його застосування в астрономії

Ефект Доплера — це зміна частоти хвиль при відносному русі джерела і приймача. Він спостерігається для всіх типів хвиль, включаючи електромагнітні.

### Формула ефекту Доплера для світла

Для світла формула ефекту Доплера в нерелятивістському наближенні має вигляд:

$$\frac{\Delta \lambda}{\lambda_0} = \frac{v_r}{c}$$

де:
- $\Delta \lambda = \lambda - \lambda_0$ — зміна довжини хвилі
- $\lambda_0$ — довжина хвилі у спокої
- $v_r$ — радіальна швидкість (додатна при віддаленні)
- $c$ — швидкість світла

При релятивістських швидкостях формула ускладнюється.

### Червоне та блакитне зміщення

Залежно від напрямку руху джерела відносно спостерігача, ефект Доплера призводить до:

1. **Червоного зміщення** — якщо джерело віддаляється, довжина хвилі збільшується, і спектральні лінії зміщуються в червону (довгохвильову) частину спектра

2. **Блакитного зміщення** — якщо джерело наближається, довжина хвилі зменшується, і спектральні лінії зміщуються в синю (короткохвильову) частину спектра

### Застосування в астрономії

Ефект Доплера має численні застосування в астрономії:

1. **Вимірювання радіальних швидкостей зір** — за зміщенням спектральних ліній можна визначити, наскільки швидко зоря наближається до нас або віддаляється

2. **Виявлення зоряних пульсацій** — періодичні зміни розмірів зорі призводять до періодичного доплерівського зміщення ліній

3. **Відкриття екзопланет** — планета, обертаючись навколо зорі, спричиняє періодичні коливання її радіальної швидкості, які можна виявити за доплерівським зміщенням

4. **Визначення швидкості обертання зір** — різні частини поверхні зорі мають різні радіальні швидкості, що призводить до розширення спектральних ліній

5. **Виявлення розширення Всесвіту** — червоне зміщення в спектрах далеких галактик свідчить про їхнє віддалення від нас, причому чим далі галактика, тим більше червоне зміщення (закон Габбла):

$$v = H_0 \cdot d$$

де $v$ — швидкість віддалення, $d$ — відстань, а $H_0$ — стала Габбла.

## Висновки

Спектральний аналіз є одним із найпотужніших методів вивчення Всесвіту. Він дозволяє нам визначати хімічний склад, температуру, швидкість і багато інших параметрів космічних об'єктів, навіть якщо вони знаходяться на відстані мільйонів і мільярдів світлових років від нас.

Розуміння природи спектрів, від найпростіших атомних до складних молекулярних і суцільних, а також знання законів, які керують їхнім утворенням і зміною, є фундаментальним для сучасної астрофізики.

Завдяки спектральному аналізу ми можемо "читати" світло, що приходить до нас від далеких зір і галактик, розкриваючи таємниці Всесвіту, недоступні для безпосереднього спостереження.




## 7. Фраунгоферові лінії

Фраунгоферові лінії — темні лінії поглинання в спектрі Сонця, відкриті Йозефом Фраунгофером на початку XIX століття.

- Виникають, коли світло від гарячого ядра Сонця проходить через більш холодні зовнішні шари
- Хімічні елементи в сонячній атмосфері поглинають світло на характерних для них довжинах хвиль
- Найвідоміші лінії: D (натрій), H та K (кальцій), G (залізо)

Фраунгоферові лінії дозволили визначити хімічний склад Сонця без необхідності відвідувати його, що стало революцією в астрономії.

## 8. Як виміряти температуру зорі на основі спектрів

### Абсолютно чорне тіло

Абсолютно чорне тіло (АЧТ) — ідеальний об'єкт, який поглинає все падаюче на нього випромінювання. Спектр його випромінювання описується законом Планка:

$$I(\lambda, T) = \frac{2hc^2}{\lambda^5} \cdot \frac{1}{e^{\frac{hc}{\lambda kT}} - 1}$$

де:
- $I(\lambda, T)$ — спектральна густина потоку випромінювання
- $T$ — абсолютна температура
- $k$ — стала Больцмана ($1.38 \times 10^{-23}$ Дж/К)

### Закон зміщення Віна

Максимум випромінювання АЧТ зміщується при зміні температури відповідно до закону Віна:

$$\lambda_{max} \cdot T = b$$

де $b = 2.9 \times 10^{-3}$ м·К — стала Віна.

### Визначення температури зорі

1. **Метод колірної температури**:
   - Вимірюється інтенсивність випромінювання на двох довжинах хвиль
   - Співвідношення інтенсивностей порівнюється з теоретичними для АЧТ різних температур

2. **Метод спектральної класифікації**:
   - Зорі класифікуються за спектральними типами (O, B, A, F, G, K, M)
   - Кожному типу відповідає певний діапазон температур
   - Визначається наявність характерних ліній елементів

3. **Метод максимуму випромінювання**:
   - Визначається довжина хвилі, на якій інтенсивність випромінювання максимальна
   - Застосовується закон Віна: $T = \frac{b}{\lambda_{max}}$

## 9. Ефект Доплера та червоне зміщення

### Ефект Доплера

Ефект Доплера — зміна спостережуваної частоти хвиль при відносному русі джерела і спостерігача.

Формула для світла (нерелятивістське наближення):

$$\frac{\Delta \lambda}{\lambda} = \frac{v_r}{c}$$

де:
- $\Delta \lambda$ — зміна довжини хвилі
- $\lambda$ — початкова довжина хвилі
- $v_r$ — променева швидкість (проекція швидкості на лінію "спостерігач-джерело")
- $c$ — швидкість світла

### Червоне зміщення

- Виникає, коли джерело віддаляється від спостерігача
- Довжина хвилі збільшується (зсувається до червоного кінця спектра)
- $v_r > 0$, $\Delta \lambda > 0$

### Блакитне зміщення

- Виникає, коли джерело наближається до спостерігача
- Довжина хвилі зменшується (зсувається до синього кінця спектра)
- $v_r < 0$, $\Delta \lambda < 0$

### Космологічне червоне зміщення

Окремий випадок — космологічне червоне зміщення, пов'язане з розширенням Всесвіту:

$$z = \frac{\lambda_{спостережувана} - \lambda_{випромінена}}{\lambda_{випромінена}} = \frac{a_{тепер}}{a_{тоді}} - 1$$

де:
- $z$ — параметр червоного зміщення
- $a$ — масштабний фактор Всесвіту

## Висновки

Спектральний аналіз є потужним інструментом дослідження природи — від визначення складу речовин на Землі до вивчення далеких зір і галактик. Він дозволяє:

- Визначати хімічний склад об'єктів
- Вимірювати температуру віддалених тіл
- Визначати швидкість руху космічних об'єктів
- Вивчати фізичні процеси в атомах і молекулах

Розуміння спектрів є фундаментальним для сучасної астрофізики, хімії та фізики, а спектроскопія залишається одним із найважливіших методів дослідження у природничих науках.
