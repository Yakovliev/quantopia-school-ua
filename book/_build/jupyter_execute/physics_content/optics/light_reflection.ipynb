# Закони відбивання світла

## Принцип Ферма

Фундаментальним принципом геометричної оптики є принцип найменшого часу, названий на честь французького математика П'єра де Ферма (1601-1665). Згідно з цим принципом:

*Світло завжди обирає таку траєкторію, щоб на подолання відстані між двома точками витратити найменший час.*

Застосування принципу Ферма:
- В однорідному середовищі світло поширюється прямолінійно, оскільки це найкоротший шлях
- У неоднорідному середовищі траєкторія світла викривляється, але все одно забезпечує мінімальний час поширення

На основі цього принципу можна математично вивести всі закони геометричної оптики (однак для цього потрібно гарно розуміти інтегрування).

## Закони геометричної оптики

> Геометрична оптика - це розділ оптики, який вивчає закони поширення світла в прозорих середовищах і принципи побудови зображень в оптичних системах без урахування хвильових властивостей світла. 

Основне поняття геометричної оптики - світловий промінь, який визначається як лінія, вздовж якої поширюється потік світлової енергії. Це суто геометричне поняття використовується для схематичного зображення світлових пучків.

Основні закони геометричної оптики:

1. Закон прямолінійного поширення світла:
   - В однорідному прозорому середовищі світло поширюється прямолінійно

2. Закон незалежного поширення світла:
   - Окремі пучки світла не впливають один на одного і поширюються незалежно

3. Закон відбивання та заломлення світла

## Закони відбивання світла

**Явище відбивання світла**

В однорідному середовищі світло поширюється прямолінійно, доки не досягне межі з іншим середовищем (наприклад, поверхні тіла). На межі поділу середовищ частина світлової енергії повертається в перше середовище - це явище називають відбиванням світла.

При падінні вузького пучка світла на дзеркало можна спостерігати:
- Падаючий промінь - промінь, що задає напрямок пучка світла, який падає на поверхню
- Відбитий промінь - промінь, що задає напрямок відбитого пучка світла

**Закони відбивання світла**

1. Падаючий промінь, відбитий промінь і перпендикуляр до поверхні відбивання, проведений із точки падіння променя, лежать в одній площині.

2. Кут відбивання світла ($\beta$) дорівнює куту його падіння ($\alpha$): $\beta = \alpha$

Із закону відбивання світла випливає важлива властивість - оборотність світлових променів: якщо падаючий промінь спрямувати шляхом відбитого променя, то відбитий промінь пройде шляхом падаючого.


**Чому ми бачимо предмети навколо**

Більшість предметів, що нас оточують, не є джерелами світла. Ми бачимо їх завдяки тому, що вони відбивають світло від джерел освітлення. При цьому:
- Світло не тільки відбивається від фізичних тіл
- Частина світла поглинається ними
- Найкращі відбивачі - дзеркала і тіла білого кольору (відбивають до 95% падаючого світла)

**Види відбивання світла**

1. Дзеркальне відбивання:
  - Відбувається від гладеньких поверхонь
  - Паралельний пучок світлових променів після відбивання залишається паралельним
  - Продовження відбитих променів перетинаються в одній точці $S_1$ (ця точка є уявним зображенням джерела світла $S$)
  - Зображення сукупності точок предмета дає зображення предмета
  - Можливе тільки від дуже гладеньких поверхонь (дзеркальних поверхонь)
  - Плоску дзеркальну поверхню називають плоским дзеркалом

2. Дифузне (розсіяне) відбивання:
  - Відбувається від шорстких поверхонь
  - Паралельні промені після відбивання поширюються в різних напрямках
  - Продовження відбитих променів не перетинаються в одній точці
  - Зображення не утворюється
  - Дозволяє бачити освітлений предмет з будь-якого боку
  - Характерне для більшості поверхонь

Завдяки цим видам відбивання ми можемо бачити як чіткі зображення в дзеркалах, так і всі інші предмети навколо нас у повсякденному житті.

Примітка: невдовзі тут будуть додані рисунки із прикладами.
