# Молекулярно-кінетична теорія. Задачі

## Основи молекулярно кінетичної теорії

**Задача 1**

Для приготування кави бариста взяв 180 грамів чистої води ($H_2O$). Обчисліть, яку кількість речовини (в молях) він використав. Молярна маса води дорівнює $18 \cdot 10^{-3} \frac{\text{кг}}{\text{моль}}$.

**Задача 2**

Кремній (Si) є основою сучасної мікроелектроніки. Для виготовлення одного мікрочипа було використано кристалик чистого кремнію масою 0.056 г.
1. Визначте молярну масу Кремнію, скориставшись періодичною таблицею ($A_r(Si) \approx 28$).
2. Обчисліть, скільки атомів Кремнію міститься в цьому кристалику.

**Задача 3**

Вуглекислий газ ($CO_2$) – один із головних парникових газів, що впливає на клімат Землі. Вчені-екологи постійно моніторять його концентрацію в атмосфері. Обчисліть масу однієї молекули вуглекислого газу в кілограмах.

**Задача 4**

Для нормального функціонування організму людині потрібен мікроелемент Ферум (Fe), який входить до складу гемоглобіну. В одній таблетці вітамінного комплексу міститься $3 \cdot 10^{20}$ атомів Феруму. Яка маса заліза у цій таблетці?

**Задача 5**

У медицині для діагностики використовують магнітно-резонансну томографію (МРТ), де для охолодження надпровідних магнітів застосовують рідкий гелій (He). У контейнері міститься 50 літрів (0,05 $м^3$) рідкого гелію. Густина рідкого гелію становить 125 $\frac{\text{кг}}{м^3}$.
1. Обчисліть масу гелію в контейнері.
2. Обчисліть кількість атомів гелію в контейнері.

**Задача 6**

Вважається, що повітря – це суміш газів, у якій приблизно 78% становить азот ($N_2$). Густина повітря за нормальних умов дорівнює 1,29 $\frac{\text{кг}}{м^3}$.
1. Знайдіть масу повітря у кімнаті об’ємом 20 $м^3$.
2. Враховуючи, що маса азоту становить 78% від маси повітря, обчисліть масу азоту в кімнаті.
3. Скільки молекул азоту знаходиться в цій кімнаті?

**Задача 7**

Для охолодження компонентів потужних ігрових комп'ютерів іноді використовують системи з рідким азотом. Знайдіть кількість речовини (в молях), що міститься в 100 г рідкого азоту, формула якого $N_2$. Відносна атомна маса Нітрогену $A_r(N) = 14$.

**Задача 8**

Літій-іонні акумулятори, які живлять наші смартфони та ноутбуки, використовують графіт ($C$) в якості анода. Скільки атомів Карбону міститься в шматочку графіту масою 6 грамів? Відносна атомна маса Карбону $A_r(C) = 12$, стала Авогадро $N_A \approx 6.02 \cdot 10^{23} \text{ моль}^{-1}$.

**Задача 9**

Для виробництва сонячних панелей використовують кремній ($Si$), а для з'єднувальних дротів - мідь ($Cu$). Що має більшу кількість атомів: зразок чистого кремнію масою 280 г чи зразок міді масою 320 г? У скільки разів більше? Відповідь обґрунтуйте розрахунками.
Відносні атомні маси: $A_r(Si) = 28$, $A_r(Cu) = 64$.

**Задача 1**

**Задача 2**

Середня кінетична енергія кожного атома дорівнює $8 \cdot 10^{-21}$ Дж. Який тиск створює аргон у цій точці?

**Задача 3**

Атмосфера Марса дуже розріджена і складається переважно з вуглекислого газу ($CO_2$). Марсохід "Perseverance" зафіксував біля поверхні планети тиск 600 Па. Яка густина марсіанської атмосфери в місці вимірювання, якщо середня квадратична швидкість молекул $CO_2$ там становить приблизно 390 м/с?

**Задача 4**

Сучасні автомобілі можуть використовувати як паливо стиснений природний газ (метан, $CH_4$), який зберігається у балонах під високим тиском.
1.  Обчисліть, який тиск створює метан у балоні, якщо його концентрація становить $8 \cdot 10^{27} м^{-3}$, а середня квадратична швидкість молекул – 550 м/с. Маса однієї молекули метану $2,66 \cdot 10^{-26}$ кг.
2.  Знайдіть густину газу в балоні за цих умов.

**Задача 5**

Медичні кисневі концентратори подають пацієнтам повітря зі збільшеною часткою кисню. У балоні з таким киснем тиск становить $2 \cdot 10^5$ Па, а середня кінетична енергія молекул $6,5 \cdot 10^{-21}$ Дж. Коли кисень потрапляє в легені, його тиск падає до $1 \cdot 10^5$ Па, а середня кінетична енергія молекул стає $6,2 \cdot 10^{-21}$ Дж.
1.  Знайдіть концентрацію молекул кисню в балоні.
2.  Знайдіть концентрацію молекул кисню в легенях.
3.  У скільки разів змінилася концентрація молекул?

**Задача 6**

В іонних двигунах, що використовуються для корекції орбіти космічних апаратів, робочим тілом часто є інертний газ, наприклад, ксенон. У камері двигуна цей газ перебуває під тиском 150 Па і має густину 0,2 кг/м³. Маса одного атома ксенону дорівнює $2,18 \cdot 10^{-25}$ кг.
1.  Обчисліть середню квадратичну швидкість атомів ксенону.
2.  Знайдіть середню кінетичну енергію поступального руху одного атома ксенону.

**Задача 7**

Для підвищення довговічності та стабільності тиску автомобільні шини часто накачують не звичайним повітрям, а чистим азотом. У шині об'ємом 30 л (0,03 м³) міститься 75 г азоту, який створює тиск 250 кПа.
1.  Спочатку обчисліть густину азоту в шині.
2.  Використовуючи знайдену густину та відомий тиск, визначте середню квадратичну швидкість молекул азоту.

**Задача 8**

Балон для дайвінгу містить повітря під тиском $2 \cdot 10^7$ Па. Під час занурення дайвер витратив половину повітря з балона. Після цього балон, що був нагрітий сонцем, охолов у воді, внаслідок чого середня кінетична енергія руху молекул повітря, що залишилось, зменшилась на 15%.
1.  Як змінився тиск у балоні одразу після того, як дайвер витратив половину повітря (вважайте, що температура не змінилася)?
2.  Яким став кінцевий тиск у балоні після його охолодження у воді?

**Задача 9**

На висоті орбіти Міжнародної космічної станції (МКС) залишкові частинки атмосфери (переважно атомарний кисень) сильно іонізовані сонячним випромінюванням, тому їхня середня кінетична енергія досить висока - близько $1,2 \cdot 10^{-19}$ Дж. Тиск, який створюють ці частинки, надзвичайно низький - приблизно $1 \cdot 10^{-9}$ Па. Середня маса частинки становить $2,66 \cdot 10^{-26}$ кг (атом кисню).
1.  Знайдіть концентрацію частинок на цій висоті.
2.  Обчисліть густину залишкової атмосфери на висоті орбіти МКС.
3.  Визначте середню квадратичну швидкість цих частинок. Порівняйте її зі швидкістю руху самої МКС (близько 7.7 км/с).
