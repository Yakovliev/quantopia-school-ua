# Затухаючі електромагнітні коливання в реальному контурі

У попередньому матеріалі ми розглядали ідеальний LC-контур. Ми уявляли, що дроти, якими з'єднані конденсатор і котушка, не мають опору. Але в будь-якому реальному провіднику є **активний опір**. Котушка індуктивності, яка є довгою мідною дротиною, також має свій власний опір.

> **Активний опір** $R$ - це фізична величина, що характеризує властивість провідника протидіяти проходженню електричного струму, перетворюючи електричну енергію на теплову. Одиниця вимірювання в СІ: **Ом ($\Omega$)**.

Цей опір діє подібно до тертя в механічних системах. Якщо ми штовхнемо гойдалку, тертя в підвісі та опір повітря поступово "з'їдять" її енергію, і зрештою вона зупиниться. Так само і в коливальному контурі: активний опір призводить до того, що при протіканні струму частина електромагнітної енергії безповоротно перетворюється на тепло (за законом Джоуля-Ленца, $Q = I^2Rt$).

Внаслідок цих втрат енергії амплітуда коливань заряду, напруги та струму з часом зменшується. Такі коливання називаються **затухаючими**.

> **Затухаючі коливання** - це коливання, амплітуда яких з часом зменшується через втрати енергії в системі.

## Реальний коливальний контур (RLC-контур)

Тепер наш коливальний контур складається з трьох елементів, з'єднаних послідовно: конденсатора (C), котушки індуктивності (L) та резистора (R). Такий контур називають **RLC-контуром**.

Давайте виведемо рівняння, яке описує процеси в такому контурі. Цього разу ми не можемо використати закон збереження енергії, бо енергія не зберігається. Замість цього ми скористаємося **узагальненим законом Ома для замкненого кола** (який є наслідком другого правила Кірхгофа). Він говорить, що сума всіх спадів напруги на елементах кола дорівнює сумі електрорушійних сил (ЕРС), що діють у цьому колі.

У нашому замкненому RLC-контурі єдиним джерелом ЕРС є сама котушка, в якій виникає **ЕРС самоіндукції** $\mathcal{E}_L$ при зміні струму.

$$\mathcal{E}_L = -L \frac{\Delta I}{\Delta t}$$

Спадами напруги є:
1.  Спад напруги на активному опорі: $U_R = I R$
2.  Спад напруги на конденсаторі: $U_C = \frac{q}{C}$

Запишемо узагальнений закон Ома для нашого контуру: cума спадів напруги = ЕРС.

$$U_R + U_C = \mathcal{E}_L$$

Підставимо вирази для кожної величини:

$$IR + \frac{q}{C} = -L \frac{\Delta I}{\Delta t}$$

Це і є основне рівняння, що описує процеси в RLC-контурі. Тепер, як і минулого разу, переведемо його на мову похідних, щоб отримати диференціальне рівняння.

### Диференціальне рівняння затухаючих коливань

Згадаймо зв'язок між зарядом та струмом:
*   Сила струму $I$ - це швидкість зміни заряду: $I = \frac{\Delta q}{\Delta t}$. У наших позначеннях $I = q'$.
*   Швидкість зміни сили струму $\frac{\Delta I}{\Delta t}$ - це "прискорення" заряду, або його друга похідна: $\frac{\Delta I}{\Delta t} = I' = q''$.
*   Напруга на конденсаторі: $U = \frac{q}{C}$.

Підставимо ці вирази в наше рівняння $IR + \frac{q}{C} = -L \frac{\Delta I}{\Delta t}$:

$$
q'R + \frac{1}{C}q = -Lq''
$$

Тепер перенесемо всі доданки в один бік вліво і розташуємо їх у порядку спадання похідних (від другої до нульової):

$$
Lq'' + Rq' + \frac{1}{C}q = 0
$$

Щоб привести рівняння до стандартного вигляду, поділимо всі його члени на індуктивність $L$:

$$
q'' + \frac{R}{L}q' + \frac{1}{LC}q = 0
$$

Це і є **диференціальне рівняння затухаючих електромагнітних коливань**. Давайте порівняємо його з рівнянням для ідеального контуру, яке ми отримали раніше: $q'' + \frac{1}{LC}q = 0$. Ми бачимо, що з'явився новий доданок $\frac{R}{L}q'$, який містить активний опір $R$ та першу похідну заряду $q'$ (тобто силу струму). Саме цей член описує "тертя" в системі, яке призводить до затухання.

### Нові параметри: коефіцієнт затухання та власна частота

Щоб зробити наше рівняння більш компактним і зрозумілим, введемо два нових параметри.

1.  Величину $\frac{1}{LC}$ ми вже зустрічали. Це квадрат **власної циклічної частоти** $\omega_0$ - частоти, з якою відбувалися б коливання в цьому ж контурі, якби в ньому не було опору ($R=0$).
    > **Власна циклічна частота** $\omega_0$ - це циклічна частота коливань в ідеальному LC-контурі (без опору).
    >
    > $$ \omega_0 = \frac{1}{\sqrt{LC}} \quad \text{або} \quad \omega_0^2 = \frac{1}{LC} $$
    >
    > Одиниця вимірювання в СІ: **радіан за секунду (рад/с)**.

2.  Весь новий "затухаючий" ефект міститься у множнику $\frac{R}{L}$. Введемо величину, яка називається **коефіцієнтом затухання**.
    > **Коефіцієнт затухання** $\beta$ - це величина, що характеризує швидкість зменшення амплітуди коливань. Він визначається як:
    >
    > $$\beta = \frac{R}{2L}$$
    >
    > Одиниця вимірювання в СІ: **одиниця на секунду ($1/c$ або $c^{-1}$)**.
    >
    > *Примітка: множник 2 у знаменнику вводиться для зручності подальших математичних перетворень.*

Тепер наше диференціальне рівняння набуває елегантного і стандартного вигляду:

$$q'' + 2\beta q' + \omega_0^2 q = 0$$

Це рівняння дуже схоже на рівняння затухаючих механічних коливань.

## Розв'язок рівняння: як змінюється заряд?

Розв'язання такого диференціального рівняння є складнішою задачею, ніж для ідеального контуру, і виходить за межі шкільного курсу математики. Тому ми приймемо його розв'язок як готовий факт, але детально проаналізуємо його фізичний зміст.

Розв'язком цього рівняння (за умови, що затухання не надто велике) є функція:

$$q(t) = q_{m0} e^{-\beta t} \cos(\omega t + \varphi_0)$$

Давайте розберемо цю формулу на частини:
*   $q_{m0} e^{-\beta t}$ - це **амплітуда затухаючих коливань**. Вона не є сталою, як у ідеальному контурі, а зменшується з часом за експоненціальним законом.
    *   $q_{m0}$ - амплітуда в початковий момент часу ($t=0$).
    *   $e = 2.718...$ - експонента, основа натурального логарифма.
    *   $e^{-\beta t}$ - експоненціальна функція, яка швидко спадає з часом. Чим більший коефіцієнт затухання $\beta$, тим швидше спадає амплітуда.
*   $\cos(\omega t + \varphi_0)$ - це **коливальна частина**. Вона показує, що заряд все ще змінюється за гармонічним законом, але "вписаний" у спадаючу експоненту.
*   $\omega$ - це **циклічна частота затухаючих коливань**. Важливо: це вже **не** власна частота $\omega_0$.
*   $\varphi_0$ - початкова фаза коливань.

Отже, ми можемо сказати, що амплітуда затухаючих коливань $A(t)$ змінюється за законом:

$$A(t) = q_0 e^{-\beta t}$$

Графік функції $q(t)$ виглядає як косинусоїда, амплітуда якої обмежена двома кривими $A_1(t) = q_{m0} e^{-\beta t}$ та $A_2(t) = -q_{m0} e^{-\beta t}$.

## Період та частота затухаючих коливань

Виявляється, що наявність опору не лише зменшує амплітуду, але й трохи "сповільнює" самі коливання. Циклічна частота затухаючих коливань $\omega$ менша за власну частоту $\omega_0$ і пов'язана з нею наступним співвідношенням:

$$
\omega = \sqrt{\omega_0^2 - \beta^2}
$$

З цієї формули видно, що $\omega < \omega_0$. Це логічно: "тертя" в системі заважає їй коливатися так само швидко, як і без нього.

Тепер ми можемо легко знайти **період затухаючих коливань** $T$. За означенням, $T = \frac{2\pi}{\omega}$. Підставимо наш вираз для $\omega$:

$$
T = \frac{2\pi}{\sqrt{\omega_0^2 - \beta^2}}
$$

А тепер повернемося до початкових параметрів контуру $R, L, C$, підставивши $\omega_0^2 = \frac{1}{LC}$ та $\beta = \frac{R}{2L}$:

$$
T = \frac{2\pi}{\sqrt{\frac{1}{LC} - \left(\frac{R}{2L}\right)^2}} = \frac{2\pi}{\sqrt{\frac{1}{LC} - \frac{R^2}{4L^2}}}
$$


Ця формула є узагальненням формули Томсона для реального контуру. Ми бачимо, що період затухаючих коливань залежить не тільки від $L$ та $C$, але й від активного опору $R$. Чим більший опір, тим більший період коливань.

Якщо ж опір $R$ дорівнює нулю (ідеальний контур), то $\beta = 0$, і ця формула перетворюється на знайому нам формулу Томсона:

$$
T = \frac{2\pi}{\sqrt{\frac{1}{LC}}} = 2\pi\sqrt{LC}
$$

**Висновок:** активний опір в коливальному контурі призводить до двох наслідків:
1.  Амплітуда коливань експоненціально зменшується з часом.
2.  Період коливань збільшується (частота зменшується).

## Умови виникнення коливань

А чи завжди в RLC-контурі будуть виникати коливання? Подивімось ще раз на формулу для частоти: $\omega = \sqrt{\omega_0^2 - \beta^2}$.

Коливання (зміна знаку величини) можливі лише тоді, коли частота $\omega$ є дійсним числом. Для цього вираз під коренем має бути додатним:

$$
\omega_0^2 - \beta^2 > 0
$$

$$
\omega_0^2 > \beta^2
$$

Підставимо визначення $\omega_0$ і $\beta$:

$$
\frac{1}{LC} > \left(\frac{R}{2L}\right)^2
$$

$$
\frac{1}{LC} > \frac{R^2}{4L^2}
$$

Давайте виразимо звідси умову для опору $R$. Помножимо обидві частини на $4L^2$:

$$
\frac{4L^2}{LC} > R^2 \implies \frac{4L}{C} > R^2
$$

Звідси отримуємо **умову виникнення коливального процесу**:

$$
R < 2\sqrt{\frac{L}{C}}
$$

Тобто, коливання в контурі виникатимуть лише тоді, коли активний опір є достатньо малим.

А що станеться, якщо ця умова не виконується?

### Режими роботи RLC-контуру

Розглянемо граничний випадок, коли $\omega_0^2 = \beta^2$. Це відбувається при певному значенні опору, яке називають **критичним опором**.

> **Критичний опір** $R_{кр}$ - це найменше значення активного опору, за якого вільні коливання в контурі стають неможливими.
>
> $$ R_{кр} = 2\sqrt{\frac{L}{C}} $$

Залежно від співвідношення між $R, L$ та $C$ можливі три режими роботи контуру.

1.  **Коливальний (слабке затухання)**: $R < 2\sqrt{\frac{L}{C}}$.

    Це випадок, який ми щойно детально розглянули. У контурі відбуваються затухаючі коливання. Енергія поступово розсіюється, але система встигає зробити кілька коливань, перш ніж зупинитись.

2.  **Критичний**: $R = R_{кр} = 2\sqrt{\frac{L}{C}}$.

    Якщо опір дорівнює цьому **критичному значенню**, то вираз під коренем для частоти стає рівним нулю. Коливань як таких уже немає. Система повертається до стану рівноваги (конденсатор розряджається) **найшвидшим можливим способом**, не переходячи через нуль (тобто не перезаряджаючись). Такий режим важливий у системах, де потрібно швидко погасити коливання, наприклад, у стрілочних амперметрах або автомобільних амортизаторах.

3.  **Аперіодичний (сильне затухання)**: $R > 2\sqrt{\frac{L}{C}}$.

    Якщо опір дуже великий, "тертя" в системі настільки сильне, що коливання не можуть виникнути в принципі. Конденсатор просто повільно розряджається через котушку та резистор, і заряд плавно спадає до нуля. Цей процес називається **аперіодичним**. Чим більший опір, тим повільніше відбувається розрядка. Вся початкова енергія дуже швидко перетворюється на тепло, "не встигаючи" перейти в енергію магнітного поля і назад.

## Характеристики затухання

Для кількісної оцінки швидкості затухання коливань використовують дві важливі характеристики.

### Логарифмічний декремент затухання

Наскільки сильно зменшується амплітуда за один період? Щоб відповісти на це питання, вводять поняття логарифмічного декремента затухання.

> **Логарифмічний декремент затухання** $\lambda$ (лямбда) - це натуральний логарифм відношення двох послідовних амплітуд коливань, моменти часу яких відрізняються на один період.
>
> $$ \lambda = \ln \frac{A(t)}{A(t+T)} $$

Давайте виведемо, чому він дорівнює. Ми знаємо, що $A(t) = q_0 e^{-\beta t}$ і $A(t+T) = q_0 e^{-\beta (t+T)}$. Підставимо це у формулу:

$$\lambda = \ln \frac{q_0 e^{-\beta t}}{q_0 e^{-\beta (t+T)}} = \ln \frac{e^{-\beta t}}{e^{-\beta t} \cdot e^{-\beta T}} = \ln \frac{1}{e^{-\beta T}} = \ln(e^{\beta T})$$

Оскільки натуральний логарифм і експонента є взаємно оберненими функціями, ми отримуємо:

$$\lambda = \beta T$$

Логарифмічний декремент є безрозмірнісною величиною і показує, у скільки разів (за експоненційною шкалою $e$) зменшується амплітуда за один період. Чим він більший, тим швидше затухають коливання.

### Добротність коливального контуру

Інша, ще більш поширена характеристика, - це добротність. Вона показує, наскільки "якісною" є коливальна система, тобто наскільки повільно вона втрачає енергію.

> **Добротність** $Q$ (або Q-фактор) - це величина, що характеризує резонансні властивості системи та показує, у скільки разів енергія, накопичена в контурі, більша за втрати енергії за одне коливання. Висока добротність означає низькі втрати та велику кількість коливань до повного затухання.

Для слабкого затухання (коли $\beta \ll \omega_0$, що виконується для більшості радіотехнічних контурів) добротність пов'язана з логарифмічним декрементом простим співвідношенням:

$$Q \approx \frac{\pi}{\lambda}$$

Підставивши $\lambda = \beta T$ і $T \approx T_0 = \frac{2\pi}{\omega_0}$, отримуємо:

$$
Q = \frac{\pi}{\beta \cdot \frac{2\pi}{\omega_0}} = \frac{\omega_0}{2\beta}
$$

Тепер підставимо вирази для $\omega_0$ і $\beta$:
$$
Q = \frac{1/\sqrt{LC}}{2 \cdot (R/2L)} = \frac{1/\sqrt{LC}}{R/L} = \frac{L}{R\sqrt{LC}} = \frac{\sqrt{L^2}}{R\sqrt{LC}} = \frac{1}{R}\sqrt{\frac{L}{C}}
$$
> $$ Q = \frac{1}{R}\sqrt{\frac{L}{C}} $$

Добротність - безрозмірнісна величина.
*   **Висока добротність ($Q \gg 1$)** означає малий опір $R$, повільне затухання і гострий резонанс. Такі контури добре "відгукуються" на коливання лише на одній певній частоті, що є критично важливим для радіоприймачів.
*   **Низька добротність ($Q \approx 1$)** означає сильне затухання, коливання зникають дуже швидко.

Добротність показує, скільки приблизно коливань здійснить система, перш ніж її амплітуда зменшиться в $e^\pi \approx 23$ рази.

## Підсумок

| Термін / Твердження | Формула / Позначення | Опис |
| --- | --- | --- |
| **Активний опір** | $R$ | Характеризує властивість провідника перетворювати електричну енергію на теплову, що призводить до затухання коливань. |
| **RLC-контур** | - | Реальний коливальний контур, що складається з резистора R, котушки індуктивності L та конденсатора C. |
| **Диференціальне рівняння затухаючих коливань** | $q'' + \frac{R}{L}q' + \frac{1}{LC}q = 0$ | Описує зміну заряду в RLC-контурі з часом. |
| **Власна циклічна частота** | $\omega_0 = \frac{1}{\sqrt{LC}}$ | Частота коливань в контурі за відсутності опору (R=0 Ом). |
| **Коефіцієнт затухання** | $\beta = \frac{R}{2L}$ | Характеризує швидкість зменшення амплітуди коливань. |
| **Стандартний вигляд рівняння коливань** | $q'' + 2\beta q' + \omega_0^2 q = 0$ | Компактний запис диференціального рівняння через введені параметри. |
| **Заряд при затухаючих коливаннях** | $q(t) = q_{m0} e^{-\beta t} \cos(\omega t + \varphi_0)$ | Показує, що заряд змінюється за гармонічним законом з амплітудою, що спадає експоненційно. |
| **Амплітуда затухаючих коливань** | $A(t) = q_{m0} e^{-\beta t}$ | Зменшується з часом за експоненціальним законом. |
| **Циклічна частота затухаючих коливань** | $\omega = \sqrt{\omega_0^2 - \beta^2}$ | Частота коливань в реальному контурі; вона менша за власну частоту. |
| **Період затухаючих коливань** | $T = \frac{2\pi}{\sqrt{\omega_0^2 - \beta^2}} = \frac{2\pi}{\sqrt{\frac{1}{LC} - \frac{R^2}{4L^2}}}$ | Залежить не тільки від L і C, але й від R. Більший за період вільних коливань. |
| **Умова виникнення коливань** | $R < 2\sqrt{\frac{L}{C}}$ | Коливання виникають лише при достатньо малому опорі. |
| **Критичний опір** | $R_{кр} = 2\sqrt{\frac{L}{C}}$ | Найменший опір, за якого коливальний процес стає неможливим. |
| **Коливальний режим** | $R < R_{кр}$ | У контурі відбуваються затухаючі коливання. |
| **Критичний режим** | $R = R_{кр}$ | Система повертається до рівноваги найшвидше, без коливань. |
| **Аперіодичний режим** | $R > R_{кр}$ | Конденсатор повільно розряджається без коливань. |
| **Логарифмічний декремент затухання** | $\lambda = \ln \frac{A(t)}{A(t+T)} = \beta T$ | Безрозмірна величина, що показує, у скільки разів ($e^\lambda$) зменшується амплітуда за один період. |
| **Добротність контуру (Q-фактор)** | $Q = \frac{1}{R}\sqrt{\frac{L}{C}}$ | Характеризує якість коливальної системи; показує, наскільки повільно вона втрачає енергію. Висока добротність означає слабке затухання. |
| **Зв'язок добротності та затухання** | $Q \approx \frac{\pi}{\lambda} = \frac{\omega_0}{2\beta}$ | Добротність обернено пропорційна коефіцієнту затухання. |
