# Закон всесвітнього тяжіння

## Гравітаційна взаємодія

З давніх-давен люди спостерігали, що предмети падають на землю. Але лише Ісаак Ньютон першим усвідомив, що сила, яка змушує яблуко падати на землю, і сила, яка утримує Місяць на орбіті навколо Землі, - це одна й та сама сила.

**Гравітаційна взаємодія** - це універсальна взаємодія, яка існує між будь-якими об'єктами, що мають масу. Простими словами, всі тіла у Всесвіті притягуються одне до одного. Ця сила завжди є силою притягання, вона ніколи не буває силою відштовхування.

### Гравітаційне поле

Як же тіла "відчувають" одне одного на відстані? Сонце знаходиться за 150 мільйонів кілометрів, але Земля відчуває його притягання. Для пояснення цього явища ми використовуємо поняття **поля**.

**Гравітаційне поле** - це особливий вид матерії, який створюється будь-яким тілом, що має масу, і проявляється в тому, що діє з певною силою на будь-яке інше тіло, внесене в це поле.

**Аналогія:** Уявіть, що ми натягнули велике гумове полотно. Якщо ми покладемо на нього важку кулю (наприклад, для боулінгу), полотно прогнеться. Це викривлення і є своєрідним аналогом гравітаційного поля. Тепер, якщо ми запустимо поблизу цієї великої кулі маленьку кульку (наприклад, скляну), вона не буде рухатись прямо, а скочуватиметься у викривлену ямку. Саме так масивні тіла "диктують" іншим тілам, як їм рухатись (хоча і менш масивні тіла так само впливають на більш масивні, їхній вплив зазвичай є меншим через меншу масу). Однак варто пам'ятати, що це лише двовимірна аналогія для тривимірного простору. Крім того, в цій аналогії викривлення створюється силою тяжіння Землі чи іншим масивним тілом, тобто ми пояснюємо гравітацію через гравітацію. Проте, як візуальна модель, ця аналогія є надзвичайно ефективною.

## Закон всесвітнього тяжіння

### Встановлення залежностей

Закон всесвітнього тяжіння - це фундаментальний закон, який кількісно описує гравітаційну взаємодію. Давайте розберемося, як Ньютон дійшов до нього.

Довгий час люди спостерігали за рухом планет та їх супутників і падінням тіл на Землі, але вважали це абсолютно різними явищами. Геніальність Ісаака Ньютона полягала в тому, що він об'єднав ці два явища одним універсальним законом.

Легенда про яблуко, що впало на голову Ньютону, є спрощенням, але вона чудово ілюструє хід його думок. Ньютон поставив собі питання: якщо яблуко падає на Землю, то чи "падає" на неї й Місяць? Відповідь - так! Якби на Місяць не діяла сила, він би за першим законом Ньютона рухався прямолінійно і рівномірно і давно полетів би від Землі. Оскільки він рухається по криволінійній (майже коловій) орбіті, це означає, що на нього постійно діє сила, спрямована до центру Землі. Ця сила змушує його постійно "падати" на Землю, але через свою величезну швидкість він ніколи не досягає поверхні, а залишається на орбіті.

Далі Ньютон почав міркувати математично.

1.  **Залежність від відстані.** Він порівняв прискорення, з яким падає яблуко на поверхні Землі $g \approx 9.8 \text{ м/с}^2$, з доцентровим прискоренням Місяця на його орбіті.

    * Місяць рухається по коловій орбіті, отже, на нього діє доцентрова сила, яка і створює доцентрове прискорення $a_{\text{доц}}$. Це прискорення можна обчислити за формулою:

        $$a_{\text{доц}} = \frac{v^2}{R_M}$$

        де $v$ - швидкість Місяця, а $R_M$ - радіус його орбіти. Швидкість можна знайти як довжину орбіти, поділену на період обертання $T$: 
        
        $$v = \frac{2\pi R_M}{T}$$

        Підставивши це у формулу прискорення, отримуємо:
        
        $$a_{\text{доц}} = \frac{(2\pi R_M/T)^2}{R_M} = \frac{4\pi^2 R_M}{T^2}$$

    * Ньютон знав, що радіус орбіти Місяця приблизно в 60 разів більший за радіус Землі $R_M \approx 60 R_З$, а період обертання Місяця $T \approx 27.3$ доби. Провівши розрахунки, він отримав, що доцентрове прискорення Місяця $a_{\text{доц}} \approx 0.00272 \, \text{м}/\text{с}^2$.

    * Далі він порівняв це прискорення з прискоренням вільного падіння біля поверхні Землі (тобто з прискоренням вільного падіння на відстані радіусу Землі $R_З$ від центру Землі):

        $$\frac{g}{a_{\text{доц}}} \approx \frac{9.8}{0.00272} \approx 3600$$

        Оскільки $3600 = 60^2$, це означає, що

        $$\frac{a_{\text{доц}}}{g} = \frac{1}{60^2} = \left(\frac{R_З}{R_M}\right)^2$$

        тобто прискорення (а отже, і сила) зменшується обернено пропорційно **квадрату відстані** від центру Землі!


    * Підсумовуємо цей результат. Обчисливши доцентрове прискорення Місяця (знаючи радіус його орбіти та період обертання), Ньютон виявив, що воно приблизно в $3600$ разів менше за прискорення яблука біля поверхні Землі. Він помітив, що $3600 = 60^2$. Це привело його до висновку: сила гравітації зменшується обернено пропорційно **квадрату відстані** між тілами:

        $$F \propto \frac{1}{r^2}$$

2.  **Залежність від маси.** Згідно з другим законом Ньютона, сила пропорційна масі тіла, на яке вона діє $F \propto m_1$. Але згідно з третім законом Ньютона, дія дорівнює протидії. Якщо Земля притягує яблуко, то і яблуко з такою ж силою притягує Землю. Отже, сила має бути пропорційна і масі другого тіла $F \propto m_2$.

    Об'єднавши ці два висновки, Ньютон отримав, що сила пропорційна добутку мас і обернено пропорційна квадрату відстані:

    $$F \propto \frac{m_1 m_2}{r^2}$$

    Щоб перетворити пропорційність на рівність, ми вводимо коефіцієнт пропорційності - гравітаційну сталу $G$.

### Формулювання та формула

**Закон всесвітнього тяжіння:** Два будь-які тіла притягуються одне до одного із силою, прямо пропорційною добутку їхніх мас і обернено пропорційною квадрату відстані між ними.

Математично цей закон записується так:

$$
F = G \frac{m_1 m_2}{r^2}
$$

де:
*   $F$ - сила всесвітнього тяжіння. Одиниця вимірювання в СІ: **ньютон (Н)**.
*   $m_1$ та $m_2$ - маси тіл, що взаємодіють. Одиниця вимірювання в СІ: **кілограм (кг)**.
*   $r$ - відстань між центрами мас тіл. Одиниця вимірювання в СІ: **метр (м)**.
*   $G$ - гравітаційна стала. Одиниця вимірювання в СІ: $\frac{\text{Н} \cdot \text{м}^2}{\text{кг}^2}$

### Гравітаційна стала та її фізичний зміст

**Гравітаційна стала** $G$ - це одна з фундаментальних фізичних констант. Її значення було вперше точно виміряно Генрі Кавендішем у 1798 році за допомогою геніального експерименту з крутильними терезами, який дозволив виміряти надзвичайно слабку силу притягання між свинцевими кулями в лабораторії.

Її числове значення:

$$
G \approx 6.67 \cdot 10^{-11} \frac{\text{Н} \cdot \text{м}^2}{\text{кг}^2}
$$

**Фізичний зміст гравітаційної сталої:** гравітаційна стала чисельно дорівнює силі, з якою притягуються два тіла масою 1 кг кожне, перебуваючи на відстані 1 м одне від одного.

Маленьке значення $G$ пояснює, чому ми не помічаємо гравітаційного притягання між повсякденними предметами (наприклад, між двома людьми або між вами та вашим столом). Ця сила надзвичайно мала. Вона стає значною лише тоді, коли хоча б одне з тіл має астрономічну масу (як планета чи зірка).

### Дослід Генрі Кавендіша

Значення гравітаційної сталої було вперше експериментально визначено англійським вченим Генрі Кавендішем у 1798 році. Він використав для цього дуже чутливий прилад - крутильні терези. У його експерименті дві маленькі свинцеві кульки, закріплені на коромислі, підвішеному на тонкій нитці, притягувалися до двох великих нерухомих свинцевих куль. За кутом закручування нитки Кавендіш зміг обчислити силу притягання і, знаючи маси куль та відстань між ними, визначити значення $G$. Цей експеримент часто називають "зважуванням Землі", оскільки, знаючи $G$, стало можливим обчислити масу нашої планети.

### Межі застосування закону

Закон всесвітнього тяжіння працює з неймовірною точністю в більшості випадків, але має свої межі:

1. Закон сформульовано для **матеріальних точок**. Він також є точним для тіл, що мають форму кулі з рівномірним розподілом маси (або розподілом, що залежить лише від радіуса). У такому випадку $r$ - це відстань між центрами цих куль. Саме тому він добре описує рух планет.
2. Закон всесвітнього тяжіння Ньютона є частинним випадком більш загальної теорії - **загальної теорії відносності** Альберта Ейнштейна. Теорія Ейнштейна описує гравітацію не як силу, а як викривлення простору-часу масивними об'єктами. Однак в умовах слабких гравітаційних полів і швидкостей, набагато менших за швидкість світла (тобто в майже всіх ситуаціях, з якими ми стикаємось), закон Ньютона дає надзвичайно точні результати.
3. У дуже сильних гравітаційних полях (біля чорних дір або поблизу масивних зір) або при швидкостях, близьких до швидкості світла, розрахунки за законом Ньютона стають неточними. У таких випадках використовується **Загальна теорія відносності** Альберта Ейнштейна.

## Сила тяжіння та прискорення вільного падіння

Давайте тепер розглянемо найближчий до нас прояв закону всесвітнього тяжіння.

### Сила тяжіння

**Сила тяжіння** $\vec{F}_{\text{тяж}}$ - це сила, з якою планета (або інше масивне небесне тіло) притягує до себе тіла, що знаходяться на її поверхні або поблизу неї.

По суті, це просто окремий випадок закону всесвітнього тяжіння. Давайте виведемо формулу.

Нехай маємо тіло масою $m$, яке знаходиться на висоті $h$ над поверхнею Землі. Маса Землі - $M_З$, а її радіус - $R_З$. Відстань від центру Землі до тіла буде $r = R_З + h$.

Підставимо ці значення в закон всесвітнього тяжіння:

$$
F_{\text{тяж}} = G \frac{m M_З}{(R_З + h)^2}
$$

Це і є формула для розрахунку сили тяжіння, що діє на тіло на будь-якій висоті $h$ над поверхнею Землі.

### Прискорення вільного падіння

З другого закону Ньютона ми знаємо, що сила надає тілу прискорення $F = ma$. Сила тяжіння також надає тілу прискорення. Це прискорення називається **прискоренням вільного падіння** і позначається як $g$.

**Прискорення вільного падіння** $g$ - це прискорення, якого набуває тіло, рухаючись під дією лише сили тяжіння. Одиниця вимірювання в СІ: **метр на секунду в квадраті ($\text{м}\text{с}^2$)**.

Отже, ми можемо записати $F_{\text{тяж}} = mg$. Тепер прирівняємо два вирази для сили тяжіння:

$$
mg = G \frac{M_З m}{(R_З + h)^2}
$$

Як ми бачимо, маса тіла $m$ скорочується! Це надзвичайно важливий висновок: **прискорення вільного падіння не залежить від маси падаючого тіла**. Це означає, що в вакуумі (де немає опору повітря) і пір'їнка, і камінь падатимуть з однаковим прискоренням.

Після скорочення маси $m$ ми отримуємо формулу для прискорення вільного падіння на висоті $h$:

$$
g = G \frac{M_З}{(R_З + h)^2}
$$

Якщо тіло знаходиться поблизу поверхні Землі, то висота $h$ є дуже малою в порівнянні з радіусом Землі $R_З$ ($R_З \approx 6370 \text{ км}$), тому ми можемо знехтувати $h$ і вважати $h \approx 0$. Тоді формула спрощується:

$$
g_0 = G \frac{M_З}{R_З^2}
$$

Це формула для прискорення вільного падіння біля поверхні Землі. Підставивши значення $G$, $M_З$ та $R_З$, ми отримаємо добре відоме нам значення:

$$g_0 \approx 9.8 \text{ м/с}^2$$

Примітка: значення $9.8 \text{ м/с}^2$ є усередненим. У конкретній точці Землі прискорення вільного падіння може трохи відрізнятись від цього значення.

Аналогічно отримується формула для прискорення вільного падіння для довільної планети:

$$
g_0 = G \frac{M}{(R + h)^2}
$$

або за умови $h \approx 0$

$$
g_0 = G \frac{M}{R^2}
$$

де $M$ - маса планети та $R$ - радіус планети.

### Як ми "зважили" Землю?

А тепер цікавий наслідок. Подивіться на останню формулу. Ми можемо виміряти $g_0$, ми можемо виміряти радіус Землі $R_З = 6370 \text{ км}$ (це зробили ще в давнину), а $G$ ми знаємо з досліду Кавендіша. Отже, ми можемо обчислити масу Землі, не ставлячи її на ваги!

З формули $g_0 = G \frac{M_З}{R_З^2}$ виразимо масу Землі $M_З$:

$$
M_З = \frac{g_0 R_З^2}{G}
$$

Саме тому дослід Кавендіша часто називають "зважуванням Землі".

## Перша космічна швидкість

Гравітація не тільки змушує тіла падати, але й дозволяє їм літати - по орбіті. Давайте розберемося, як це можливо. Уявимо собі високу гору і гармату на її вершині (це уявний експеримент Ньютона).

*   Якщо вистрілити з невеликою швидкістю, ядро пролетить невелику відстань і впаде на Землю.
*   Якщо збільшити швидкість, воно полетить далі.
*   При досягненні певної швидкості, ядро полетить так далеко, що траєкторія його падіння "повторить" кривизну земної поверхні. Воно почне падати, але поверхня Землі буде "втікати" від нього з тією ж швидкістю. Ядро вийде на колову орбіту і стане штучним супутником Землі.

> **Перша космічна швидкість** $v$ - це мінімальна швидкість, яку необхідно надати тілу на певній висоті над поверхнею планети, щоб воно стало її супутником, рухаючись по коловій орбіті.

Щоб тіло рухалось по колу, на нього має діяти доцентрова сила. У випадку супутника на орбіті роль доцентрової сили виконує сила тяжіння. Доцентрова сила визначається як добуток маси тіла $m$ та доцентрового прискорення $a = v^2 / r$:

$$F_{доц} = \frac{mv^2}{r}$$

Сила тяжіння на висоті $h$ (де $r = R + h$) визначається як 

$$F_{\text{тяж}} = G \frac{m M}{(R + h)^2}$$

де $M$ - маса планети та $R$ - радіус планети.

Прирівняємо ці дві сили:

$$
\frac{mv^2}{R + h} = G \frac{m M}{(R + h)^2}
$$

Скорочуємо масу супутника $m$ і один раз скорочуємо $(R + h)$:

$$
v^2 = G \frac{M}{R + h}
$$

Звідси отримуємо формулу для першої космічної швидкості на висоті $h$ над поверхнею планети (довільної планети, не обов'язково Землі):

$$
v = \sqrt{G \frac{M}{R + h}}
$$

Для випадків руху поблизу поверхні планети $h \approx 0$ формула для першої космічної швидкості набуває вигляду:

$$
v = \sqrt{G \frac{M}{R}}
$$

### Спрощена формула та значення для Землі

Ми можемо зробити ще одне перетворення. Згадаймо, що прискорення вільного падіння біля поверхні планети $g_0 = G \frac{M}{R^2}$. Звідси $GM = g_0 R^2$. Підставимо це в нашу формулу для першої космічної швидкості $v$:

$$
v = \sqrt{\frac{g_0 R^2}{R}} = \sqrt{g_0 R}
$$

Це дуже зручна формула для обчислень. Давайте розрахуємо першу космічну швидкість для Землі:

$$
v = \sqrt{g_0 R_З}
$$

$$v = \sqrt{9.8 \text{ м}/\text{с}^2 \cdot 6.37 \cdot 10^6 \text{ м}} \approx 7900 \text{ м/с} \approx 28440 \text{ км/год}$$

Отже, перша космічна швидкість для Землі становить приблизно **7.9 км/с**.

## Друга та третя космічні швидкості

Якщо ми будемо збільшувати швидкість тіла, його орбіта з колової стане еліптичною. При досягненні певної швидкості тіло зможе подолати гравітаційне тяжіння планети і назавжди покинути її, перетворившись на супутник Сонця.

> **Друга космічна швидкість** $v_2$ - це найменша швидкість, яку необхідно надати тілу, щоб воно подолало гравітаційне поле планети.

Вона пов'язана з першою космічною швидкістю простою формулою:

$$
v_2 = \sqrt{2} \cdot v_1
$$

де $v_1$ - перша космічна швидкість. Ця формула виводиться із закону збереження енергії, за яким кінетична енергія тіла має бути достатньою для подолання гравітаційної потенціальної енергії планети (на наступних сторінках ми розглянемо, що таке кінетична та потенціальна енергії).

Для Землі значення другої космічної швидкості становить 

$$v_2 \approx 1.4142 \cdot 7.9 \text{ км/с} \approx \textbf{11.2 км/с}$$

А що, якщо ми хочемо покинути не тільки Землю, але і всю Сонячну систему? Для цього потрібно подолати тяжіння Сонця.

> **Третя космічна швидкість** $v_3$ - це мінімальна швидкість, яку необхідно надати тілу біля поверхні Землі, щоб воно покинуло Сонячну систему.

Її значення залежить від напрямку запуску тіла з планети. При оптимальному запуску тіла із Землі значення третьої космічної швидкості становить приблизно **16.7 км/с**.
