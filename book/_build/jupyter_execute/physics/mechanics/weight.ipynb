# Вага тіла

Після вивчення всесвітнього тяжіння та сили тяжіння, саме час перейти до поняття ваги. На перший погляд, вага і сила тяжіння - це одне й те саме, але насправді між ними є принципова різниця. Сьогодні ми детально розберемо, що таке вага, яка її фізична природа, і як вона може змінюватись, створюючи такі явища, як невагомість та перевантаження.

У повсякденному житті ми часто плутаємо поняття "маса" та "вага", кажучи, наприклад, "моя вага 60 кілограмів". З точки зору фізики це некоректно, адже маса вимірюється в кілограмах, а вага, як і будь-яка інша сила, - в ньютонах. Але ще важливіша плутанина виникає між вагою та силою тяжіння. Давайте розставимо все по місцях.


## Що таке вага тіла?

Уявімо, що на столі лежить книжка. Ми вже знаємо, що на неї діє **сила тяжіння** $m\vec{g}$ - це сила, з якою Земля притягує книжку до себе. Ця сила прикладена до центру мас книжки і напрямлена вертикально вниз.

Але чому книжка не падає крізь стіл? Тому що стіл чинить опір. Згідно з третім законом Ньютона, якщо книжка тисне на стіл, то і стіл тисне на книжку. Сила, з якою опора (стіл) діє на тіло (книжку), називається **силою нормальної реакції опори** $\vec{N}$. Вона напрямлена перпендикулярно (нормально) до поверхні опори, тобто в нашому випадку - вертикально вгору.

> **Сила нормальної реакції опори** $\vec{N}$ - це сила, що діє на тіло з боку опори і напрямлена перпендикулярно до поверхні дотику. Одиниця вимірювання в СІ: **ньютон (Н)**.

Тепер найголовніше. Книжка, внаслідок притягання до Землі, тисне на стіл. Ось ця сила і є вагою.

> **Вага тіла** $\vec{P}$ - це сила, з якою тіло внаслідок гравітаційного притягання діє на горизонтальну опору або вертикальний підвіс. Одиниця вимірювання в СІ: **ньютон (Н)**.

Аналогічно, якщо ми підвісимо на нитці якийсь вантаж, на нього діятиме сила тяжіння $m\vec{g}$ (вниз) і **сила натягу підвісу** $\vec{T}$ (вгору). Вагою в цьому випадку буде сила, з якою вантаж розтягує нитку.

> **Сила натягу підвісу** $\vec{T}$ - це сила, що діє на тіло з боку підвісу (нитки, троса) і напрямлена вздовж нього. Одиниця вимірювання в СІ: **ньютон (Н)**.

## Вага у стані спокою: причина плутанини

Давайте розглянемо тіло (наприклад, нашу книжку на столі), що перебуває у стані спокою. Прискорення тіла дорівнює нулю $\vec{a} = 0$. Запишемо для нього другий закон Ньютона у векторній формі:

$$
m\vec{a} = \vec{F}_{\text{тяж}} + \vec{N}
$$

Оскільки $\vec{a} = 0$, то $\vec{F}_{\text{тяж}} + \vec{N} = 0$, звідки випливає, що

$$
\vec{N} = - \vec{F}_{\text{тяж}} = -m\vec{g}
$$

Це означає, що сила нормальної реакції опори за модулем дорівнює силі тяжіння, але напрямлена в протилежний бік.

А що ж таке вага? Вага $\vec{P}$ - це сила, з якою *тіло діє на опору*. Сила нормальної реакції $\vec{N}$ - це сила, з якою *опора діє на тіло*. Згідно з третім законом Ньютона, ці дві сили рівні за модулем і протилежні за напрямком:

$$
\vec{P} = -\vec{N}
$$

Тепер об'єднаємо ці два результати. Якщо $\vec{N} = -m\vec{g}$, то:

$$
\vec{P} = -(-m\vec{g}) = m\vec{g}
$$

Отже, для нерухомого тіла (або тіла, що рухається рівномірно і прямолінійно) його вага $\vec{P}$ чисельно дорівнює силі тяжіння $m\vec{g}$. Саме тому ці поняття часто плутають.

Аналогічна ситуація виникає, коли тіло підвішене на нитці, тросі або пружині. Тіло розтягує підвіс (це його вага), а підвіс, у свою чергу, тягне тіло вгору. За третім законом Ньютона, сила, з якою тіло тягне підвіс (вага $\vec{P}$), і сила, з якою підвіс тягне тіло (сила натягу $\vec{T}$), є парою "дія-протидія". Тому

$$
\vec{P} = -\vec{T}
$$

**Важливий висновок:** Щоб знайти вагу тіла $\vec{P}$, нам потрібно знайти силу реакції опори $\vec{N}$ або силу натягу підвісу $\vec{T}$, що діє на саме тіло. Модуль ваги буде дорівнювати модулю цих сил $P = N$ або $P = T$.

### Ключова відмінність і природа сил

Незважаючи на рівність у стані спокою, вага і сила тяжіння - це принципово різні сили. Давайте підсумуємо їхні відмінності.

| Характеристика | Сила тяжіння $\vec{F}_{\text{тяж}} = m\vec{g}$ | Вага тіла $\vec{P}$ |
| :--- | :--- | :--- |
| **Точка прикладання** | Прикладена до **центру мас тіла** (Земля притягує тіло). | Прикладена до **опори або підвісу** (тіло діє на опору). |
| **Фізична природа** | **Гравітаційна**. Це прояв закону всесвітнього тяжіння. | **Електромагнітна**. Це сила, що виникає через деформацію опори/підвісу на атомному рівні. Наприклад, коли тіло тисне на опору, атоми тіла та опори зближуються, і виникають сили відштовхування між їхніми електронними оболонками. |
| **Умова існування** | Діє на будь-яке тіло поблизу планети, незалежно від його стану руху. | Виникає лише тоді, коли є опора або підвіс, тобто є контакт. |

**Аналогія**: Уявіть, що ви тиснете рукою на стіну. Сила, з якою ви тиснете на стіну, - це аналог ваги. Сила, з якою стіна тисне у відповідь на вашу руку, - це аналог сили реакції опори. А сила тяжіння - це щось зовсім інше, наприклад, невидима мотузка, що тягне вас до центру Землі, незалежно від того, чи торкаєтесь ви стіни.

## Вага в русі

Найцікавіше починається, коли тіло рухається з прискоренням. Розглянемо найкласичніший приклад: людина масою $m$ стоїть у ліфті, що рухається з прискоренням $\vec{a}$.

Запишемо другий закон Ньютона для людини. На неї діють дві сили: сила тяжіння $m\vec{g}$ (завжди вниз) і сила нормальної реакції з боку підлоги ліфта $\vec{N}$ (завжди вгору).

$$
m\vec{a} = m\vec{g} + \vec{N}
$$

Для зручності оберемо вісь OY, напрямлену вертикально вгору. Тоді проекції сил і прискорення на цю вісь будуть:
*   Проекція прискорення: $a_y = a$ (якщо прискорення напрямлене вгору) або $a_y = -a$ (якщо вниз).
*   Проекція сили нормальної реакції: $N_y = N$.
*   Проекція сили тяжіння: $F_{\text{тяж},y} = -mg$.

Рівняння другого закону Ньютона в проекціях набуває вигляду:

$$
ma_y = N - mg
$$

Звідси ми можемо виразити силу нормальної реакції опори:

$$
N = mg + ma_y = m(g + a_y)
$$

Оскільки вага за модулем дорівнює силі нормальної реакції $P = N$, ми отримуємо загальну формулу для ваги тіла, що рухається з прискоренням по вертикалі:

$$
P = m(g + a_y)
$$

А тепер проаналізуємо різні випадки.

### Випадок 1: Ліфт прискорюється вгору (Перевантаження)

Ліфт починає рух вгору або гальмує під час руху вниз. У будь-якому з цих випадків вектор прискорення $\vec{a}$ напрямлений вгору. Тому його проекція на вісь OY є додатною: $a_y = a > 0$.

Підставляємо це в нашу формулу:

$$
P = m(g + a)
$$

Оскільки $a > 0$, то $(g + a) > g$, а отже, $P > mg$.

**Висновок:** Вага людини збільшилася. Цей стан називається **перевантаженням**.

Перевантаження відчувають пілоти та космонавти під час старту ракети, тільки набагато сильніше, ніж людина у ліфті.

Для характеристики перевантаження використовують коефіцієнт перевантаження $n$, який показує, у скільки разів вага тіла перевищує "нормальну" вагу (силу тяжіння):

$$
n = \frac{P}{mg} = \frac{m(g+a)}{mg} = 1 + \frac{a}{g}
$$

Якщо кажуть "пілот зазнав перевантаження 5g", це означає, що його вага в 5 разів перевищувала звичайну.

### Випадок 2: Ліфт прискорюється вниз

Ліфт починає рух вниз або гальмує під час руху вгору. Тепер вектор прискорення $\vec{a}$ напрямлений вниз.

Його проекція на вісь OY (яка напрямлена вгору) є від'ємною: $a_y = -a < 0$.

Підставляємо у формулу:

$$
P = m(g + (-a)) = m(g - a)
$$

Оскільки $a > 0$, то $(g - a) < g$, а отже, $P < mg$.

**Висновок:** Вага людини зменшилася. Коли ви в ліфті починаєте спускатись, ви відчуваєте короткочасне відчуття легкості.

### Випадок 3: Невагомість

Якщо трос ліфта обірветься, ліфт і всі тіла в ньому будуть падати вільно. Людина всередині не буде тиснути на підлогу, а предмети будуть "літати" по кабіні. Це і є невагомість. У такому випадку ліфт і все, що в ньому, почне **вільне падіння**. Прискорення ліфта буде дорівнювати прискоренню вільного падіння: $\vec{a} = \vec{g}$.

Вектор прискорення $\vec{a}$ напрямлений вниз, і його модуль дорівнює $g$. Проекція на вісь OY буде $a_y = -g$. Підставимо це значення в нашу формулу:

$$
P = m(g + (-g)) = m(g - g) = 0
$$

**Висновок:** Вага людини дорівнює нулю! Цей стан називається **невагомістю**.

> **Невагомість** - це стан, за якого вага тіла дорівнює нулю. Тіло перебуває в стані невагомості, коли воно рухається під дією лише сили тяжіння (тобто вільно падає).

**Важливо:** У стані невагомості сила тяжіння нікуди не зникає! Вона продовжує діяти. Саме вона і надає тілу прискорення $\vec{g}$. Просто тіло і його опора (кабіна ліфта, космічний корабель) падають разом, з однаковим прискоренням. Через це тіло не тисне на опору, і його вага стає нульовою. Космічні станції на орбіті також перебувають у стані постійного вільного падіння на Землю (вони постійно "падають" і "промахуються" повз неї через свою величезну горизонтальну швидкість), тому всередині них і панує невагомість.

## Загальна векторна формула для ваги

Ми можемо узагальнити всі наші міркування в одній елегантній векторній формулі. Повернімося до другого закону Ньютона:

$$
m\vec{a} = m\vec{g} + \vec{N}
$$

Виразимо звідси силу нормальної реакції опори:

$$
\vec{N} = m\vec{a} - m\vec{g}
$$

Оскільки вага $\vec{P}$ пов'язана з силою реакції співвідношенням $\vec{P} = -\vec{N}$, то:

$$
\vec{P} = -(m\vec{a} - m\vec{g}) = m\vec{g} - m\vec{a}
$$

Таким чином, ми отримали загальну векторну формулу для ваги:

$$
\vec{P} = m(\vec{g} - \vec{a})
$$

Ця формула працює для будь-якого напрямку прискорення. Давайте перевіримо:
*   **Спокій ($\vec{a}=0$):** $\vec{P} = m(\vec{g} - 0) = m\vec{g}$.
*   **Прискорення вгору:** вектор $\vec{a}$ протилежний до $\vec{g}$. Різниця векторів $(\vec{g} - \vec{a})$ дасть вектор, більший за модулем, ніж $\vec{g}$. Вага збільшується.
*   **Прискорення вниз:** вектор $\vec{a}$ співнапрямлений з $\vec{g}$. Різниця векторів $(\vec{g} - \vec{a})$ дасть вектор, менший за модулем, ніж $\vec{g}$. Вага зменшується.
*   **Вільне падіння ($\vec{a}=\vec{g}$):** $\vec{P} = m(\vec{g} - \vec{g}) = 0$. Невагомість.

## Підсумок

| Поняття / Ситуація | Опис / Характеристика | Формула |
| :--- | :--- | :--- |
| **Сила тяжіння** | Сила, з якою Земля притягує тіло. Прикладена до центру мас тіла, має гравітаційну природу. | $\vec{F}_{\text{тяж}} = m\vec{g}$ |
| **Вага тіла** | Сила, з якою тіло діє на опору або підвіс. Прикладена до опори/підвісу, має електромагнітну природу. | $\vec{P}$ |
| **Зв'язок ваги та реакції опори** | За третім законом Ньютона, вага $\vec{P}$ і сила реакції опори $\vec{N}$ (або сила натягу $\vec{T}$) рівні за модулем і протилежні за напрямком. | $\vec{P} = -\vec{N}$ |
| **Стан спокою або рівномірного руху** | Прискорення дорівнює нулю $\vec{a} = 0$. Вага тіла за модулем дорівнює силі тяжіння. | $P = mg$ |
| **Рух з прискоренням (загальний випадок)** | Узагальнена векторна формула, що працює для будь-якого напрямку прискорення $\vec{a}$. | $\vec{P} = m(\vec{g} - \vec{a})$ |
| **Перевантаження (рух вгору з прискоренням)** | Вага тіла більша за силу тяжіння. Виникає, коли вектор прискорення $\vec{a}$ напрямлений протилежно до $\vec{g}$. | $P = m(g + a)$ |
| **Зменшення ваги (рух вниз з прискоренням)** | Вага тіла менша за силу тяжіння. Виникає, коли вектор прискорення $\vec{a}$ співнапрямлений з $\vec{g}$. | $P = m(g - a)$ |
| **Невагомість (вільне падіння)** | Стан, за якого вага тіла дорівнює нулю. Настає, коли тіло рухається лише під дією сили тяжіння. | $\vec{a} = \vec{g} \implies P = 0$ |
