# Механічна робота, кінетична енергія та потужність

Уявіть, що ми хочемо описати, як сили впливають на рух тіл. Ми вже знаємо про закони Ньютона, які пов'язують силу, масу та прискорення. Але є й інший, надзвичайно потужний підхід для опису руху - через поняття **роботи** та **енергії**. Цей підхід дозволяє нам не враховувати час або конкретний вид руху (наприклад, чи він був рівноприскореним), а дивитися на "початковий" та "кінцевий" стани системи.

Давайте почнемо з першого ключового поняття.

## Механічна робота

У повсякденному житті ми називаємо "роботою" будь-яку діяльність, що вимагає зусиль - розумових чи фізичних. Однак у фізиці це поняття має дуже чітке визначення.

### Що таке механічна робота?

> **Механічна робота $A$** -  це скалярна фізична величина, яка характеризує зміну стану тіла, що відбувається під дією сили і дорівнює скалярному добутку вектора сили $\vec{F}$ на вектор переміщення $\vec{s}$ точки прикладання цієї сили.

Якщо ви штовхаєте стіну, ви докладаєте силу, втомлюєтеся, але стіна не рухається. З точки зору фізики, ви не виконуєте жодної роботи, оскільки немає переміщення. А от якщо ви штовхаєте візок, і він котиться, ваша сила виконує роботу.

Робота залежить від трьох речей: величини прикладеної сили $F$, величини переміщення тіла $s$, кута $\alpha$ між напрямком сили та напрямком переміщення. Формула для роботи постійної сили виглядає так:

$$A = \vec{F} \cdot \vec{s}$$

або, якщо розписати скалярний добуток, ця формула набуває вигляду

$$
A = Fs\cos\alpha
$$

де:
*   $A$ - механічна робота.
*   $F$ - модуль сили, що діє на тіло.
*   $s$ - модуль переміщення тіла.
*   $\alpha$ - кут між вектором сили $\vec{F}$ та вектором переміщення $\vec{s}$.

Іншими словами, робота залежить не від усієї величини сили, а лише від тієї її частини (проєкції), яка спрямована вздовж руху тіла.

Одиницею вимірювання роботи в системі СІ є **Джоуль (Дж)**.

$$
1 \text{ Дж} = 1 \text{ Н} \cdot \text{м}
$$

Один джоуль - це робота, яку виконує сила в 1 Ньютон, переміщуючи тіло на 1 метр у напрямку дії сили.

### Коли робота додатна, від'ємна або дорівнює нулю?

Значення косинуса визначає знак роботи, що має глибокий фізичний зміст. Розглянемо три випадки.

**Додатна робота $A > 0$**

Робота є додатною, коли кут $\alpha$ є гострим ($0^\circ \le \alpha < 90^\circ$). У цьому випадку $\cos\alpha > 0$.
*   **Фізичний зміст**: Сила "допомагає" руху, спрямована в ту ж сторону, що й переміщення. Вона сприяє збільшенню швидкості тіла.
*   **Приклад**: Коли ми тягнемо санчата за мотузку, сила натягу мотузки виконує додатну роботу.

**Від'ємна робота $A < 0$**

Робота є від'ємною, коли кут $\alpha$ є тупим ($90^\circ < \alpha \le 180^\circ$). У цьому випадку $\cos\alpha < 0$.
*   **Фізичний зміст**: Сила "заважає" руху, спрямована протилежно до переміщення. Вона гальмує тіло, зменшуючи його швидкість.
*   **Приклад**: Сила тертя ковзання завжди спрямована проти руху, тому її робота завжди від'ємна. Вона "забирає" енергію руху.

**Робота дорівнює нулю $A = 0$**

Робота дорівнює нулю у двох основних випадках:
*   **Кут $\alpha = 90^\circ$**. У цьому випадку $\cos90^\circ = 0$.
    *   **Фізичний зміст**: Сила перпендикулярна до переміщення і ніяк не впливає на зміну швидкості.
    *   **Приклад**: Коли ми несемо сумку, тримаючи її в руці, і йдемо горизонтально. Сила, з якою ми підтримуємо сумку, спрямована вгору, а переміщення - горизонтально. Кут між ними $90^\circ$, тому робота цієї сили дорівнює нулю. Сила тяжіння, що діє на супутник на коловій орбіті, також не виконує роботи, бо вона завжди перпендикулярна до вектора швидкості (а отже, і до переміщення).
*   **Переміщення $s = 0$**.
    *   **Фізичний зміст**: Тіло не рухається.
    *   **Приклад**: Вже згаданий випадок, коли ми штовхаємо стіну. Сила є, а переміщення немає.

### Геометричний зміст роботи

Давайте поглянемо на формулу $A = Fs\cos\alpha$ інакше. Вираз $F\cos\alpha$ є **проекцією вектора сили на напрямок переміщення**. Позначимо цю проекцію як $F_s$ (або $F_x$, якщо рух відбувається вздовж осі Ox).

Тоді формулу роботи можна записати простіше:

$$
A = F_s \cdot s
$$

Це означає, що роботу виконує лише та складова сили, яка спрямована вздовж переміщення.

А тепер уявімо, що ми побудували графік залежності цієї проекції сили $F_s$ від переміщення $s$. Для найпростішого випадку, коли сила є сталою, цей графік буде прямою лінією, паралельною осі переміщення.

Площа прямокутника під цим графіком дорівнює добутку $F_s \cdot s$. Але ж це і є наша формула для роботи!

**Геометричний зміст роботи**: робота сили чисельно дорівнює площі фігури під графіком залежності проекції сили на напрямок руху від переміщення.

Цей висновок є надзвичайно корисним, коли сила не є сталою (наприклад, при розтягуванні пружини). У такому випадку робота все одно буде дорівнювати площі під криволінійним графіком $F_s(s)$.

## Кінетична енергія

Ми з'ясували, що виконання роботи над тілом може змінювати його швидкість. Величина, яка пов'язана зі швидкістю тіла, називається енергією.

> **Енергія** - це скалярна фізична величина, яка є загальною мірою руху та взаємодії тіл і характеризує здатність тіла (або системи тіл) виконувати роботу.

Енергія буває різних видів: теплова, електрична, ядерна, хімічна. У механіці ми виділяємо два основних види:
*   **Кінетична енергія** - енергія руху.
*   **Потенціальна енергія** - енергія взаємодії.

Зараз ми детально зупинимося на кінетичній енергії.

### Кінетична енергія – енергія руху

> **Кінетична енергія** - це скалярна фізична величина, що є частиною повної енергії системи і характеризує енергію її руху.

Будь-яке рухоме тіло має **кінетичну енергію** $E_k$. Ця енергія залежить від маси тіла та швидкості його руху. Чим більша маса і чим більша швидкість, тим більша кінетична енергія.

Одиницею вимірювання енергії в СІ, так само як і роботи, є **Джоуль (Дж)**. Це не випадковість, адже, як ми скоро побачимо, робота і енергія тісно пов'язані.

### Виведення формули кінетичної енергії

А звідки ж взялася знаменита формула $E_k = \frac{mv^2}{2}$? Її можна отримати, використовуючи поняття роботи. Давайте це зробимо.

Уявімо тіло масою $m$, яке перебуває у стані спокою (початкова швидкість дорівнює нулю $v_0 = 0$). На нього починає діяти постійна рівнодійна сила $F$, яка змушує його рухатись рівноприскорено вздовж прямої. Нехай під дією цієї сили тіло проходить шлях $s$ і набуває швидкості $v$.

1.  Робота, яку виконала сила $F$, дорівнює $A = F \cdot s$ (вважаємо, що сила і переміщення співнапрямлені, тому $\cos\alpha=1$).
2.  Згідно з другим законом Ньютона, $F = ma$. Підставимо це у формулу роботи:

    $$
    A = (ma)s = m(as)
    $$

3.  Нам потрібно позбутися прискорення $a$ та шляху $s$ і виразити роботу через швидкість $v$. Згадаймо формулу з кінематики для рівноприскореного руху без часу:

    $$
    s = \frac{v^2 - v_0^2}{2a}
    $$

4.  Оскільки початкова швидкість $v_0 = 0$, формула спрощується:

    $$
    s = \frac{v^2}{2a} \implies as = \frac{v^2}{2}
    $$

5.  Тепер підставимо вираз для $as$ у нашу формулу для роботи:

    $$
    A = m \cdot \left(\frac{v^2}{2}\right) = \frac{mv^2}{2}
    $$

Ми отримали, що робота, виконана для розгону тіла від стану спокою до швидкості $v$, дорівнює $\frac{mv^2}{2}$. Ця робота "перейшла" в енергію руху тіла. Саме цю величину і називають **кінетичною енергією**.

Отже, кінетична енергія тіла масою $m$, що рухається зі швидкістю $v$, визначається формулою

$$
E_k = \frac{mv^2}{2}
$$

## Теорема про кінетичну енергію

Ми підійшли до центральної ідеї, яка пов'язує роботу та енергію.

> **Теорема про кінетичну енергію**: Робота рівнодійної всіх сил, що діють на тіло, дорівнює зміні кінетичної енергії цього тіла.

$$
A_{\text{сумарна}} = \Delta E_k = E_k - E_{k0}
$$

де $E_k$ - кінцева кінетична енергія, а $E_{k0}$ - початкова кінетична енергія.

Давайте доведемо цю теорему. Доведення дуже схоже на виведення формули кінетичної енергії $E_k$, але тепер ми розглядаємо загальний випадок, коли початкова швидкість $v_0$ не дорівнює нулю.

1.  Знову розглядаємо тіло масою $m$, на яке діє рівнодійна сила $F$. Під її дією швидкість тіла змінюється від $v_0$ до $v$ на шляху $s$.
2.  Робота цієї сили дорівнює $A = Fs$.
3.  За другим законом Ньютона, $F = ma$, отже $A = mas$.
4.  Знову використовуємо кінематичну формулу $s = \frac{v^2 - v_0^2}{2a}$. Звідси, $as = \frac{v^2 - v_0^2}{2}$.
5.  Підставляємо вираз для $as$ у формулу для роботи:

    $$
    A = m \cdot \left(\frac{v^2 - v_0^2}{2}\right) = \frac{mv^2}{2} - \frac{mv_0^2}{2}
    $$

6.  Ми бачимо, що перший доданок - це кінцева кінетична енергія $E_k$, а другий - початкова $E_{k0}$. Отже, ми довели теорему:

    $$
    A = E_k - E_{k0} = \Delta E_k
    $$

Ця теорема є надзвичайно важливою. Вона каже нам:
*   Якщо сумарна робота над тілом **додатна** ($A > 0$), його кінетична енергія зростає ($\Delta E_k > 0$), тобто тіло **розганяється**.
*   Якщо сумарна робота **від'ємна** ($A < 0$), його кінетична енергія зменшується ($\Delta E_k < 0$), тобто тіло **гальмує**.
*   Якщо сумарна робота **дорівнює нулю** ($A = 0$), його кінетична енергія не змінюється ($\Delta E_k = 0$), тобто тіло рухається зі **сталою швидкістю**.

Робота - це процес передачі енергії від одного тіла до іншого за допомогою сили.

## Потужність

Ми вже знаємо, як обчислити роботу. Але часто важливо знати не тільки, яка робота виконана, а й **наскільки швидко** вона виконана. Для характеристики швидкості виконання роботи вводять поняття потужності.

> **Потужність** - це фізична величина, яка дорівнює відношенню виконаної роботи до проміжку часу, за який цю роботу було виконано.

*   **Аналогія**: Двоє людей перенесли однакові ящики на один і той самий поверх. Вони виконали однакову роботу. Але якщо один впорався за хвилину, а інший за п'ять, то перший був більш "потужним".

Формула для середньої потужності $P$:

$$
P = \frac{A}{t}
$$

де $A$ - виконана робота, $t$ - час виконання роботи.

Одиницею вимірювання потужності в СІ є **Ват (Вт)**, названа на честь Джеймса Ватта.

$$
1 \text{ Вт} = 1 \frac{\text{Дж}}{\text{с}}
$$

Потужність в 1 Вт розвиває механізм, який за 1 секунду виконує роботу в 1 Джоуль.

На практиці також використовують позасистемну одиницю - **кінську силу (к. с.)**. Її часто застосовують для характеристики потужності двигунів автомобілів.

$$
1 \text{ к. с.} \approx 735.5 \text{ Вт}
$$

Існує ще одна зручна формула для потужності. Давайте її виведемо.

1.  Візьмемо основну формулу: $P = \frac{A}{t}$.
2.  Підставимо формулу роботи $A = Fs\cos\alpha$:

    $$
    P = \frac{Fs\cos\alpha}{t}
    $$

3.  Згрупуємо множники інакше:

    $$
    P = (F\cos\alpha) \cdot \frac{s}{t}
    $$

4.  Ми знаємо, що відношення шляху до часу $\frac{s}{t}$ - це швидкість руху $v$ (для рівномірного руху).
5.  Тоді, якщо сила співнапрямлена з рухом ($\alpha=0$, $\cos\alpha=1$), отримуємо просту і корисну формулу:

    $$
    P = Fv
    $$

Ця формула показує, що потужність, яку розвиває, наприклад, двигун автомобіля, дорівнює добутку сили тяги, яку він створює, на швидкість руху автомобіля. Щоб підтримувати високу швидкість, потрібно розвивати велику потужність, щоб долати сили опору (тертя, опір повітря).
