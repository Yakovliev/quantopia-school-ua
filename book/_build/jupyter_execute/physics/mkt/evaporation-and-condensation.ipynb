# Пароутворення і конденсація

Ми знаємо, що речовина може існувати в трьох основних агрегатних станах: твердому, рідкому та газоподібному. Перехід з одного стану в інший називається **фазовим переходом**. Сьогодні нас цікавить шлях між рідким та газоподібним станами.

> **Пароутворення** - це процес переходу речовини з рідкого стану в газоподібний (пару).

> **Конденсація** - це зворотний процес перетворення пари на рідину.

Існує два види пароутворення: випаровування та кипіння. Давайте розглянемо кожен з них.

## Випаровування

> **Випаровування** - це пароутворення, що відбувається з вільної поверхні рідини.

Ми знаємо, що калюжа на асфальті висохне і в спекотний літній день, і в прохолодний осінній, просто з різною швидкістю. Це означає, що для випаровування не потрібне нагрівання до якоїсь конкретної температури. Як це можливо з точки зору МКТ?

Механізм випаровування з точки зору МКТ:
1.  **Молекули рухаються:** Ми знаємо, що молекули рідини перебувають у безперервному хаотичному русі. Їхні швидкості не однакові: є "повільні", "середні" та дуже "швидкі" молекули, кінетична енергія яких значно перевищує середню. Температура рідини визначає їхню *середню* кінетичну енергію.
2.  **Подолання притягання:** Молекули в рідині утримуються разом силами міжмолекулярного притягання.
3.  **"Втеча" з поверхні:** Коли одна з таких "швидких" молекул опиняється біля поверхні рідини, її кінетичної енергії може вистачити, щоб подолати сили притягання сусідніх молекул і вилетіти з рідини.

**Чому випаровування призводить до охолодження?**

Це один із найцікавіших наслідків випаровування. Оскільки рідину залишають молекули з найбільшою кінетичною енергією, середня кінетична енергія молекул, що залишилися, зменшується. А ми пам'ятаємо, що температура є мірою середньої кінетичної енергії. Отже, **рідина під час випаровування охолоджується** (за відсутності надходження тепла ззовні).

* **Аналогія:** Уявіть, що з групи людей виходять лише люди з найбільшою висотою. Середня висота групи людей після цього знизиться. Так само і з рідиною: коли "високі" (найшвидші молекули) її покидають, середня "висота" (температура) решти падає.

Саме тому ми відчуваємо прохолоду, коли виходимо з води в спекотний день - вода, випаровуючись з поверхні нашого тіла, забирає тепло.

### Від чого залежить швидкість випаровування?

Швидкість цього процесу не є сталою і залежить від кількох факторів:
*   **Температура:** Чим вища температура рідини, тим більше в ній "швидких" молекул, і тим інтенсивніше відбувається випаровування.
*   **Площа поверхні:** Випаровування відбувається з поверхні, тому чим більша площа цієї поверхні, тим більше молекул одночасно можуть покинути рідину. Чай у блюдці остигає і випаровується швидше, ніж у чашці.
*   **Рух повітря (вітер):** Над поверхнею рідини утворюється "хмарка" пари. Якщо її не прибирати, частина молекул повертається назад у рідину (конденсується). Вітер відносить цю пару, зменшуючи ймовірність повернення молекул, і тим самим прискорює випаровування.
*   **Рід речовини:** Різні рідини випаровуються з різною швидкістю. Наприклад, ефір випаровується набагато швидше за воду, а вода - швидше за олію. Це пов'язано з тим, що сили притягання між молекулами ефіру слабші, ніж між молекулами води.

## Конденсація: повернення додому

> **Конденсація** - це процес переходу речовини з газоподібного стану в рідкий.

Одночасно з випаровуванням відбувається і зворотний процес. Молекули пари над рідиною рухаються хаотично. Деякі з них, рухаючись у напрямку до поверхні рідини, можуть бути знову "захоплені" силами притягання і повернутися до рідкого стану.

При конденсації виділяється енергія. Коли молекула повертається в рідину, потенціальна енергія її взаємодії з іншими молекулами зменшується, а надлишок енергії виділяється у вигляді тепла. Тут може бути наступна аналогія: астероїд захоплюється гравітаційним полем Землі і падає на Землі. При цьому цей астероїд розганяється якраз за рахунок того, що його потенціальна енергія в гравітаційному полі Землі починає зменшуватися при наближенні до Землі (те ж саме відбувається із будь-яким тілом, яке падає на Землі: швидкість та кінетична енергія зростає, а потенціальна енергія зменшується).

Саме тому, коли ми заходимо у ванну, наповнену гарячою парою, ми відчуваємо тепло - це пара конденсується на нашій шкірі, виділяючи енергію.

## Динамічна рівновага. Насичена та ненасичена пара

Уявімо собі експеримент. Наллємо трохи води у герметично закриту посудину і почнемо спостерігати.
1.  **Початковий момент:** Вода починає випаровуватися. Молекули переходять у простір над рідиною. Оскільки пари ще мало, процес конденсації практично не відбувається. Швидкість випаровування значно більша за швидкість конденсації. Пара, яка знаходиться в такому стані, називається **ненасиченою**.
2.  **Проміжний стан:** Концентрація молекул пари над рідиною зростає. Це означає, що все більше молекул пари вдаряються об поверхню води і повертаються в неї. Швидкість конденсації зростає.
3.  **Стан рівноваги:** Нарешті, настає момент, коли кількість молекул, що покидають рідину за секунду, стає рівною кількості молекул, що повертаються в неї за ту ж секунду:
    
    Швидкість випаровування = Швидкість конденсації

    Хоча на мікрорівні процеси випаровування та конденсації тривають, на макрорівні кількість рідини та пари залишаються незмінними. Такий стан називається **динамічною рівновагою**.

> **Насичена пара** - це пара, що перебуває в стані динамічної рівноваги зі своєю рідиною.
> **Ненасичена пара** - це пара, яка не перебуває в стані динамічної рівноваги зі своєю рідиною (її тиск і густина менші, ніж у насиченої пари за даної температури).

У відкритій посудині (наприклад, у склянці на столі) стан динамічної рівноваги зазвичай не досягається, оскільки молекули пари розлітаються по кімнаті (дифундують), і швидкість випаровування завжди переважає. Тому рідина поступово випаровується повністю.

Існує також особливий, нестійкий стан, який називається **перенасиченою парою**. Це пара, тиск якої вищий за тиск насиченої пари за даної температури. Такий стан можна отримати, наприклад, при швидкому охолодженні насиченої пари за відсутності центрів конденсації (пилу, іонів). Перенасичена пара є нестабільною, і за найменшої можливості надлишок пари миттєво конденсується. Саме це явище ми спостерігаємо у вигляді туману або хмар.

### Властивості насиченої пари

Насичена пара має дві цікаві властивості, які відрізняють її від ідеального газу.

**1. Тиск $p$ і густина $\rho$ насиченої пари НЕ залежать від об'єму**

Уявімо, що в нас є циліндр із поршнем, в якому під поршнем знаходиться рідина та її насичена пара.
*   **Стискаємо пару (зменшуємо об'єм).** Що станеться? На мить концентрація молекул пари зросте. Це призведе до того, що швидкість конденсації стане більшою за швидкість випаровування. Частина пари перетвориться на рідину, і концентрація молекул (а отже, і густина та тиск) повернеться до початкового значення, характерного для насиченої пари при цій температурі.
*   **Розширюємо пару (збільшуємо об'єм).** На мить концентрація молекул пари зменшиться. Швидкість конденсації стане меншою за швидкість випаровування. Додаткова кількість рідини випарується, доки концентрація молекул пари (а отже, і густина та тиск) знову не досягне значення насичення.

Те ж саме можна продемонструвати за допомогою формул $p = nkT$ та $\rho = n m_0$, де $m_0$ - маса однієї молекули. У випадку із ідеальним газом, коли ми зменшуємо (збільшуємо) об'єм, концентрація збільшується (зменшується), оскільки кількість атомів/молекул ідеального газу не змінюється. Саме тому й існує ця залежність тиску від об'єму для ідеального газу. Однак у випадку із насиченою парою, коли ми змінюємо об'єм, концентрація залишається сталою (оскільки кількість атомів/молекул змінюється), як пояснено вище.

**Висновок:** При незмінній температурі тиск насиченої пари $p_н$ та густина насиченої пари $\rho_н$ є сталими величинами і не залежать від об'єму, який займає пара.

**2. Тиск і густина насиченої пари залежать від температури**

А що буде, якщо ми почнемо нагрівати нашу закриту посудину з рідиною та її насиченою парою?
1.  Підвищення температури збільшує середню кінетичну енергію молекул рідини.
2.  Це призводить до збільшення швидкості випаровування (більше молекул мають енергію для "втечі").
3.  Щоб встановилася нова динамічна рівновага, швидкість конденсації також має зрости. А це можливо лише за умови, що концентрація молекул у парі стане більшою.

Тобто, зростання температури насиченої пари призводить до зростання концентрації. Таким чином, згідно із формулами $p = nkT$ та $\rho = n m_0$, тиск та густина зростають при збільшенні температури (зокрема, бо зростає концентрація).

Зростання тиску насиченої пари при збільшенні температури є швидшим, ніж у випадку із ідеальним газом:
- При збільшенні температури та за фіксованого об'єму, концентрація ідеального газу залишається сталою і тиск лінійно залежить від температури.
- При збільшенні температури та за фіксованого об'єму, концентрація насиченої пари зростає, що призводить до нелінійної залежності тиску від температури згідно з формулою $p = nkT$, де концентрація $n = n(T)$ залежить від температури (а саме, збільшується при збільшенні температури). Як результат, тиск насиченої пари зростає швидше при збільшенні температури, ніж тиск ідеального газу.

**Висновок:** Зі збільшенням температури концентрація, густина і тиск насиченої пари **зростають**. І ця залежність є нелінійною (досить стрімкою). Для кожної температури існує своє, чітко визначене значення тиску насиченої пари.

На відміну від насиченої пари, ненасичена пара за своїми властивостями дуже схожа на ідеальний газ. Для неї справедливі газові закони (Бойля-Маріотта, Шарля, Гей-Люссака), і її тиск залежить і від об'єму, і від температури.

## Кипіння: випаровування зсередини

Ми звикли, що при нагріванні води в чайнику за певної температури вона починає бурлити - це і є кипіння.

> **Кипіння** - це інтенсивний процес пароутворення, який відбувається не тільки з поверхні, а й по всьому об'єму рідини з утворенням бульбашок пари.

На відміну від випаровування, кипіння відбувається за **певної, строго визначеної температури**, яка називається **температурою кипіння**.

### Механізм кипіння

Щоб зрозуміти, як відбувається кипіння, прослідкуймо за долею однієї бульбашки всередині рідини, яку ми нагріваємо.
1.  **Звідки беруться бульбашки?** Бульбашки пари не можуть утворитися в ідеально чистій однорідній рідині. Вони завжди утворюються на так званих **центрах пароутворення**. Це можуть бути:
    *   Мікроскопічні бульбашки повітря, розчинені у воді.
    *   Подряпини та нерівності на стінках і дні посудини.
    *   Частинки пилу та інші домішки.

2.  **Що всередині бульбашки?** Бульбашка, що утворилася, заповнена **насиченою парою** самої рідини. Ця пара створює тиск зсередини, який намагається розширити бульбашку. Позначимо його $p_н$.

3.  **Що тисне на бульбашку ззовні?** На бульбашку діє зовнішній тиск, який намагається її стиснути. Цей тиск складається з трьох частин:
    *   **Атмосферний тиск** $p_{\text{атм}}$, який тисне на вільну поверхню рідини.
    *   **Гідростатичний тиск** стовпа рідини над бульбашкою $p_{\text{від}} = \rho g h$, де $\rho$ - густина рідини, $g$ - прискорення вільного падіння, $h$ - глибина занурення бульбашки.
    *   **Тиск Лапласа** $p_{\text{л}}$, створений силами поверхневого натягу самої рідини, який особливо значний для маленьких бульбашок ($p_{\text{л}} \sim r^{-1}$, де $r$ - радіус бульбашки). Ми не будемо його детально розглядати, але важливо знати, що він існує. Аналогією тиску Лапласа для бульбашки може слугувати пружність стінок гумової кульки, що ускладнює надування цієї кульки.

    Отже, повний зовнішній тиск на бульбашку: $p_{\text{зовн}} = p_{\text{атм}} + \rho g h + p_{\text{л}}$.

4.  **Нагрівання і ріст бульбашок.** Коли ми починаємо нагрівати воду, її температура зростає. Разом з нею зростає і тиск насиченої пари $p_н$ всередині бульбашок.
    *   **Початковий етап (вода "шумить").** Бульбашки, що утворюються на гарячому дні ($p_н \approx p_{\text{зовн}}$), збільшуються в об'ємі і за рахунок архімедової виштовхувальної сили починають підніматися у холодніші верхні шари. Холодніша температура верхніх шарів води спричиняє суттєве падіння тиску $p_н$ і пара в них починає кондинсуватись, за рахунок чого вони **"схлопуються"**, створюючи характерний шум.
    *   **Момент закипання.** Нагрівання усього об'єму води триває, і нарешті настає момент, коли рідина повністю прогрівається. Коли рідина рівномірно прогріта і тиск насиченої пари всередині бульбашки зрівнюється із зовнішнім тиском (або стає трохи більшим за нього), розпочинається кипіння

        $$p_н(T_{\text{кип}}) \approx p_{\text{зовн}}$$

        Тепер бульбашки не схлопуються, а піднімаються на поверхню, зростають ще більше в об'ємі і лопаються, викидаючи пару.

### Температура кипіння та її залежність від тиску

З умови кипіння випливає найважливіший висновок. Оскільки гідростатичний тиск і тиск Лапласа зазвичай малі порівняно з атмосферним, умову кипіння можна спростити:

> **Рідина кипить за такої температури, за якої тиск її насиченої пари дорівнює зовнішньому (атмосферному) тиску.**
>
> $$ p_н(T_{\text{кип}}) \approx p_{\text{атм}}$$

Це пояснює, чому температура кипіння не є сталою величиною, а залежить від зовнішнього тиску:
*   **У горах.** На великій висоті атмосферний тиск нижчий. Тому тиску насиченої пари води не потрібно досягати великих значень, щоб зрівнятися із зовнішнім. Вода в горах закипає при нижчій температурі (наприклад, 80-90 °C). Зварити в такій воді м'ясо або картоплю дуже складно.
*   **В автоклаві (скороварці).** Це герметично закрита каструля, в якій при нагріванні тиск пари зростає значно вище атмосферного. Щоб вода закипіла, тиск її насиченої пари має досягти цього підвищеного тиску. Це відбувається за температури 110-120 °C. За такої температури їжа готується набагато швидше. Цей принцип використовується для стерилізації медичних інструментів в автоклавах.

**Важливо:** Під час кипіння, незважаючи на постійне підведення тепла, температура рідини **не змінюється**. Вся енергія, що підводиться, йде на перетворення рідини в пару.

| Поняття / Процес | Ключові характеристики та опис | Умова / Формула |
| :--- | :--- | :--- |
| **Пароутворення** | Загальна назва процесу переходу речовини з рідкого стану в газоподібний. | |
| **Випаровування** | Пароутворення, що відбувається з вільної поверхні рідини. Відбувається за будь-якої температури та призводить до охолодження рідини. Швидкість залежить від температури, площі поверхні, руху повітря та роду речовини. | |
| **Конденсація** | Процес переходу речовини з газоподібного стану в рідкий. Це зворотний до пароутворення процес, який супроводжується виділенням енергії. | |
| **Динамічна рівновага** | Стан, у якому швидкість випаровування дорівнює швидкості конденсації. Макроскопічні параметри (кількість рідини та пари) залишаються незмінними. | Швидкість випаровування = Швидкість конденсації |
| **Ненасичена пара** | Пара, тиск і густина якої менші, ніж у насиченої пари за даної температури. Не перебуває в динамічній рівновазі зі своєю рідиною. | |
| **Насичена пара** | Пара, що перебуває в стані динамічної рівноваги зі своєю рідиною. | |
| *Властивість 1* | При сталій температурі тиск та густина насиченої пари **не залежать** від об'єму, який вона займає. | $p_н = \text{const}$ та $\rho_н = \text{const}$ (при $T = \text{const}$) |
| *Властивість 2* | Зі збільшенням температури, концентрація, густина і тиск насиченої пари **зростають** значно швидше, ніж у ідеального газу. | $p = nkT$, де концентрація $n$ зростає з температурою $T$ |
| **Кипіння** | Інтенсивний процес пароутворення, який відбувається по всьому об'єму рідини з утворенням бульбашок пари. | |
| *Температура кипіння* | Відбувається за строго визначеної температури. Під час кипіння, попри підведення тепла, температура рідини **не змінюється**. | $T_{\text{кип}} = \text{const}$ |
| *Умова кипіння* | Рідина кипить за температури, при якій тиск її насиченої пари всередині бульбашок приблизно дорівнює зовнішньому тиску. | $p_н(T_{\text{кип}}) \approx p_{\text{зовн}}$ |
| *Залежність від тиску* | Температура кипіння залежить від зовнішнього тиску: вищий тиск - вища температура кипіння, і навпаки. | $p_{\text{зовн}} = p_{\text{атм}} + \rho g h + p_{\text{л}}$ |
