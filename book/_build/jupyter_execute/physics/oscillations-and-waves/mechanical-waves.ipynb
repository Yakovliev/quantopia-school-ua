# Механічні хвилі

Уявімо собі спокійну поверхню озера. Якщо кинути у воду камінь, від місця його падіння побіжать кола, що розширюються. Це і є хвилі. Або згадаємо, як на стадіоні вболівальники запускають "хвилю": кожен з них лише встає і сідає, але по трибунах біжить видимий рух. В обох випадках ми спостерігаємо поширення коливань, але сама речовина (вода чи люди) не переміщується на великі відстані. Це і є ключова ідея хвильового руху.

> **Хвиля** - це процес поширення коливань у просторі з плином часу.

За своєю фізичною природою хвилі бувають різними. Найбільш відомими є **електромагнітні хвилі** (світло, радіохвилі, рентгенівські промені), які можуть поширюватися навіть у вакуумі. Але ми зосередимося на **механічних хвилях**.

> **Механічною хвилею** називають процес поширення механічних коливань у пружному середовищі.

Для існування та поширення механічної хвилі необхідні дві ключові умови:

1.  **Джерело хвилі:** Це тіло, що коливається і створює початкову деформацію або збурення у середовищі. Наприклад, камінь, що впав у воду, або голосові зв'язки, що коливаються і створюють звук.
2.  **Пружне середовище:** Це середовище, в якому виникають сили пружності при його деформації. Частинки цього середовища пов'язані між собою (наприклад, молекули води або атоми у твердому тілі). Саме завдяки цим зв'язкам коливання від однієї частинки передаються до сусідньої, і так далі. Без середовища (у вакуумі) механічна хвиля поширюватися не може.

Отже, **хвильовий рух** - це процес передачі **пружних коливань** від одних частинок середовища до інших.

Найважливіша властивість будь-якої хвилі полягає в тому, що **хвилі переносять енергію без перенесення речовини**. Вболівальник на стадіоні залишається на своєму місці, але енергія руху (і сам рух) передається далі. Частинки води в озері у найпростішому випадку лише коливаються вгору-вниз, але енергія від падіння каменя поширюється по всій поверхні.

## Класифікація механічних хвиль

Залежно від напрямку коливань частинок середовища відносно напрямку поширення хвилі, розрізняють два основних типи хвиль.

### Поперечні хвилі

> **Поперечна хвиля** - це хвиля, у якій частинки середовища коливаються в напрямку, перпендикулярному до напрямку її поширення.

Класичний приклад - хвиля, що біжить по натягнутій мотузці, якщо один її кінець смикнути вгору-вниз. Кожна точка мотузки рухається лише вертикально, а сама хвиля (горб) біжить горизонтально.

Поперечні хвилі можуть поширюватися лише в середовищах, де виникає пружність при деформації зсуву. Це характерно для **твердих тіл**. У рідинах та газах поперечні хвилі, як правило, не виникають, оскільки шари рідини чи газу легко зсуваються один відносно одного. Винятком є хвилі на поверхні рідини, які мають поперечну складову, що виникає завдяки силам поверхневого натягу, а також поперечні (зсувні) хвилі у дуже в'язких рідинах.

### Поздовжні хвилі

> **Поздовжня хвиля** - це хвиля, у якій частинки середовища коливаються вздовж напрямку її поширення.

Найкращою аналогією є стиснення та розтягнення пружини. Ми побачимо, як по пружині біжать області згущення та розрідження витків. Кожен виток при цьому коливається лише вперед-назад, вздовж пружини.

Поздовжні хвилі пов'язані з деформаціями стиску та розтягу. Оскільки такі деформації викликають появу сил пружності у **всіх середовищах - твердих, рідких та газоподібних**, поздовжні хвилі можуть поширюватися в них усіх. **Звук є важливим прикладом поздовжньої хвилі**.

### Хвилі змішаного типу

Існують і більш складні випадки. Наприклад, хвилі на поверхні рідини є змішаними: частинки води в них рухаються по колах або еліпсах, тобто в їхньому русі є як поперечна, так і поздовжня складові.

## Основні характеристики хвилі

Як і коливання, хвилі мають набір фізичних величин, що їх описують. Частота та період хвилі визначаються частотою та періодом коливань джерела, тому ці поняття нам вже знайомі. Розглянемо нові, специфічні для хвиль характеристики.

> **Швидкість поширення хвилі** $v$ - це швидкість, з якою в просторі переміщується певна фаза коливань (наприклад, гребінь хвилі). Вона залежить виключно від властивостей середовища (його пружності, густини, температури). **Одиниця в СІ:** метр за секунду (м/с).

> **Довжина хвилі** $\lambda$ - це просторова характеристика хвилі. Це відстань, на яку поширюється хвиля за час, що дорівнює одному періоду коливань $T$. Іншими словами, це відстань між двома найближчими точками середовища, які коливаються в однаковій фазі (наприклад, між двома сусідніми гребенями або западинами). **Одиниця в СІ:** метр (м).

### Зв'язок між швидкістю, довжиною та періодом (частотою) хвилі

Ці три фундаментальні характеристики пов'язані простою і важливою формулою. Виведемо її.

З означення довжини хвилі випливає, що за час $T$ (один період) хвиля проходить відстань $\lambda$. Використовуючи базову формулу кінематики для рівномірного руху ($\text{відстань} = \text{швидкість} \cdot \text{час}$), ми можемо записати:

$$ \lambda = v T $$

Це і є **формула зв'язку довжини хвилі та періоду**.

Оскільки період $T$ і частота $\nu$ пов'язані співвідношенням $T = 1/\nu$, ми можемо підставити це у нашу формулу:

$$
\lambda = v \frac{1}{\nu}
$$

Звідки отримуємо ще одну, не менш важливу формулу, яку часто називають **формулою хвилі**:

$$
v = \lambda \nu
$$

Ця формула показує, що швидкість поширення хвилі дорівнює добутку її довжини на частоту коливань. Вона є універсальною для хвиль будь-якої природи.

## Періодичність хвилі в часі та просторі

Хвиля є процесом, що повторюється як у часі, так і в просторі.
*   **Періодичність у часі:** Якщо ми будемо спостерігати за однією конкретною точкою середовища (наприклад, за поплавком на воді), ми побачимо, що вона здійснює гармонічні коливання з періодом $T$. Її стан повністю повторюється через проміжки часу, кратні $T$.
*   **Періодичність у просторі:** Якщо ми зробимо "миттєвий знімок" хвилі, то побачимо, що її форма (профіль) періодично повторюється у просторі. Відстань, через яку форма хвилі повторюється, і є довжина хвилі $\lambda$.

## Рівняння плоскої гармонічної хвилі

Давайте спробуємо математично описати хвилю. Розглянемо найпростіший випадок - **плоску гармонічну хвилю**, що поширюється вздовж осі $x$. Плоска хвиля - це така хвиля, в якої всі точки на будь-якій площині, перпендикулярній до напрямку поширення, коливаються в однаковій фазі.

Нехай джерело хвилі знаходиться в точці $x=0$ і здійснює гармонічні коливання за законом косинуса (припускаємо, що початкова фаза дорівнює нулю для простоти):

$$u(0, t) = A \cos(\omega t)$$

Тут $u(0, t)$ - зміщення точки $x=0$ в момент часу $t$, $A$ - амплітуда, $\omega$ - циклічна частота.

Тепер розглянемо довільну точку середовища з координатою $x$. Коливання до неї дійдуть не миттєво. Хвилі потрібен час, щоб подолати відстань $x$. Цей час "запізнення" $\Delta t$ дорівнює:

$$\Delta t = \frac{x}{v}$$

де $v$ - швидкість поширення хвилі.

Це означає, що точка з координатою $x$ в момент часу $t$ буде здійснювати точно такі ж коливання, які точка $x=0$ здійснювала в більш ранній момент часу $t - \Delta t$. Щоб знайти зміщення $u(x, t)$, нам потрібно просто підставити час $t - \Delta t$ у рівняння коливань джерела:

$$u(x, t) = u(0, t - \Delta t) = A \cos(\omega(t - \Delta t))$$

Підставимо вираз для $\Delta t$:

$$
u(x, t) = A \cos\left(\omega\left(t - \frac{x}{v}\right)\right)
$$

Розкриємо дужки у фазі коливань:

$$u(x, t) = A \cos\left(\omega t - \frac{\omega}{v}x\right)$$

Вираз $\frac{\omega}{v}$ зустрічається настільки часто, що для нього ввели спеціальне позначення і назву - **хвильове число** $k$. **Одиниця в СІ:** радіан на метр (рад/м).

Давайте розберемося з його фізичним змістом. Згадаємо, що $\omega = \frac{2\pi}{T}$ і $v = \frac{\lambda}{T}$. Підставимо це у вираз для $k$:

$$k = \frac{\omega}{v} = \frac{2\pi/T}{\lambda/T} = \frac{2\pi}{\lambda}$$

Отже, хвильове число $k$ показує, скільки разів фаза коливань змінюється на $2\pi$ радіан на відрізку довжиною $\lambda$. Якщо циклічна частота $\omega$ характеризує періодичність процесу в часі, то хвильове число $k$ характеризує його періодичність у просторі. Інакше кажучи, хвильове число показує, скільки довжин хвиль укладається на відрізку довжиною $2 \pi$ метрів. Це аналог циклічної частоти $\omega$ (яка є часовою частотою), але для простору.

Тепер ми можемо записати **рівняння плоскої гармонічної хвилі** в його остаточному, стандартному вигляді:

$$
u(x, t) = A \cos(\omega t - kx)
$$

Це рівняння описує зміщення $u$ будь-якої точки середовища з координатою $x$ в будь-який момент часу $t$. Знак "мінус" перед $kx$ вказує, що хвиля поширюється в додатному напрямку осі $x$. Для хвилі, що біжить у від'ємному напрямку, рівняння матиме вигляд $u(x, t) = A \cos(\omega t + kx)$.

## Фронт хвилі та хвильові поверхні

> **Хвильова поверхня** - це поверхня, всі точки якої коливаються в однаковій фазі.

> **Фронт хвилі** - це хвильова поверхня, що відокремлює частину простору, вже охоплену хвильовим процесом, від тієї, де коливання ще не почалися.

Форма хвильових поверхонь залежить від форми джерела.
*   **Сферична хвиля:** Виникає від точкового джерела. Хвильові поверхні - концентричні сфери.
*   **Циліндрична хвиля:** Виникає від джерела у формі лінії (наприклад, нитки, що коливається). Хвильові поверхні - коаксіальні циліндри.
*   **Плоска хвиля:** Виникає, коли джерело є нескінченною площиною, або як наближення для сферичної/циліндричної хвилі на дуже великій відстані від джерела. Хвильові поверхні - паралельні площини.

## Відбивання механічних хвиль та стоячі хвилі

Що станеться, коли хвиля дійде до межі середовища або перешкоди? Вона відіб'ється. Характер відбивання залежить від властивостей перешкоди.

1.  **Відбивання від "більш щільної" (жорстко закріпленої) межі:** Якщо кінець мотузки, по якій біжить хвиля, жорстко закріплений, то відбита хвиля повернеться "догори дриґом". Кажуть, що її фаза змінюється на протилежну (на $\pi$ радіан). Гребінь стає западиною, і навпаки.
2.  **Відбивання від "менш щільної" (вільної) межі:** Якщо кінець мотузки прикріплений до легкого кільця, що може вільно ковзати по стержню, то відбита хвиля повернеться без зміни фази. Гребінь відіб'ється гребенем.

Коли первинна (падаюча) хвиля зустрічається з відбитою, вони накладаються одна на одну. Цей процес називається інтерференцією. В особливому випадку, коли дві однакові за частотою та амплітудою хвилі поширюються назустріч одна одній, виникає унікальне явище.

> **Стояча хвиля** - це коливальний рух, що виникає в результаті накладання двох зустрічних когерентних хвиль, і в якому положення вузлів та пучностей залишається незмінним у просторі.

На відміну від біжучої хвилі, стояча хвиля не переносить енергію. Енергія "замикається" між вузлами.

У стоячій хвилі є точки, які залишаються нерухомими.
*   **Вузли** - це точки, в яких амплітуда коливань завжди дорівнює нулю.

Також є точки, що коливаються з максимально можливою амплітудою.
*   **Пучності** (або анти-вузли) - це точки, в яких амплітуда коливань є максимальною (дорівнює $2A$).

Математичний аналіз показує (і це можна спостерігати на досліді), що:
*   Відстань між двома сусідніми вузлами дорівнює половині довжини хвилі $\lambda/2$.
*   Відстань між двома сусідніми пучностями також дорівнює $\lambda/2$.
*   Відстань між сусідніми вузлом та пучністю дорівнює $\lambda/4$.

Стоячі хвилі є основою звучання всіх струнних та духових музичних інструментів. Коли ми граємо на гітарі, в струні виникає саме стояча хвиля, довжина якої визначає висоту тону.

| **Зв'язок характеристик** | Основні співвідношення, що пов'язують швидкість, довжину, період $T$ та частоту $\nu$:<br>Формула довжини хвилі $\lambda = vT$<br>Формула хвилі $v = \lambda\nu$ |
| **Хвильове число** $k$ | Характеризує просторову періодичність хвилі; аналог циклічної частоти для простору. <br> $k = \frac{2\pi}{\lambda}$ |
| **Рівняння плоскої хвилі** | Математичний опис зміщення $u$ точки з координатою $x$ в момент часу $t$:<br>$u(x, t) = A \cos(\omega t - kx)$ <br> (знак "мінус" для хвилі, що рухається в додатному напрямку осі $x$) |
| **Хвильова поверхня** | Поверхня, усі точки якої коливаються в однаковій фазі. |
| **Фронт хвилі** | Хвильова поверхня, що відокремлює область, охоплену коливаннями, від області спокою. |
| **Стояча хвиля** | Результат накладання (інтерференції) двох однакових зустрічних хвиль. **Властивість:** не переносить енергію. |
| **Вузли стоячої хвилі** | Точки, в яких амплітуда коливань завжди дорівнює нулю. Відстань між сусідніми вузлами дорівнює $\lambda/2$. |
| **Пучності стоячої хвилі** | Точки, в яких амплітуда коливань максимальна (дорівнює $2A$). Відстань між сусідніми пучностями дорівнює $\lambda/2$. |
