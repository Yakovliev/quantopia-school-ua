# Механічні коливання

У навколишньому світі ми постійно спостерігаємо механічні коливання: хитання гілок дерев під вітром, вібрації струн у музичних інструментах, рух поплавця на водній поверхні, коливання маятника годинника, пульсація серця тощо. Коливальний рух належить до найрозповсюджених явищ природи та має певні характерні властивості, які ми розглянемо в цій лекції.


## Коливальний рух

Під механічними коливаннями розуміють рухи фізичного тіла або системи тіл навколо певної точки рівноваги, що повторюються точно або приблизно через однакові проміжки часу.

Як і будь-який інший вид руху, коливальний рух описується фундаментальними фізичними параметрами: **швидкістю**, **прискоренням** та **координатою** (або **зміщенням**).

> Зміщення $x$ визначається як відстань між положенням рівноваги та поточним положенням тіла, що здійснює коливальний рух, у конкретний момент часу.

Під час коливань механічний стан тіла постійно змінюється. Коливання називають **періодичними**, коли координата, величина та напрямок швидкості руху тіла регулярно повторюються через однакові інтервали часу.

Для характеристики періодичних коливань використовують кілька специфічних фізичних величин: **амплітуду**, **період** та **частоту**.

> **Амплітуда коливань** ($A$) - це максимальне відхилення тіла від положення рівноваги, тобто $A = x_{max}$. Амплітуда механічних коливань вимірюється в метрах (м).

> **Період коливань** ($T$) - це час, за який відбувається одне повне коливання. Період обчислюється за формулою $T = \frac{t}{N}$, де $t$ - загальний час спостереження, а $N$ - кількість коливань за цей час. Одиницею періоду в СІ є секунда (с).

> **Частота коливань** ($\nu$) - кількість коливань за одиницю часу. Частота розраховується за формулою $\nu = \frac{N}{t}$. Одиницею частоти в СІ є герц (Гц), що відповідає одному коливанню за секунду.

Частота і період пов'язані між собою оберненою залежністю: 

$$\nu = \frac{1}{T}$$

## Незатухаючі та затухаючі коливання

Проаналізуємо коливання вантажу на пружині. В ідеальній системі «вантаж - пружина - Земля», де відсутні втрати механічної енергії, коливання тривали б нескінченно довго з незмінною амплітудою.

> Коливання, при яких амплітуда залишається постійною з плином часу, класифікують як **незатухаючі**.

Однак у реальних фізичних системах завжди присутні втрати енергії. Енергія витрачається на подолання сил тертя та деформації тіл під час коливального руху. Внаслідок цього механічна енергія поступово перетворюється на внутрішню теплову енергію. Тому, якщо система не отримує енергію із зовнішніх джерел, амплітуда коливань поступово зменшується, і з часом коливання припиняються.

> Коливання, амплітуда яких з часом зменшується, називаються **затухаючими**.

## Вільні та вимушені коливання, автоколивання

Існує декілька типів механічних коливань з різними характеристиками. Деякі коливання можуть відбуватися без періодичного зовнішнього впливу. Прикладом таких коливань є рух кульки, підвішеної на нитці або на пружині, після її початкового відхилення від положення рівноваги і відпускання. Такі коливання називають **вільними**.

> **Вільні коливання** - це коливання, що виникають внаслідок дії внутрішніх сил системи після виведення її з положення рівноваги.

Частота вільних коливань залежить від конструктивних особливостей та параметрів самої системи.

Фізичну систему, здатну здійснювати вільні коливання, називають **коливальною системою**. Ключовою характеристикою будь-якої коливальної системи є наявність **положення стійкої рівноваги**. Саме навколо цього положення відбуваються вільні коливання. Для виникнення вільних коливань необхідно забезпечити дві умови:
* надати системі додаткову енергію (потенціальну або кінетичну);
* забезпечити достатньо малий коефіцієнт тертя в системі, інакше коливання швидко згаснуть або взагалі не виникнуть.

Оскільки при вільних коливаннях система не отримує додаткової енергії ззовні, **вільні коливання завжди є затухаючими**. Інтенсивність затухання прямо пропорційна величині тертя в системі. Наприклад, якщо змусити одне й те саме тіло коливатися в повітрі та у воді, то в повітрі коливання триватимуть значно довше через менший опір середовища, а у воді вони швидко затухнуть. Цей принцип використовується в конструкції гідравлічних амортизаторів автомобілів.

Існують також коливання, які можуть відбуватись лише під дією зовнішніх періодично змінних сил. Прикладами є рух повітря в духових інструментах або поршня в двигуні внутрішнього згоряння. Такі коливання називають **вимушеними**.

> **Вимушені коливання** - це коливання, які виникають у системі під дією зовнішньої сили, що змінюється періодично.

Вимушені коливання зазвичай є **незатухаючими**, а їхня частота відповідає частоті зміни зовнішньої сили, що спричиняє коливання.

Існують також особливі системи, в яких незатухаючі коливання підтримуються не за рахунок періодичного зовнішнього впливу, а завдяки здатності самої системи регулювати надходження енергії від постійного джерела. Такі системи називають **автоколивальними**, а незатухаючі коливання в них - **автоколиваннями**.

> **Автоколивання** - це незатухаючі коливання, які підтримуються за рахунок енергії від постійного джерела, причому надходження цієї енергії регулюється самою коливальною системою.

Частота автоколивань, як і у випадку вільних коливань, визначається властивостями самої системи. Класичним прикладом механічної автоколивальної системи є храповий механізм маятникового годинника. Практично будь-яка автоколивальна система складається з трьох основних компонентів: **коливальної системи**, в якій можуть виникати вільні коливання (наприклад, маятник годинника), **джерела енергії** (піднята гиря, що обертає храпове колесо), і **пристрою зворотного зв'язку**, який регулює подачу енергії від джерела певними порціями (анкерний механізм, через який маятник контролює момент передачі енергії від гирі до храпового колеса).

## Гармонічні коливання

Залежно від характеру зміни зміщення (координати) тіла з часом, коливання поділяють на **гармонічні** та **негармонічні**. У більшості випадків залежність $x(t)$ має досить складний характер.

> **Гармонічними коливаннями** називають коливання, при яких координата $x$ тіла змінюється з часом $t$ за законом косинуса або синуса:

$$x = A\cos(\omega t + \varphi_0) \text{ або } x = A\sin(\omega t + \varphi_0)$$

Ці рівняння називають **рівняннями гармонічних коливань**. Розглянемо значення кожного параметра в цих рівняннях.

**Амплітуда коливань** $A$ визначає максимальне відхилення від положення рівноваги: $x_{max} = A$, оскільки максимальне значення функцій косинуса і синуса дорівнює одиниці.

**Фаза коливань** $\varphi = \omega t + \varphi_0$ однозначно визначає механічний стан тіла в будь-який момент часу.

**Початкова фаза коливань** $\varphi_0$ - це фаза коливань у початковий момент відліку часу, коли $t = 0$, тобто $\varphi = \varphi_0$.

**Циклічна частота коливань** $\omega$ пов'язана з періодом коливань $T$ співвідношенням $\omega = \frac{2\pi}{T}$. Це випливає з того, що косинус і синус є періодичними функціями: 

$$\cos(\omega t + \varphi_0) = \cos(\omega t + \varphi_0 + 2\pi)$$

Оскільки коливання повністю повторюються через час, рівний періоду $T$, то 

$$\cos(\omega t + \varphi_0) = \cos(\omega(t + T) + \varphi_0)$$

звідки 

$$\omega t + \varphi_0 + 2\pi = \omega t + \omega T + \varphi_0$$

і отримуємо 

$$\omega = \frac{2\pi}{T}$$

Одиницею вимірювання циклічної частоти в СІ є радіан за секунду (рад/с або с$^{-1}$).

Можна математично довести, що коли координата тіла змінюється за гармонічним законом, то швидкість і прискорення руху тіла також змінюються гармонічно. При цьому виконуються такі співвідношення:

$$v_{max} = \omega x_{max}; \quad a_{max} = \omega^2 x_{max}; \quad a_x = -\omega^2 x$$

Існує і зворотне твердження: **якщо в довільний момент часу прискорення тіла прямо пропорційне його зміщенню і спрямоване у протилежний бік відносно зміщення, то такий рух є гармонічними коливаннями**.

## Зверніть увагу!

* Якщо початковий момент відліку часу $(t = 0)$ співпадає з моментом максимального відхилення тіла від положення рівноваги $(x_0 = x_{max} = A)$, то рівняння коливань доцільно записувати у формі: $x_0 = A\cos\omega t$.
* Якщо початковий момент відліку часу $(t = 0)$ співпадає з моментом проходження тілом положення рівноваги $(x_0 = 0)$, то рівняння коливань краще записувати у формі: $x = A\sin\omega t$.
* За допомогою графіка коливань або рівняння коливань можна легко визначити всі фізичні величини, що характеризують коливальний рух.
