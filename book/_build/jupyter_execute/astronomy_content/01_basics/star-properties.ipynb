# Фізичні характеристики зір


Ось переклад запитаних фізичних та астрономічних термінів на англійську мову з урахуванням можливих контекстів:

Видима зоряна величина (астрономія): Apparent magnitude

Це міра яскравості зорі або іншого небесного об'єкта, яку ми спостерігаємо з Землі.
Абсолютна зоряна величина (астрономія): Absolute magnitude

Це міра внутрішньої світності небесного об'єкта, визначена як видима зоряна величина, яку б мав об'єкт, перебуваючи на стандартній відстані 10 парсеків без врахування поглинання світла.
Світловий потік (фотометрія, вимірюється в люменах): Luminous flux

Це повна кількість видимого світла, що випромінюється джерелом світла за одиницю часу. Одиниця вимірювання в системі SI - люмен (lm).
Сила світла (фотометрія, одиниця вимірювання - кандела): Luminous intensity

Це світловий потік, що випромінюється в певному напрямку в межах одиничного тілесного кута. Одиниця вимірювання в системі SI - кандела (cd).
Освітленість (фотометрія, одиниця вимірювання - люкс): Illuminance

Це світловий потік, що падає на одиницю площі поверхні. Одиниця вимірювання в системі SI - люкс (lx), що дорівнює лм/м².
Світність зорі (астрономія): Luminosity (часто використовується термін без уточнення "stellar", але для чіткості можна додати Stellar luminosity)

Це повна енергія, випромінювана зорею (або іншим астрономічним об'єктом) за одиницю часу в усьому діапазоні електромагнітного спектра. Не слід плутати зі світловим потоком, який стосується лише видимого світла і є фотометричною величиною.
Яскравість (фотометрія, одиниця вимірювання - кандела на квадратний метр):

У фотометрії для величини, що вимірюється в канделах на квадратний метр (cd/m²), використовується термін Luminance. Вона характеризує світність поверхні, що світиться або відбиває світло, в певному напрямку.
Загальний термін Brightness може використовуватися в більш суб'єктивному сенсі для опису сприйняття світла, але в точній фотометрії та для одиниці cd/m² використовується Luminance.

## Вступ

Коли ми дивимось на нічне небо, ми бачимо зорі різної яскравості. Деякі з них виглядають яскравими і привертають увагу, інші ледь помітні. Але чи справді яскравіші зорі є більшими чи потужнішими? Не завжди. Щоб зрозуміти істинну природу зір, астрономи розробили систему вимірювань, яка дозволяє нам розрізняти видиму яскравість (те, що ми бачимо з Землі) і справжню потужність випромінювання зір. У цьому лекційному матеріалі ми розглянемо ключові поняття, які допомагають нам зрозуміти зорі: видиму зоряну величину, абсолютну зоряну величину, світність та як визначити розміри зір за допомогою закону Стефана-Больцмана.

## Видима зоряна величина

Історія вимірювання яскравості зір починається ще з часів давньогрецького астронома Гіппарха (ІІ ст. до н.е.), який створив першу систему класифікації зір за їхньою яскравістю. Він розділив усі видимі зорі на шість груп, де зорі першої величини були найяскравішими, а шостої — ледь помітними неозброєним оком.

Видима зоряна величина (позначається літерою $m$) — це міра яскравості небесного об'єкта, яка спостерігається з Землі. Важливо розуміти, що ця шкала є **логарифмічною** і **оберненою**: чим менше значення величини, тим яскравіша зоря. Наприклад, зоря з видимою зоряною величиною $m = 1$ яскравіша за зорю з $m = 6$.

У сучасній астрономії ця система була формалізована і розширена. Різниця в 5 зоряних величин відповідає різниці в яскравості у 100 разів. Математично це виражається формулою:

$$\frac{F_1}{F_2} = 10^{0.4(m_2 - m_1)}$$

де $F_1$ та $F_2$ — це видимі потоки випромінювання від двох зір, а $m_1$ та $m_2$ — їхні видимі зоряні величини.

Цікаво, що шкала видимих зоряних величин не обмежується значеннями від 1 до 6. Найяскравіші об'єкти мають негативні значення: наприклад, Сіріус має $m = -1.46$, Венера може досягати $m = -4.6$, а повний Місяць має $m \approx -12.7$. Сонце, найяскравіший об'єкт на нашому небі, має видиму зоряну величину $m \approx -26.7$.

З іншого боку, сучасні телескопи дозволяють спостерігати дуже тьмяні об'єкти з величинами +30 і більше.

## Абсолютна зоряна величина

Проблема видимої зоряної величини полягає в тому, що вона залежить від відстані до об'єкта. Тьмяна, але близька зоря може мати таку ж видиму яскравість, як і дуже яскрава, але далека зоря. Щоб мати можливість порівнювати справжню яскравість різних зір, астрономи ввели поняття абсолютної зоряної величини.

Абсолютна зоряна величина (позначається літерою $M$) — це видима зоряна величина, яку мала б зоря, якби знаходилась на стандартній відстані 10 парсеків (близько 32.6 світлових років) від спостерігача.

Зв'язок між видимою ($m$) та абсолютною ($M$) зоряними величинами виражається формулою:

$$M = m - 5\log_{10}\left(\frac{d}{10}\right)$$

або в спрощеному вигляді:

$$M = m - 5\log_{10}(d) + 5$$

де $d$ — відстань до зорі у парсеках.

Різниця між видимою та абсолютною зоряними величинами називається модулем відстані:

$$m - M = 5\log_{10}(d) - 5$$

Абсолютна зоряна величина дозволяє нам порівнювати справжню яскравість зір незалежно від їхньої відстані від нас. Наприклад, Сонце має абсолютну зоряну величину $M \approx +4.83$, що означає, що з відстані 10 парсеків воно виглядало б як досить тьмяна зоря, яку можна було б побачити неозброєним оком за темного неба.

Для порівняння, надгігант Рігель у сузір'ї Оріона має $M \approx -7.84$, що робить його приблизно в 40 000 разів яскравішим за Сонце!

## Світність зорі

Рухаючись далі до більш фундаментальних характеристик зір, ми приходимо до поняття світності.

Світність зорі (позначається літерою $L$) — це повна кількість енергії, яку випромінює зоря за одиницю часу в усіх напрямках і на всіх довжинах хвиль. Зазвичай світність вимірюється у ватах (Вт) або як відношення до світності Сонця ($L_{\odot} \approx 3.828 \times 10^{26}$ Вт).

Світність є більш фундаментальною характеристикою зорі, ніж зоряна величина, оскільки вона безпосередньо пов'язана з фізичними процесами, що відбуваються у зорі, насамперед з термоядерними реакціями в її ядрі.

Зв'язок між абсолютною зоряною величиною та світністю виражається формулою:

$$\frac{L}{L_{\odot}} = 10^{0.4(M_{\odot} - M)}$$

де $M_{\odot} \approx +4.83$ — абсолютна зоряна величина Сонця.

У логарифмічному масштабі:

$$\log_{10}\left(\frac{L}{L_{\odot}}\right) = 0.4(M_{\odot} - M) \approx 0.4(4.83 - M)$$

Світність зорі залежить від двох основних параметрів: її радіуса та температури поверхні. Це приводить нас до наступного важливого закону в астрофізиці.

## Закон Стефана-Больцмана та визначення радіусів зір

Закон Стефана-Больцмана встановлює зв'язок між температурою абсолютно чорного тіла та енергією, яку воно випромінює з одиниці площі за одиницю часу. Хоча зорі не є ідеальними чорними тілами, вони досить добре описуються цією моделлю.

Згідно з законом Стефана-Больцмана, енергія, яку випромінює одиниця площі поверхні зорі за одиницю часу (потік випромінювання), пропорційна четвертій степені її ефективної температури:

$$F = \sigma T^4$$

де $F$ — потік випромінювання (Вт/м²), $T$ — ефективна температура поверхні зорі (К), а $\sigma \approx 5.67 \times 10^{-8}$ Вт·м⁻²·К⁻⁴ — стала Стефана-Больцмана.

Загальна світність зорі дорівнює потоку випромінювання, помноженому на площу поверхні зорі:

$$L = 4\pi R^2 \cdot \sigma T^4$$

де $R$ — радіус зорі.

З цієї формули ми можемо виразити радіус зорі:

$$R = \sqrt{\frac{L}{4\pi\sigma T^4}}$$

Часто зручніше виражати радіус зорі у відношенні до радіуса Сонця ($R_{\odot}$). Тоді:

$$\frac{R}{R_{\odot}} = \sqrt{\frac{L/L_{\odot}}{(T/T_{\odot})^4}}$$

де $T_{\odot} \approx 5778$ К — ефективна температура поверхні Сонця.

Оскільки світність залежить від абсолютної зоряної величини, ми можемо об'єднати попередні формули і виразити радіус зорі через її абсолютну зоряну величину та ефективну температуру:

$$\frac{R}{R_{\odot}} = \sqrt{10^{0.4(M_{\odot} - M)}} \cdot \left(\frac{T_{\odot}}{T}\right)^2$$

або

$$\frac{R}{R_{\odot}} = 10^{0.2(M_{\odot} - M)} \cdot \left(\frac{T_{\odot}}{T}\right)^2$$

Ця формула дає нам потужний інструмент для визначення розмірів зір. Знаючи абсолютну зоряну величину (яку можна обчислити з видимої величини та відстані) та ефективну температуру (яку можна визначити зі спектральних спостережень), ми можемо обчислити радіус зорі.

## Практичне застосування

Розглянемо приклад. Припустимо, ми спостерігаємо зорю з видимою зоряною величиною $m = 3.5$ на відстані $d = 50$ парсеків, і визначили, що її ефективна температура становить $T = 10000$ К.

1. Спочатку обчислимо абсолютну зоряну величину:
   $$M = m - 5\log_{10}(d) + 5 = 3.5 - 5\log_{10}(50) + 5 = 3.5 - 5 \cdot 1.699 + 5 = 3.5 - 8.495 + 5 = 0.005 \approx 0$$

2. Тепер обчислимо світність відносно Сонця:
   $$\frac{L}{L_{\odot}} = 10^{0.4(M_{\odot} - M)} = 10^{0.4(4.83 - 0)} \approx 10^{1.932} \approx 85.5$$
   Тобто ця зоря випромінює у 85.5 разів більше енергії, ніж Сонце.

3. Нарешті, обчислимо радіус:
   $$\frac{R}{R_{\odot}} = 10^{0.2(M_{\odot} - M)} \cdot \left(\frac{T_{\odot}}{T}\right)^2 = 10^{0.2 \cdot 4.83} \cdot \left(\frac{5778}{10000}\right)^2 \approx 10^{0.966} \cdot \left(0.5778\right)^2 \approx 9.25 \cdot 0.334 \approx 3.09$$
   
   Отже, радіус цієї зорі приблизно у 3.09 рази більший за радіус Сонця.

## Висновки

Розглянуті в цьому лекційному матеріалі поняття є фундаментальними для розуміння природи зір. Шлях від видимої яскравості, яку ми спостерігаємо з Землі, до фізичних характеристик зір (світності та радіуса) демонструє потужність астрономічних методів дослідження.

Зоряні величини (видима та абсолютна) дозволяють нам кількісно описувати яскравість зір у зручній логарифмічній шкалі. Світність зорі є мірою її енергетичної потужності. А закон Стефана-Больцмана разом з даними про світність і температуру дозволяє нам визначати розміри зір, які неможливо спостерігати безпосередньо через величезні відстані до них.

Ці знання є основою для побудови моделей еволюції зір, розуміння їхньої внутрішньої структури та процесів, що в них відбуваються, а також для класифікації зір на діаграмі Герцшпрунга-Рассела, яка є одним з найважливіших інструментів у сучасній астрофізиці.
