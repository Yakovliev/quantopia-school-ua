# Сонячна активність

Сонце – не статичний об'єкт. Його активність змінюється з часом, найвідомішим проявом чого є 11-річний сонячний цикл (іноді 22-річний, якщо враховувати зміну магнітної полярності).


Сонячний Цикл: Протягом приблизно 11 років кількість сонячних плям, спалахів, протуберанців та інших проявів активності зростає до максимуму, а потім спадає до мінімуму. Цей цикл пов'язаний зі складними процесами генерації та переорганізації магнітного поля Сонця всередині конвективної зони, що посилюється диференціальним обертанням.

Космічна Погода: Сонячна активність (спалахи, КВМ, сонячний вітер) безпосередньо впливає на Землю та космічний простір навколо неї. Це явище називають космічною погодою. Потужні викиди можуть викликати:
Полярні сяйва (Аврори).
Порушення радіозв'язку та роботи супутників GPS.
Збої в роботі енергосистем на Землі.
Підвищення рівня радіації для космонавтів та пасажирів літаків на великих висотах.

Феномен сонячної активності характеризується варіаціями кількості плям на поверхні Сонця та сумарною площею, яку вони займають. Наукові спостереження демонструють, що внутрішня частина плями відзначається доволі високою температурою, яка досягає показника 4500 К. Водночас, на тлі значно гарячішої фотосфери з температурним режимом 5780 К, ці утворення візуально сприймаються як темні області. Виникає логічне запитання: який механізм спричиняє зниження температурних показників усередині плями? Враховуючи, що сонячні плями здатні зберігати свою структуру впродовж декількох місяців, науковцями була висунута теорія про існування певного процесу, що перешкоджає конвективному руху плазми в межах плями та забезпечує підтримку температурного градієнту. Сучасні дослідження підтверджують, що роль такого "ізолятора" виконує потужне магнітне поле, яке, взаємодіючи з електрично зарядженими компонентами плазми, пригнічує конвекційні потоки в області плями.

Додатковою загадкою сонячної активності залишається її циклічність - послідовність змін кількісних показників плям, що повторюється з періодичністю приблизно 11 років.

Структурно сонячні плями об'єднані між собою магнітними силовими лініями, подібно до полюсів у магнітних системах, причому кожне таке утворення характеризується власною полярністю. Аналогічно до неможливості фізичного розділення північного та південного полюсів магніту, сонячні плями функціонують виключно як парні структури з протилежними магнітними полярностями. При врахуванні полярності плям, повний цикл сонячної активності становить близько 22 років.

## Вплив сонячної активності на Землю

Астрономічні дослідження Сонця, проведені за допомогою супутникових систем та автоматичних міжпланетних станцій, виявили інтенсивне корпускулярне випромінювання - потік фундаментальних частинок (протонів, нейтронів, електронів). Зокрема, хромосферні спалахи, що виникають у безпосередній близькості до плям, супроводжуються вивільненням колосального обсягу енергії, порівнянного з випромінюванням цілісної фотосфери Сонця. Важливо диференціювати спалахи від протуберанців. **Протуберанці** є постійними елементами сонячної активності - це компактні холодні водневі хмаровидні утворення, що піднімаються до корони та переміщуються вздовж магнітних силових ліній. Протуберанці забезпечують циркуляцію речовини між хромосферним шаром та короною.

**Хромосферний спалах** представляє собою короткочасне суттєве підвищення світлової інтенсивності локалізованої ділянки хромосфери Сонця, що супроводжується вибуховим вивільненням речовини та енергії, акумульованої в магнітному полі сонячних плям.

Спалахові явища формуються в просторі між двома плямами з протилежною полярністю, коли протягом кількох годин відбувається стрімке підвищення температури в цій зоні до рівня $5 \cdot 10^6$ К, що супроводжується генерацією енергії порядку $10^21-10^25$ Дж - показника, майже еквівалентного світимості Сонця у видимому діапазоні спектра. В процесі спалаху енергія випромінюється переважно в невидимих спектральних діапазонах (радіо-, ультрафіолетовому та рентгенівському). Одночасно з цим у міжпланетний простір відбувається викид потоків заряджених частинок, що рухаються зі швидкістю до 20 000 км/с. Після кількагодинного польоту такі корпускулярні потоки можуть досягати земної атмосфери, викликаючи збурення магнітного поля планети та іонізацію верхніх шарів атмосфери, що візуалізується у формі яскравих полярних сяйв.

**Магнітна буря** характеризується збуренням магнітного поля Землі внаслідок сонячного спалаху. Такі періоди супроводжуються неполадками радіокомунікаційних систем та електронного обладнання, а також погіршенням фізіологічного стану людей.

**Увага!** При проведенні спостережень Сонця категорично забороняється безпосередній візуальний контакт із сонячним диском як неозброєним оком, так і через телескопічні прилади без застосування спеціалізованих світлофільтрів!

Сонячні плями
Сонячні плями — це темні області на фотосфері Сонця з сильним магнітним полем. Напруженість магнітного поля в плямах у тисячі разів перевищує напруженість поля в спокійних областях фотосфери. Через сильне магнітне поле конвекція в цих областях пригнічується, що призводить до зниження температури (приблизно 4 000 К порівняно з 5 700 К у навколишній фотосфері) і, відповідно, до зменшення яскравості.


Сонячні плями зазвичай з'являються групами і можуть існувати від кількох днів до кількох місяців. Кількість і площа плям змінюються з циклом приблизно в 11 років.

Сонячні спалахи і корональні викиди маси
Сонячні спалахи — це потужні вибухи на поверхні Сонця, під час яких вивільняється енергія, еквівалентна мільярдам мегатонн тротилу. Спалахи тривають від кількох хвилин до кількох годин і супроводжуються викидом високоенергетичного випромінювання (від радіохвиль до гамма-променів) та прискоренням заряджених частинок.
Корональні викиди маси (КВМ) — це масштабні викиди плазми і магнітного поля з сонячної корони в міжпланетний простір. КВМ можуть виносити мільярди тонн сонячної речовини зі швидкістю до 3 000 км/с. Якщо КВМ спрямований у бік Землі, він може спричинити геомагнітні бурі, полярні сяйва та різноманітні порушення в роботі технічних систем.

Сонячний вітер
Сонячний вітер — це потік заряджених частинок (переважно протонів та електронів), що постійно витікає з сонячної корони. Швидкість сонячного вітру варіюється від 300 до 800 км/с залежно від активності Сонця та області, з якої він витікає:

Повільний сонячний вітер (300-400 км/с) походить з областей із замкнутими магнітними петлями
Швидкий сонячний вітер (600-800 км/с) витікає з корональних дір — областей з відкритими магнітними лініями

Сонячний вітер формує геліосферу і взаємодіє з магнітними полями планет, створюючи магнітосфери. Взаємодія сонячного вітру з магнітосферою Землі призводить до різноманітних геофізичних явищ, включаючи полярні сяйва та геомагнітні бурі.

Механізми сонячного динамо — як саме генерується магнітне поле Сонця і що визначає 11-річний цикл сонячної активності?
