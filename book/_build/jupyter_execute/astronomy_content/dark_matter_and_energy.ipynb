## Чому вчені припускають існування темної матерії?

1. Криві обертання галактик:
Спостереження показують, що зорі в зовнішніх областях спіральних галактик обертаються практично з тією ж швидкістю, як і зорі ближче до центру. Якщо маса галактики була б визначена лише за видимою матерією, швидкість обертання мала б спадати із збільшенням відстані. Наявність додаткової, невидимої маси (тобто темної матерії) дозволяє зберегти практично постійну швидкість обертання на великих відстанях від центру.

2. Гравітаційне лінзування:
Коли світло від далеких об’єктів (наприклад, квазарів) проходить поруч із великими скупченнями галактик, воно викривляється під впливом гравітаційного поля. Аналізуючи ці ефекти, вчені приходять до висновку, що маса скупчення значно перевищує суму мас видимих об’єктів, що вказує на наявність великої кількості невидимої матерії.


## Які кандидати на роль частинок темної матерії?

Оскільки темна матерія не взаємодіє електромагнітно, її пряме виявлення є дуже складним. Сьогодні основні кандидати включають:

1. WIMPs (Weakly Interacting Massive Particles):
Це слабко взаємодіючі масивні частинки, які за масою можуть бути в десятки чи навіть сотні разів важчими за протон. Вони взаємодіють гравітаційно та через слабку ядерну взаємодію, але майже не взаємодіють зі звичайною матерією. Пошуки WIMP ведуться за допомогою спеціалізованих підземних детекторів.

2. Аксіони:
Дуже легкі гіпотетичні частинки, які виникли як розв’язання проблеми сильного CP-інваріантності в квантовій хромодинаміці. Незважаючи на малу масу, вони можуть накопичуватися у великій кількості та діяти як холодна темна матерія. Деякі експерименти, наприклад XENON1T, шукали сигнали, які могли б бути пов’язані з аксіонами.

Також існують альтернативні теорії, що пояснюють аномалії у гравітаційному впливі не за рахунок невидимої матерії, а через модифікації законів гравітації. Проте ці підходи поки не отримали широкого експериментального підтвердження.


## Чому важливо, щоб темна матерія була "холодною"?

"Холодна" темна матерія означає, що частинки рухаються повільно порівняно з швидкістю світла. Такий стан дозволяє їм накопичуватись у гравітаційних гало та сприяти формуванню структур – від карликових галактик до надскупчень. Якщо б темна матерія була "гарячою" (тобто реліктові нейтрино, що рухаються з великою швидкістю), вона згладжувала б маленькі структури і не дала б змоги формуватись галактикам так, як ми їх спостерігаємо.

## Чорні діри - не темна матерія

Чорні діри – це об’єкти, що утворилися внаслідок гравітаційного колапсу звичайної (баріонної) матерії, і їх можна виявити через гравітаційне лінзування, випромінювання акреційних дисків або вплив на рух навколишніх об’єктів. На відміну від темної матерії, яка, за визначенням, не випромінює, не поглинає і не відбиває електромагнітне випромінювання, чорні діри активно взаємодіють із світлом – вони його поглинають, що і призводить до появи горизонту подій. Таким чином, хоча чорні діри "чорні" в тому сенсі, що світло, що потрапляє до них, не може вийти, вони не підпадають під категорію темної матерії. Темна матерія, як правило, є розподіленою невидимою субстанцією, яка не взаємодіє електромагнітно, і її існування намагаються встановити виключно через гравітаційні ефекти, а не через пряме спостереження випромінювання.
