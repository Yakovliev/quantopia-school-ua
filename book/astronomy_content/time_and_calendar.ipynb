**Час та календар: Історія вимірювання часу та розвиток календарних систем**

Розвиток систем вимірювання часу та створення календарів є одним із найважливіших досягнень людства. Сучасний календар, яким користуються європейські країни, базується на тропічному році. Однак створення ідеального календаря виявилося складним завданням через те, що тропічний рік не містить цілого числа діб.

Історично першою системою, яка набула широкого поширення в Європі, був юліанський календар. Цей календар, запроваджений Юлієм Цезарем у 46 році до нашої ери, встановлював тривалість року у 365 діб і 6 годин. Для компенсації цих додаткових годин було впроваджено систему високосних років: три роки по 365 діб, а четвертий - 366 діб. 

Проте юліанський календар мав певну неточність. Середня тривалість календарного року перевищувала тропічний рік на 11 хвилин і 14 секунд. За чотири століття різниця складала майже три доби, що призводило до поступового зміщення календарних дат відносно астрономічних явищ.

У 1582 році було здійснено важливу реформу календаря за ініціативою Папи Григорія XIII. Новий календар, який отримав назву григоріанський, виправив накопичену помилку та запровадив більш точну систему високосних років. За григоріанським календарем, рік вважається високосним, якщо він ділиться на 4, за винятком років, які закінчуються двома нулями і не діляться на 400.

**Системи вимірювання часу**

Вимірювання часу завжди було тісно пов'язане з астрономічними спостереженнями. Основною одиницею виміру часу стала доба - період повного обертання Землі навколо своєї осі. Для точного визначення часу використовують два типи відліку:

1. Зоряний час - базується на спостереженні за зірками
2. Сонячний час - заснований на видимому русі Сонця

Справжня сонячна доба визначається як проміжок часу між двома послідовними однойменними кульмінаціями центра сонячного диска. У повсякденному житті використовується поясний час, який було запроваджено для зручності. Земля поділена на 24 часових пояси, і в межах кожного поясу встановлюється єдиний час, що відповідає часу центрального меридіана даного поясу.

Нульовий пояс проходить через Гринвіцький меридіан, і час за цим меридіаном називається всесвітнім часом. Цей час використовується в астрономії для фіксації моментів різних космічних явищ.

# Визначення часу з астрономічних спостережень. Календарні системи

Людство здавна спостерігало за небесними тілами, щоб вимірювати час. Найбільш помітним для спостерігачів було обертання небесної сфери та пов'язані з цим явища сходу і заходу світил. Саме ці спостереження дали нам природну одиницю часу - добу. При цьому астрономи розрізняють два типи діб: зоряну та сонячну.

**Зоряний час і доба**

Зоряна доба визначається як період між двома послідовними верхніми кульмінаціями точки весняного рівнодення на одному меридіані. При цьому зоряний час ($s$) - це часовий інтервал, який пройшов від моменту верхньої кульмінації точки весняного рівнодення.

Важливо розуміти, що через обертання Землі навколо своєї осі, на різних географічних меридіанах кульмінація точки весняного рівнодення відбувається в різні моменти. Якщо позначити через $s_0$ зоряний час на нульовому (Грінвіцькому) меридіані, то для спостерігача, який знаходиться на схід від Грінвіча з географічною довготою λ (виражена в годинах), зоряний час обчислюється за формулою: $s = s_0 + λ$

**Сонячний час**

Повсякденне життя людини тісно пов'язане з видимим рухом Сонця - його сходом, кульмінацією та заходом. Фактично, ми живемо за сонячним часом. Справжня сонячна доба - це інтервал між двома послідовними однойменними кульмінаціями центру сонячного диска на одному географічному меридіані.

За початок справжньої сонячної доби приймається момент нижньої кульмінації Сонця. Проте тривалість справжньої сонячної доби не є сталою величиною через два фактори:
- Земля рухається навколо Сонця по еліптичній орбіті нерівномірно
- Сонце рухається по екліптиці, яка нахилена до небесного екватора

Для практичного використання було введено поняття середнього сонця - фіктивної точки, яка рівномірно рухається вздовж небесного екватора і здійснює повний оберт за той самий час, що й Сонце повертається до точки весняного рівнодення.

**Системи часу**

Місцевий час ($T_м$) вимірюється на певному географічному меридіані. Для зручності у 1884 році було прийнято систему поясного часу, розділивши земну кулю на 24 годинних пояси з шириною 15°. Центральний меридіан кожного поясу визначає поясний час ($T_п$).

Всесвітній час ($T_0$) - це місцевий середній час Грінвіцького меридіана. Для будь-якого меридіана східної довготи λ кульмінації світил відбуваються на λ годин раніше, ніж на Грінвіцькому. Тому місцевий середній сонячний час пов'язаний з $T_0$ формулою: $T_м = T_0 + λ$

**Тропічний і зоряний роки**

Тропічний рік (365,2422 доби) - це період повного оберту Сонця відносно точки весняного рівнодення. Цей цикл включає всі сезони року. Зоряний рік (365,2561 доби) - період повного оберту Сонця відносно зір.

У сучасному календарі використовується тропічний рік. Григоріанський календар, введений у 1582 році папою Григорієм XIII, враховує різницю між календарним і тропічним роком за допомогою особливої системи високосних років. Цей календар дає похибку приблизно одну добу за 33 століття.

# Типи календарів та визначення часу в астрономії

**Еволюція календарних систем**

Сучасне літочислення у європейських країнах базується на тропічному році, проте створення ідеального календаря завжди було складним завданням через те, що тропічний рік не містить цілого числа діб.

Історично важливим етапом розвитку календарних систем став юліанський календар, запроваджений Юлієм Цезарем у 46 році до нашої ери. Цей календар враховував, що тривалість тропічного року становить 365 діб 6 годин 00 хвилин 00 секунд. Для компенсації додаткових 6 годин було встановлено правило: три роки поспіль мали по 365 діб, а четвертий рік містив 366 діб (високосний рік). 

Однак реальна тривалість тропічного року виявилася коротшою - 365 діб 5 годин 48 хвилин 46 секунд. Через цю різницю в 11 хвилин 14 секунд календар поступово відставав від астрономічних явищ. За чотири століття накопичувалось близько 3 діб розбіжності.

**Григоріанська реформа**

Для виправлення цієї неточності у 1582 році за ініціативою Папи Григорія XIII було проведено календарну реформу. Новий календар отримав назву григоріанського (або нового стилю). Суть реформи полягала у наступному:
- У високосних роках, які закінчуються двома нулями (1700, 1800, 1900, 2000, 2100 тощо), високосним вважається тільки той рік, число сотень якого ділиться на 4 без остачі
- 1582 рік було скорочено на 10 діб для компенсації накопиченої різниці

**Вимірювання часу в астрономії**

Час є фундаментальною категорією, що має філософське, фізичне та соціальне значення. Точне вимірювання часу залишається однією з найважливіших задач сучасної науки. З повсякденного досвіду ми знаємо, що час плине рівномірно, подібно до течії спокійної річки.

Розвиток технологій вимірювання часу пройшов довгий шлях:
- Водяні та пісочні годинники у давнину
- Механічні годинники з маятником
- Сучасні електронні годинники, що використовують коливання кристалів в електромагнітному полі
- Атомні годинники найвищої точності

**Системи відліку часу**

В астрономії використовують дві основні системи відліку часу:
- Зоряний час - використовується переважно астрономами
- Сонячний час - застосовується у повсякденному житті

Важливим поняттям є доба - період одного повного обертання Землі навколо своєї осі. В астрономії використовують поняття справжньої сонячної доби - проміжок часу між двома послідовними однойменними кульмінаціями центра сонячного диска.

**Часові пояси**

У XIX столітті було впроваджено систему часових поясів - Землю поділили меридіанами на 24 годинні пояси. У межах кожного поясу всі годинники показують однаковий час. При перетині межі часового поясу час змінюється на одну годину.

Нульовий пояс проходить через Гринвіцький меридіан, тому час у Великій Британії називають всесвітнім часом. Цей час використовується в астрономії для визначення моментів космічних подій.

# Зоряне небо та основи астрономічних спостережень

**Сонячний час і зодіакальні сузір'я**

Протягом тисячоліть люди спостерігали за рухом небесних тіл і помітили, що Сонце, Місяць та планети рухаються вздовж певного шляху на небесній сфері. Цей шлях проходить через 12 сузір'їв, які отримали назву зодіакальних. Термін "зодіак" походить від грецького слова, що означає "коло зображень тварин", оскільки більшість цих сузір'їв названі на честь живих істот.

Спостерігаючи за річним рухом Землі навколо Сонця, можна помітити, що наша планета проходить через різні зодіакальні сузір'я. Якщо провести уявну лінію від центру Землі до Сонця та продовжити її до зоряного неба, то ця лінія перетне небесну сферу в точці, яка знаходиться в одному із зодіакальних сузір'їв. 

Екліптика - це велике коло небесної сфери, яке утворюється внаслідок перетину площини орбіти Землі з небесною сферою. Математично екліптика визначається як лінія перетину площини орбіти Землі з небесною сферою. Ця площина зберігає майже незмінне положення відносно зір протягом тривалого часу.

**Прецесія земної осі**

Важливим явищем, яке впливає на спостереження небесних об'єктів, є прецесія - поступова зміна напрямку осі обертання Землі. Вісь обертання Землі описує в просторі конус з періодом близько 26000 років. Внаслідок цього явища положення полюсів світу на небесній сфері змінюється. Якщо зараз Полярна зоря знаходиться поблизу північного полюса світу, то через приблизно 13000 років полюс переміститься в напрямку сузір'я Ліри.

**Орієнтування на місцевості**

Для визначення свого положення на поверхні Землі астрономи використовують систему координат, основану на поняттях горизонту та прямовисної лінії. Напрямок прямовисної лінії визначається силою тяжіння Землі в точці спостереження. Горизонт являє собою площину, перпендикулярну до прямовисної лінії.

На горизонті виділяють чотири головні точки: північ (N), південь (S), схід (E) і захід (W). Точка зеніту (Z) - це точка перетину прямовисної лінії з небесною сферою над головою спостерігача. Протилежна їй точка називається надиром.

**Кульмінація світил**

Внаслідок добового обертання Землі навколо своєї осі небесні світила здійснюють видимий рух по небу. Момент проходження світила через найвищу точку над горизонтом називається верхньою кульмінацією. Цей момент настає, коли світило перетинає площину меридіана.

У весняний та осінній періоди Сонце сходить близько точки сходу та заходить біля точки заходу. Влітку точки сходу та заходу Сонця зміщуються до півночі, а взимку - до півдня. Деякі зорі в наших широтах ніколи не заходять за горизонт, тому для них можна спостерігати як верхню, так і нижню кульмінацію.
