# Сила пружності

Уявімо, що ми взяли до рук звичайну пружинку від ручки. Якщо ми її трохи розтягнемо, її довжина збільшиться. Якщо стиснемо - зменшиться. Ця зміна форми або розмірів тіла під дією зовнішніх сил і називається **деформацією**.

## Якою буває деформація?

Не всі деформації однакові. Головна відмінність полягає в тому, чи повертається тіло до своєї початкової форми. За цією ознакою ми виділяємо два основні типи деформацій.

> **Пружна деформація** - це така деформація, яка повністю зникає після припинення дії зовнішньої сили. Тіло повертається до своїх початкових розмірів і форми.

**Приклад:** Уявіть гумовий м'яч. Коли він вдаряється об підлогу, він на мить сплющується (деформується), але одразу ж відновлює свою сферичну форму. Це і є пружна деформація. Те саме відбувається з пружинами, батутом, струнами гітари.

> **Пластична деформація** - це деформація, яка не зникає після припинення дії зовнішньої сили. Тіло назавжди змінює свою форму.

**Приклад:** Візьмемо шматок пластиліну. Коли ми його стискаємо, він змінює форму. Але якщо ми приберемо руку, він так і залишиться сплющеним. Те саме станеться, якщо зігнути металеву скріпку - вона не розігнеться сама по собі.

Важливо розуміти, що одне й те саме тіло може зазнавати як пружної, так і пластичної деформації. Якщо ми ледь-ледь розтягнемо пружину, вона повернеться у вихідне положення (пружна деформація). Але якщо ми потягнемо її з усієї сили, вона може розтягнутися і вже ніколи не стиснутися назад (пластична деформація). Межа, за якою пружна деформація переходить у пластичну, називається **межею пружності**.

## Види деформацій за характером дії сили

Залежно від того, як саме прикладена сила, ми можемо класифікувати деформації на декілька основних видів.

*   **Деформація розтягнення та стиснення:** Виникають, коли сили діють уздовж однієї прямої, намагаючись або збільшити (розтягнення), або зменшити (стиснення) довжину тіла. Приклади: розтягнення троса під вагою вантажу, стиснення колон будівлі.

*   **Деформація згину:** Це складніший вид деформації. Уявіть, що ми кладемо дошку між двома стільцями і стаємо на неї посередині. Верхні шари дошки стискаються, а нижні - розтягуються. Це і є згин.

*   **Деформація кручення:** Виникає, коли до тіла прикладають сили, що намагаються його "скрутити" навколо своєї осі. Класичний приклад - викручування мокрої ганчірки. Вали двигунів, свердла - всі вони зазнають деформації кручення.

*   **Деформація зсуву:** Виникає, коли шари тіла зміщуються один відносно одного. Уявіть колоду карт: якщо ми натиснемо на верхню карту збоку, вся колода перекоситься - це наочна деформація зсуву. Заклепки та болти, що з'єднують металеві листи, працюють саме на зсув.



## Як виміряти деформацію? Сила пружності

Тепер, коли ми знаємо, що таке деформація, давайте навчимося її вимірювати і зрозуміємо, яка сила їй протидіє. Для простоти ми будемо розглядати найпростіший випадок - деформацію розтягнення або стиснення.

### Абсолютне видовження

Кількісною мірою деформації розтягнення/стиснення є **абсолютне видовження**.

> **Абсолютне видовження** - це фізична величина, що показує, наскільки змінилася довжина тіла.

Позначається вона зазвичай літерою $x$ або $\Delta l$ (читається "дельта ель"). У різних джерелах можуть позначати по різному.

Якщо початкова довжина тіла (наприклад, нерозтягнутої пружини) була $l_0$, а після розтягнення стала $l$, то абсолютне видовження обчислюється за формулою:

$$
x = \Delta l = l - l_0
$$

**Одиниці вимірювання в СІ:** метр (м).

Якщо тіло розтягують, то $l > l_0$ і видовження $x$ буде додатним. Якщо тіло стискають, то $l < l_0$ і видовження буде від'ємним.

### Сила пружності

Коли ми намагаємося деформувати пружне тіло, ми відчуваємо опір. Цей опір створює внутрішня сила, яка виникає в тілі й намагається повернути його до початкового стану.

> **Сила пружності** $\vec{F}_{\text{пруж}}$ - це сила, що виникає всередині тіла під час його деформації і напрямлена в бік, протилежний напрямку зміщення частинок тіла.

**Одиниці вимірювання в СІ:** Ньютон (Н).

Якщо ми розтягуємо пружину вправо, то сила пружності буде тягнути її вліво. Якщо стискаємо пружину вліво, сила пружності буде штовхати її вправо. Вона завжди прагне відновити початковий стан і повернути все, як було.

## Закон Гука

Ще в 17 столітті англійський вчений Роберт Гук експериментально встановив дивовижно простий зв'язок між силою пружності та величиною деформації. Цей зв'язок відомий нам як **закон Гука**.

> **Закон Гука:** Для малих пружних деформацій сила пружності прямо пропорційна видовженню тіла і напрямлена протилежно до нього.

Математично цей закон записується у векторній формі так:

$$
\vec{F}_{\text{пруж}} = -k\vec{x}
$$

Давайте розберемо цю формулу:
*   $\vec{F}_{\text{пруж}}$ - вектор сили пружності.
*   $\vec{x}$ - вектор видовження (зміщення від положення рівноваги).
*   Знак "мінус" - найважливіша частина! Він математично показує те, про що ми говорили: сила пружності **завжди** напрямлена в протилежний бік до деформації. Якщо $\vec{x}$ напрямлений вправо, то $\vec{F}_{\text{пруж}}$ - вліво.
*   $k$ - коефіцієнт пропорційності, який називається **жорсткістю** тіла.

Для розв'язування задач, де нам важлива лише величина (модуль) сили, ми можемо записати закон Гука в скалярній формі, без векторів та мінуса:

$$
F_{\text{пруж}} = k|x|
$$

Тут $|x|$ - це модуль видовження, тобто абсолютна зміна довжини.

> **Жорсткість** $k$ - це фізична величина, яка характеризує пружні властивості тіла. Вона показує, яку силу потрібно прикласти до тіла, щоб змінити його довжину на одиницю.

З формули закону Гука ми можемо виразити жорсткість:

$$
k = \frac{F_{\text{пруж}}}{|x|}
$$

**Одиниця вимірювання в СІ:** Ньютон на метр (Н/м).

**Аналогія:** Уявіть дві пружини. Одна - від підвіски вантажівки, а інша - від кулькової ручки. Щоб розтягнути пружину від вантажівки на 1 см, потрібна величезна сила. А пружину від ручки можна розтягнути на той самий 1 см легким рухом пальців. Це означає, що жорсткість пружини вантажівки (її значення $k$) у багато разів більша, ніж жорсткість пружини від ручки.

Жорсткість залежить від матеріалу, з якого виготовлено тіло, його форми та геометричних розмірів (наприклад, для дротини - від її довжини та площі перерізу).

## Природа сили пружності

А звідки ж узагалі береться ця сила? Чому тіло хоче повернутись до початкової форми? Відповідь криється на мікроскопічному рівні.

Всі тіла складаються з атомів і молекул, які перебувають на певних відстанях один від одного і взаємодіють між собою. Цю взаємодію можна уявити як мільярди крихітних пружинок, що з'єднують сусідні частинки. Це, звісно, аналогія, але вона добре пояснює суть. Насправді ця взаємодія має **електромагнітну природу**. Атоми та молекули в речовині одночасно притягуються і одночасно відштовхуються, але мій цим притяганням та відштовхуванням встановлюється рівновага. Однак рівновага порушується, коли ми починаємо деформувати тіло:
*   Коли ми **розтягуємо** тіло, ми збільшуємо відстань між атомами/молекулами. Сили притягання між ними стають сильнішими за сили відштовхування й намагаються повернути їх на місце. Сума всіх цих мікроскопічних сил притягання і створює макроскопічну силу пружності.
*   Коли ми **стискаємо** тіло, ми зближуємо атоми. На малих відстанях починають домінувати сили відштовхування між їхніми електронними оболонками. Ці сили намагаються "розштовхати" атоми назад. Сума цих сил відштовхування також створює силу пружності.

Отже, сила пружності - це не якась окрема фундаментальна сила природи, а прояв електромагнітної взаємодії між частинками речовини.

## З'єднання пружин

На практиці часто доводиться використовувати не одну, а кілька пружин, з'єднаних разом. Розглянемо два основні способи з'єднання та виведемо формули для загальної жорсткості системи.

### Послідовне з'єднання

Пружини з'єднані послідовно, якщо вони прикріплені одна до одної "ланцюжком".

Давайте проаналізуємо цю систему.
1.  **Сила:** Якщо ми потягнемо за кінець другої пружини з силою $F$, то така сама сила діятиме на кожну пружину в ланцюжку. Це схоже на натяг мотузки - він однаковий по всій довжині. Отже, $F_{\text{пруж}} = F_1 = F_2$.
2.  **Видовження:** Загальне видовження всієї системи дорівнює сумі видовжень кожної окремої пружини. $x = x_1 + x_2$.

Тепер використаємо ці два факти для виведення формули загальної жорсткості $k$.

Із закону Гука ми знаємо, що $x = \frac{F}{k}$. Підставимо це у формулу для видовження:

$$
\frac{F_{\text{пруж}}}{k} = \frac{F_1}{k_1} + \frac{F_2}{k_2}
$$

Оскільки всі сили однакові ($F_{\text{пруж}} = F_1 = F_2$), ми можемо скоротити їх:

$$
\frac{1}{k} = \frac{1}{k_1} + \frac{1}{k_2}
$$

Зводимо до спільного знаменника і отримуємо фінальну формулу:

$$
\frac{1}{k} = \frac{k_2}{k_1 k_2} + \frac{k_1}{k_1 k_2} = \frac{k_1 + k_2}{k_1 k_2}
$$

$$
k = \frac{k_1 k_2}{k_1 + k_2}
$$

Це і є формула для розрахунку загальної жорсткості при послідовному з'єднанні. Із цієї формули можна зробити висновок, що при послідовному з'єднанні загальна жорсткість системи **зменшується** і стає меншою за жорсткість найменш жорсткої пружини. Системі стає "легше" розтягуватися.

### Паралельне з'єднання

При паралельному з'єднанні пружини кріпляться до спільної основи та спільного вантажу. Вони працюють разом, "пліч-о-пліч".

Проаналізуємо цю систему:
1.  **Видовження:** Оскільки пружини закріплені між тими самими основою та вантажем, вони розтягуються на однакову величину. $x = x_1 = x_2$.
2.  **Сила:** Загальна сила, що розтягує систему, розподіляється між пружинами. Кожна пружина робить свій "внесок". Тому загальна сила пружності дорівнює сумі сил пружності кожної пружини. $F_{\text{пруж}} = F_1 + F_2$.

Використаємо ці факти для виведення формули. Із закону Гука $F = k x$. Підставимо це у формулу для сил:

$$
k x = k_1 x_1 + k_2 x_2
$$
Оскільки всі видовження однакові ($x_{\text{заг}} = x_1 = x_2$), ми можемо скоротити їх:

$$
k = k_1 + k_2
$$

**Висновок:** При паралельному з'єднанні загальна жорсткість системи **збільшується** і дорівнює сумі жорсткостей окремих пружин і завжди є більшою за найбільшу з них. Система стає значно "жорсткішою".
