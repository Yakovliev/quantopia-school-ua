# Потенціальна енергія та закон збереження енергії

## Потенціальна енергія

Уявімо собі яблуко, що висить на гілці. Воно нерухоме, тому його кінетична енергія дорівнює нулю. Але якщо гілка зламається, яблуко почне падати, набираючи швидкість, а отже, і кінетичну енергію. Звідки взялася ця енергія? Вона була "накопичена" в яблуці завдяки його положенню відносно Землі. Або візьмемо стрілу, натягнуту на тятиві лука. Стріла нерухома, але варто відпустити тятиву, і стріла полетить з величезною швидкістю. Енергія була накопичена у вигнутій дузі лука.

Ця "накопичена" енергія, яка залежить від взаємного розташування тіл або частин одного тіла, і є потенціальною енергією.

> **Потенціальна енергія $E_p$** - це частина енергії фізичної системи, що виникає завдяки взаємодії між тілами, які складають систему, або внаслідок взаємодії із зовнішніми відносно цієї системи тілами.

Як і будь-який інший вид енергії, в системі СІ потенціальна енергія вимірюється в **Джоулях (Дж)**.

### Консервативні (потенціальні) сили

Поняття потенціальної енергії має сенс лише для певного класу сил, які називаються **консервативними** або **потенціальними**.

> **Консервативні сили** - це сили, робота яких не залежить від траєкторії руху тіла, а визначається лише його початковим і кінцевим положенням.

Ключова властивість таких сил: **робота консервативної сили на будь-якій замкненій траєкторії дорівнює нулю.**

**Приклади консервативних сил:**
*   **Сила тяжіння:** Робота, яку виконує сила тяжіння при переміщенні тіла з точки А в точку Б, не залежить від того, чи кинули ми тіло по параболі, чи спустили по похилій площині. Вона залежить лише від різниці висот початковою точкою та кінцевою точкою.
*   **Сила пружності:** Робота, яку виконує пружина при розтягуванні чи стисканні, залежить лише від початкової та кінцевої деформації, а не від того, як швидко чи повільно ми її розтягували.

**А які сили не є консервативними?**

Найяскравіший приклад - **сила тертя ковзання**. Її робота *дуже* залежить від пройденого шляху. Якщо ми пересунемо ящик з точки А в точку В по прямій, робота сили тертя буде одна. А якщо ми пересунемо його по довгій звивистій траєкторії, то робота сили тертя буде значно більшою. Робота сили тертя по замкненому шляху ніколи не дорівнює нулю - вона завжди від'ємна, оскільки спрямована проти руху тіла. Для таких неконсервативних сил поняття потенціальної енергії не вводять.

## Потенціальна енергія тіла в полі тяжіння Землі

Давайте розглянемо найпростіший випадок: тіло масою $m$ переміщується з початкової точки А, що знаходиться на висоті $h_0$, до кінцевої точки В, що знаходиться на висоті $h$ над поверхнею Землі. Ми розглядаємо загальний випадок, тому точка В не обов'язково буде рівно під точкою А, ця точка В може бути трохи в стороні. Таким чином, у нас є безліч різних різних шляхів, якими можна потрапити з точки А в точку В. Ми розглянемо два шляхи і порівняємо для них результат. Нехай маємо прямокутний трикутник АВС, де АВ - це гіпотенуза. Точки В та С знаходяться на одній висоті $h$.

Перша траєкторія: рух від точки А вертикально вниз до точки С і потім горизонтально до точки В. Друга траєкторія - рух по гіпотенузі АВ з точки А в точку В. На тіло діє стала сила тяжіння $\vec{F_т}$, напрямлена вертикально вниз. Знайдемо роботу цієї сили для обох траєкторій.

### Траєкторія 1: рух по ламаній АСВ

Роботу на цій траєкторії ми можемо розбити на дві частини: роботу на вертикальній ділянці АС та на горизонтальній ділянці СВ.

1.  **Ділянка АС (вертикальний рух):**
    *   Переміщення $s_{AC} = h_0 - h$ напрямлене вертикально вниз.
    *   Сила тяжіння $\vec{F_т}$ також напрямлена вертикально вниз.
    *   Оскільки напрямки сили і переміщення збігаються, кут $\alpha = 0^\circ$, а $\cos(0^\circ) = 1$.
    *   Робота на цій ділянці: $A_{AC} = F_т \cdot s_{AC} \cdot \cos(0^\circ) = mg(h_0 - h)$.

2.  **Ділянка СВ (горизонтальний рух):**
    *   Переміщення $s_{CB}$ напрямлене горизонтально.
    *   Сила тяжіння $\vec{F_т}$ напрямлена вертикально вниз.
    *   Кут між вектором сили і вектором переміщення становить $\alpha = 90^\circ$, а $\cos(90^\circ) = 0$.
    *   Робота на цій ділянці: $A_{CB} = F_т \cdot s_{CB} \cdot \cos(90^\circ) = 0$. Сила тяжіння не виконує роботи при горизонтальному русі.

3.  **Повна робота** на траєкторії АСВ дорівнює сумі робіт на ділянках:

    $$
    A_1 = A_{AC} + A_{CB} = mg(h_0 - h) + 0 = mgh_0 - mgh
    $$


### Траєкторія 2: рух по гіпотенузі АВ

1.  На цій траєкторії тіло рухається по прямій з точки А в точку В. Довжину цього шляху позначимо як $s_{AB}$.
2.  Сила тяжіння $\vec{F_т}$ напрямлена вертикально вниз.
3.  Робота сили тяжіння: $A_2 = F_т \cdot s_{AB} \cdot \cos\alpha$, де $\alpha$ - кут між вектором сили тяжіння (вертикаль) і вектором переміщення $\vec{s}_{AB}$.
4.  З прямокутного трикутника АВС видно, що добуток $s_{AB} \cdot \cos\alpha$ - це проєкція гіпотенузи АВ на вертикальний катет АС. А довжина цього катета якраз дорівнює різниці висот $h_0 - h$.
5.  Таким чином, $s_{AB} \cdot \cos\alpha = h_0 - h$.
6.  Підставимо це у формулу для роботи:
    $$
    A_2 = mg \cdot (s_{AB} \cos\alpha) = mg(h_0 - h) = mgh_0 - mgh
    $$

### Потенціальна енергія піднятого тіла

Ми отримали, що $A_1 = A_2$. Робота сили тяжіння виявилася однаковою для обох траєкторій. Можна довести, що вона буде такою ж для **будь-якої** траєкторії, що з'єднує точки А і В. Це і є головна ознака **консервативної сили**: її робота залежить лише від початкового та кінцевого положення тіла, а не від форми шляху.

Формула, яку ми отримали для роботи, надзвичайно важлива:

$$
A_т = mgh_0 - mgh
$$

Вона показує, що робота сили тяжіння дорівнює різниці деякої величини, що залежить від висоти. Цю величину $mgh$ ми й називаємо **потенціальною енергією тіла, піднятого над Землею**:

$$
E_p = mgh
$$

Тепер ми можемо переписати нашу формулу для роботи через потенціальну енергію: $A_т = E_{p0} - E_p$. Бачимо, що робота сили тяжіння дорівнює різниці потенціальних енергій у початковому та кінцевому станах. Цей результат є чудовою ілюстрацією загального правила, яке ми сформулюємо трохи згодом.

**Важливий момент:** Формула $E_p = mgh$ визначає потенціальну енергію відносно певного **нульового рівня**. Це рівень, на якому ми домовились вважати потенціальну енергію рівною нулю ($h=0$). Ми можемо обрати цей рівень довільно: поверхня столу, підлога кімнати, поверхня Землі. Сама по собі величина потенціальної енергії не має глибокого фізичного змісту, зміст має лише її **зміна** $\Delta E_p$.

Якщо книга лежить на столі висотою 1 м у кімнаті, її потенціальна енергія відносно підлоги дорівнює $mg \cdot (1 \text{ м})$, але відносно поверхні столу - нулю. Фізичний зміст має лише зміна енергії, наприклад, при падінні книги зі столу на підлогу або при підійманні книги зі столу на поличку.

**Важливий момент:** формула $E_p = mgh$ є точною лише поблизу поверхні планети, де поле тяжіння можна вважати однорідним, а прискорення вільного падіння $g$ - сталим. Для космічних масштабів або для випадків, коли змінна висоти є суттєвою і поле тяжіння (гравітаційне поле) не можна вважати однорідним, використовується більш загальна формула гравітаційної потенціальної енергії 

$$E_p = -G\frac{Mm}{r}$$

де $M$ - маса планети, $m$ - маса досліджуваного тіла, $r$ - відстань від центру планети до тіла. Цю формулу доволі легко отримати, якщо ви вмієте працювати із визначеними інтегралами і розумієте суть застосування цих інтегралів у фізиці. Виведення цієї формули можна поглянути наприклад ось [тут](https://en.wikipedia.org/wiki/Gravitational_energy).

## Потенціальна енергія пружно деформованого тіла

Тепер дослідимо іншу фундаментальну взаємодію - пружну деформацію. Уявімо, що ми розтягуємо або стискаємо пружину жорсткістю $k$ з початкового положення, де її деформація дорівнювала $x_0$, до кінцевого положення з деформацією $x$.

З боку пружини на тіло діє сила пружності, яка за законом Гука дорівнює $F_{\text{пр}} = -kx$. Знак "мінус" вказує на те, що сила пружності завжди намагається повернути систему в положення рівноваги (тобто напрямлена протилежно вектору деформації $\vec{x}$).

Оскільки ця сила не є сталою, а лінійно залежить від деформації $x$, для знаходження її роботи ми скористаємось уже знайомим нам геометричним методом - знайдемо площу фігури під графіком залежності сили від переміщення.

1.  Побудуємо графік залежності проєкції сили пружності $F_{\text{пр}}$ на вісь Ох від координати (деформації) $x$. Це буде пряма похила лінія, що проходить через початок координат.
2.  Робота, яку виконує сила пружності при зміні деформації від $x_0$ до $x$, чисельно дорівнює площі фігури, обмеженої графіком сили, віссю Ох та вертикальними лініями $x=x_0$ та $x=x$. Ця фігура є **трапецією**.
3.  Площа трапеції обчислюється за формулою: $S = \frac{a+b}{2} \cdot h$, де $a$ і $b$ - основи, а $h$ - висота.
    *   У нашому випадку "основами" трапеції є модулі значень сили пружності в початковій та кінцевій точках: $a = |F_{\text{пр}}(x_0)| = kx_0$ та $b = |F_{\text{пр}}(x)| = kx$.
    *   "Висотою" трапеції є модуль зміни деформації: $h = |x - x_0|$.
4.  Важливий момент зі знаком роботи: якщо ми розтягуємо пружину (наприклад, $x > x_0 > 0$), то переміщення відбувається в напрямку осі Ох, а сила пружності напрямлена протилежно. Отже, її робота має бути від'ємною, сила пружності у такому випадку заважатиме руху тіла. Робота буде додатною, коли пружина сама стискається, виконуючи роботу над якимось тілом. Об'єднати ці випадки можна, розписавши роботу через інтеграл, але ми можемо отримати результат простіше. Робота сили пружності дорівнює площі трапеції, взятій зі знаком "мінус", якщо сила і переміщення протилежно напрямлені. Давайте знайдемо загальну формулу, яка врахує всі знаки. Робота сили пружності:

    $$
    A_{\text{пр}} = \text{Площа трапеції під графіком}
    $$

    Площу трапеції можна знайти як півсуму основ на висоту. Основи - це $F_{\text{пр}}(x_0) = -kx_0$ та $F_{\text{пр}}(x) = -kx$. Висота - це $(x - x_0)$.

    $$
    A_{\text{пр}} = \frac{F_{\text{пр}}(x_0) + F_{\text{пр}}(x)}{2} \cdot (x-x_0) = \frac{-kx_0 - kx}{2} \cdot (x-x_0)
    $$

    $$
    A_{\text{пр}} = -\frac{k}{2}(x_0 + x)(x-x_0) = -\frac{k}{2}(x^2 - x_0^2)
    $$

    Розкривши дужки, ми отримаємо фінальний вигляд:

    $$
    A_{\text{пр}} = \frac{kx_0^2}{2} - \frac{kx^2}{2}
    $$

Подивімося уважно на результат. Формула для роботи сили пружності залежить **лише від початкової ($x_0$) та кінцевої ($x$) деформацій**. Вона абсолютно не залежить від того, як саме відбувався перехід між цими двома станами. Це доводить, що **сила пружності є консервативною силою**.

Ми знову бачимо знайому структуру: робота дорівнює різниці деякої величини. Цю величину, що характеризує "запас" енергії в деформованому тілі, ми називаємо **потенціальною енергією пружно деформованого тіла**:

$$
E_p = \frac{kx^2}{2}
$$

І знову ми можемо переписати формулу для роботи через потенціальну енергію: $A_{\text{пр}} = E_{p0} - E_p$. Ми вдруге переконалися в існуванні загального зв'язку між роботою консервативної сили та зміною потенціальної енергії.

Тут за нульовий рівень потенціальної енергії природно вважати стан, коли пружина не деформована ($x=0$).

## Теорема про потенціальну енергію

Тепер, розглянувши два конкретні приклади, ми можемо сформулювати загальну теорему, яка їх об'єднує.

> **Теорема про потенціальну енергію:** Робота консервативної сили дорівнює зміні потенціальної енергії системи, взятій з протилежним знаком.

$$
A = -(E_p - E_{p0}) = E_{p0} - E_p = -\Delta E_p
$$

де:
*   $A$ - робота консервативної сили.
*   $E_p$ - кінцева потенціальна енергія.
*   $E_{p0}$ - початкова потенціальна енергія.

**Фізичний зміст:** Коли консервативна сила виконує додатну роботу (наприклад, сила тяжіння при падінні тіла), потенціальна енергія системи зменшується. І навпаки, коли консервативна сила виконує від'ємну роботу (ми підіймаємо тіло вгору, а сила тяжіння напрямлена вниз), потенціальна енергія системи зростає. Робота відбувається ніби "за рахунок" накопиченої потенціальної енергії.

Існує цікава закономірність: будь-яка механічна система прагне зайняти положення, в якому її потенціальна енергія буде мінімальною. Це так званий **принцип мінімуму потенціальної енергії**. Положення, в якому знаходиться тіло при мінімумі потенціальної енергії є положенням **стійкої рівноваги**. Консервативна сила завжди напрямлена в бік зменшення потенціальної енергії.

**Приклад**: Кулька, що лежить на дні сферичної чаші, перебуває в положенні стійкої рівноваги. Її потенціальна енергія мінімальна. Якщо її трохи відхилити, вона повернеться назад.

## Повна механічна енергія та її збереження

Тепер, коли ми знаємо про два види механічної енергії - кінетичну (енергію руху) та потенціальну (енергію взаємодії), - ми можемо ввести поняття повної механічної енергії.

> **Повна механічна енергія $E$** - це сума кінетичної та потенціальної енергій тіла (або системи тіл)
>
> $$ E = E_k + E_p $$

Ця величина є дуже важливою у фізиці, оскільки за певних умов вона залишається незмінною.

### Закон збереження повної механічної енергії

Для того, щоб сформулювати цей закон, нам потрібне поняття **замкненої (або ізольованої) системи тіл**.

> **Замкнена система тіл** - це сукупність тіл, які взаємодіють лише між собою і не взаємодіють з іншими тілами, що не входять до цієї системи.

На практиці це означає, що зовнішні сили або відсутні, або їхня дія скомпенсована.

Тепер ми готові сформулювати один з фундаментальних законів природи.

> **Закон збереження повної механічної енергії:** У замкненій системі тіл, де діють лише консервативні сили (сила тяжіння та сила пружності), повна механічна енергія системи залишається сталою.
>
> $$ E = E_k + E_p = \text{const} $$

Це означає, що для двох будь-яких моментів часу повна енергія системи однакова:

$$
E_{k0} + E_{p0} = E_k + E_p
$$

### Виведення закону збереження механічної енергії

Цей закон не є постулатом, його можна вивести з уже відомих нам теорем. Давайте зробимо це.

1.  Згідно з **теоремою про кінетичну енергію**, робота всіх сил, що діють на тіло, дорівнює зміні його кінетичної енергії: $A = \Delta E_k = E_k - E_{k0}$.
2.  У нашій системі діють лише консервативні сили, отже згідно з **теоремою про потенціальну енергію**, робота консервативних сил дорівнює: $A_{\text{конс}} = -\Delta E_p = E_{p0} - E_p$.
3.  Прирівняємо вирази для роботи з пунктів 1 і 2:

    $$
    \Delta E_k = -\Delta E_p
    $$

    $$
    E_k - E_{k0} = E_{p0} - E_p
    $$

4.  Перегрупуємо доданки: перенесемо всі величини з індексом "0" (початковий стан) в один бік, а без індексу (кінцевий стан) - в інший.

    $$
    E_{k0} + E_{p0} = E_k + E_p
    $$
    
    Що й треба було довести.

### Перетворення енергії

Закон збереження енергії не означає, що кінетична та потенціальна енергії не змінюються. Навпаки, вони постійно **перетворюються** одна в одну, але їхня сума залишається незмінною.

**Приклад:** Тіло, кинуте вертикально вгору.

Уявімо, що ми кидаємо м'яч масою $m$ вертикально вгору з початковою швидкістю $v_0$ з поверхні землі (яку ми приймемо за нульовий рівень висоти, $h=0$). Нехтуємо опором повітря.

*   **У момент кидка (точка А, $h=0$):**
    *   Швидкість м'яча максимальна ($v_0$), отже його **кінетична енергія є максимальною**: $E_k = \frac{mv_0^2}{2}$.
    *   Висота дорівнює нулю, отже **потенціальна енергія дорівнює нулю**: $E_p = 0$.
    *   Повна механічна енергія системи: $E_A = E_k + E_p = \frac{mv_0^2}{2} + 0$.

*   **Під час польоту вгору (проміжна точка В):**
    *   Під дією сили тяжіння м'яч сповільнюється, тому його швидкість $v$ зменшується, а отже, **зменшується і кінетична енергія**.
    *   Одночасно м'яч набирає висоту $h$, тому його **потенціальна енергія зростає**.
    *   Відбувається перетворення кінетичної енергії в потенціальну: $E_k \to E_p$. Повна енергія залишається незмінною: $E_B = \frac{mv^2}{2} + mgh = E_A$.

*   **У найвищій точці траєкторії (точка С, $h=h_{max}$):**
    *   М'яч на мить зупиняється, його швидкість дорівнює нулю. Отже, **кінетична енергія дорівнює нулю**: $E_k = 0$.
    *   Висота є максимальною, отже **потенціальна енергія досягає свого максимуму**: $E_p = mgh_{max}$.
    *   Вся початкова кінетична енергія повністю перейшла в потенціальну: $E_C = 0 + mgh_{max} = E_A$.

*   **Під час падіння вниз:**
    *   Процес відбувається у зворотному напрямку. Висота зменшується, швидкість зростає. **Потенціальна енергія перетворюється назад на кінетичну**: $E_p \to E_k$.

У будь-якій точці траєкторії сума $E = E_k + E_p$ залишатиметься сталою і дорівнюватиме початковому значенню $\frac{mv_0^2}{2}$.

## Роль неконсервативних сил: сила тертя

А що станеться, якщо в системі діють неконсервативні сили, наприклад, сила тертя? У такому разі повна механічна енергія **не зберігається**. Вона зменшується.

Робота сили тертя $A_{\text{тертя}}$ завжди від'ємна. Вона "забирає" механічну енергію з системи. У цьому випадку зв'язок між роботою і зміною енергії виглядає так:

> Робота всіх неконсервативних сил (наприклад, сили тертя), що діють у системі, дорівнює зміні повної механічної енергії системи.
>
> $$ A_{\text{тертя}} = \Delta E = E - E_0 = (E_k + E_p) - (E_{k0} + E_{p0}) $$

Оскільки $A_{\text{тертя}} < 0$, то і $\Delta E < 0$, що означає $E < E_0$ - кінцева механічна енергія менша за початкову.

## Загальний закон збереження енергії

Куди ж зникає механічна енергія, коли діє сила тертя? Чи вона просто зникає? Ні.

**Енергія нікуди не зникає й нізвідки не з'являється, вона лише перетворюється з одного виду на інший або переходить від одного тіла до іншого.**

Це - **загальний закон збереження енергії**, один з найголовніших у всій науці.

Коли сила тертя виконує роботу, механічна енергія перетворюється в інший вид енергії - **внутрішню енергію**. Це енергія хаотичного руху та взаємодії молекул і атомів, з яких складаються тіла. Простіше кажучи, тіла нагріваються.

**Приклад:** Коли автомобіль гальмує, його величезна кінетична енергія не зникає. Вона перетворюється на внутрішню енергію гальмівних колодок та дисків, які сильно нагріваються. Так само відбувається при **непружному ударі** (наприклад, коли шматок пластиліну падає на підлогу і прилипає до неї) - кінетична енергія частково переходить у нагрівання та деформацію тіл.

Таким чином, якщо враховувати всі види енергії (механічну, внутрішню, електромагнітну, ядерну тощо), то повна енергія ізольованої системи завжди залишається незмінною. Закон збереження механічної енергії є лише окремим, хоч і дуже важливим, випадком цього всеосяжного закону природи.
