# Другий та третій закони Ньютона

Для початку давайте розберемося з двома ключовими поняттями, без яких неможливо уявити динаміку - силою та масою.

## Сила $\vec{F}$

У повсякденному житті ми постійно стикаємося з поняттям сили: сила вітру, сила м'язів, сила тяжіння. У фізиці це поняття має чітке визначення.

> **Сила** $\vec{F}$ - це векторна фізична величина, яка є мірою дії одного тіла на інше. Саме сила є причиною зміни швидкості руху тіла, тобто причиною його прискорення, а також причиною його деформації.

Оскільки сила є векторною величиною, вона характеризується не тільки числовим значенням (модулем), але й напрямком.

В Міжнародній системі одиниць (СІ) одиницею вимірювання сили є **ньютон (Н)**.

$$
1 \text{ Н} = 1 \frac{\text{кг} \cdot \text{м}}{\text{с}^2}
$$

Один ньютон - це сила, яка, діючи на тіло масою 1 кг, надає йому прискорення 1 м/с² у напрямку дії сили.

## Маса $m$ та інертність

Уявімо, що перед нами два візки: один порожній, а інший - навантажений цеглою. Щоб зрушити їх з місця і надати їм однакової швидкості за однаковий час (тобто однакового прискорення), нам доведеться докласти різних зусиль. Навантажений візок зрушити буде значно важче. Це відбувається тому, що він має більшу **інертність**. Ми вже згадували поняття інертності, коли розглядали перший закон Ньютона. Наводимо це поняття знову, щоб нагадати.

> **Інертність** - це властивість тіла опиратися спробам змінити його швидкість. Тобто, будь-яке тіло "воліє" зберігати свою поточну швидкість (і напрямок руху) незмінною. 

Фізичною величиною, яка є кількісною мірою інертності тіла, є **маса**.

> **Маса** $m$ - це скалярна фізична величина, яка є мірою інертних та гравітаційних властивостей тіла. Чим більша маса тіла, тим воно "інертніше", і тим важче змінити його швидкість.

Одиницею вимірювання маси в системі СІ є **кілограм (кг)**.

### Інертна та Гравітаційна Маса

Для кращої обізнаності ми розглянемо різні прояви маси, однак занурюватись у деталі ми будемо поступово (на наступних сторінках). Нижче наведені доволі базові поняття, задача яких лише продемонструвати, що не все так просто. Згодом, як познайомимось із законом всесвітнього тяжіння Ньютона, ми розберемо тему інтерної та гравітаційної мас детальніше.

Отже, є два прояви маси:

*   **Інертна маса** $m_i$ характеризує здатність тіла протидіяти зміні його швидкості (тобто інертність). Вона фігурує у другому законі Ньютона, який ми розглядаємо нижче.
*   **Гравітаційна маса** $m_g$ визначає силу, з якою тіло притягується до інших тіл (наприклад, до Землі) відповідно до закону всесвітнього тяжіння.

Численні експерименти, починаючи з дослідів Галілея, який спостерігав за вільним падінням тіл, показали, що інертна та гравітаційна маси пропорційні одна одній. Це дозволяє в сучасній фізиці вважати їх еквівалентними та використовувати єдине поняття "маса" і не розрізняти окремо інертну та гравітаційну маси (технічно це відбувається за рахунок вибору відповідних розмірностей фізичних величин). Цей фундаментальний факт називається **принципом еквівалентності**. Тому надалі ми будемо говорити просто про "масу", а до розгляду інертної та гравітаційної маси і до їх фізичного аналізу, який зрештою привів до того, що для цих двох мас почав використовуватись єдиний термін "маса", повернемось пізніше.

### Як Виміряти Масу?

Існує два основних способи вимірювання маси, що базуються на її різних проявах:

1.  **Зважування**: Цей метод ґрунтується на гравітаційних властивостях маси. За допомогою терезів ми порівнюємо силу тяжіння, що діє на тіло, із силою тяжіння, що діє на еталонні гирі.
2.  **Взаємодія тіл**: Цей метод ґрунтується на інертних властивостях маси. Якщо привести у взаємодію два тіла і мінімізувати сили тертя (в ідеальному випадку тертя не розглядається взагалі) (наприклад, два візки на гладкій поверхні, з'єднані стиснутою пружиною), то після вивільнення пружини вони роз'їдуться в різні боки. Досліди показують, що відношення їхніх мас обернено пропорційне відношенню модулів їхніх прискорень:

    $$
    \frac{m_1}{m_2} = \frac{a_2}{a_1}
    $$

    Таким чином, знаючи масу одного тіла (еталона), ми можемо визначити масу іншого. Ми виведемо цю формулу нижче.

### Властивості Маси

*   **Інваріантність**: У класичній механіці маса тіла вважається сталою величиною, яка не залежить від швидкості його руху. Варто зазначити, що в релятивістській фізиці маса залежить від швидкості, але для швидкостей, значно менших за швидкість світла, цією зміною можна знехтувати. Поки що ми розглядаємо лише класичну механіку, тому надалі будемо використовувати властивість інваріантності без жодних застережень.
*   **Адитивність**: У класичній механіці маса системи тіл дорівнює сумі мас тіл, що входять до цієї системи. Наприклад, маса автомобіля з пасажирами дорівнює сумі маси самого автомобіля та мас усіх пасажирів.
*   **Закон збереження маси**: У класичній механіці маса ізольованої системи тіл залишається сталою при будь-яких процесах, що відбуваються в цій системі.

## Другий Закон Ньютона


### Експериментальний підхід

Щоб встановити зв'язок між трьома величинами - силою, масою та прискоренням - нам потрібно провести серію дослідів. Ідея полягає в тому, щоб змінювати один параметр, тримаючи інший постійним, і дивитися, як при цьому змінюється третій. Уявімо, що ми маємо горизонтальну поверхню, по якій може майже без тертя рухатись візок. До візка прикріплено пружину або нитку, перекинуту через блок, на якій підвішено вантаж.

Силу, що діє на візок, ми можемо вимірювати за розтягненням пружини або визначати за вагою вантажу. Масу системи ми можемо змінювати, додаючи на візок додаткові тягарці. А прискорення можна розрахувати, вимірявши відстань, яку проходить візок, і час, за який він її долає, за формулою для рівноприскореного руху:

$$
a = \frac{2s}{t^2}
$$

Тепер розглянемо два ключові експерименти.

---

### Дослід 1: Залежність прискорення від маси (при сталій силі)

Наша мета - з'ясувати, як маса тіла впливає на його прискорення, якщо сила, що діє на нього, залишається незмінною.

1.  **Крок 1.** Беремо візок масою $m$. Діємо на нього певною сталою силою $\vec{F}$ (наприклад, забезпечуємо однакове розтягнення пружини в усіх дослідах цієї серії). Вимірюємо час руху і розраховуємо прискорення, яке позначимо як $\vec{a}$.

2.  **Крок 2.** Тепер збільшимо масу системи вдвічі. Для цього на візок ставимо ще один такий самий візок або тягарець такої ж маси. Загальна маса тепер становить $2m$. Знову діємо на систему тією ж самою силою $\vec{F}$. Проведені виміри покажуть, що прискорення системи тепер стало вдвічі меншим: $\frac{1}{2}\vec{a}$.

3.  **Крок 3.** Збільшимо масу системи втричі, до $3m$, і знову подіємо тією ж силою $\vec{F}$. Експеримент покаже, що прискорення зменшилося втричі й стало дорівнювати $\frac{1}{3}\vec{a}$.

**Результат досліду:** При сталій силі прискорення, якого набуває тіло, обернено пропорційне його масі.

$$
a \sim \frac{1}{m}
$$

Це інтуїтивно зрозуміло: чим "важче" (масивніше) тіло, тим складніше його "розігнати".

---

### Дослід 2: Залежність прискорення від сили (при сталій масі)

Тепер наше завдання - дослідити, як змінюється прискорення тіла, якщо змінювати прикладену до нього силу, але залишати його масу незмінною.

1.  **Крок 1.** Візьмемо візок масою $m$ і подіємо на нього силою $\vec{F}$. Виміряємо прискорення, яке він набув, і позначимо його $\vec{a}$.

2.  **Крок 2.** Не змінюючи маси візка, збільшимо силу, що діє на нього, вдвічі, до $2\vec{F}$ (наприклад, підвісимо два однакових вантажі замість одного). Виміри покажуть, що і прискорення візка збільшилося рівно вдвічі й стало $2\vec{a}$.

3.  **Крок 3.** Збільшимо силу втричі, до $3\vec{F}$. Експеримент покаже, що прискорення також зросло втричі й стало $3\vec{a}$.

**Результат досліду:** При сталій масі прискорення, якого набуває тіло, прямо пропорційне силі, прикладеній до нього.

$$
a \sim F
$$

Це також логічно: чим сильніше ми штовхаємо тіло, тим швидше воно розганяється.

---

### Об'єднання Результатів та Формулювання Закону

Тепер ми маємо два важливі експериментальні факти:
*   $a \sim \frac{1}{m}$ (при $F = \text{const}$)
*   $a \sim F$ (при $m = \text{const}$)

Об'єднавши ці дві пропорційності, ми можемо записати загальну залежність прискорення одночасно і від сили, і від маси:

$$
a \sim \frac{F}{m}
$$

Щоб перейти від пропорційності до точної рівності, потрібно ввести коефіцієнт пропорційності $k$:

$$
a = k \frac{F}{m}
$$

Але тут фізики зробили дуже елегантний крок. Вони домовились обрати одиницю вимірювання сили - ньютон - таким чином, щоб цей коефіцієнт $k$ дорівнював одиниці. Тобто, як ми вже казали, 1 ньютон - це саме така сила, яка тілу масою 1 кг надає прискорення 1 $\text{м}/\text{с}^2$. За такого вибору одиниць коефіцієнт $k$ стає безрозмірною одиницею, і формула набуває свого знаменитого простого вигляду:

$$
\vec{a} = \frac{\vec{F}}{m}
$$

Таким чином, ми бачимо, що другий закон Ньютона - це не просто умоглядна ідея, а результат, що безпосередньо випливає з ретельно поставлених експериментів, які вивчають, як світ навколо нас реагує на взаємодії.

Другий закон Ньютона є центральним законом динаміки. Він встановлює зв'язок між силою, що діє на тіло, його масою та прискоренням, яке воно набуває.

Щоб ще точніше сформулювати другий закон Ньютона, введемо поняття **рівнодійної сили**. Дуже рідко на тіло діє лише одна сила. Зазвичай їх декілька. Наприклад, на книгу, що лежить на столі, діє сила тяжіння (спрямована вниз) і сила реакції опори з боку столу (спрямована вгору). Якщо на тіло діє декілька сил, то їхній сукупний ефект еквівалентний дії однієї сили, яку називають рівнодійною.

> **Рівнодійна сила** $\vec{F}$ - це векторна сума всіх сил, що діють на тіло.
>
> $$\vec{F} = \vec{F_1} + \vec{F_2} + ... + \vec{F_n}$$

**Приклад:** Якщо дві сили $F_1 = 10 \text{ Н}$ і $F_2 = 15 \text{ Н}$ діють на тіло в одному напрямку, то їхня рівнодійна $R = 10 + 15 = 25 \text{ Н}$ і спрямована в той же бік. Якщо ж вони діють у протилежних напрямках, то модуль рівнодійної $R = |15 - 10| = 5 \text{ Н}$, і вона буде спрямована в бік більшої сили.

> **Другий закон Ньютона**: Прискорення $\vec{a}$, якого набуває тіло внаслідок дії сили, прямо пропорційне цій силі (точніше, рівнодійній усіх сил) та обернено пропорційне масі тіла.

Математично цей закон записується у вигляді формули:

$$
\vec{a} = \frac{\vec{F}}{m}
$$

Звідси випливає найпоширеніший запис другого закону Ньютона:

$$
\vec{F} = m\vec{a}
$$

Або, якщо на тіло діє кілька сил і ми детально розписуємо рівнодійну сил:

$$
m\vec{a} = \vec{F_1} + \vec{F_2} + ... + \vec{F_n}
$$

З формули видно, що напрямок вектора прискорення завжди збігається з напрямком вектора рівнодійної сили.

Важливо наголосити, що закони Ньютона (і другий зокрема) виконуються не в будь-якій системі відліку. Вони справедливі лише в **інерціальних системах відліку (ІСВ)**, які ми вже розглядали під час вивчення першого закону Ньютона.

З високою точністю інерціальною можна вважати геліоцентричну систему відліку, пов'язану із Сонцем та віддаленими зірками. Будь-яка інша система відліку, що рухається відносно інерціальної рівномірно і прямолінійно, також є інерціальною. Система відліку, пов'язана із Землею, є наближено інерціальною, і для багатьох задач її можна вважати такою.

## Третій закон Ньютона

Третій закон Ньютона описує фундаментальну властивість будь-якої взаємодії. Він пояснює, що сили завжди виникають парами.

> **Третій закон Ньютона** стверджує: тіла діють одне на одне з силами, рівними за модулем і протилежними за напрямком.

Математично це записується так:

$$
\vec{F_{12}} = -\vec{F_{21}}
$$

де $\vec{F_{12}}$ - сила, з якою перше тіло діє на друге, а $\vec{F_{21}}$ - сила, з якою друге тіло діє на перше.

Ключові особливості сил у третьому законі Ньютона:
1. **Сили завжди виникають парами**: Не існує сили, яка б діяла сама по собі. Якщо ви штовхаєте стіну, стіна з такою ж силою штовхає вас.
2. **Сили прикладені до різних тіл**: Це дуже важливий момент! Сила $\vec{F}_{12}$ прикладена до другого тіла, а сила $\vec{F}_{21}$ - до першого. Оскільки ці сили прикладені до різних тіл, вони **ніколи** не компенсують і не врівноважують одна одну.
3. **Сили мають однакову фізичну природу**: Якщо ви притягуєтесь до Землі з гравітаційною силою, то й Земля притягується до вас з такою ж за модулем гравітаційною силою. Якщо електрон відштовхує інший електрон з електростатичною силою, то й другий електрон відштовхує перший з такою ж електростатичною силою.

**Приклад**: Коли людина йде, вона відштовхується ногами від землі (діє на землю із силою, спрямованою назад). За третім законом Ньютона, земля діє на людину з рівною за модулем, але протилежно спрямованою силою (силою тертя спокою, спрямованою вперед), яка, власне, і рухає людину.

Приклади дії третього закону Ньютона:
*  **Ходьба:** Коли людина йде, вона відштовхується ногами від землі (діє на землю із силою, спрямованою назад). За третім законом Ньютона, земля діє на людину з рівною за модулем, але протилежно спрямованою силою (силою тертя спокою, спрямованою вперед), яка, власне, і рухає людину.
*  **Плавання:** Плавець відштовхує воду руками та ногами назад, а вода штовхає плавця вперед.
*  **Яблуко і Земля:** Яблуко падає, бо Земля притягує його. Але згідно з третім законом, яблуко так само притягує Землю. Просто через величезну масу Землі її прискорення мізерно мале і непомітне.

### Внутрішні та Зовнішні Сили

При розгляді системи, що складається з кількох тіл, зручно поділяти всі сили на дві категорії.

*   **Внутрішні сили** - це сили, з якими тіла, що входять до складу системи, взаємодіють одне з одним. Наприклад, сили притягання між Сонцем та планетами є внутрішніми для Сонячної системи.
*   **Зовнішні сили** - це сили, що діють на тіла системи з боку тіл, які не належать до цієї системи. Наприклад, сила притягання з боку сусідньої зірки буде зовнішньою для Сонячної системи.

Цей поділ є надзвичайно важливим для законів збереження. За третім законом Ньютона, векторна сума всіх внутрішніх сил у будь-якій системі завжди дорівнює нулю, оскільки для кожної внутрішньої сили є рівна за модулем і протилежна за напрямком сила протидії. Саме тому змінити рух системи як цілого (тобто її центру мас) можуть лише зовнішні сили.

### Співвідношення мас та прискорень: виведення формули

Розглянемо два тіла з масами $m_1$ та $m_2$, що взаємодіють одне з одним (наприклад, відштовхуються). Згідно з третім законом Ньютона (який ми розглянемо далі), сили, з якими вони діють одне на одного, рівні за модулем і протилежні за напрямком: $F_1 = F_2 = F$ (у цій останній формулі ми розглядаємо саме модулі сил).

Згідно з другим законом Ньютона, кожне з тіл набуде прискорення:

$$
F = m_1 a_1
$$

$$
F = m_2 a_2
$$

Оскільки ліві частини цих рівнянь рівні, то рівні й праві:

$$
m_1 a_1 = m_2 a_2
$$

Звідси ми можемо отримати важливе співвідношення, яке дозволяє порівнювати маси тіл за їхніми прискореннями під час взаємодії:

$$
\frac{m_1}{m_2} = \frac{a_2}{a_1}
$$

Це співвідношення показує, що при взаємодії двох тіл відношення їхніх мас обернено пропорційне відношенню модулів їхніх прискорень. Тіло з більшою масою набуде меншого прискорення, і навпаки.
