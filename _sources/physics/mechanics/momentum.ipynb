# Імпульс тіла

## Імпульс тіла та імпульс сили

Уявіть, що на вас котиться іграшковий візок і вантажівка. Очевидно, що зупинити їх - це два абсолютно різні завдання, навіть якщо їхні швидкості однакові. Так само, легше зловити м'яч, що летить повільно, ніж той, що летить швидко. Це інтуїтивне розуміння того, що "кількість руху" залежить і від маси, і від швидкості, підводить нас до поняття імпульсу.

### Імпульс тіла

> **Імпульс тіла** $\vec{p}$ - це векторна фізична величина, що є мірою механічного руху тіла і дорівнює добутку маси тіла $m$ на швидкість його руху $\vec{v}$.

Формула для обчислення імпульсу тіла виглядає так:

$$ \vec{p} = m\vec{v} $$

З цієї формули видно кілька важливих речей:
*   Імпульс - це **векторна величина**. Його напрямок завжди збігається з напрямком вектора швидкості тіла.
*   Чим більша маса тіла або його швидкість, тим більший його імпульс.
*   Якщо тіло перебуває у стані спокою ($\vec{v} = 0$), його імпульс дорівнює нулю.

Одиницею вимірювання імпульсу тіла в системі СІ є **кілограм-метр на секунду** ($\text{кг } \cdot \text{м/с}$).

### Другий закон Ньютона в імпульсному вигляді

Ми звикли до формулювання другого закону Ньютона як $\vec{F} = m\vec{a}$. Проте існує й інша, більш загальна форма запису цього закону, яка, власне, і є тією, яку спочатку запропонував сам Ньютон. Давайте виведемо її.

1.  Почнемо зі звичного нам другого закону Ньютона:
    $$
    \vec{F} = m\vec{a}
    $$
    Тут $\vec{F}$ - це рівнодійна всіх сил, що діють на тіло.

2.  Згадаємо визначення прискорення. Прискорення - це швидкість зміни швидкості:
    $$
    \vec{a} = \frac{\Delta \vec{v}}{\Delta t} = \frac{\vec{v} - \vec{v}_0}{\Delta t}
    $$
    де $\vec{v}_0$ і $\vec{v}$ - початкова та кінцева швидкості тіла, а $\Delta t$ - проміжок часу, за який ця зміна відбулася.

3.  Підставимо визначення прискорення у другий закон Ньютона:
    $$
    \vec{F} = m \frac{\vec{v} - \vec{v}_0}{\Delta t}
    $$

4.  Перенесемо $\Delta t$ у ліву частину та розкриємо дужки у правій:
    $$
    \vec{F}\Delta t = m\vec{v} - m\vec{v}_0
    $$

5.  Вирази $m\vec{v}$ та $m\vec{v}_0$ - це кінцевий та початковий імпульси тіла. Отже, $m\vec{v} - m\vec{v}_0 = \vec{p} - \vec{p}_0 = \Delta \vec{p}$ - це зміна імпульсу тіла.

Таким чином, ми отримуємо **другий закон Ньютона в імпульсній формі**:

$$
\vec{F}\Delta t = \vec{p} - \vec{p}_0 = \Delta \vec{p}
$$

Це рівняння пов'язує силу, що діє на тіло, та зміну його імпульсу.

### Імпульс сили

Добуток у лівій частині отриманого рівняння також є важливою фізичною величиною.

> **Імпульс сили** - це векторна фізична величина, що дорівнює добутку сили на час її дії і є мірою впливу сили на тіло за певний проміжок часу.

Формула для імпульсу постійної сили $\vec{F}$, що діє протягом часу $\Delta t$:

$$
\text{Імпульс сили} = \vec{F}\Delta t
$$

Одиниця вимірювання імпульсу сили в СІ - **Ньютон-секунда** ($\text{H} \cdot \text{с}$). Легко перевірити, що $1 \text{ Н} \cdot \text{с} = 1 (\text{кг} \cdot \text{м/с}^2) \cdot \text{с} = 1 \text{ кг} \cdot \text{м/с}$, тобто одиниці вимірювання імпульсу сили та імпульсу тіла є еквівалентними.

Тепер ми можемо сформулювати другий закон Ньютона ще одним способом:

> **Зміна імпульсу тіла дорівнює імпульсу рівнодійної сили, що діє на нього.**

Це формулювання є надзвичайно корисним при аналізі короткочасних взаємодій, таких як удари або зіткнення. Часто ми не знаємо точного значення сили удару (адже вона може змінюватися протягом дуже короткого часу), але ми можемо виміряти зміну імпульсу тіла і таким чином оцінити середню силу удару.

## Закон збереження імпульсу

Тепер ми підійшли до одного з найважливіших законів збереження у фізиці. Він дозволяє аналізувати взаємодію тіл у системі, не розглядаючи детально внутрішні сили, що діють між ними.

### Замкнена система тіл

Для формулювання закону нам знадобиться поняття замкненої, або ізольованої, системи. Ми користувались цим терміном раніше. Згадаємо його.

> **Замкнена (ізольована) система тіл** - це сукупність тіл, які взаємодіють лише між собою (внутрішніми силами) і не взаємодіють з тілами, що не входять до цієї системи (зовнішні сили відсутні або їхня дія скомпенсована).

На практиці ідеально замкнених систем не існує. На будь-яке тіло на Землі діє сила тяжіння та часто сила тертя. Однак у багатьох випадках систему можна вважати замкненою:
*   Якщо зовнішні сили зрівноважені (наприклад, сила тяжіння і сила нормальної реакції опори для тіл на горизонтальній поверхні).
*   Якщо внутрішні сили, що діють між тілами системи, набагато більші за зовнішні сили.

### Формулювання та виведення закону

> **Закон збереження імпульсу:** У замкненій системі тіл векторна сума імпульсів усіх тіл системи залишається сталою (зберігається) за будь-яких взаємодій цих тіл між собою.

$$ \vec{p}_1 + \vec{p}_2 + ... + \vec{p}_n = \text{const} $$

Це означає, що сумарний імпульс системи до взаємодії дорівнює сумарному імпульсу системи після взаємодії:

$$ m_1\vec{v}_{01} + m_2\vec{v}_{02} + ... + m_n\vec{v}_{0n} = m_1\vec{v}_1 + m_2\vec{v}_2 + ... + m_n\vec{v}_n $$

де $\vec{v}_{0i}$ - початкові швидкості тіл, а $\vec{v}_i$ - їхні швидкості після взаємодії.

**Виведення закону збереження імпульсу (на прикладі двох тіл):**

1.  Розглянемо замкнену систему, що складається з двох тіл масами $m_1$ та $m_2$. Нехай вони взаємодіють між собою із силами $\vec{F}_{12}$ (сила, що діє на перше тіло з боку другого) та $\vec{F}_{21}$ (сила, що діє на друге тіло з боку першого).

2.  Згідно з третім законом Ньютона, ці сили рівні за модулем і протилежні за напрямком:

    $$
    \vec{F}_{12} = -\vec{F}_{21} \implies \vec{F}_{12} + \vec{F}_{21} = 0
    $$

3.  Запишемо другий закон Ньютона в імпульсній формі для кожного тіла. Якщо взаємодія тривала час $\Delta t$:
    *   Для першого тіла: $\Delta \vec{p}_1 = \vec{F}_{12} \Delta t$
    *   Для другого тіла: $\Delta \vec{p}_2 = \vec{F}_{21} \Delta t$

4.  Додамо ці два рівняння:

    $$
    \Delta \vec{p}_1 + \Delta \vec{p}_2 = (\vec{F}_{12} + \vec{F}_{21}) \Delta t
    $$

5.  Оскільки сума сил $\vec{F}_{12} + \vec{F}_{21} = 0$, то і права частина рівняння дорівнює нулю:

    $$
    \Delta \vec{p}_1 + \Delta \vec{p}_2 = 0
    $$

6.  Зміна сумарного імпульсу системи $\Delta (\vec{p}_1 + \vec{p}_2)$ дорівнює нулю. Це означає, що сам сумарний імпульс не змінюється з часом:

    $$
    \vec{p}_1 + \vec{p}_2 = \text{const}
    $$
    Що й треба було довести.

Цей закон є одним із фундаментальних у фізиці, його справедливість виходить далеко за межі класичної механіки.

## Застосування закону збереження імпульсу

Розглянемо кілька яскравих прикладів, де цей закон проявляє себе.

### Реактивний рух

> **Реактивний рух** - це рух, що виникає, коли від тіла з певною швидкістю відокремлюється його частина.

**Приклади в природі та техніці:**
*   Рух кальмарів, медуз, які викидають із себе струмінь води.
*   Надувна кулька, з якої виходить повітря.
*   Політ ракети.

**Фізика руху ракети:**

Ракета - це класичний приклад застосування закону збереження імпульсу. Розглянемо систему "ракета + паливо". До старту ракета перебуває у спокої, її сумарний імпульс дорівнює нулю.

1.  У камері згоряння паливо перетворюється на гарячий газ, який з величезною швидкістю $\vec{v}_{\text{газу}}$ викидається із сопла. Цей струмінь газу має імпульс $\vec{p}_{\text{газу}} = m_{\text{газу}}\vec{v}_{\text{газу}}$.
2.  Оскільки система "ракета-газ" є замкненою (силою тяжіння на початку руху можна знехтувати порівняно з силою тяги), її сумарний імпульс має залишатися нульовим.
3.  Щоб скомпенсувати імпульс газів, оболонка ракети масою $m_{\text{ракети}}$ набуває імпульсу $\vec{p}_{\text{ракети}}$, рівного за модулем, але протилежного за напрямком:

    $$\vec{p}_{\text{ракети}} + \vec{p}_{\text{газу}} = 0 \implies m_{\text{ракети}}\vec{v}_{\text{ракети}} = -m_{\text{газу}}\vec{v}_{\text{газу}}$$

4.  Саме цей імпульс і надає ракеті швидкості. Особливість реактивного руху полягає в тому, що ракета прискорюється без взаємодії із зовнішніми тілами (їй не потрібно "відштовхуватися" від повітря чи землі), що робить його єдиним можливим способом пересування у вакуумі.

Для досягнення космічних швидкостей використовують **багатоступеневі ракети**. Коли паливо в одному ступені закінчується, він відкидається, зменшуючи загальну масу ракети. Це дозволяє наступному ступеню надавати меншій масі більшого прискорення, що робить політ значно ефективнішим.

### Зіткнення тіл

Зіткнення (або удар) - це короткочасна взаємодія тіл, у результаті якої їхні швидкості суттєво змінюються. Під час удару система тіл є замкненою, тому в ній завжди виконується **закон збереження імпульсу**. А от механічна енергія може як зберігатися, так і не зберігатися. Залежно від цього розрізняють два основні види зіткнень.

#### Непружне зіткнення

> **Непружне зіткнення** - це зіткнення, в результаті якого повна механічна енергія системи не зберігається. Частина кінетичної енергії переходить у внутрішню (нагрівання тіл) або витрачається на незворотну деформацію.

Особливим випадком є **абсолютно непружне зіткнення**, після якого тіла рухаються разом, як єдине ціле.

**Приклад:** Дві пластилінові кульки масами $m_1$ та $m_2$ рухаються назустріч одна одній зі швидкостями $\vec{v}_{01}$ та $\vec{v}_{02}$. Після зіткнення вони злипаються і рухаються разом зі швидкістю $\vec{v}$.

Запишемо закон збереження імпульсу у проєкції на вісь Ox, спрямовану вздовж руху першої кульки:

$$
m_1v_{01x} + m_2v_{02x} = (m_1+m_2)v_x
$$

Звідси можна знайти кінцеву швидкість:

$$
v_x = \frac{m_1v_{01x} + m_2v_{02x}}{m_1+m_2}
$$

#### Пружне зіткнення

> **Пружне зіткнення** - це зіткнення, в результаті якого повна механічна енергія (зокрема, кінетична) системи зберігається.

**Приклад:** Зіткнення більярдних куль або атомних частинок можна вважати майже абсолютно пружним.

При абсолютно пружному зіткненні виконуються **одночасно два закони збереження**:
1.  **Закон збереження імпульсу:**

    $$
    m_1\vec{v}_{01} + m_2\vec{v}_{02} = m_1\vec{v}_1 + m_2\vec{v}_2
    $$

2.  **Закон збереження повної механічної енергії**

    $$
    \frac{m_1v_{01}^2}{2} + \frac{m_2v_{02}^2}{2} = \frac{m_1v_1^2}{2} + \frac{m_2v_2^2}{2}
    $$

    Примітка: у цьому випадку ми не розглядаємо потенціальну енергію по двом причинам:
    * Ми вважаємо, що взаємодія між тілами відбувається лише під час зіткнення. Якщо немає зіткнення, взаємодії між тілами або немає, або вона дуже мала (наприклад, гравітаційна взаємодія двох кульок, навіть якщо вони матимуть масу десятки чи сотні тон, буде дуже слабкою).
    * Потенціальна енергія може накопичуватись, а потім витрачатись тільки під час зіткнення. У сам момент пружного удару кінетична енергія тимчасово переходить у потенціальну енергію пружної деформації, а потім повертається назад у кінетичну. Ми розглядаємо моменти "до" та "після" зіткнення, а не саме зіткнення, тому у нашому виразі є лише кінетична енергія.

Розв'язуючи цю систему рівнянь, можна знайти швидкості тіл після зіткнення.

#### Центральне зіткнення

> **Центральне зіткнення** - це зіткнення, при якому швидкості тіл до і після удару напрямлені вздовж прямої, що з'єднує центри мас цих тіл.

Розв'язання задач про центральні зіткнення спрощується, оскільки рух відбувається вздовж однієї прямої, і векторні рівняння можна замінити скалярними у проєкціях на цю пряму. Наведені вище приклади для пружних та непружних ударів також можна розглядати як приклади центральних зіткнень за умови, що швидкості тіл до та після удару напрямлені вздовж тієї самої прямої.
