# Фазові переходи

У попередніх темах ми детально розібрали, що речовина складається з частинок, які безперервно рухаються та взаємодіють. Саме співвідношення між енергією їхнього руху (кінетичною) та енергією взаємодії (потенціальною) визначає агрегатний стан речовини: твердий, рідкий чи газоподібний.

Сьогодні ми розглянемо ці стани більш формально і, що найголовніше, дослідимо процеси переходу між ними.

## Що таке фаза та рівноважний стан?

У фізиці поняття "агрегатний стан" є частиною більш загального поняття "фаза".

> **Фаза** - це однорідна за своїми фізичними властивостями частина системи, відокремлена від інших частин поверхнею поділу.

Простими словами, якщо ми бачимо чітку межу, яка розділяє речовину на частини з різними властивостями, то ми маємо справу з різними фазами.
*   Лід, що плаває у воді, - це система з двох фаз: твердої (лід) та рідкої (вода).
*   Вода та її насичена пара в закритій посудині - це також система з двох фаз.
*   Суміш води та олії - теж двофазна система, оскільки вони не змішуються і мають чітку межу поділу.
*   А от суміш азоту та кисню в повітрі - це одна газова фаза, бо гази повністю змішуються і система є однорідною.

Ми будемо розглядати **рівноважні стани**, тобто такі стани, в яких макроскопічні параметри системи (тиск $p$, об'єм $V$, температура $T$) залишаються незмінними як завгодно довго за відсутності зовнішніх впливів. Перехід з одного такого рівноважного стану в інший називається **фазовим переходом**.

> **Фазовий перехід** - це процес переходу речовини з однієї фази в іншу при зміні зовнішніх умов (зазвичай температури або тиску).

## Термодинаміка фазових переходів

Розглянемо найважливіші фазові переходи, пов'язані зі зміною агрегатного стану. Ключовим моментом тут є робота з енергією.

### Плавлення та кристалізація (тверде тіло - рідина)

Уявімо, що ми взяли шматок льоду і почали його повільно нагрівати.
1.  Спочатку температура льоду зростає. Ми надаємо тепло, і воно йде на збільшення кінетичної енергії молекул води - вони починають коливатися сильніше у вузлах кристалічної ґратки.
2.  Коли температура досягає 0 °C (за нормального тиску), лід починає плавитися. І тут відбувається найцікавіше: ми продовжуємо підводити тепло, але температура суміші льоду та води **не змінюється**! Вона залишається 0 °C, доки весь лід не розплавиться.

**Куди ж дівається енергія?**

З точки зору МКТ, плавлення - це процес руйнування кристалічної ґратки. Енергія, яку ми підводимо, йде не на збільшення швидкості руху молекул (тобто не на збільшення їхньої кінетичної енергії), а на **збільшення їхньої потенціальної енергії**. Ця енергія потрібна, щоб подолати сили притягання між молекулами і "вирвати" їх із жорстких вузлів ґратки, давши їм свободу переміщатися одна відносно одної, як у рідині.

> **Температура плавлення** - це температура, за якої тверда кристалічна речовина переходить у рідкий стан. Під час плавлення температура речовини не змінюється.

Для кількісного опису цього процесу вводиться спеціальна фізична величина.

> **Питома теплота плавлення** $\lambda$ (лямбда) - це фізична величина, що показує, яку кількість теплоти необхідно надати 1 кг кристалічної речовини, щоб за температури плавлення повністю перетворити її на рідину.

Одиниця в СІ: **Джоуль на кілограм (Дж/кг)**.

Кількість теплоти $Q$, необхідна для плавлення тіла масою $m$, розраховується за формулою:

$$Q = \lambda m$$

**Кристалізація (тверднення)** - це зворотний процес переходу з рідкого стану в твердий. Він відбувається за тієї ж температури, що й плавлення. При кристалізації, навпаки, **виділяється** точно така ж кількість теплоти. Молекули, "шикуючись" у кристалічну ґратку, зменшують свою потенціальну енергію, і цей надлишок виділяється у вигляді тепла.

### Пароутворення та конденсація (рідина - газ)

Ці процеси ми вже детально обговорювали, коли розглядали насичену пару та вологість. Давайте тепер зосередимось на енергетичному аспекті.

Як і при плавленні, при кипінні рідини вся енергія, що підводиться, йде не на нагрівання, а на сам фазовий перехід.

> **Температура кипіння** - це температура, за якої рідина інтенсивно перетворюється на пару по всьому об'єму. Під час кипіння температура рідини не змінюється.

Енергія, що підводиться під час кипіння, йде на:
1.  Подолання сил міжмолекулярного притягання в рідині, щоб "відірвати" молекули одна від одної.
2.  Виконання роботи проти сил зовнішнього тиску, оскільки пара займає значно більший об'єм, ніж рідина.

> **Питома теплота пароутворення** $L$ - це фізична величина, що показує, яку кількість теплоти необхідно надати 1 кг рідини, щоб за температури кипіння повністю перетворити її на пару.

Одиниця в СІ: **Джоуль на кілограм (Дж/кг)**.

Кількість теплоти $Q$, необхідна для перетворення на пару рідини масою $m$, розраховується за формулою:

$$Q = L m$$

**Конденсація** - це зворотний процес. При конденсації пара виділяє точно таку ж кількість теплоти.

**Цікавий факт:** Питома теплота пароутворення води ($L \approx 2.3 \cdot 10^6$ Дж/кг) набагато більша за питому теплоту плавлення льоду ($\lambda \approx 3.3 \cdot 10^5$ Дж/кг). Чому?
*   **Аналогія:** Уявіть собі театр, повний глядачів на своїх місцях (тверде тіло).
    *   **Плавлення:** Щоб глядачі могли вільно ходити по залу, але не виходити з нього, потрібно подолати "порядок" сидіння. Це вимагає певної енергії ($\lambda$).
    *   **Пароутворення:** Щоб глядачі не просто ходили по залу, а повністю покинули будівлю театру і розійшлися по місту, потрібно набагато більше енергії ($L$).
    
Так само і з молекулами: при плавленні ми лише руйнуємо дальній порядок, але молекули залишаються близько одна до одної. При пароутворенні ми практично повністю розриваємо зв'язки між ними, розводячи їх на великі відстані. Це вимагає значно більше енергії.

## Фазова діаграма: карта станів речовини

Ми звикли думати, що вода замерзає при 0 °C і кипить при 100 °C. Але ми вже знаємо, що температура кипіння залежить від тиску. Виявляється, температура плавлення також залежить від тиску, хоч і значно слабше.

Щоб наочно представити, в якому фазовому стані буде знаходитись речовина за певних умов, використовують **фазові діаграми**.

> **Фазова діаграма** - це графік, що показує рівноважні стани речовини (фази) залежно від макроскопічних параметрів, зазвичай тиску $p$ та температури $T$.

### Ключові елементи фазової діаграми

**Осі координат:** По вертикалі відкладено тиск $p$, по горизонталі - абсолютну температуру $T$.

На діаграмі ми бачимо три області, що відповідають трьом фазам: **Тверда (Лід)**, **Рідка (Вода)**, **Газоподібна (Пара)**. Якщо точка з координатами ($T, p$) лежить всередині однієї з цих областей, то при таких умовах речовина існуватиме саме в цій фазі. Ці області розділені трьома кривими.

1.  **Крива плавлення:** Ця лінія розділяє тверду та рідку фази (Рис. 1). Будь-яка точка на цій кривій відповідає умовам (тиску та температурі), за яких лід і вода можуть співіснувати в рівновазі.
    *   **Особливість води:** Зверніть увагу, що крива AB для води має **від'ємний нахил**. Це означає, що **зі збільшенням тиску температура плавлення льоду зменшується!** Це аномальна властивість (у більшості речовин навпаки). Саме завдяки цьому ковзани легко ковзають по льоду: під великим тиском леза ковзана лід плавиться, утворюючи тонкий шар води, який зменшує силу тертя.

2.  **Крива пароутворення:** Ця лінія розділяє рідку та газоподібну фази (Рис. 1). Точки на цій кривій показують, за якої температури кипить вода при заданому тиску (або, що те ж саме, який тиск має насичена пара при заданій температурі).

3.  **Крива сублімації:** Ця лінія розділяє тверду та газоподібну фази (Рис. 1). Вона показує умови, за яких тверде тіло може переходити безпосередньо в газ, минаючи рідку фазу. Цей процес називається **сублімацією** (або **випаровуванням твердих тіл**).
    *   **Приклад:** Сухий лід (твердий вуглекислий газ) за звичайного тиску не плавиться, а одразу випаровується. Мокра білизна висихає на морозі - це теж сублімація льоду.

### Особливі точки на діаграмі

На фазовій діаграмі є дві унікальні точки.

**Потрійна точка**

> **Потрійна точка** - це унікальна для кожної речовини точка на фазовій діаграмі, що відповідає таким значенням тиску і температури, за яких у рівновазі можуть одночасно існувати всі три фази: тверда, рідка і газоподібна.

Для води потрійна точка має координати:
*   $T \approx 273.16 \text{ K}$ (або +0.01 °C)
*   $p \approx 611.657 \text{ Па}$ (це дуже низький тиск, приблизно в 165 разів менший за нормальний атмосферний).

**Критична точка**

Рухаючись по кривій пароутворення BC вгору, ми бачимо, що зі зростанням температури і тиску густина рідини зменшується (вона розширюється), а густина насиченої пари зростає. У точці C густини рідини і пари стають однаковими.

> **Критична точка** - це точка на фазовій діаграмі, в якій зникає різниця у фізичних властивостях між рідиною та її насиченою парою. Межа поділу між ними зникає.

Параметри, що відповідають цій точці, називаються критичними:
*   **Критична температура ($T_{\text{кр}}$)** - це температура, вище якої речовина не може існувати в рідкому стані, яким би високим не був тиск.
*   **Критичний тиск ($p_{\text{кр}}$)** - максимальний тиск, при якому може спостерігатися фазова границя між рідиною та газом.

Для води $T_{\text{кр}} \approx 647 \text{ K}$ (374 °C), $p_{\text{кр}} \approx 22.1 \text{ МПа}$ (у 218 разів більше атмосферного).

Стан речовини при $T > T_{\text{кр}}$ і $p > p_{\text{кр}}$ називається **надкритичною рідиною (надкритичним флюїдом)**. Це особливий стан, який має густину, як у рідини, але в'язкість і здатність до дифузії, як у газу.

## Особливі фазові стани та переходи

Окрім класичних трьох станів, існують і інші, не менш цікаві фазові стани.

### Рідкі кристали (РК)

Це особливий проміжний стан (фаза), в якому деякі речовини перебувають між твердим кристалічним та рідким станами.

> **Рідкі кристали** - це фазовий стан, що поєднує в собі властивості як рідини, так і кристала.

*   Від рідин вони взяли **плинність** - здатність текти і приймати форму посудини.
*   Від кристалів - **анізотропію**, тобто залежність фізичних властивостей (оптичних, електричних) від напрямку. Примітка: звичайна рідина є **ізотропною** - її властивості однакові в усіх напрямках.

Як це можливо? Молекули рідких кристалів зазвичай мають видовжену, паличкоподібну форму.
*   У твердому стані вони впорядковані і за положенням (утворюють кристалічну ґратку), і за орієнтацією.
*   У рідкому стані вони рухаються хаотично і не мають жодного порядку.
*   У рідкокристалічному стані молекули втрачають порядок у розташуванні (немає ґратки, вони можуть вільно переміщатися), але зберігають **дальній орієнтаційний порядок**. Це означає, що осі більшості молекул у всьому об'ємі напрямлені переважно в один бік.

**Аналогія:** Уявіть коробку, наповнену олівцями.
*   **Твердий кристал:** Олівці акуратно складені в ряди, нерухомі.
*   **Рідина:** Олівці хаотично накидані і перемішані.
*   **Рідкий кристал:** Олівці можуть вільно ковзати один відносно одного, але всі вони орієнтовані в одному напрямку.

Найбільш поширеними є **нематичні рідкі кристали** (нематики). У них молекули хаотично розташовані в просторі, але їхні довгі осі здебільшого орієнтовані вздовж одного напрямку. Кажуть, що в них є **дальній орієнтаційний порядок**.

Найвідоміше застосування рідких кристалів - це **РК-екрани (рідкокристалічні екрани)** (LCD - Liquid Crystal Display) в моніторах, телевізорах, калькуляторах. Принцип їхньої роботи ґрунтується на тому, що під дією електричного поля молекули РК можуть змінювати свою орієнтацію. Це, в свою чергу, змінює їхні оптичні властивості - вони можуть або пропускати поляризоване світло, або блокувати його. Керуючи електричним полем у кожній комірці (пікселі), можна створювати зображення.

### Інші фазові переходи

Поняття фазового переходу є набагато ширшим, ніж просто зміна агрегатного стану.

Деякі матеріали, як-от залізо, кобальт, нікель, є **феромагнетиками**. Вони можуть сильно намагнічуватися і створювати власне магнітне поле. Ця властивість пов'язана з упорядкованою орієнтацією мікроскопічних магнітних моментів атомів.

Якщо нагрівати феромагнетик, то при досягненні певної температури, яка називається **точкою Кюрі**, тепловий рух руйнує цей порядок. Речовина втрачає свої феромагнітні властивості і переходить в іншу магнітну фазу - **парамагнітну** (здатна лише дуже слабо намагнічуватися). Цей перехід також є фазовим.

При дуже низьких температурах деякі метали та сплави переходять у **надпровідний стан**. Це також фазовий перехід. Нижче певної критичної температури електричний опір матеріалу стрибком падає до нуля. Це означає, що одного разу запущений в такому надпровідному кільці електричний струм може циркулювати в ньому практично вічно без втрат.
