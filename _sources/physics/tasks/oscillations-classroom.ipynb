# Механічні коливання. Задачі

## Механічні коливання

**Задача 1**

Вібромотор у смартфоні для створення відчуття вібрації здійснює 3600 коливань за 30 секунд. Визначте період та частоту цих коливань.

**Задача 2**

Сучасна ультразвукова зубна щітка генерує коливання щетинок із частотою 30 кГц (30 000 Гц). Скільки повних коливань здійснять щетинки за 10 секунд роботи?

**Задача 3**

Фітнес-трекер на руці людини фіксує кроки, аналізуючи періодичний рух руки. Під час спокійної ходьби рука людини робить одне повне коливання (вперед-назад) приблизно за 1.25 с. Яку частоту коливань зафіксує пристрій?

**Задача 4**

Фітнес-трекер на руці спортсмена під час бігу фіксує 90 повних коливань руки (вперед-назад) за 30 секунд. Амплітуда цих коливань становить 25 см. Визначте частоту та період коливань руки спортсмена. Який шлях проходить кисть руки за один повний період коливань?

**Задача 5**

Квадрокоптер для аерофотозйомки стабільно завис у повітрі. Одна з його лопатей робить 12 000 обертів за хвилину. Хоча це обертальний рух, його можна характеризувати тими ж параметрами, що й коливальний.

**Задача 6**

Коливання мембрани (дифузора) динаміка, що відтворює чистий тон, описуються рівнянням: $x = 0.002 \cos(400\pi t)$, де всі величини вказані в СІ.
1.  Чому дорівнює амплітуда коливань?
2.  Визначте циклічну частоту.
3.  Обчисліть період та частоту звуку, який відтворює динамік.

**Задача 7**

Чутливий елемент сейсмографа може здійснювати гармонічні коливання. Під час слабкого підземного поштовху він почав коливатися з періодом 0.5 с. Амплітуда його коливань становила 1 мм (0.001 м). Вважаючи, що коливання почалися з максимального відхилення, запишіть рівняння цих коливань $x(t)$ (усі величини записуйте в системі СІ). Обчисліть максимальну швидкість та максимальне прискорення чутливого елемента.

**Задача 8**

Кристал в голівці ультразвукового медичного сканера для медичної діагностики коливається з частотою 2 МГц ($2 \cdot 10^6$ Гц). Амплітуда цих коливань надзвичайно мала і становить 0.1 мкм ($10^{-7}$ м).
1.  Визначте циклічну частоту коливань кристала.
2.  З якою максимальною швидкістю рухаються частинки кристала?
3.  Знайдіть максимальне прискорення.

## Пружинний та математичний маятники

Примітка: У задачах, де це необхідно, прискорення вільного падіння $g$ вважайте рівним 9.8 $\text{м}/\text{с}^2$.

**Задача 1**

Амортизатор у підвісці автомобіля можна змоделювати як пружинний маятник. Після проїзду нерівності частина автомобіля масою 300 кг починає коливатися на пружині підвіски. Визначте жорсткість пружини амортизатора, якщо період цих коливань становить 1.2 с.

**Задача 2**

Для розваги в парку атракціонів встановили гігантську гойдалку, яку можна вважати математичним маятником. Довжина тросів, на яких підвішене сидіння, становить 10 м. З яким періодом буде коливатися ця гойдалка при малих кутах відхилення?

**Задача 3**

Майбутня місія на Марс планує дослідити місцеве прискорення вільного падіння за допомогою маятника. Яким буде період коливань маятника довжиною 1 м на Марсі, якщо відомо, що прискорення вільного падіння там становить приблизно 3.7 $\text{м}/\text{с}^2$?

**Задача 4**

Амортизатор у підвісці сучасного електромобіля можна змоделювати як пружинний маятник. Нехай маса, що припадає на один амортизатор, дорівнює 400 кг. Жорсткість пружини амортизатора становить 40 000 Н/м. Обчисліть період вільних коливань автомобіля на цій пружині після проїзду нерівності.

**Задача 5**

Під час будівельних робіт хмарочоса вантаж масою 2 тонни підвішений на сталевому тросі баштового крана довжиною 81 м. Коливання цього вантажу можна змоделювати як математичний маятник. Визначте період малих коливань вантажу. Вважайте, що прискорення вільного падіння дорівнює 10 $\text{м}/\text{с}^2$.

**Задача 6**

На Міжнародній космічній станції (МКС) для вимірювання маси космонавтів в умовах невагомості використовують спеціальне крісло, що коливається на пружинах.
1.  Під час тренування космонавт сів у крісло, і система почала коливатися з періодом 2.5 с. Визначте масу космонавта, якщо сумарна жорсткість пружин крісла дорівнює 1600 Н/м. Масою крісла можна знехтувати.
2.  Через місяць тренувань у невагомості період коливань цього ж космонавта у кріслі зменшився. Це свідчить про збільшення чи зменшення його м'язової маси? Поясніть свою відповідь.

**Задача 7**

Будівельний кран підіймає вантаж масою 2 тонни на тросі довжиною 20 м. Через порив вітру вантаж починає розгойдуватися як математичний маятник. Максимальна швидкість вантажу при проходженні положення рівноваги склала 1.5 м/с. Визначте повну механічну енергію коливань вантажу та максимальну висоту, на яку він підіймається відносно положення рівноваги.

**Задача 8**

Геологи для пошуку родовищ корисних копалин використовують гравіметри - прилади для точного вимірювання прискорення вільного падіння. Один із таких приладів є математичним маятником. Під час вимірювань у певній місцевості маятник довжиною 0.5 м здійснив 50 повних коливань за 71 с. Обчисліть значення прискорення вільного падіння в цій точці. Чи може це свідчити про наявність покладів щільних руд?

**Задача 9**

Безпілотний дрон доставляє невеликий вантаж масою 2 кг, підвішений на тросі довжиною 1.6 м. Під час польоту вантаж починає розгойдуватися. Максимальний кут відхилення троса від вертикалі становить $30^\circ$.
1. На яку максимальну висоту $h_{max}$ піднімається вантаж відносно свого найнижчого положення? (Підказка: використайте тригонометрію. $h = l - l \cos\alpha$).
2. Якою буде швидкість вантажу в момент, коли кут відхилення троса від вертикалі становитиме 15°? Вважайте $g = 10$ $\text{м}/\text{с}^2$.
