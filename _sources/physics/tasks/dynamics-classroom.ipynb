# Динаміка. Задачі

## Закони Ньютона

**Задача 1**

Сучасний квадрокоптер для доставки важить 4 кг. Щоб вертикально піднятися вгору, його двигуни повинні створити силу тяги, більшу за силу тяжіння. Припустимо, що сумарна сила тяги двигунів становить 60 Н, а сила тяжіння — 40 Н.
1.  Знайдіть рівнодійну силу, що діє на квадрокоптер під час зльоту.
2.  Обчисліть прискорення, з яким квадрокоптер починає підніматися.

**Задача 2**

На краш-тесті автомобіль масою 1500 кг зупиняється, врізавшись у перешкоду. Датчики показують, що в момент удару рівнодійна сила, яка діяла на автомобіль, дорівнювала 300000 Н. Якого прискорення (за модулем) зазнав автомобіль у цей момент?

**Задача 3**

Смартфон масою 200 г випадково падає з ваших рук. Під час падіння він притягується до Землі.
1.  З якою силою Земля притягує смартфон? (Вважайте, що прискорення вільного падіння дорівнює $g = 10 \text{ м}/\text{с}^2$).
2.  Згідно з третім законом Ньютона, смартфон також притягує до себе Землю. З якою силою смартфон притягує Землю?
3.  Маса Землі приблизно дорівнює $6 \cdot 10^{24}$ кг. Якого прискорення надає Землі сила притягання смартфона? Порівняйте це прискорення з прискоренням смартфона і поясніть, чому ми не помічаємо руху Землі.

**Задача 4**

Ліфт у сучасному хмарочосі масою 800 кг починає рух вгору зі стану спокою і за 3 секунди досягає швидкості 6 м/с. Рух вважайте рівноприскореним.
1.  Знайдіть прискорення ліфта.
2.  Обчисліть рівнодійну силу, яка надає ліфту цього прискорення.
3.  Знайдіть силу натягу троса, що піднімає ліфт (сила тяжіння, що діє на ліфт, дорівнює 8000 Н).

**Задача 5**

Двоє космонавтів у відкритому космосі, де немає тертя та опору, вирішили перевірити закони фізики. Перший космонавт масою 80 кг відштовхується від другого, масою 100 кг. У момент поштовху перший космонавт набуває прискорення 0.5 $\text{ м}/\text{с}^2$.
1.  Яку силу приклав перший космонавт до другого?
2.  Якого прискорення набув другий космонавт?

**Задача 6**

Кур'єр на електросамокаті масою 100 кг (разом із самокатом) рухається зі швидкістю 18 км/год. Він помічає перешкоду і починає гальмувати, зупиняючись за 2.5 секунди.
1.  Переведіть початкову швидкість у м/с.
2.  Знайдіть прискорення самоката під час гальмування (не забудьте про його напрямок).
3.  Визначте модуль гальмівної сили, що діяла на кур'єра із самокатом.

**Задача 7**

Інженер тестує новий іонний двигун для космічного зонда. Зонд масою 500 кг перебуває у глибокому космосі, де зовнішніми силами можна знехтувати.
1.  Перший тест: двигун працює в номінальному режимі, створюючи тягу 0.5 Н. Якого прискорення набуває зонд? Яку швидкість він набере за 24 години роботи двигуна?
2.  Другий тест: інженер вмикає форсажний режим, і прискорення зонда зростає до 0.003 $\text{ м}/\text{с}^2$. Яку силу тяги створює двигун у цьому режимі?
3.  Уявіть, що для корекції курсу одночасно вмикаються два однакові двигуни, розташовані під кутом 90° один до одного, кожен із силою тяги 0.5 Н. Яким буде модуль рівнодійної сили, що діє на зонд?

## Закон всесвітнього тяжіння

**Задача 1**

Уявіть, що два новітні електропікапи Tesla Cybertruck, масою 3000 кг кожен, припарковані так, що відстань між їхніми центрами мас становить 4 метри. Обчисліть силу гравітаційного притягання між ними. Порівняйте цю силу із силою, з якою Земля притягує один із цих автомобілів.

**Задача 2**
Міжнародна космічна станція (МКС) обертається на висоті приблизно 400 км над поверхнею Землі. Маса станції становить близько 450 000 кг.
1. Яка сила тяжіння діє на МКС з боку Землі на цій висоті?
2. Чому астронавти на борту перебувають у стані невагомості, якщо сила тяжіння є такою значною?

(Для розрахунків візьміть масу Землі $M_З = 6 \cdot 10^{24}$ кг, радіус Землі $R_З = 6400$ км, $G = 6.67 \cdot 10^{-11} \frac{\text{Н} \cdot \text{м}^2}{\text{кг}^2}$).

**Задача 3**

Марсохід "Perseverance" працює на поверхні Марса. Використовуючи довідкові дані, обчисліть прискорення вільного падіння на поверхні Марса. Наскільки воно менше за земне? (Маса Марса $M_M \approx 6.4 \cdot 10^{23}$ кг, радіус Марса $R_M \approx 3400$ км).

**Задача 4**
У рамках програми "Артеміда" NASA планує нові місії на Місяць. Щоб вивести супутник на низьку навколомісячну орбіту, йому потрібно надати першу космічну швидкість. Обчисліть, чому дорівнює перша космічна швидкість для Місяця. (Маса Місяця $M_{Міс} \approx 7.35 \cdot 10^{22}$ кг, радіус Місяця $R_{Міс} \approx 1740$ км).

**Задача 5**

Супутник зв'язку Starlink рухається по коловій орбіті на висоті 550 км. Згодом, для корекції орбіти, його двигуни переміщують його на вищу орбіту — 1100 км. У скільки разів змінилася сила гравітаційного притягання, що діє на супутник з боку Землі? Сила збільшилася чи зменшилася? (Радіус Землі $R_З = 6400$ км).

**Задача 6**

Астрономи відкрили екзопланету Kepler-186f, яку вважають однією з найбільш схожих на Землю. Її радіус становить приблизно $1.1$ радіуса Землі, а маса — $1.5$ маси Землі.
1. Обчисліть прискорення вільного падіння на поверхні Kepler-186f.
2. Якщо ваша маса становить 60 кг, якою буде ваша вага на цій екзопланеті? (Прискорення вільного падіння на Землі вважайте рівним $g_0 \approx 9.8 \text{ м/с}^2$).

**Задача 7**

Геостаціонарні супутники, які використовуються для телевізійного мовлення, "висять" над однією точкою земної поверхні. Це можливо тому, що вони рухаються по орбіті з періодом, що дорівнює періоду обертання Землі. Прискорення вільного падіння на такій орбіті значно менше, ніж на поверхні. На якій висоті над поверхнею Землі прискорення вільного падіння буде в 4 рази меншим, ніж на її поверхні? (Радіус Землі $R_З = 6400$ км).

**Задача 8**

Уявіть, що ви астроном, який вивчає Юпітер. За допомогою спостережень за його супутниками ви визначили, що прискорення вільного падіння на умовній "поверхні" хмар Юпітера становить $g_Ю \approx 24.8 \text{ м/с}^2$. Радіус Юпітера становить приблизно $R_Ю \approx 70000$ км. Використовуючи ці дані, виконайте те, що колись зробив Кавендіш для Землі — "зважте" Юпітер, тобто обчисліть його масу.

**Задача 9**

Між Землею та Місяцем існує точка (точка Лагранжа L1), в якій сили гравітаційного притягання від обох небесних тіл зрівноважують одна одну. Космічний апарат, розміщений у цій точці, буде "зависати" між ними. Знайдіть відстань від центру Землі до цієї точки. (Маса Землі $M_З \approx 81 \cdot M_{Міс}$, відстань між центрами Землі та Місяця $r \approx 384000$ км).

**Задача 10**

Супутники системи глобального позиціонування (GPS) обертаються навколо Землі на висоті приблизно 20 200 км.
1. Обчисліть швидкість, з якою рухається такий супутник по своїй орбіті.
2. Визначте період обертання супутника навколо Землі (час, за який він робить один повний оберт).
(Маса Землі $M_З = 6 \cdot 10^{24}$ кг, радіус Землі $R_З = 6400$ км).

**Задача 11**

Астрономи активно шукають "Супер-Землі" — планети, масивніші за Землю, але легші за газових гігантів. Розглянемо гіпотетичну планету, радіус якої вдвічі більший за радіус Землі ($R_п = 2R_З$), а маса — у 10 разів більша за масу Землі ($M_п = 10M_З$).
1. У скільки разів прискорення вільного падіння на поверхні цієї планети більше, ніж на поверхні Землі?
2. У скільки разів перша космічна швидкість для цієї планети більша за першу космічну швидкість для Землі?


## Механічна робота, кінетична енергія та потужність

**Задача 1**

Малогабаритний квадрокоптер піднімає вертикально вгору пакунок масою 400 г на висоту 15 м для доставки на балкон. Яку роботу виконує підіймальна сила двигунів квадрокоптера? Вважайте, що пакунок рухається зі сталою швидкістю, а прискорення вільного падіння дорівнює $g_0 \approx 10 \text{ м/с}^2$.

**Задача 2**

Електросамокат разом з водієм має масу 90 кг і рухається зі швидкістю 20 км/год. Обчисліть кінетичну енергію самоката.

**Задача 3**

Ліфт у сучасному хмарочосі піднімає кабіну масою 1200 кг на висоту 50 м за 10 секунд. Яку середню потужність розвиває двигун ліфта під час цього підйому? Вважайте рух рівномірним, а прискорення вільного падіння $g_0 \approx 10 \text{ м/с}^2$.

**Задача 4**

Велосипедист, що рухався зі швидкістю 18 км/год, починає гальмувати до повної зупинки. Яку роботу виконала сила тертя? Маса велосипедиста разом з велосипедом становить 80 кг.

**Задача 5**

Автомобіль Tesla Model 3 масою 1800 кг рухався зі швидкістю 108 км/год, після чого водій застосував екстрене гальмування до повної зупинки. Яку роботу виконала сумарна сила гальмування (включаючи тертя та опір повітря)?

**Задача 6**

Робот-доставник тягне за собою по горизонтальній поверхні контейнер з посилками, прикладаючи силу 100 Н під кутом 30° до горизонту. Контейнер переміщується на відстань 20 метрів. Сила тертя, що діє на контейнер, становить 30 Н. Обчисліть:
1.  Роботу, яку виконує робот.
2.  Роботу, яку виконує сила тертя.
3.  Роботу рівнодійної сили.

**Задача 7**

Високошвидкісний потяг "Інтерсіті+" масою 450 тонн рухається зі швидкістю 144 км/год. У разі виникнення перешкоди на колії, автоматична система вмикає гальмування із середньою силою 300 кН. На якій мінімальній відстані до перешкоди має спрацювати система, щоб потяг встиг зупинитися?

**Задача 8**

Електромобіль, рухаючись рівномірно і прямолінійно зі швидкістю 90 км/год, долає силу опору повітря та тертя, сумарна величина якої становить 800 Н. Яку потужність розвиває двигун електромобіля в цей момент? Відповідь дайте у Ватах та кінських силах.

**Задача 9**

Професійний бігун масою 75 кг біжить зі швидкістю 6 м/с. Водночас він тримає в руці смартфон масою 200 г. У скільки разів кінетична енергія бігуна більша за кінетичну енергію смартфона, що рухається разом з ним?

**Задача 10**

Міжпланетний зонд масою 500 кг рухається у відкритому космосі зі швидкістю 8 км/с відносно Сонця. Для корекції траєкторії на короткий час вмикаються іонні двигуни, які виконують роботу в $2.25 \cdot 10^9$ Дж у напрямку руху зонда. Якою стане швидкість зонда після завершення роботи двигунів?

**Задача 11**

Лопаті сучасної вітрової турбіни за 1 секунду перехоплюють потік повітря масою 12 000 кг, що рухається зі швидкістю 15 м/с.
1.  Яка кінетична енергія цього об'єму повітря?
2.  Яку максимальну потужність міг би розвивати вітрогенератор, якби вся кінетична енергія повітря перетворювалася на електричну?
3.  Обчисліть реальну електричну потужність турбіни, якщо її коефіцієнт корисної дії (ККД) становить 45%.

**Задача 12**

Електромобіль Rivian R1T масою 3200 кг розганяється з місця до швидкості 108 км/год за 4 секунди. Середня сила опору руху під час розгону становить 1000 Н.
1.  Яку силу тяги повинен розвивати двигун, щоб забезпечити такий розгін?
2.  Яку миттєву потужність розвиває двигун в кінці розгону (при швидкості 108 км/год)?

**Задача 13**

Турист везе валізу масою 20 кг на коліщатках по горизонтальній підлозі аеропорту, прикладаючи силу 50 Н за допомогою ручки, нахиленої під кутом 60° до горизонту.
1.  Яку роботу виконав турист, перемістивши валізу на відстань 30 м?
2.  Яку роботу виконала сила тяжіння на цьому шляху?

**Задача 14**

Електромобіль Tesla Model S масою 2200 кг розганяється від 0 до 100 км/год за 3,2 секунди.
1.  Яку роботу виконала рівнодійна всіх сил, що діяли на автомобіль під час розгону?
2.  Яку середню потужність розвивали двигуни автомобіля під час цього розгону? (Силами опору знехтувати).
