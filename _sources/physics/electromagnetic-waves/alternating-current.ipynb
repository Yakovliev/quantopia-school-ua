# Змінний струм

У попередніх темах ми детально розглядали вільні та затухаючі електромагнітні коливання в коливальному контурі. Ми бачили, що якщо зарядити конденсатор і надати системі "поштовх", вона почне коливатися, але через неминучі втрати енергії (наприклад, на нагрівання провідників) ці коливання з часом згасають.

Але майже вся електроенергія, яку ми використовуємо в побуті та промисловості - від розетки у вашій кімнаті до потужних заводських верстатів - це енергія **змінного струму**. А змінний струм - це приклад **незатухаючих вимушених електромагнітних коливань**.

> **Вимушені електромагнітні коливання** - це незатухаючі коливання заряду, напруги та сили струму, які спричинені дією в електричному колі зовнішньої електрорушійної сили (ЕРС), що періодично змінюється.

Пристроєм, який створює таку змінну ЕРС і є джерелом змінного струму, є **генератор**. Сьогодні ми детально розберемося, що таке змінний струм і як працює його головне джерело - індукційний генератор.

## Фундамент: явище електромагнітної індукції

В основі роботи будь-якого електродинамічного генератора лежить одне з найкрасивіших явищ у фізиці - **явище електромагнітної індукції**, відкрите Майклом Фарадеєм. Давайте коротко повторимо матеріал про це явище. Суть його проста:

> Якщо змінювати магнітне поле, що пронизує замкнений провідний контур, то в цьому контурі виникає електричний струм. Такий струм називають **індукційним**.

Щоб кількісно описати це явище, нам потрібно ввести поняття **магнітного потоку**.

### Магнітний потік

Уявіть собі дощ. Кількість води, що пройде крізь відчинене вікно, залежить від трьох речей: інтенсивності дощу, площі вікна і кута, під яким падають краплі. Якщо дощ перпендикулярний до вікна - води потрапить максимум. Якщо він іде паралельно - жодна крапля не залетить.

Магнітний потік - це дуже схожа аналогія. Він описує, наскільки "інтенсивно" магнітне поле "пронизує" певну поверхню.

> **Магнітний потік** $\Phi$ - це фізична скалярна величина, яка є потоком вектора магнітної індукції через задану поверхню.

Примітка: У математиці є таке поняття як потік векторного поля. Магнітний потік - це і є потік векторного поля, де у якості векторного поля є реальне фізичне магнітне векторне (бо має напрямок) поле. Подальші формули - це не просто результат роботи фізиків, це частина цілого математичного розділу, де розглядають векторні поля і описують їх математично.

Для однорідного магнітного поля з індукцією $\vec{B}$ та плоскої рамки площею $S$, магнітний потік розраховується за формулою:

$$\Phi = BS \cos \alpha$$

де:
*   $B$ - модуль вектора магнітної індукції. Вимірюється в **Теслах (Тл)**.
*   $S$ - площа контуру (рамки). Вимірюється в **квадратних метрах ($\text{м}^2$)**.
*   $\alpha$ - кут між вектором магнітної індукції $\vec{B}$ і вектором нормалі $\vec{n}$ (перпендикуляром) до площини рамки.

Одиниця вимірювання магнітного потоку в СІ - **Вебер (Вб)**. $1 \text{ Вб} = 1 \text{ Тл} \cdot \text{м}^2$.

### Закон Фарадея

Фарадей встановив, що причина виникнення струму - це не сам магнітний потік, а його **зміна з часом**.

> **Закон електромагнітної індукції (Закон Фарадея)**: Електрорушійна сила (ЕРС) індукції, що виникає в замкненому контурі, дорівнює за модулем швидкості зміни магнітного потоку, що пронизує цей контур.

Математично це записується так:

$$
\mathcal{E}_i = -\frac{\Delta\Phi}{\Delta t} \quad \text{або в формі похідної:} \quad \mathcal{E}_i = -\Phi'(t)
$$

де:
*   $\mathcal{E}_i$ - **електрорушійна сила (ЕРС) індукції**. Це не силова характеристика, а енергетична. Вона показує, яку роботу виконують сторонні (не кулонівські) сили при переміщенні одиничного позитивного заряду вздовж контуру. Вимірюється у **Вольтах (В)**.
*   $\Delta\Phi$ - зміна магнітного потоку за проміжок часу $\Delta t$.
*   Знак "мінус" відображає **правило Ленца**: індукційний струм, що виникає, завжди має такий напрямок, що його власне магнітне поле протидіє тій зміні магнітного потоку, яка цей струм породила.

Отже, щоб створити ЕРС, нам потрібно змусити магнітний потік змінюватися. Як це зробити, виходячи з формули $\Phi = BS \cos \alpha$? Є три шляхи:
1.  Змінювати величину магнітного поля $B$.
2.  Змінювати площу контуру $S$.
3.  **Змінювати кут $\alpha$ між полем і рамкою, тобто обертати рамку.**

Саме третій спосіб і є основою роботи генераторів змінного струму.

## Принцип роботи ідеального генератора

Уявімо собі найпростішу модель генератора: прямокутна провідна рамка з $N$ витків дроту, яка з постійною кутовою швидкістю обертається в однорідному магнітному полі.

### Виведення рівняння ЕРС

Давайте крок за кроком виведемо закон, за яким буде змінюватися ЕРС у такій рамці.

1.  **Запишемо залежність кута від часу.**

    Якщо рамка обертається рівномірно з кутовою швидкістю $\omega$, то кут повороту $\alpha$ в момент часу $t$ дорівнює:

    $$\alpha = \omega t$$

    > **Циклічна (кутова) частота** $\omega$ показує, на який кут повертається рамка за одиницю часу. Вимірюється в **радіанах за секунду (рад/с)**.

2.  **Запишемо залежність магнітного потоку від часу.**

    Підставимо залежність кута $\alpha = \omega t$ у формулу магнітного потоку $\Phi = BS \cos \alpha$. Отримаємо закон, за яким магнітний потік, що пронизує рамку, змінюється з часом:

    $$
    \Phi(t) = BS \cos(\omega t)
    $$

    Ми бачимо, що магнітний потік змінюється за гармонічним законом (за законом косинуса).

3.  **Знайдемо ЕРС індукції за законом Фарадея.**

    Тепер найважливіший крок. Щоб знайти ЕРС, нам потрібно взяти похідну від магнітного потоку по часу зі знаком "мінус":

    $$
    \mathcal{E}_i(t) = -\frac{d\Phi}{dt} = -(BS \cos(\omega t))'
    $$

    Величини $B$ і $S$ є сталими, тому ми виносимо їх за знак похідної:

    $$
    \mathcal{E}_i(t) = -BS \cdot (\cos(\omega t))'
    $$

    Похідна від складеної функції $\cos(\omega t)$ дорівнює $-\omega \sin(\omega t)$. Підставляємо:

    $$
    \mathcal{E}_i(t) = -BS \cdot (-\omega \sin(\omega t))
    $$

    Після спрощення отримуємо:

    $$
    \varepsilon(t) = BS\omega \sin(\omega t)
    $$

    *Примітка: ми замінили позначення $\mathcal{E}_i$ на $\varepsilon$, оскільки в темі змінного струму миттєві значення величин прийнято позначати маленькими літерами.*

    Якщо наша рамка (обмотка) має не один, а $N$ витків, то ЕРС в них будуть додаватися, і загальна ЕРС буде в $N$ разів більшою:
    $$
    \varepsilon(t) = NBS\omega \sin(\omega t)
    $$
    Це і є **основне рівняння ЕРС генератора змінного струму**.

### Аналіз отриманого рівняння

Давайте проаналізуємо формулу $\varepsilon(t) = NBS\omega \sin(\omega t)$.
*   Ми бачимо, що ЕРС змінюється з часом за законом **синуса**. Це означає, що вона є **гармонічним коливанням**.
*   ЕРС періодично змінює своє значення і знак. Це і є ознакою **змінного струму**.

Тепер введемо ключові характеристики цих коливань.

> **Миттєве значення ЕРС** $\varepsilon(t)$ - це значення електрорушійної сили в даний момент часу.

> **Амплітудне значення ЕРС** $\mathcal{E}_{\max}$ - це максимальне за модулем значення ЕРС.
>
> $$\mathcal{E}_{\max} = NBS\omega$$
>
> Амплітуда ЕРС залежить від конструкції генератора ($N, S$), від сили магнітного поля ($B$) і від швидкості обертання ($\omega$).

Тепер наше рівняння можна записати у більш компактному вигляді:

$$
e = \mathcal{E}_{\max} \sin(\omega t)
$$

### Зв'язок параметрів коливань

Згадаймо зв'язок між циклічною частотою, звичайною частотою та періодом.

*   **Період коливань** $T$ - це час одного повного коливання (одного повного оберту рамки). Вимірюється в **секундах (с)**.
*   **Частота коливань** $\nu$ - це кількість повних коливань за одну секунду. Вимірюється в **Герцах (Гц)**.

Вони пов'язані з циклічною частотою $\omega$ наступними співвідношеннями:
$$
\omega = 2\pi\nu = \frac{2\pi}{T}
$$

Стандартна частота змінного струму в електромережах України та Європи становить $\nu = 50$ Гц. Це означає, що за одну секунду струм 50 разів змінює свій напрямок на протилежний (здійснює 50 повних коливань).

## Змінний струм у колі з активним навантаженням

Ми отримали джерело змінної ЕРС. Тепер підключимо до нього навантаження, наприклад, звичайну лампочку розжарення або обігрівач. Такі елементи, в яких електрична енергія перетворюється лише на теплову, називають **активним навантаженням**. Їх основна характеристика - це **активний опір** $R$.

> **Змінний електричний струм** - це вимушені коливання сили струму і напруги, зумовлені змінною ЕРС.

Знайдемо закон, за яким буде змінюватися сила струму в такому колі. Для цього використаємо **закон Ома для повного кола**:

$$
i = \frac{\text{ЕРС}}{\text{Повний опір}}
$$

Нехай наш генератор має внутрішній опір $r$, а зовнішнє навантаження - опір $R$. Тоді повний опір кола буде $R+r$. Підставимо миттєве значення ЕРС $\varepsilon(t) = \mathcal{E}_{\max} \sin(\omega t)$:

$$
i(t) = \frac{\mathcal{E}_{\max} \sin(\omega t)}{R+r}
$$

Величина $\frac{\mathcal{E}_{\max}}{R+r}$ є максимально можливим значенням сили струму. Позначимо її як амплітуду сили струму $I_{\max}$.

> **Амплітудне значення сили струму** $I_{\max}$ - це максимальне за модулем значення сили струму в колі.
>
> $$I_{\max} = \frac{\mathcal{E}_{\max}}{R+r}$$

Тепер рівняння для миттєвого значення сили струму набуває вигляду:

$$
i = I_{\max} \sin(\omega t)
$$

Аналогічно, напруга на зовнішній ділянці кола (на нашому навантаженні $R$) буде:

$$
u = iR = (I_{\max} \sin(\omega t))R = I_{\max}R \sin(\omega t)
$$

Позначивши $U_{\max} = I_{\max}R$ як амплітуду напруги, отримаємо:
$$
u = U_{\max} \sin(\omega t)
$$

**Важливий висновок:** у колі з чисто активним навантаженням коливання сили струму $i(t)$ і напруги $u(t)$ **збігаються за фазою**. Це означає, що вони одночасно досягають своїх максимальних значень і одночасно проходять через нуль.

У загальному випадку, якщо коло містить ще й конденсатори або котушки індуктивності, може виникати **зсув фаз** $\varphi_0$ між коливаннями струму та напруги. Тоді рівняння для струму матиме вигляд:
$$
i = I_{\max} \sin(\omega t + \varphi_0)
$$

## Будова та принцип дії промислових генераторів

Розглянута нами модель з обертовою рамкою добре ілюструє фізичний принцип, але в реальних промислових генераторах все влаштовано дещо інакше і значно ефективніше.

> **Генератор змінного струму (альтернатор)** - це пристрій, що перетворює механічну енергію на електричну енергію змінного струму на основі явища електромагнітної індукції.

Промисловий генератор складається з трьох основних частин:

1.  **Статор** - нерухома частина генератора. У потужних промислових генераторах на статорі розміщують обмотки, в яких індукується ЕРС і з яких знімається струм. Це робиться для зручності, оскільки з нерухомої частини набагато легше відводити велику потужність, ніж з обертової. Осердя статора виготовляють не суцільним, а з тонких, ізольованих одна від одної сталевих пластин. Це робиться для боротьби зі шкідливим явищем - **струмами Фуко**.
    > **Струми Фуко** (або вихрові струми) - це індукційні струми, що виникають у масивних провідниках при зміні магнітного потоку, що їх пронизує. Вони спричиняють нагрівання провідника (за законом Джоуля-Ленца) і призводять до втрат енергії. Ламіноване осердя сильно збільшує опір для цих струмів, тим самим зменшуючи їх.

2.  **Ротор** (індуктор) - обертова частина генератора. У промислових генераторах ротор є потужним електромагнітом, який створює магнітне поле. Обертаючись, це магнітне поле "пробігає" по витках обмотки статора, змінюючи в них магнітний потік і тим самим індукуючи ЕРС.

3.  **Система збудження** - для живлення електромагніту ротора потрібен постійний струм. Цей струм подається на ротор через **контактні (ковзні) кільця** та **щітки**. Щітки (зазвичай графітові) притискаються до кілець, що обертаються разом з ротором, забезпечуючи надійний електричний контакт.

**Принцип дії:**
1.  Зовнішня механічна сила (наприклад, від турбіни на гідро-, тепло- чи атомній електростанції) обертає ротор генератора.
2.  Електромагніт ротора створює потужне магнітне поле, яке обертається разом з ним.
3.  Обертове магнітне поле перетинає провідники обмотки статора.
4.  Внаслідок цього магнітний потік через обмотку статора безперервно змінюється за синусоїдальним законом.
5.  Відповідно до закону електромагнітної індукції, у витках обмотки статора виникає синусоїдальна ЕРС.
6.  Якщо до виводів обмотки статора підключити навантаження (споживача), в колі потече змінний електричний струм.

Отже, ми перетворили механічну енергію обертання на електричну енергію.


### Частота струму промислового генератора

У нашому простому прикладі з одним магнітом (однією парою полюсів N-S) за один повний оберт рамки відбувається один повний цикл зміни ЕРС. Отже, частота струму дорівнює частоті обертання рамки.

У промислових генераторах ротор може мати не одну, а кілька пар магнітних полюсів (кількість пар позначається як $p$). Це дозволяє отримати стандартну частоту струму (в Україні та Європі - 50 Гц) при меншій швидкості обертання ротора, що є важливим з інженерної точки зору.

Частота змінного струму $\nu$, що виробляється генератором, пов'язана зі швидкістю обертання ротора $n$ (в обертах за секунду) та кількістю пар магнітних полюсів $p$ наступною формулою:

$$
\nu = pn
$$

Наприклад, щоб отримати струм частотою 50 Гц за допомогою двополюсного генератора ($p=1$), ротор має обертатися зі швидкістю $n = \nu/p = 50/1 = 50$ об/с, або 3000 об/хв. Якщо ж генератор має дві пари полюсів ($p=2$), то для тієї ж частоти 50 Гц потрібна швидкість обертання вже 25 об/с, або 1500 об/хв.
