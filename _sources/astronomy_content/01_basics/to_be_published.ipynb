Дійсно, існування темної матерії було запідозрено саме завдяки спостереженням за рухом галактик та їхніх скупчень, які не відповідали прогнозам, заснованим лише на видимій матерії. Давайте розберемося, як це відбувалося.

### 1. Перші "дзвіночки": Фріц Цвіккі і скупчення галактик (1930-ті роки)

Перші натяки на існування невидимої матерії з'явилися ще в 1930-х роках завдяки швейцарському астроному **Фріцу Цвіккі**. Він вивчав **скупчення галактик Волосся Вероніки**.

* **Що він спостерігав:** Цвіккі виміряв швидкості, з якими окремі галактики рухалися всередині цього скупчення. Використовуючи ефект Доплера (зміщення спектральних ліній світла від рухомих об'єктів), він зміг визначити радіальні швидкості галактик (тобто швидкості, з якими вони рухаються до нас або від нас).
* **У чому була проблема:** Згідно з законами гравітації Ньютона, щоб скупчення галактик залишалося гравітаційно зв'язаним і не розліталося, його сумарна маса повинна бути достатньо великою, щоб утримувати галактики. Цвіккі підрахував масу скупчення, виходячи з яскравості видимих галактик (які є індикатором їхньої маси). Однак, виявилося, що обчислені швидкості галактик були набагато вищими, ніж це дозволяла б маса, отримана з видимої матерії.
* **Висновок Цвіккі:** Щоб пояснити, чому скупчення не розпадається, Цвіккі припустив, що у ньому має бути значно більше невидимої маси, ніж тієї, яку ми бачимо у вигляді зірок, газу та пилу. Він назвав цю невидиму матерію "темною матерією" (Dunkle Materie). Його розрахунки показали, що видимої матерії було лише близько 1-2% від необхідної для стабільності скупчення.

### 2. Ключові докази: Віра Рубін і криві обертання галактик (1970-ті роки)

Роботи Цвіккі довгий час залишалися без належної уваги, доки не були підтверджені та поглиблені дослідженнями американської астрономки **Віри Рубін** та її колег у 1970-х роках. Вона зосередилася на вивченні **кривих обертання спіральних галактик**.

* **Що вона спостерігала:** Віра Рубін вимірювала швидкості обертання зірок і газу на різних відстанях від центру спіральних галактик. Знову ж таки, вона використовувала ефект Доплера, аналізуючи спектри світла від об'єктів у галактичному диску.
* **У чому була проблема (розбіжність з очікуваннями):**
    * **Очікування:** Згідно з класичною фізикою, більша частина маси спіральної галактики зосереджена в її центрі (балдже та внутрішніх частинах диска), де сконцентровано більшість зірок. Отже, швидкість обертання зірок мала б зростати від центру до певного піку, а потім зменшуватися з відстанню, подібно до того, як швидкість планет зменшується, чим далі вони від Сонця.
    * **Реальність:** Спостереження Рубін показали, що швидкість обертання зірок та газу залишалася майже **постійною або навіть зростала** на великих відстанях від центру галактики, далеко за межами видимої частини галактичного диска. Це було дуже дивно, оскільки там не було достатньо видимої матерії, щоб пояснити таку швидкість гравітацією.
* **Висновок Віри Рубін:** Єдиним поясненням для такої "плоскої" кривої обертання було припущення, що галактики оточені величезним, невидимим **гало темної матерії**, яке простягається набагато далі, ніж видимий диск. Ця темна матерія додає додаткову гравітаційну силу, яка "тримає" зовнішні частини галактики, не дозволяючи їм відірватися.

### Інші докази, які підтвердили існування темної матерії:

Крім обертання галактик і динаміки скупчень, є й інші вагомі докази:

* **Гравітаційне лінзування:** Скупчення галактик та їхні гало темної матерії викривляють простір-час і, відповідно, світло від далеких об'єктів, що проходять повз них. Цей ефект називається гравітаційним лінзуванням. Спостереження за викривленням світла від далеких галактик дозволяють астрономам "зважити" загальну масу скупчень, і ці виміри також показують, що маси скупчень значно перевищують маси видимої матерії. Особливо вражаючим є спостереження за **скупченням Куля (Bullet Cluster)**, де видно, як дві галактичні скупчення зіткнулися. При зіткненні видима матерія (газ) взаємодіє і гальмується, тоді як темна матерія проходить крізь одна одну без значної взаємодії. Карти розподілу маси (отримані з гравітаційного лінзування) показують, що більша частина маси (темна матерія) знаходиться там, де мав би бути центр мас, а не там, де сконцентрований гарячий газ. Це є дуже переконливим прямим доказом існування темної матерії.
* **Космічний мікрохвильовий фон (CMB):** Аналіз флуктуацій температури в реліктовому випромінюванні (відлуння Великого Вибуху) також свідчить про необхідність існування темної матерії для пояснення великомасштабної структури Всесвіту. Моделі без темної матерії не можуть відтворити спостережувану картину анізотропій CMB.
* **Формування великомасштабної структури:** Моделювання формування галактик, скупчень і ниток у Всесвіті (космічної павутини) показує, що без гравітаційного впливу темної матерії ці структури не могли б утворитися так швидко і мати такий вигляд, як ми їх спостерігаємо. Темна матерія є своєрідним "каркасом", на якому "наростає" видима матерія.

Таким чином, від перших натяків Цвіккі до систематичних досліджень Рубін і подальших спостережень, астрономи та фізики прийшли до висновку, що темна матерія є невід'ємним компонентом Всесвіту. Хоча ми не бачимо її безпосередньо і не знаємо, з яких частинок вона складається, її гравітаційний вплив є беззаперечним.

Космологія — це галузь астрономії, яка вивчає Всесвіт як єдине ціле: його походження, еволюцію, великомасштабну структуру та кінцеву долю. Це надзвичайно широка і складна дисципліна, яка поєднує спостереження з телескопів, теоретичну фізику (особливо загальну теорію відносності Ейнштейна), а також елементи філософії.

### Основні концепції сучасної космології:

1.  **Принцип Коперника (або Космологічний принцип):** Стверджує, що Земля не є якимось особливим або привілейованим місцем у Всесвіті. У більш широкому сенсі, Всесвіт є **ізотропним** (виглядає однаково в усіх напрямках) та **однорідним** (виглядає однаково в усіх місцях) на великих масштабах. Цей принцип значно спрощує математичні моделі Всесвіту.
2.  **Розширення Всесвіту:** Одним з найважливіших відкриттів 20-го століття (Хаббл, 1929) є те, що Всесвіт розширюється. Галактики віддаляються одна від одної, і чим далі галактика, тим швидше вона віддаляється. Це розширення не є "розширенням у щось", а радше розтягуванням самого простору.
3.  **Модель Великого Вибуху (Big Bang):** Це домінуюча космологічна модель, яка описує, як Всесвіт розвивався з надзвичайно гарячого, щільного стану близько 13.8 мільярдів років тому. Вона успішно пояснює:
    * **Розширення Всесвіту (закон Хаббла):** Виміряні швидкості віддалення галактик прямо пропорційні їхнім відстаням.
    * **Існування космічного мікрохвильового фону (CMB):** Це реліктове випромінювання, що залишилося від раннього гарячого Всесвіту, яке є однорідним і ізотропним, з невеликими анізотропіями, що є "відбитками" початкових флуктуацій.
    * **Поширеність легких елементів (первинний нуклеосинтез):** Модель Великого Вибуху точно передбачає спостережувані співвідношення водню, гелію та літію, які утворилися в перші хвилини після Вибуху.
4.  **Темна матерія та Темна енергія:** Ці загадкові компоненти становлять близько 95% маси-енергії Всесвіту.
    * **Темна матерія (близько 27%):** Була запідозрена через її гравітаційний вплив на видиму матерію (як ми вже обговорювали щодо обертання галактик та скупчень). Вона не взаємодіє зі світлом, тому її неможливо безпосередньо спостерігати.
    * **Темна енергія (близько 68%):** Була відкрита наприкінці 1990-х років, коли з'ясувалося, що розширення Всесвіту не сповільнюється, а **прискорюється**. Темна енергія є гіпотетичною формою енергії, яка, як вважається, відповідає за це прискорення, діючи як свого роду антигравітація. Її природа залишається однією з найбільших загадок сучасної фізики.

### Космологічні парадокси

Космологічні парадокси — це уявні суперечності, які виникають при застосуванні певних фізичних принципів або повсякденного досвіду до Всесвіту в цілому, особливо до його нескінченної або вічної природи, доки не були знайдені адекватні космологічні пояснення. Вони відіграли важливу роль у розвитку сучасної космології.

#### 1. Парадокс Ольберса (Парадокс темного нічного неба)

* **Суть парадоксу:** Якщо Всесвіт нескінченний, однорідний і наповнений зірками, які розподілені більш-менш рівномірно, то, куди б ми не подивилися вночі, наш погляд зрештою мав би впиратися в якусь зірку. Отже, нічне небо мало б бути яскравим, як Сонце, а не темним.
* **Чому це парадокс:** Наше нічне небо *темне*. Це спостереження явно суперечить посилкам.
* **Розв'язання парадоксу:** Сучасна космологія пропонує два основні пояснення:
    * **Скінченний вік Всесвіту:** Це головне пояснення. Всесвіт не є нескінченним і вічним. Він має кінцевий вік (близько 13.8 мільярдів років). Це означає, що світло від зірок, що знаходяться далі 13.8 мільярдів світлових років, ще не встигло досягти нас. Таким чином, існує "горизонт подій", за яким ми не можемо бачити джерела світла, навіть якби вони там були.
    * **Розширення Всесвіту та червоне зміщення:** Світло від дуже віддалених галактик, що швидко віддаляються від нас (через розширення Всесвіту), зазнає значного червоного зміщення. Це означає, що їхні довжини хвиль розтягуються, і світло стає менш енергійним і може навіть переходити з видимого діапазону в інфрачервоний або радіодіапазон, стаючи невидимим для наших очей.
    * *Додаткові фактори (менш значущі):* Поглинання світла міжзоряним пилом та газом також зменшує яскравість далеких джерел, але це не було б достатньо для розв'язання парадоксу само по собі.

#### 2. Гравітаційний парадокс (Парадокс Ньютона-Зелігера)

* **Суть парадоксу:** Якщо Всесвіт нескінченний і наповнений матерією, розподіленою більш-менш однорідно, то, згідно з ньютонівською теорією гравітації, кожна точка у Всесвіті зазнавала б нескінченно великої гравітаційної сили в усіх напрямках. Це означало б, що жодна стабільна структура не могла б існувати, і Всесвіт мав би колапсувати або бути розірваним на частини.
* **Чому це парадокс:** Ми бачимо, що Всесвіт містить стабільні структури (зірки, галактики, скупчення), і він не колапсує.
* **Розв'язання парадоксу:** Цей парадокс був вирішений з появою **загальної теорії відносності (ЗТВ) Ейнштейна**.
    * **Обмеженість ньютонівської гравітації:** Ньютонівська гравітація не є адекватною для опису Всесвіту в цілому, особливо на космологічних масштабах, де просторовий час викривляється.
    * **Розширення Всесвіту:** Як і у випадку з Ольберсом, розширення Всесвіту грає ключову роль. Активне розширення простору "розтягує" гравітаційні взаємодії на великих відстанях, перешкоджаючи нескінченному накопиченню гравітаційного потенціалу.
    * **Рішення Фрідмана-Леметра-Робертсона-Вокера (FLRW):** ЗТВ дозволяє існування динамічних, розширюваних Всесвітів (що описуються метрикою FLRW), які не підпадають під умови, що призводять до парадоксу в ньютонівській моделі. У ЗТВ гравітація не є миттєвою силою, що діє на відстані, а є викривленням простору-часу.
    * **Темна енергія:** На додаток, існування темної енергії, яка спричиняє прискорене розширення, ще більше допомагає уникнути гравітаційного колапсу на великих масштабах.

#### 3. Теплова смерть Всесвіту (не зовсім парадокс, скоріше прогноз)

Хоча це не "парадокс" у тому ж сенсі, як Ольберса або Ньютона-Зелігера, концепція теплової смерті Всесвіту є важливим космологічним прогнозом, який базується на законах термодинаміки.

* **Суть:** Згідно з другим законом термодинаміки, ентропія (міра безладу або розсіювання енергії) замкненої системи завжди зростає. Якщо Всесвіт є замкненою системою, то з часом вся енергія в ньому буде рівномірно розподілена, температури вирівняються, і не залишиться доступної енергії для виконання роботи. Це призведе до стану максимальної ентропії, де не відбуватиметься жодних процесів – Всесвіт стане "мертвим" і холодним.
* **Прогнозований сценарій:**
    * Зірки вигорять, чорні діри випаруються (через випромінювання Гокінга).
    * Матерія розпадеться на елементарні частинки.
    * Всесвіт буде продовжувати розширюватися, стаючи все більш холодним і порожнім.
* **Актуальність:** Цей сценарій є одним з можливих "кінців" Всесвіту, якщо його розширення буде продовжуватися необмежено (що підтримується існуванням темної енергії). Існують і інші сценарії, такі як "Велике стиснення" (якщо темна енергія зникне або зміниться) або "Великий розрив" (якщо темна енергія стане настільки сильною, що розірве все, аж до атомів).

Космологічні парадокси, хоч і здавалися суперечливими, насправді слугували важливими каталізаторами для розвитку сучасних моделей Всесвіту, спонукаючи вчених шукати нові, більш точні та повні теорії, які б відповідали спостереженням.
Чудово! Космологія — це захоплива наука, яка вивчає Всесвіт як єдине ціле: його походження, еволюцію, будову та долю. Вона прагне відповісти на фундаментальні питання про наше місце у Всесвіті.

### Що таке Космологія?

**Космологія** (від грецького *космос* – світ, порядок, і *логос* – вчення) — це розділ астрономії та фізики, що вивчає Всесвіт як ціле. Вона займається наймасштабнішими явищами, що охоплюють мільярди світлових років, і досліджує:

* **Походження Всесвіту:** Теорія Великого Вибуху є наріжним каменем сучасної космології, описуючи початковий стан Всесвіту та його розширення.
* **Еволюція Всесвіту:** Як Всесвіт змінювався з часом, від гарячої, щільної плазми до формування зірок, галактик, скупчень галактик та великомасштабної структури.
* **Склад Всесвіту:** Визначення пропорцій видимої матерії (зірок, газу, пилу), темної матерії та темної енергії.
* **Геометрія Всесвіту:** Чи є Всесвіт плоским, викривленим (як поверхня кулі) або гіперболічним (як сідло).
* **Доля Всесвіту:** Яким буде кінцевий результат розширення Всесвіту – чи продовжить він розширюватися нескінченно, чи сповільниться, чи навіть почне стискатися.

Сучасна космологія значною мірою ґрунтується на **Загальній теорії відносності Ейнштейна**, яка описує гравітацію як викривлення простору-часу. Це дозволяє моделювати динаміку Всесвіту в цілому.

### Космологічні парадокси

Космологічні парадокси – це суперечності, які виникають при спробі застосувати відомі фізичні закони та принципи до Всесвіту в цілому, особливо до його ранніх моделей, що базувалися на уявленнях про нескінченний, статичний та однорідний Всесвіт. Ці парадокси були важливим двигуном для розвитку сучасної космології та привели до розуміння, що Всесвіт не є ні статичним, ні нескінченним у часі.

Найбільш відомі космологічні парадокси:

#### 1. Фотометричний парадокс (Парадокс Ольберса)

* **Хто сформулював:** Вперше його порушив Жан-Філіпп де Шезо (1744), а пізніше більш детально проаналізував Генріх Вільгельм Ольберс (1823).
* **Суть парадоксу:** Якщо Всесвіт нескінченний, однорідний (зірки рівномірно розподілені) і стаціонарний (не змінюється з часом), то в будь-якому напрямку, куди б ми не подивилися, наш промінь зору рано чи пізно мав би натрапити на зірку. Це означало б, що нічне небо мало б бути таким же яскравим, як поверхня Сонця, або принаймні дуже яскравим, а не темним.
* **Чому виникає:**
    * Яскравість зірки зменшується обернено пропорційно квадрату відстані ($1/r^2$).
    * Але кількість зірок у сферичній оболонці (на певній відстані $r$) збільшується пропорційно квадрату відстані ($r^2$, оскільки об'єм оболонки збільшується як $r^2 \cdot \Delta r$).
    * Таким чином, освітленість від кожної оболонки (нехтуючи поглинанням світла) була б однаковою, незалежно від відстані. Якщо таких оболонок нескінченна кількість, то сумарна освітленість була б нескінченною.
* **Розв'язання парадоксу сучасною космологією:**
    1.  **Скінченний вік Всесвіту:** Найважливіший фактор. Всесвіт не є нескінченно старим. Він виник близько 13.8 мільярдів років тому. Це означає, що світло від зірок, що знаходяться далі, ніж **горизонт частинок** (максимальна відстань, з якої світло встигло досягти нас з моменту Великого Вибуху), просто не встигло до нас дійти. Тобто ми бачимо лише скінченну частину Всесвіту.
    2.  **Розширення Всесвіту та червоне зміщення:** Світло від дуже далеких галактик, які віддаляються від нас (через розширення Всесвіту), зазнає значного **червоного зміщення**. Це означає, що їхні фотони втрачають енергію, і їхнє випромінювання зміщується в бік довших хвиль (наприклад, з видимого світла в інфрачервоний діапазон), роблячи їх менш яскравими або зовсім невидимими для людського ока.
    3.  **Еволюція зірок:** Зірки не існують вічно. Вони народжуються, еволюціонують і помирають. Світло від найстаріших зірок вже згасло.

#### 2. Гравітаційний парадокс (Парадокс Неймана-Зелігера)

* **Хто сформулював:** Карл Нейман (1894) та Гуго фон Зелігер (1896) незалежно один від одного.
* **Суть парадоксу:** Якщо Всесвіт нескінченний, однорідний і заповнений матерією з ненульовою середньою густиною, то, згідно з ньютонівською теорією гравітації, гравітаційна сила, що діє на будь-яке тіло, мала б бути нескінченно великою і спрямованою в усі боки одночасно. Це робило б неможливим стабільне існування будь-яких об'єктів (зірок, галактик) або навіть визначення чіткого напрямку сили.
* **Чому виникає:**
    * Гравітаційна сила віддаленої маси зменшується обернено пропорційно квадрату відстані ($1/r^2$).
    * Але кількість маси у сферичній оболонці збільшується пропорційно квадрату відстані ($r^2$).
    * Таким чином, сума гравітаційних впливів від нескінченно віддалених шарів матерії розходиться (прямує до нескінченності).
* **Розв'язання парадоксу сучасною космологією:**
    1.  **Загальна теорія відносності Ейнштейна:** Цей парадокс є одним з найважливіших свідчень обмеженості ньютонівської теорії гравітації для опису Всесвіту в цілому. Загальна теорія відносності, яка лежить в основі сучасної космології, по-іншому описує гравітацію. Вона дозволяє існування нескінченних всесвітів з однорідним розподілом матерії, які не стикаються з цим парадоксом, оскільки гравітація описується не силою, а викривленням простору-часу.
    2.  **Розширення Всесвіту:** Розширення Всесвіту також відіграє роль. Воно ефективно "розбавляє" матерію, зменшуючи її середню густину з часом і ускладнюючи формування нескінченних сумарних сил.
    3.  **Неоднорідності на великих масштабах:** Хоча Всесвіт є однорідним на дуже великих масштабах, на менших масштабах (до мільярдів світлових років) існують кластеризації матерії (галактики, скупчення, нитки), які є гравітаційно зв'язаними системами.

#### 3. Парадокс теплової смерті Всесвіту (Термодинамічний парадокс)

* **Суть парадоксу:** Згідно з другим законом термодинаміки, в замкненій системі ентропія (міра хаосу або розсіювання енергії) постійно зростає. Якщо Всесвіт є замкненою системою і існує нескінченно довго, то він повинен був би вже досягти стану максимальної ентропії, тобто "теплової смерті". У цьому стані всі види енергії перетворилися б на рівномірно розподілене тепло, не було б жодних градієнтів температур чи енергії, а отже, не відбувалося б жодних процесів, життя, чи руху. Оскільки ми спостерігаємо активний Всесвіт, це свідчить про те, що він не досяг стану теплової смерті.
* **Розв'язання парадоксу сучасною космологією:**
    1.  **Скінченний вік Всесвіту:** Як і у випадку з парадоксом Ольберса, ключовим є те, що Всесвіт має скінченний вік. Він просто не існував достатньо довго, щоб досягти стану термодинамічної рівноваги.
    2.  **Розширення Всесвіту:** Розширення Всесвіту постійно створює новий простір і "розбавляє" енергію, що уповільнює досягнення термодинамічної рівноваги. Ефективно, Всесвіт не є повністю замкненою системою в класичному сенсі, оскільки він розширюється.
    3.  **Наявність темної енергії:** Сучасні спостереження показують, що розширення Всесвіту прискорюється завдяки темній енергії. Це означає, що Всесвіт, ймовірно, ніколи не досягне повного теплового рівноваги, оскільки його компоненти будуть розлітатися все далі і далі, стаючи все більш розрідженими. Кінцевий стан Всесвіту, ймовірно, буде холодним, порожнім і безжиттєвим, але це не буде класична "теплова смерть" в сенсі максимальної ентропії в обмеженому об'ємі.

Ці парадокси, хоча і були сформульовані на основі уявлень про Всесвіт, які виявилися помилковими, відіграли надзвичайно важливу роль у розвитку космології. Вони змусили вчених переосмислити фундаментальні припущення про природу Всесвіту і проклали шлях до сучасної моделі Великого Вибуху, розширення Всесвіту, його скінченного віку та наявності темної матерії і темної енергії.
