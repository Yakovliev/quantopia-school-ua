**Перший закон термодинаміки та адіабатний процес**

Один із фундаментальних законів природи - закон збереження та перетворення енергії - має цікаву історію відкриття. Німецький лікар і фізик Юліус Роберт фон Маєр (1814-1878) зробив важливе спостереження під час своєї практики. Він помітив, що венозна кров у жителів тропічних країн має світліший колір, ніж у європейців. Це спостереження привело його до розуміння зв'язку між кількістю спожитого кисню та процесами в організмі. Незалежно від Маєра, до схожих висновків прийшли англійський промисловець Джеймс Прескотт Джоуль (1818-1889) та німецький фізик і фізіолог Герман Людвіг Фердинанд фон Гельмгольц (1821-1894).

**Формулювання першого закону термодинаміки**

У термодинаміці розглядаються системи, де механічна енергія при переході між станами може змінюватися. При цьому, якщо зовнішні сили виконують роботу $A'$ і системі передається певна кількість теплоти $Q$, то вся ця енергія йде на зміну внутрішньої енергії системи $\Delta U$. Математично це записується як:

$\Delta U = A' + Q$

Важливо зазначити: якщо система отримує теплоту, то $Q$ береться зі знаком "+", якщо віддає - зі знаком "-". На практиці частіше розглядають роботу $A$, яку виконує сама система проти зовнішніх сил. Оскільки $A' = -A$, перший закон термодинаміки можна записати у формі:

$Q = \Delta U + A$

**Перший закон термодинаміки для різних процесів**

1. Ізохорний процес $(V = const)$:
   - Об'єм газу не змінюється $(∆V = 0)$
   - Робота газу дорівнює нулю $(A = 0)$
   - Рівняння набуває вигляду: $Q = \Delta U$
   - Для ідеального одноатомного газу: $Q = \Delta U = \frac{3}{2} \frac{m}{M}R\Delta T$

2. Ізотермічний процес $(T = const)$:
   - Температура не змінюється
   - Внутрішня енергія залишається сталою $(\Delta U = 0)$
   - Рівняння набуває вигляду: $Q = A$
   - Вся передана теплота йде на виконання роботи

3. Ізобарний процес $(p = const)$:
   - Тиск залишається сталим
   - Рівняння має вигляд: $Q = \Delta U + A$
   - Для ідеального одноатомного газу: $Q = \frac{5}{2}p\Delta V$ або $Q = \frac{5}{2}\frac{m}{M}R\Delta T$

**Адіабатний процес**

Адіабатним називається процес, який відбувається без теплообміну з навколишнім середовищем $(Q = 0)$. У цьому випадку перший закон термодинаміки набуває вигляду:

$\Delta U + A = 0$ або $A = -\Delta U$

При адіабатному розширенні газ виконує додатну роботу за рахунок зменшення внутрішньої енергії, що призводить до зниження його температури. При адіабатному стисненні спостерігається протилежний ефект.

В реальних умовах процеси, близькі до адіабатних, можна спостерігати:
- У посудині з хорошою теплоізоляцією
- При дуже швидких процесах, коли теплообмін не встигає відбутися
- При поширенні звукових хвиль
- При швидкому розширенні газу під час вибуху

Практичне застосування адіабатного стиснення можна побачити в дизельних двигунах, де саме різке стиснення повітря призводить до значного підвищення температури, достатньої для самозаймання паливної суміші.

### **Перший закон термодинаміки**

Перший закон термодинаміки формулює закон збереження енергії для термодинамічних систем. При переході системи з одного стану в інший, якщо зовнішні сили виконують роботу $ A' $, а системі передається певна кількість теплоти $ Q $, то зміна внутрішньої енергії $ \Delta U $ визначається як:
$$
\Delta U = A' + Q,
$$
де:
- $ A' $ — робота зовнішніх сил,
- $ Q $ — кількість теплоти.

З огляду на те, що робота, яку виконує система проти зовнішніх сил, дорівнює $ A = -A' $, рівняння набуває вигляду:
$$
Q = \Delta U + A.
$$

**Зверніть увагу:** Якщо теплота передається системі, то $ Q > 0 $. Якщо система втрачає теплоту, $ Q < 0 $.

---

### **Варіанти першого закону для ізопроцесів**

1. **Ізохорний процес** ($ V = const $):
   $$
   Q = \Delta U.
   $$
   Уся теплота витрачається на збільшення внутрішньої енергії газу.

2. **Ізотермічний процес** ($ T = const $):
   $$
   Q = A.
   $$
   Внутрішня енергія не змінюється ($ \Delta U = 0 $), вся теплота йде на виконання роботи.

3. **Ізобарний процес** ($ p = const $):
   $$
   Q = \Delta U + A.
   $$
   Теплота витрачається як на зміну внутрішньої енергії, так і на виконання роботи:
   $$
   Q = \frac{5}{2} \frac{m}{M} RT.
   $$

---

### **Адіабатний процес**

Адіабатний процес — це термодинамічний процес, що відбувається без теплообміну ($ Q = 0 $):
$$
\Delta U = -A.
$$

- **Розширення:** Газ виконує роботу, його внутрішня енергія зменшується, а температура падає.
- **Стиснення:** Газ втрачає об’єм, його внутрішня енергія збільшується, температура зростає.

**Особливості:**
- Адіабатне стиснення відбувається швидше, ніж ізотермічне, через відсутність обміну теплом з навколишнім середовищем.
- Для опису використовується рівняння Пуассона:
  $$
  pV^\gamma = const,
  $$
  де $ \gamma = \frac{C_p}{C_v} $ — показник адіабати.

**Застосування:** Адіабатні процеси використовуються в дизельних двигунах, де адіабатне стиснення призводить до займання паливної суміші.

---

### **Підсумки**

1. Перший закон термодинаміки описує збереження енергії в термодинамічних процесах.
2. У різних ізопроцесах теплота або витрачається на зміну внутрішньої енергії, або йде на виконання роботи.
3. Адіабатний процес є прикладом ефективного використання енергії без теплообміну.
